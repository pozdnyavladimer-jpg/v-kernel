```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D37_SPECTRAL_ORBITAL_MODES
# Intent-clock scan for spectral orbital memory
#
# Key format:
#   D{direction}_O{octave}_M{mode}
#
# Example:
#   D4_O5_M3
#
# Meaning:
#   D = spectral direction / color
#   O = octave / energy level
#   M = orbital mode / standing-wave shape
#
# A valid write requires:
#   1) spectral resonance: direction + octave
#   2) orbital mode match: M0..M5
#   3) stable 6-phase intent clock scan: S0..S5
#   4) receiver expected intent match
# ============================================================

np.random.seed(37)

STATE = "D37_SPECTRAL_ORBITAL_MODES"

directions = 6
octaves = 7
spectral_states = directions * octaves

orbital_modes = 6
clock_phases = 6

resonance_threshold = 0.62
ambiguity_margin = 0.12

intent_score_threshold = 0.82
intent_margin_threshold = 0.10
phase_scan_tolerance = 0.22

PHI = (1 + np.sqrt(5)) / 2

phase_names = [
    "S0_EXPAND_1",
    "S1_EXPAND_PHI",
    "S2_EXPAND_PHI2",
    "S3_COLLAPSE_PHI2",
    "S4_COLLAPSE_PHI",
    "S5_COLLAPSE_1",
]

orbital_names = {
    0: "CORE",
    1: "TWO_LOBE",
    2: "FOUR_LOBE",
    3: "FLOWER",
    4: "RING",
    5: "SPIRAL",
}

# 6-phase breathing signatures.
# This is the "intent shape" scanned by clock.
orbital_templates = {
    0: np.array([0.92, 0.96, 1.00, 0.96, 0.92, 0.96]),
    1: np.array([0.18, 0.62, 1.00, 0.62, 0.18, 0.62]),
    2: np.array([0.25, 0.85, 0.55, 0.85, 0.25, 0.55]),
    3: np.array([0.35, 0.75, 1.00, 0.82, 0.45, 0.68]),
    4: np.array([0.62, 0.92, 0.62, 0.92, 0.62, 0.92]),
    5: np.array([0.22, 0.42, 0.68, 1.00, 0.72, 0.46]),
}

receiver_filter_sequence = [
    "D4_O5_M3",
    "D1_O2_M1",
    "D3_O6_M5",
    "D5_O4_M2",
    "D0_O1_M0",
    "D2_O3_M4",
    "D4_O5_M3",
    "D1_O2_M1",
]

receiver_memory = ["EMPTY"] * len(receiver_filter_sequence)


# ------------------------------------------------------------
# Key utilities
# ------------------------------------------------------------

def parse_intent_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE"]:
        return None, None, None

    parts = str(key).split("_")

    if len(parts) == 2:
        d_part, o_part = parts
        m_part = "M0"
    elif len(parts) == 3:
        d_part, o_part, m_part = parts
    else:
        raise ValueError(f"Bad intent key format: {key}")

    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))
    mode = int(m_part.replace("M", ""))

    return direction, octave, mode


def make_intent_key(direction, octave, mode):
    return f"D{direction}_O{octave}_M{mode}"


def make_spectral_key(direction, octave):
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    return octave * directions + direction


def key_phase(direction):
    return round(direction / directions, 3)


def orbital_token(key):
    d, o, m = parse_intent_key(key)

    if d is None:
        return "NONE"

    frequency = o + 1
    phase = key_phase(d)
    mode_name = orbital_names[m]

    return f"ORBITAL[D{d}|O{o}|M{m}|{mode_name}|F{frequency}|P{phase}]"


def key_color(key, strength=1.0):
    if key in [None, "EMPTY", "NONE"]:
        return "black"

    if key == "WHITE":
        return "white"

    d, o, m = parse_intent_key(key)

    if d is None:
        return "black"

    hue_value = ((d + 1) / (directions + 1) + 0.055 * (o + 1) + 0.025 * (m + 1)) % 1.0
    return plt.cm.hsv(hue_value)


# ------------------------------------------------------------
# Packet generation / decoding
# ------------------------------------------------------------

def normalize_trace(trace):
    trace = np.asarray(trace, dtype=float)
    mx = float(np.max(trace))

    if mx <= 1e-9:
        return trace

    return trace / mx


def mode_score(trace, template):
    trace_n = normalize_trace(trace)
    template_n = normalize_trace(template)

    dist = np.linalg.norm(trace_n - template_n) / np.sqrt(len(trace_n))
    return max(0.0, 1.0 - float(dist))


def make_orbital_trace(
    key,
    noise=0.015,
    blend_key=None,
    blend_weight=0.5,
    roll=0,
    scale=1.0,
):
    _, _, mode = parse_intent_key(key)

    base = orbital_templates[mode].copy()

    if blend_key not in [None, "EMPTY", "NONE", "WHITE"]:
        _, _, blend_mode = parse_intent_key(blend_key)
        blend = orbital_templates[blend_mode].copy()
        base = (1.0 - blend_weight) * base + blend_weight * blend

    if roll != 0:
        base = np.roll(base, roll)

    trace = scale * base + np.random.normal(0.0, noise, size=clock_phases)
    return np.clip(trace, 0.0, 1.0)


def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0,
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.18 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o, _ = parse_intent_key(target_key)
        packet[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d2, o2, _ = parse_intent_key(ambiguous_key)
        packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_spectral_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_spectral_key(best_direction, best_octave)
    second_key = make_spectral_key(second_direction, second_octave)

    margin = best_score - second_score
    spectral_ok = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_spectral_key": best_key,
        "best_state_id": best_id,
        "best_direction": best_direction,
        "best_octave": best_octave,
        "best_score": best_score,
        "second_spectral_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "spectral_margin": margin,
        "spectral_ok": spectral_ok,
    }


def decode_orbital_mode(trace):
    scores = []

    for mode in range(orbital_modes):
        s = mode_score(trace, orbital_templates[mode])
        scores.append((mode, s))

    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    best_mode, best_score = scores[0]
    second_mode, second_score = scores[1]

    margin = best_score - second_score
    intent_scan_ok = best_score >= intent_score_threshold and margin >= intent_margin_threshold

    return {
        "best_mode": best_mode,
        "best_mode_name": orbital_names[best_mode],
        "best_mode_score": best_score,
        "second_mode": second_mode,
        "second_mode_name": orbital_names[second_mode],
        "second_mode_score": second_score,
        "mode_margin": margin,
        "intent_scan_ok": intent_scan_ok,
    }


def decode_intent(packet, trace):
    spectral = decode_spectral_packet(packet)
    orbital = decode_orbital_mode(trace)

    d = spectral["best_direction"]
    o = spectral["best_octave"]
    m = orbital["best_mode"]

    decoded_intent_key = make_intent_key(d, o, m)
    intent_locked = bool(spectral["spectral_ok"] and orbital["intent_scan_ok"])

    return {
        **spectral,
        **orbital,
        "decoded_intent_key": decoded_intent_key,
        "intent_locked": intent_locked,
    }


# ------------------------------------------------------------
# Visualization helpers
# ------------------------------------------------------------

def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.25

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=13,
        linewidth=1.8,
        color=color,
        alpha=alpha,
    )
    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.14
    gap = 0.03

    ax.text(x0, y0 + 1.38, title, fontsize=6.5, ha="left", va="bottom", color="black")

    selected = (None, None)

    if selected_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o, _ = parse_intent_key(selected_key)
        selected = (d, o)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = packet[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)
            edge = "lime" if is_selected else "white"
            lw = 1.5 if is_selected else 0.25

            ax.add_patch(Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw))


def draw_memory_slots(ax, x0, y0, seq, title, active_slot=None):
    cell = 0.18
    gap = 0.045

    ax.text(x0, y0 + 1.85, title, fontsize=6.5, ha="left", va="bottom", color="black")

    for i, key in enumerate(seq):
        x = x0
        y = y0 + (len(seq) - 1 - i) * (cell + gap)

        color = key_color(key)
        edge = "lime" if i == active_slot else "black"
        lw = 1.5 if i == active_slot else 0.45

        ax.add_patch(Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw))
        ax.text(x + 0.27, y + cell / 2, str(i), fontsize=5.5, ha="left", va="center", color="black")


def draw_phase_clock(ax, x0, y0, trace, locked):
    ax.text(x0, y0 + 0.48, "intent clock scan", fontsize=6.5, ha="left", color="black")

    max_v = max(float(np.max(trace)), 1e-9)

    for i, v in enumerate(trace):
        x = x0 + i * 0.28
        color = plt.cm.viridis(float(v) / max_v)
        edge = "lime" if locked else "red"

        ax.add_patch(Circle((x, y0), 0.10, facecolor=color, edgecolor=edge, linewidth=0.8))
        ax.text(x, y0 - 0.22, f"S{i}", fontsize=5.5, ha="center", color="black")


def draw_orbital_shape(ax, cx, cy, key, trace, accepted):
    if key in [None, "EMPTY", "NONE", "WHITE"]:
        return

    d, o, m = parse_intent_key(key)

    theta = np.linspace(0.0, 2.0 * np.pi, 280)
    phase_shift = (d / directions) * 2.0 * np.pi
    envelope = float(np.mean(normalize_trace(trace)))

    if m == 0:
        r = 0.28 + 0.04 * np.cos(theta * 6)
    elif m == 1:
        r = 0.18 + 0.22 * (np.cos(theta - phase_shift) ** 2)
    elif m == 2:
        r = 0.14 + 0.24 * (np.cos(2 * (theta - phase_shift)) ** 2)
    elif m == 3:
        r = 0.12 + 0.22 * (np.cos(3 * (theta - phase_shift)) ** 2)
    elif m == 4:
        r = 0.26 + 0.05 * np.sin(6 * theta + phase_shift)
    else:
        r = 0.08 + 0.25 * (theta / (2 * np.pi))

    r = r * (0.80 + 0.35 * envelope)

    x = cx + r * np.cos(theta)
    y = cy + r * np.sin(theta)

    color = key_color(key)

    ax.plot(x, y, color=color, linewidth=1.5)
    ax.add_patch(Circle((cx, cy), 0.04, facecolor=color, edgecolor="black", linewidth=0.5))

    edge = "lime" if accepted else "red"
    ax.add_patch(Circle((cx, cy), 0.48, fill=False, edgecolor=edge, linewidth=1.2))
    ax.text(cx, cy - 0.72, orbital_names[m], fontsize=6, ha="center", color="black")


def draw_snapshot(ax, case_row, packet, trace, memory_state):
    slot = int(case_row["MEMORY_SLOT"])
    source_key = case_row["SOURCE_INTENT_KEY"]
    expected_key = case_row["EXPECTED_INTENT_KEY"]
    decoded_key = case_row["DECODED_INTENT_KEY"]

    intent_locked = bool(case_row["INTENT_LOCKED"])
    absorbed = bool(case_row["ABSORBED"])

    ax.set_facecolor("white")
    ax.set_xlim(0, 14.2)
    ax.set_ylim(0.5, 5.4)
    ax.set_aspect("equal")
    ax.axis("off")

    source_color = key_color(source_key)
    decoded_color = key_color(decoded_key)
    mem_color = key_color(memory_state[slot] if absorbed else "EMPTY")

    ax.add_patch(Circle((0.75, 3.25), 0.34, facecolor=source_color, edgecolor="black", linewidth=1.0))
    ax.text(0.75, 2.62, f"SRC\n{source_key}", fontsize=6.5, ha="center", color="black")

    ax.add_patch(Circle((1.70, 3.25), 0.23, facecolor=source_color, edgecolor="black", linewidth=1.0))
    ax.text(1.70, 2.62, "WAVE", fontsize=6.5, ha="center", color="black")

    ax.add_patch(Rectangle((2.35, 3.02), 0.42, 0.42, facecolor="limegreen", edgecolor="black", linewidth=1.0))
    ax.text(2.56, 2.62, "FIELD", fontsize=6.5, ha="center", color="black")

    clock_color = "limegreen" if intent_locked else "red"
    ax.add_patch(Rectangle((3.25, 3.02), 0.42, 0.42, facecolor=clock_color, edgecolor="black", linewidth=1.0))
    ax.text(3.46, 2.62, "CLOCK", fontsize=6.5, ha="center", color="black")

    ax.add_patch(Circle((4.45, 3.25), 0.34, facecolor=decoded_color, edgecolor="lime" if intent_locked else "red", linewidth=2.0))
    ax.text(4.45, 2.48, f"DECODE\n{decoded_key}", fontsize=6.2, ha="center", color="black")

    ax.add_patch(Circle((5.75, 3.25), 0.34, facecolor=mem_color, edgecolor="lime" if absorbed else "red", linewidth=2.0))
    ax.text(5.75, 2.48, f"MEM\nslot {slot}", fontsize=6.5, ha="center", color="black")

    rejected = not absorbed

    draw_arrow(ax, (1.08, 3.25), (1.47, 3.25), active=True, rejected=False)
    draw_arrow(ax, (1.93, 3.25), (2.35, 3.25), active=True, rejected=False)
    draw_arrow(ax, (2.77, 3.25), (3.25, 3.25), active=True, rejected=rejected)
    draw_arrow(ax, (3.67, 3.25), (4.12, 3.25), active=True, rejected=rejected)
    draw_arrow(ax, (4.80, 3.25), (5.40, 3.25), active=True, rejected=rejected)

    draw_phase_clock(ax, 0.62, 1.52, trace, intent_locked)
    draw_orbital_shape(ax, 6.95, 3.30, decoded_key, trace, absorbed)
    draw_packet_grid(ax, 8.05, 2.42, packet, "42-channel spectral packet", selected_key=decoded_key)
    draw_memory_slots(ax, 11.55, 1.85, memory_state, "orbital memory", active_slot=slot)

    ax.text(
        0.50,
        0.72,
        f"case={case_row['CASE']} slot={slot} expected={expected_key} decoded={decoded_key} "
        f"mode={case_row['MODE_NAME']} score={case_row['MODE_SCORE']} "
        f"action={case_row['ACTION']}\n{case_row['OUTPUT_TOKEN']}",
        fontsize=6.2,
        ha="left",
        color="black",
    )

    ax.set_title(
        f"D37 snapshot t={case_row['T']} — {case_row['CASE']} / {case_row['ACTION']}",
        fontsize=9,
        color="black",
    )


# ------------------------------------------------------------
# Test cases
# ------------------------------------------------------------

cases = [
    {
        "case": "INTENT_MATCH_FLOWER",
        "slot": 0,
        "source_key": "D4_O5_M3",
        "emitted_key": "D4_O5_M3",
        "expected_key": "D4_O5_M3",
        "packet": spectral_packet("D4_O5_M3", strength=0.92),
        "trace": make_orbital_trace("D4_O5_M3"),
    },
    {
        "case": "SAME_COLOR_WRONG_MODE",
        "slot": 6,
        "source_key": "D4_O5_M1",
        "emitted_key": "D4_O5_M1",
        "expected_key": "D4_O5_M3",
        "packet": spectral_packet("D4_O5_M1", strength=0.91),
        "trace": make_orbital_trace("D4_O5_M1"),
    },
    {
        "case": "WRONG_DIRECTION_REJECT",
        "slot": 1,
        "source_key": "D0_O2_M1",
        "emitted_key": "D0_O2_M1",
        "expected_key": "D1_O2_M1",
        "packet": spectral_packet("D0_O2_M1", strength=0.90),
        "trace": make_orbital_trace("D0_O2_M1"),
    },
    {
        "case": "WRONG_OCTAVE_REJECT",
        "slot": 2,
        "source_key": "D3_O3_M5",
        "emitted_key": "D3_O3_M5",
        "expected_key": "D3_O6_M5",
        "packet": spectral_packet("D3_O3_M5", strength=0.89),
        "trace": make_orbital_trace("D3_O3_M5"),
    },
    {
        "case": "CLOCK_DRIFT_INTENT_REJECT",
        "slot": 3,
        "source_key": "D5_O4_M2",
        "emitted_key": "D5_O4_M2",
        "expected_key": "D5_O4_M2",
        "packet": spectral_packet("D5_O4_M2", strength=0.90),
        "trace": make_orbital_trace("D5_O4_M2", roll=2),
    },
    {
        "case": "WEAK_ORBITAL_REJECT",
        "slot": 4,
        "source_key": "D0_O1_M0",
        "emitted_key": "D0_O1_M0",
        "expected_key": "D0_O1_M0",
        "packet": spectral_packet("D0_O1_M0", strength=0.39),
        "trace": make_orbital_trace("D0_O1_M0"),
    },
    {
        "case": "AMBIGUOUS_MODE_HOLD",
        "slot": 5,
        "source_key": "D2_O3_M4",
        "emitted_key": "D2_O3_M4",
        "expected_key": "D2_O3_M4",
        "packet": spectral_packet("D2_O3_M4", strength=0.88),
        "trace": make_orbital_trace("D2_O3_M4", blend_key="D2_O3_M2", blend_weight=0.48),
    },
    {
        "case": "WHITE_WITH_TARGET_INTENT",
        "slot": 7,
        "source_key": "WHITE",
        "emitted_key": "D1_O2_M1",
        "expected_key": "D1_O2_M1",
        "packet": spectral_packet("D1_O2_M1", strength=0.90, broad=True),
        "trace": make_orbital_trace("D1_O2_M1"),
    },
]


# ------------------------------------------------------------
# Simulation
# ------------------------------------------------------------

decision_rows = []
scan_rows = []
memory_rows = []
snapshots = []

amplitude_accepted = 0
intent_locked_count = 0
absorbed_count = 0
rejected_count = 0

for t, item in enumerate(cases):
    decoded = decode_intent(item["packet"], item["trace"])

    expected_key = item["expected_key"]
    decoded_key = decoded["decoded_intent_key"]

    spectral_ok = bool(decoded["spectral_ok"])
    intent_scan_ok = bool(decoded["intent_scan_ok"])
    intent_locked = bool(decoded["intent_locked"])

    receiver_match = decoded_key == expected_key
    absorbed = bool(intent_locked and receiver_match)

    before_key = receiver_memory[item["slot"]]

    if absorbed:
        receiver_memory[item["slot"]] = decoded_key
        absorbed_count += 1

    if spectral_ok:
        amplitude_accepted += 1

    if intent_locked:
        intent_locked_count += 1

    if not absorbed:
        rejected_count += 1

    if not spectral_ok:
        action = "REJECT_WEAK_OR_AMBIGUOUS"
    elif not intent_scan_ok:
        action = "REJECT_INTENT_CLOCK_UNLOCKED"
    elif not receiver_match:
        action = "REJECT_INTENT_MISMATCH"
    else:
        action = "ORBITAL_INTENT_WRITE"

    token = orbital_token(decoded_key) if intent_locked else "NONE"

    row = {
        "STATE": STATE,
        "T": t * 120,
        "CASE": item["case"],
        "MEMORY_SLOT": item["slot"],
        "SOURCE_INTENT_KEY": item["source_key"],
        "EMITTED_INTENT_KEY": item["emitted_key"],
        "EXPECTED_INTENT_KEY": expected_key,
        "DECODED_INTENT_KEY": decoded_key,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_SPECTRAL_KEY": decoded["second_spectral_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "SPECTRAL_MARGIN": round(decoded["spectral_margin"], 3),
        "SPECTRAL_OK": int(spectral_ok),
        "MODE": decoded["best_mode"],
        "MODE_NAME": decoded["best_mode_name"],
        "MODE_SCORE": round(decoded["best_mode_score"], 3),
        "SECOND_MODE": decoded["second_mode"],
        "SECOND_MODE_NAME": decoded["second_mode_name"],
        "SECOND_MODE_SCORE": round(decoded["second_mode_score"], 3),
        "MODE_MARGIN": round(decoded["mode_margin"], 3),
        "INTENT_SCAN_OK": int(intent_scan_ok),
        "INTENT_LOCKED": int(intent_locked),
        "RECEIVER_MATCH": int(receiver_match),
        "ABSORBED": int(absorbed),
        "BEFORE_KEY": before_key,
        "AFTER_KEY": receiver_memory[item["slot"]],
        "ACTION": action,
        "OUTPUT_TOKEN": token,
    }

    decision_rows.append(row)

    expected_trace = orbital_templates[parse_intent_key(expected_key)[2]]
    trace_n = normalize_trace(item["trace"])
    expected_n = normalize_trace(expected_trace)

    for phase_idx, phase_name in enumerate(phase_names):
        err = abs(float(trace_n[phase_idx] - expected_n[phase_idx]))

        scan_rows.append({
            "STATE": STATE,
            "CASE": item["case"],
            "MEMORY_SLOT": item["slot"],
            "PHASE_INDEX": phase_idx,
            "PHASE": phase_name,
            "OBSERVED": round(float(trace_n[phase_idx]), 3),
            "EXPECTED": round(float(expected_n[phase_idx]), 3),
            "ABS_ERROR": round(err, 3),
            "PHASE_SCAN_OK": int(err <= phase_scan_tolerance),
        })

    memory_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "MEMORY_SLOT": item["slot"],
        "EXPECTED_KEY": expected_key,
        "DECODED_KEY": decoded_key,
        "BEFORE_KEY": before_key,
        "AFTER_KEY": receiver_memory[item["slot"]],
        "ABSORBED": int(absorbed),
        "HELD_IF_REJECTED": int((not absorbed) and before_key == receiver_memory[item["slot"]]),
        "MATCH_IF_ABSORBED": int(receiver_memory[item["slot"]] == expected_key) if absorbed else "not_written",
    })

    snapshots.append((row, item["packet"], item["trace"], receiver_memory.copy()))

df_decision = pd.DataFrame(decision_rows)
df_scan = pd.DataFrame(scan_rows)
df_memory = pd.DataFrame(memory_rows)

expected_final = ["EMPTY"] * len(receiver_filter_sequence)

for row in decision_rows:
    if row["ABSORBED"] == 1:
        expected_final[row["MEMORY_SLOT"]] = row["EXPECTED_INTENT_KEY"]

orbital_memory_accuracy = float(np.mean(np.array(receiver_memory) == np.array(expected_final)))

intent_safety_accuracy = float(np.mean([
    1 if row["ABSORBED"] == 1 and row["AFTER_KEY"] == row["EXPECTED_KEY"] else
    1 if row["ABSORBED"] == 0 and row["AFTER_KEY"] == row["BEFORE_KEY"] else
    0
    for row in memory_rows
]))

result = (
    "SPECTRAL_ORBITAL_INTENT_LOCKED"
    if orbital_memory_accuracy == 1.0 and intent_safety_accuracy == 1.0
    else "SPECTRAL_ORBITAL_INTENT_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "ORBITAL_MODES": orbital_modes,
    "CLOCK_PHASES": clock_phases,
    "MEMORY_SLOTS": len(receiver_memory),
    "RECEIVER_FILTER_SEQUENCE": " ".join(receiver_filter_sequence),
    "RECEIVER_FINAL_SEQUENCE": " ".join(receiver_memory),
    "EXPECTED_FINAL_SEQUENCE": " ".join(expected_final),
    "CASES_TESTED": len(cases),
    "AMPLITUDE_ACCEPTED_PACKETS": amplitude_accepted,
    "INTENT_LOCKED_PACKETS": intent_locked_count,
    "ABSORBED_PACKETS": absorbed_count,
    "REJECTED_PACKETS": rejected_count,
    "ORBITAL_MEMORY_ACCURACY": round(orbital_memory_accuracy, 3),
    "INTENT_SAFETY_ACCURACY": round(intent_safety_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "INTENT_SCORE_THRESHOLD": intent_score_threshold,
    "INTENT_MARGIN_THRESHOLD": intent_margin_threshold,
    "PHASE_SCAN_TOLERANCE": phase_scan_tolerance,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(9, 3.4 * len(snapshots)))
axes = np.atleast_1d(axes)

for ax, (row, packet, trace, memory_state) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, trace, memory_state)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"orbital modes:                {orbital_modes}")
print(f"clock phases:                 {clock_phases}")
print(f"memory slots:                 {len(receiver_memory)}")
print(f"receiver filter sequence:     {' '.join(receiver_filter_sequence)}")
print(f"receiver final sequence:      {' '.join(receiver_memory)}")
print(f"expected final sequence:      {' '.join(expected_final)}")
print(f"cases tested:                 {len(cases)}")
print(f"amplitude accepted packets:   {amplitude_accepted}")
print(f"intent locked packets:        {intent_locked_count}")
print(f"absorbed packets:             {absorbed_count}")
print(f"rejected packets:             {rejected_count}")
print(f"orbital memory accuracy:      {round(orbital_memory_accuracy, 3)}")
print(f"intent safety accuracy:       {round(intent_safety_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"intent score threshold:       {intent_score_threshold}")
print(f"intent margin threshold:      {intent_margin_threshold}")
print(f"phase scan tolerance:         {phase_scan_tolerance}")

print("\nOrbital intent decision table:")
display(df_decision)

print("\nIntent clock scan table:")
display(df_scan)

print("\nOrbital memory table:")
display(df_memory)

print("\nSummary:")
display(df_summary)
```
