```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D35_SPECTRAL_COHERENCE_CLOCK
# Phase-coherence clock for photon-like spectral memory packets
#
# Goal:
# spectral packet must satisfy:
#   1) correct direction+octave key
#   2) sufficient amplitude
#   3) non-ambiguous margin
#   4) phase aligned with expected spectral phase
#   5) low jitter / coherent clock window
#
# If phase drift is recoverable, clock resync can relock packet.
# ============================================================

np.random.seed(35)

STATE = "D35_SPECTRAL_COHERENCE_CLOCK"

directions = 6
octaves = 7
spectral_states = directions * octaves
memory_slots = 8

resonance_threshold = 0.62
ambiguity_margin = 0.12
phase_tolerance = 0.18
jitter_tolerance = 0.16
retention_threshold = 0.50

receiver_target_sequence = [
    "D4_O5",
    "D1_O2",
    "D3_O6",
    "D5_O4",
    "D0_O1",
    "D2_O3",
    "D4_O5",
    "D1_O2",
]

receiver_initial_sequence = [
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
]


# ------------------------------------------------------------
# Core spectral helpers
# ------------------------------------------------------------

def parse_key(key):
    if key in ["EMPTY", "NONE", None]:
        return None, None
    if "_" not in key:
        return None, None
    d_part, o_part = key.split("_")
    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))
    return direction, octave


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    return octave * directions + direction


def expected_phase_for_key(key):
    d, o = parse_key(key)
    if d is None:
        return None
    return round(d / directions, 3)


def phase_distance(a, b):
    if a is None or b is None:
        return 1.0
    diff = abs(a - b) % 1.0
    return min(diff, 1.0 - diff)


def wave_token(key, phase=None, amp=None):
    d, o = parse_key(key)
    if d is None:
        return "NONE"
    frequency = o + 1
    if phase is None:
        phase = expected_phase_for_key(key)
    if amp is None:
        return f"COHERENCE_PHOTON[D{d}|O{o}|F{frequency}|P{round(phase, 3)}]"
    return f"COHERENCE_PHOTON[D{d}|O{o}|F{frequency}|P{round(phase, 3)}|A{round(amp, 3)}]"


def key_color(key, strength=1.0):
    if key in ["EMPTY", "NONE", None]:
        return "black"
    d, o = parse_key(key)
    if d is None:
        return "white"
    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)
    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0,
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.16 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE"]:
        d, o = parse_key(target_key)
        packet[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE"]:
        d2, o2 = parse_key(ambiguous_key)
        packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score

    amplitude_ok = best_score >= resonance_threshold
    margin_ok = margin >= ambiguity_margin
    accepted = amplitude_ok and margin_ok

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "amplitude_ok": amplitude_ok,
        "margin_ok": margin_ok,
        "accepted": accepted,
    }


def measured_phase_for_case(decoded_key, phase_offset, jitter):
    base_phase = expected_phase_for_key(decoded_key)
    if base_phase is None:
        return None
    jitter_noise = np.random.normal(0.0, jitter / 5.0)
    return round((base_phase + phase_offset + jitter_noise) % 1.0, 3)


def resync_phase(decoded_key, residual=0.025):
    base_phase = expected_phase_for_key(decoded_key)
    if base_phase is None:
        return None
    correction_noise = np.random.normal(0.0, residual / 4.0)
    return round((base_phase + correction_noise) % 1.0, 3)


# ------------------------------------------------------------
# Visualization helpers
# ------------------------------------------------------------

def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.25

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=13,
        linewidth=1.8,
        color=color,
        alpha=alpha,
    )
    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.16
    gap = 0.032

    ax.text(x0, y0 + 1.48, title, fontsize=7, ha="left", va="bottom", color="black")

    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE"] else (None, None)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = packet[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)
            edge = "lime" if is_selected else "white"
            lw = 1.4 if is_selected else 0.3

            ax.add_patch(Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw))


def draw_memory_slots(ax, x0, y0, seq, title, active_slot=None):
    cell = 0.20
    gap = 0.055

    ax.text(x0, y0 + 2.05, title, fontsize=7, ha="left", va="bottom", color="black")

    for i, key in enumerate(seq):
        x = x0
        y = y0 + (len(seq) - 1 - i) * (cell + gap)

        color = key_color(key)
        edge = "lime" if i == active_slot else "black"
        lw = 1.6 if i == active_slot else 0.45

        ax.add_patch(Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw))
        ax.text(x + 0.30, y + cell / 2, str(i), fontsize=6, ha="left", va="center", color="black")


def draw_phase_clock(ax, cx, cy, expected_phase, measured_phase, corrected_phase, phase_ok, resync):
    radius = 0.47

    ax.add_patch(Circle((cx, cy), radius, facecolor="white", edgecolor="black", linewidth=1.0))
    ax.text(cx, cy + 0.68, "phase clock", fontsize=7, ha="center", color="black")

    for p, color, label in [
        (expected_phase, "limegreen", "EXP"),
        (measured_phase, "red", "MEAS"),
        (corrected_phase, "orange" if resync else "gray", "SYNC"),
    ]:
        if p is None:
            continue

        angle = 2 * np.pi * p
        x = cx + radius * np.cos(angle)
        y = cy + radius * np.sin(angle)

        ax.plot([cx, x], [cy, y], color=color, linewidth=1.5)
        ax.add_patch(Circle((x, y), 0.055, facecolor=color, edgecolor="black", linewidth=0.4))
        ax.text(x, y - 0.16, label, fontsize=5.5, ha="center", color="black")

    status = "LOCK" if phase_ok else "UNLOCK"
    ax.text(cx, cy - 0.72, status, fontsize=7, ha="center", color="limegreen" if phase_ok else "red")


def draw_snapshot(ax, case_row, packet, receiver_memory):
    slot = int(case_row["MEMORY_SLOT"])
    source_key = case_row["SOURCE_KEY"]
    decoded_key = case_row["DECODED_KEY"]
    expected_key = case_row["EXPECTED_KEY"]
    final_key = case_row["AFTER_KEY"]

    amplitude_ok = bool(case_row["AMPLITUDE_OK"])
    phase_ok = bool(case_row["PHASE_OK"])
    jitter_ok = bool(case_row["JITTER_OK"])
    absorbed = bool(case_row["ABSORBED"])
    resync = bool(case_row["CLOCK_RESYNC"])

    expected_phase = case_row["EXPECTED_PHASE"]
    measured_phase = case_row["MEASURED_PHASE"]
    corrected_phase = case_row["CORRECTED_PHASE"]

    ax.set_facecolor("white")
    ax.set_xlim(0, 12.4)
    ax.set_ylim(0.5, 5.15)
    ax.set_aspect("equal")
    ax.axis("off")

    source_color = key_color(source_key)
    decoded_color = key_color(decoded_key)
    final_color = key_color(final_key)

    rejected = not absorbed

    ax.add_patch(Circle((0.9, 3.25), 0.40, facecolor=source_color, edgecolor="black", linewidth=1.0))
    ax.text(0.9, 2.58, f"SRC\n{source_key}", fontsize=7, ha="center", color="black")

    ax.add_patch(Circle((2.0, 3.25), 0.28, facecolor=decoded_color, edgecolor="black", linewidth=1.0))
    ax.text(2.0, 2.58, "PHOTON", fontsize=7, ha="center", color="black")

    ax.add_patch(Rectangle((2.95, 3.02), 0.44, 0.44, facecolor="limegreen", edgecolor="black", linewidth=1.0))
    ax.text(3.17, 2.58, "FIELD", fontsize=7, ha="center", color="black")

    ax.add_patch(Rectangle((4.10, 2.98), 0.52, 0.52, facecolor="orange" if phase_ok else "red", edgecolor="black", linewidth=1.0))
    ax.text(4.36, 2.58, "CLOCK", fontsize=7, ha="center", color="black")

    ax.add_patch(Circle((5.35, 3.25), 0.40, facecolor=decoded_color, edgecolor="lime" if amplitude_ok else "red", linewidth=2.0))
    ax.text(5.35, 2.45, f"DECODE\n{decoded_key}", fontsize=7, ha="center", color="black")

    ax.add_patch(Circle((6.80, 3.25), 0.40, facecolor=final_color, edgecolor="lime" if absorbed else "red", linewidth=2.0))
    ax.text(6.80, 2.45, f"MEM\nslot {slot}", fontsize=7, ha="center", color="black")

    draw_arrow(ax, (1.30, 3.25), (1.72, 3.25), active=True, rejected=False)
    draw_arrow(ax, (2.28, 3.25), (2.95, 3.25), active=True, rejected=False)
    draw_arrow(ax, (3.39, 3.25), (4.10, 3.25), active=True, rejected=rejected)
    draw_arrow(ax, (4.62, 3.25), (4.95, 3.25), active=True, rejected=rejected)
    draw_arrow(ax, (5.75, 3.25), (6.40, 3.25), active=True, rejected=rejected)

    draw_phase_clock(ax, 8.10, 3.25, expected_phase, measured_phase, corrected_phase, phase_ok, resync)
    draw_packet_grid(ax, 9.0, 2.20, packet, "42-channel phase packet", selected_key=decoded_key)
    draw_memory_slots(ax, 11.45, 1.42, receiver_memory, "clock memory", active_slot=slot)

    ax.text(
        0.45,
        0.82,
        f"case={case_row['CASE']} slot={slot} expected={expected_key} decoded={decoded_key} "
        f"phase_err={case_row['PHASE_ERROR']} jitter={case_row['JITTER']} action={case_row['ACTION']}\n"
        f"{case_row['OUTPUT_TOKEN']}",
        fontsize=6.7,
        ha="left",
        color="black",
    )

    ax.set_title(
        f"D35 snapshot t={case_row['T']} — {case_row['CASE']} / {case_row['ACTION']}",
        fontsize=10,
        color="black",
    )


# ------------------------------------------------------------
# Test cases
# ------------------------------------------------------------

cases = [
    {
        "case": "LOCKED_PHASE_WRITE",
        "slot": 0,
        "source_key": "D4_O5",
        "expected_key": "D4_O5",
        "strength": 0.92,
        "phase_offset": 0.00,
        "jitter": 0.02,
        "resync": False,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "SMALL_DRIFT_TOLERATED",
        "slot": 1,
        "source_key": "D1_O2",
        "expected_key": "D1_O2",
        "strength": 0.90,
        "phase_offset": 0.12,
        "jitter": 0.03,
        "resync": False,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "LARGE_DRIFT_REJECT",
        "slot": 2,
        "source_key": "D3_O6",
        "expected_key": "D3_O6",
        "strength": 0.91,
        "phase_offset": 0.35,
        "jitter": 0.03,
        "resync": False,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "RESYNC_CORRECTION_ACCEPT",
        "slot": 3,
        "source_key": "D5_O4",
        "expected_key": "D5_O4",
        "strength": 0.89,
        "phase_offset": 0.31,
        "jitter": 0.04,
        "resync": True,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "DECOHERENCE_JITTER_REJECT",
        "slot": 4,
        "source_key": "D0_O1",
        "expected_key": "D0_O1",
        "strength": 0.90,
        "phase_offset": 0.02,
        "jitter": 0.28,
        "resync": False,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "WEAK_CLOCK_PACKET_REJECT",
        "slot": 5,
        "source_key": "D2_O3",
        "expected_key": "D2_O3",
        "strength": 0.39,
        "phase_offset": 0.01,
        "jitter": 0.02,
        "resync": False,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "AMBIGUOUS_PHASE_HOLD",
        "slot": 6,
        "source_key": "D4_O5",
        "expected_key": "D4_O5",
        "strength": 0.86,
        "phase_offset": 0.02,
        "jitter": 0.02,
        "resync": False,
        "broad": False,
        "ambiguous_key": "D1_O2",
        "ambiguous_strength": 0.80,
    },
    {
        "case": "NEGATIVE_DRIFT_TOLERATED",
        "slot": 7,
        "source_key": "D1_O2",
        "expected_key": "D1_O2",
        "strength": 0.88,
        "phase_offset": -0.16,
        "jitter": 0.03,
        "resync": False,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
]

receiver_memory = receiver_initial_sequence.copy()

decision_rows = []
clock_rows = []
memory_rows = []
snapshots = []

accepted_by_amplitude = 0
phase_locked = 0
resync_count = 0
absorbed_count = 0
rejected_count = 0

for idx, item in enumerate(cases):
    packet = spectral_packet(
        target_key=item["source_key"],
        strength=item["strength"],
        broad=item["broad"],
        ambiguous_key=item["ambiguous_key"],
        ambiguous_strength=item["ambiguous_strength"],
    )

    decoded = decode_packet(packet)
    decoded_key = decoded["best_key"]
    expected_key = item["expected_key"]

    expected_phase = expected_phase_for_key(expected_key)
    measured_phase = measured_phase_for_case(decoded_key, item["phase_offset"], item["jitter"])
    raw_phase_error = phase_distance(measured_phase, expected_phase)

    corrected_phase = measured_phase
    clock_resync = bool(item["resync"])

    if clock_resync:
        corrected_phase = resync_phase(decoded_key, residual=0.025)
        resync_count += 1

    corrected_phase_error = phase_distance(corrected_phase, expected_phase)

    amplitude_ok = bool(decoded["accepted"])
    phase_ok = corrected_phase_error <= phase_tolerance
    jitter_ok = item["jitter"] <= jitter_tolerance
    resonance_match = decoded_key == expected_key

    absorbed = amplitude_ok and phase_ok and jitter_ok and resonance_match

    before_key = receiver_memory[item["slot"]]

    if absorbed:
        receiver_memory[item["slot"]] = decoded_key
        absorbed_count += 1
    else:
        rejected_count += 1

    after_key = receiver_memory[item["slot"]]

    if amplitude_ok:
        accepted_by_amplitude += 1

    if phase_ok and jitter_ok:
        phase_locked += 1

    if not amplitude_ok:
        action = "REJECT_WEAK_OR_AMBIGUOUS"
    elif not resonance_match:
        action = "REJECT_FILTER_MISMATCH"
    elif not jitter_ok:
        action = "REJECT_DECOHERENCE"
    elif not phase_ok:
        action = "REJECT_PHASE_UNLOCKED"
    elif clock_resync:
        action = "RESYNC_AND_WRITE"
    else:
        action = "CLOCK_ACCEPT_WRITE"

    output_token = wave_token(decoded_key, corrected_phase, decoded["best_score"]) if amplitude_ok else "NONE"

    row = {
        "STATE": STATE,
        "T": idx * 120,
        "CASE": item["case"],
        "MEMORY_SLOT": item["slot"],
        "SOURCE_KEY": item["source_key"],
        "EXPECTED_KEY": expected_key,
        "DECODED_KEY": decoded_key,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "AMPLITUDE_OK": int(amplitude_ok),
        "EXPECTED_PHASE": expected_phase,
        "MEASURED_PHASE": measured_phase,
        "CORRECTED_PHASE": corrected_phase,
        "PHASE_ERROR": round(corrected_phase_error, 3),
        "RAW_PHASE_ERROR": round(raw_phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "JITTER": round(item["jitter"], 3),
        "JITTER_OK": int(jitter_ok),
        "CLOCK_RESYNC": int(clock_resync),
        "RESONANCE_MATCH": int(resonance_match),
        "ABSORBED": int(absorbed),
        "BEFORE_KEY": before_key,
        "AFTER_KEY": after_key,
        "ACTION": action,
        "OUTPUT_TOKEN": output_token,
    }

    decision_rows.append(row)

    clock_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "EXPECTED_KEY": expected_key,
        "EXPECTED_PHASE": expected_phase,
        "MEASURED_PHASE": measured_phase,
        "RAW_PHASE_ERROR": round(raw_phase_error, 3),
        "CLOCK_RESYNC": int(clock_resync),
        "CORRECTED_PHASE": corrected_phase,
        "CORRECTED_PHASE_ERROR": round(corrected_phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "JITTER": round(item["jitter"], 3),
        "JITTER_OK": int(jitter_ok),
    })

    memory_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "MEMORY_SLOT": item["slot"],
        "BEFORE_KEY": before_key,
        "EXPECTED_KEY": expected_key,
        "DECODED_KEY": decoded_key,
        "AFTER_KEY": after_key,
        "ABSORBED": int(absorbed),
        "HELD_IF_REJECTED": int((not absorbed) and after_key == before_key),
        "MATCH_IF_ABSORBED": int(after_key == expected_key) if absorbed else "not_written",
    })

    snapshots.append((row, packet, receiver_memory.copy()))


df_decision = pd.DataFrame(decision_rows)
df_clock = pd.DataFrame(clock_rows)
df_memory = pd.DataFrame(memory_rows)

expected_final = receiver_initial_sequence.copy()
for item in cases:
    packet = spectral_packet(
        target_key=item["source_key"],
        strength=item["strength"],
        broad=item["broad"],
        ambiguous_key=item["ambiguous_key"],
        ambiguous_strength=item["ambiguous_strength"],
    )
    decoded = decode_packet(packet)

    # deterministic expected behavior based on case design
    if item["case"] in [
        "LOCKED_PHASE_WRITE",
        "SMALL_DRIFT_TOLERATED",
        "RESYNC_CORRECTION_ACCEPT",
        "NEGATIVE_DRIFT_TOLERATED",
    ]:
        expected_final[item["slot"]] = item["expected_key"]

coherence_accuracy = float(np.mean(np.array(receiver_memory) == np.array(expected_final)))

clock_safety_accuracy = float(np.mean([
    1 if row["ABSORBED"] == 1 and row["AFTER_KEY"] == row["EXPECTED_KEY"] else
    1 if row["ABSORBED"] == 0 and row["AFTER_KEY"] == row["BEFORE_KEY"] else
    0
    for row in memory_rows
]))

result = (
    "SPECTRAL_COHERENCE_CLOCK_LOCKED"
    if coherence_accuracy == 1.0 and clock_safety_accuracy == 1.0
    else "SPECTRAL_COHERENCE_CLOCK_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": memory_slots,
    "RECEIVER_TARGET_SEQUENCE": " ".join(receiver_target_sequence),
    "RECEIVER_FINAL_SEQUENCE": " ".join(receiver_memory),
    "EXPECTED_FINAL_SEQUENCE": " ".join(expected_final),
    "CASES_TESTED": len(cases),
    "AMPLITUDE_ACCEPTED_PACKETS": accepted_by_amplitude,
    "PHASE_LOCKED_PACKETS": phase_locked,
    "CLOCK_RESYNC_EVENTS": resync_count,
    "ABSORBED_PACKETS": absorbed_count,
    "REJECTED_PACKETS": rejected_count,
    "COHERENCE_ACCURACY": round(coherence_accuracy, 3),
    "CLOCK_SAFETY_ACCURACY": round(clock_safety_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "PHASE_TOLERANCE": phase_tolerance,
    "JITTER_TOLERANCE": jitter_tolerance,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8.8, 3.35 * len(snapshots)))

if len(snapshots) == 1:
    axes = [axes]

for ax, (row, packet, memory_state) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, memory_state)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"memory slots:                 {memory_slots}")
print(f"receiver target sequence:     {' '.join(receiver_target_sequence)}")
print(f"receiver final sequence:      {' '.join(receiver_memory)}")
print(f"expected final sequence:      {' '.join(expected_final)}")
print(f"cases tested:                 {len(cases)}")
print(f"amplitude accepted packets:   {accepted_by_amplitude}")
print(f"phase locked packets:         {phase_locked}")
print(f"clock resync events:          {resync_count}")
print(f"absorbed packets:             {absorbed_count}")
print(f"rejected packets:             {rejected_count}")
print(f"coherence accuracy:           {round(coherence_accuracy, 3)}")
print(f"clock safety accuracy:        {round(clock_safety_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"phase tolerance:              {phase_tolerance}")
print(f"jitter tolerance:             {jitter_tolerance}")

print("\nCoherence decision table:")
display(df_decision)

print("\nClock correction table:")
display(df_clock)

print("\nMemory coherence table:")
display(df_memory)

print("\nSummary:")
display(df_summary)
```
