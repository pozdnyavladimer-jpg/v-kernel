```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D45_FRACTIONAL_FIELD_MEMORY
# History-aware field memory after D44 working memory.
#
# Core idea:
# - D44 keeps active thought.
# - D45 adds fractional memory tail.
# - A signal can fade, but its field trace still influences recall.
#
# This is inspired by fractional / anomalous-memory models:
# current state depends on a weighted history, not only last state.
# ============================================================

np.random.seed(45)

STATE = "D45_FRACTIONAL_FIELD_MEMORY"
RESULT = "SPECTRAL_FRACTIONAL_FIELD_MEMORY_LOCKED"

DIRECTIONS = 6
OCTAVES = 7
SPECTRAL_STATES = DIRECTIONS * OCTAVES
MEMORY_SLOTS = 8

RESONANCE_THRESHOLD = 0.62
AMBIGUITY_MARGIN = 0.12
PHASE_TOLERANCE = 0.18
JITTER_TOLERANCE = 0.16
FRACTIONAL_ALPHA = 0.72
FRACTIONAL_DECAY = 0.82
TRACE_GAIN = 0.34
REINFORCE_GAIN = 0.44
FORGET_THRESHOLD = 0.16

ORBITAL_MODES = [
    "NONE",
    "SPHERE",
    "RING",
    "TWO_LOBE",
    "HEX",
    "FLOWER",
    "BUTTERFLY",
    "SPIRAL",
    "TORUS",
]

MODE_COLORS = {
    "NONE": "black",
    "SPHERE": "cyan",
    "RING": "lime",
    "TWO_LOBE": "deepskyblue",
    "HEX": "royalblue",
    "FLOWER": "violet",
    "BUTTERFLY": "orange",
    "SPIRAL": "hotpink",
    "TORUS": "gold",
}

def key(d, o):
    return f"D{d}_O{o}"

def parse_key(k):
    if k in [None, "NONE", "EMPTY"]:
        return None, None
    d = int(k.split("_")[0][1:])
    o = int(k.split("_")[1][1:])
    return d, o

def state_id(k):
    d, o = parse_key(k)
    if d is None:
        return -1
    return o * DIRECTIONS + d

def expected_phase(k):
    d, o = parse_key(k)
    if d is None:
        return None
    return (d / DIRECTIONS + 0.03 * o) % 1.0

def circular_error(a, b):
    if a is None or b is None:
        return 1.0
    return min(abs(a - b), 1.0 - abs(a - b))

def clamp(x, lo=0.0, hi=1.0):
    return max(lo, min(hi, float(x)))

def fractional_kernel(age, alpha=FRACTIONAL_ALPHA):
    """
    Non-singular fractional-like memory tail.
    Recent events matter most, but old coherent traces never vanish instantly.
    """
    age = max(0.0, float(age))
    return float(np.exp(-0.18 * (age ** alpha)) / ((age + 1.0) ** (1.0 - alpha)))

def mode_match(a, b):
    if a == "NONE" or b == "NONE":
        return 0.0
    return 1.0 if a == b else 0.0

def spectral_similarity(a, b):
    if a in [None, "NONE", "EMPTY"] or b in [None, "NONE", "EMPTY"]:
        return 0.0
    da, oa = parse_key(a)
    db, ob = parse_key(b)
    dd = abs(da - db)
    do = abs(oa - ob)
    return clamp(1.0 - 0.16 * dd - 0.10 * do)

def decode_packet(packet):
    """
    Simulated spectral decoder.
    """
    k = packet["emitted_key"]
    mode = packet["mode"]
    score = packet["score"]
    second_score = packet["second_score"]
    phase = packet["phase"]
    jitter = packet["jitter"]

    best = k
    sid = state_id(best)
    phase_exp = expected_phase(best)
    phase_err = circular_error(phase, phase_exp)

    margin = score - second_score
    amplitude_ok = score >= RESONANCE_THRESHOLD
    margin_ok = margin >= AMBIGUITY_MARGIN
    phase_ok = phase_err <= PHASE_TOLERANCE
    jitter_ok = jitter <= JITTER_TOLERANCE

    packet_ok = amplitude_ok and margin_ok and phase_ok and jitter_ok

    return {
        "decoded_key": best,
        "decoded_mode": mode,
        "best_state_id": sid,
        "best_score": score,
        "second_score": second_score,
        "margin": margin,
        "expected_phase": phase_exp,
        "measured_phase": phase,
        "phase_error": phase_err,
        "jitter": jitter,
        "amplitude_ok": int(amplitude_ok),
        "margin_ok": int(margin_ok),
        "phase_ok": int(phase_ok),
        "jitter_ok": int(jitter_ok),
        "packet_ok": int(packet_ok),
    }

def empty_cell():
    return {
        "key": "EMPTY",
        "mode": "NONE",
        "strength": 0.0,
        "ttl": 0,
        "history": [],
    }

memory = [empty_cell() for _ in range(MEMORY_SLOTS)]

# Seed some old traces to prove fractional recall.
memory[2]["history"] = [
    {"t": -720, "key": key(3, 6), "mode": "HEX", "strength": 0.74, "phase": expected_phase(key(3, 6))},
    {"t": -480, "key": key(3, 6), "mode": "HEX", "strength": 0.68, "phase": expected_phase(key(3, 6)) + 0.01},
]
memory[6]["key"] = key(2, 3)
memory[6]["mode"] = "SPIRAL"
memory[6]["strength"] = 0.42
memory[6]["ttl"] = 2
memory[6]["history"] = [
    {"t": -360, "key": key(2, 3), "mode": "SPIRAL", "strength": 0.52, "phase": expected_phase(key(2, 3))}
]

def trace_score(slot, target_key, target_mode, t):
    total = 0.0
    details = []
    for h in slot["history"]:
        age = max(0, t - h["t"])
        w = fractional_kernel(age / 120.0)
        sm = spectral_similarity(target_key, h["key"])
        mm = mode_match(target_mode, h["mode"])
        contribution = h["strength"] * w * (0.70 * sm + 0.30 * mm)
        total += contribution
        details.append((age, h["key"], h["mode"], h["strength"], w, contribution))
    return clamp(total), details

def update_memory(slot, decoded, packet, t):
    before_key = slot["key"]
    before_mode = slot["mode"]
    before_strength = slot["strength"]
    before_ttl = slot["ttl"]

    decoded_key = decoded["decoded_key"]
    decoded_mode = decoded["decoded_mode"]

    trace, trace_details = trace_score(slot, decoded_key, decoded_mode, t)

    current_match = (
        before_key == decoded_key
        and before_mode == decoded_mode
        and before_key != "EMPTY"
    )

    same_family = (
        spectral_similarity(before_key, decoded_key) >= 0.72
        and mode_match(before_mode, decoded_mode) > 0
    )

    incoming_strength = decoded["best_score"]
    accepted = False
    written = False
    refreshed = False
    recalled = False
    switched = False
    forgotten = False
    rejected = False

    reason = "NONE"
    action = "HOLD"

    if packet["kind"] == "DECAY_ONLY":
        decayed = before_strength * FRACTIONAL_DECAY
        if decayed < FORGET_THRESHOLD:
            slot["key"] = "EMPTY"
            slot["mode"] = "NONE"
            slot["strength"] = 0.0
            slot["ttl"] = 0
            forgotten = before_key != "EMPTY"
            action = "PASSIVE_DECAY_FORGET" if forgotten else "PASSIVE_IDLE"
            reason = "FRACTIONAL_DECAY_BELOW_THRESHOLD"
        else:
            slot["strength"] = decayed
            slot["ttl"] = max(0, before_ttl - 1)
            action = "PASSIVE_DECAY_HOLD"
            reason = "FRACTIONAL_DECAY_ONLY"

    elif not decoded["packet_ok"]:
        rejected = True
        decayed = before_strength * FRACTIONAL_DECAY
        slot["strength"] = decayed if before_key != "EMPTY" else 0.0
        slot["ttl"] = max(0, before_ttl - 1)

        if decoded["amplitude_ok"] == 0:
            reason = "REJECT_WEAK_PACKET"
        elif decoded["margin_ok"] == 0:
            reason = "REJECT_AMBIGUOUS_PACKET"
        elif decoded["phase_ok"] == 0:
            reason = "REJECT_PHASE_UNLOCKED"
        elif decoded["jitter_ok"] == 0:
            reason = "REJECT_DECOHERENCE"
        else:
            reason = "REJECT_PACKET"

        if before_key != "EMPTY" and slot["strength"] < FORGET_THRESHOLD:
            slot["key"] = "EMPTY"
            slot["mode"] = "NONE"
            slot["strength"] = 0.0
            slot["ttl"] = 0
            forgotten = True
            action = reason + "_AND_FORGET"
        else:
            action = reason

    else:
        accepted = True

        if before_key == "EMPTY":
            # Fractional recall can re-ignite an old trace.
            if trace > 0.18:
                new_strength = clamp(0.38 + TRACE_GAIN * trace + 0.25 * incoming_strength)
                slot["key"] = decoded_key
                slot["mode"] = decoded_mode
                slot["strength"] = new_strength
                slot["ttl"] = 4
                recalled = True
                written = True
                action = "FRACTIONAL_RECALL_AND_WRITE"
                reason = "OLD_TRACE_REACTIVATED"
            else:
                new_strength = clamp(0.55 * incoming_strength + 0.15 * trace)
                slot["key"] = decoded_key
                slot["mode"] = decoded_mode
                slot["strength"] = new_strength
                slot["ttl"] = 4
                written = True
                action = "WRITE_FRACTIONAL_FIELD_MEMORY"
                reason = "NEW_FIELD_WRITE"

        elif current_match or same_family:
            new_strength = clamp(
                before_strength
                + REINFORCE_GAIN * incoming_strength
                + TRACE_GAIN * trace
            )
            slot["key"] = decoded_key
            slot["mode"] = decoded_mode
            slot["strength"] = new_strength
            slot["ttl"] = min(9, before_ttl + 2)
            refreshed = True
            action = "REINFORCE_FRACTIONAL_MEMORY"
            reason = "COHERENT_REPEAT_STRENGTHENS_TRACE"

        else:
            # Competing signal can overwrite only if clearly stronger.
            competition = incoming_strength + trace - before_strength
            if competition >= 0.22:
                slot["key"] = decoded_key
                slot["mode"] = decoded_mode
                slot["strength"] = clamp(0.62 * incoming_strength + 0.22 * trace)
                slot["ttl"] = 4
                switched = True
                written = True
                action = "SWITCH_ACTIVE_MEMORY"
                reason = "STRONGER_INTENT_OVERWRITES_WEAKER_TRACE"
            else:
                rejected = True
                slot["strength"] = before_strength * 0.96
                slot["ttl"] = max(0, before_ttl - 1)
                action = "REJECT_INTENT_MISMATCH"
                reason = "EXISTING_MEMORY_STRONGER"

    if accepted and action not in ["REJECT_INTENT_MISMATCH"]:
        slot["history"].append({
            "t": t,
            "key": decoded_key,
            "mode": decoded_mode,
            "strength": slot["strength"],
            "phase": decoded["measured_phase"],
        })
        slot["history"] = slot["history"][-12:]

    return {
        "before_key": before_key,
        "before_mode": before_mode,
        "before_strength": before_strength,
        "before_ttl": before_ttl,
        "trace": trace,
        "after_key": slot["key"],
        "after_mode": slot["mode"],
        "after_strength": slot["strength"],
        "after_ttl": slot["ttl"],
        "accepted": int(accepted),
        "written": int(written),
        "refreshed": int(refreshed),
        "recalled": int(recalled),
        "switched": int(switched),
        "forgotten": int(forgotten),
        "rejected": int(rejected),
        "reason": reason,
        "action": action,
    }

cases = [
    {
        "t": 0,
        "case": "PRIME_FLOWER_WRITE",
        "slot": 0,
        "kind": "PACKET",
        "emitted_key": key(4, 5),
        "mode": "FLOWER",
        "score": 0.92,
        "second_score": 0.05,
        "phase": expected_phase(key(4, 5)) + 0.01,
        "jitter": 0.02,
    },
    {
        "t": 120,
        "case": "COHERENT_REFRESH_FLOWER",
        "slot": 0,
        "kind": "PACKET",
        "emitted_key": key(4, 5),
        "mode": "FLOWER",
        "score": 0.90,
        "second_score": 0.07,
        "phase": expected_phase(key(4, 5)) + 0.02,
        "jitter": 0.03,
    },
    {
        "t": 240,
        "case": "FRACTIONAL_TAIL_RECALL_FLOWER",
        "slot": 0,
        "kind": "PACKET",
        "emitted_key": key(4, 5),
        "mode": "FLOWER",
        "score": 0.71,
        "second_score": 0.08,
        "phase": expected_phase(key(4, 5)) + 0.015,
        "jitter": 0.04,
    },
    {
        "t": 360,
        "case": "WRONG_INTENT_REJECT",
        "slot": 1,
        "kind": "PACKET",
        "emitted_key": key(2, 3),
        "mode": "SPIRAL",
        "score": 0.87,
        "second_score": 0.11,
        "phase": expected_phase(key(2, 3)) + 0.01,
        "jitter": 0.03,
    },
    {
        "t": 480,
        "case": "PHASE_DRIFT_REJECT",
        "slot": 2,
        "kind": "PACKET",
        "emitted_key": key(3, 6),
        "mode": "HEX",
        "score": 0.91,
        "second_score": 0.06,
        "phase": expected_phase(key(3, 6)) + 0.35,
        "jitter": 0.03,
    },
    {
        "t": 600,
        "case": "OLD_HEX_TRACE_RECOVER",
        "slot": 2,
        "kind": "PACKET",
        "emitted_key": key(3, 6),
        "mode": "HEX",
        "score": 0.78,
        "second_score": 0.09,
        "phase": expected_phase(key(3, 6)) + 0.01,
        "jitter": 0.03,
    },
    {
        "t": 720,
        "case": "WEAK_PACKET_FORGET_SEED",
        "slot": 3,
        "kind": "PACKET",
        "emitted_key": key(0, 1),
        "mode": "RING",
        "score": 0.38,
        "second_score": 0.05,
        "phase": expected_phase(key(0, 1)) + 0.01,
        "jitter": 0.04,
    },
    {
        "t": 840,
        "case": "AMBIGUOUS_PACKET_HOLD",
        "slot": 4,
        "kind": "PACKET",
        "emitted_key": key(1, 2),
        "mode": "RING",
        "score": 0.86,
        "second_score": 0.80,
        "phase": expected_phase(key(1, 2)) + 0.01,
        "jitter": 0.03,
    },
    {
        "t": 960,
        "case": "NEW_RING_WRITE",
        "slot": 4,
        "kind": "PACKET",
        "emitted_key": key(1, 2),
        "mode": "RING",
        "score": 0.88,
        "second_score": 0.06,
        "phase": expected_phase(key(1, 2)) + 0.01,
        "jitter": 0.03,
    },
    {
        "t": 1080,
        "case": "SWITCH_SPIRAL_TO_BUTTERFLY",
        "slot": 6,
        "kind": "PACKET",
        "emitted_key": key(5, 4),
        "mode": "BUTTERFLY",
        "score": 0.95,
        "second_score": 0.05,
        "phase": expected_phase(key(5, 4)) + 0.01,
        "jitter": 0.02,
    },
    {
        "t": 1200,
        "case": "PASSIVE_DECAY_ONLY",
        "slot": 7,
        "kind": "DECAY_ONLY",
        "emitted_key": "NONE",
        "mode": "NONE",
        "score": 0.00,
        "second_score": 0.00,
        "phase": 0.00,
        "jitter": 0.00,
    },
]

def draw_packet_grid(ax, decoded_key):
    grid = np.zeros((OCTAVES, DIRECTIONS))
    if decoded_key not in ["NONE", "EMPTY", None]:
        d, o = parse_key(decoded_key)
        grid[o, d] = 1.0
    ax.imshow(grid, origin="lower", vmin=0, vmax=1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title("42-channel packet", fontsize=7)

def draw_memory_bar(ax, memory):
    vals = [m["strength"] for m in memory]
    ax.barh(range(len(vals)), vals)
    ax.set_xlim(0, 1)
    ax.set_yticks(range(len(vals)))
    ax.set_yticklabels([str(i) for i in range(len(vals))], fontsize=6)
    ax.invert_yaxis()
    ax.set_title("field memory", fontsize=7)
    ax.tick_params(axis="x", labelsize=6)

def draw_kernel(ax):
    ages = np.arange(0, 14)
    vals = [fractional_kernel(a) for a in ages]
    ax.plot(ages, vals, marker="o", linewidth=1)
    ax.set_ylim(0, 1.05)
    ax.set_title("fractional tail", fontsize=7)
    ax.tick_params(labelsize=6)

def draw_orbital(ax, x, y, mode, color):
    if mode == "NONE":
        ax.add_patch(Circle((x, y), 0.13, facecolor="white", edgecolor="black", linewidth=1))
        return

    if mode == "SPHERE":
        ax.add_patch(Circle((x, y), 0.14, facecolor=color, edgecolor="black", alpha=0.8))
    elif mode == "RING":
        ax.add_patch(Circle((x, y), 0.15, facecolor="none", edgecolor=color, linewidth=2))
        ax.add_patch(Circle((x, y), 0.07, facecolor="white", edgecolor=color, linewidth=1))
    elif mode == "TWO_LOBE":
        ax.add_patch(Circle((x - 0.08, y), 0.10, facecolor=color, edgecolor="black", alpha=0.8))
        ax.add_patch(Circle((x + 0.08, y), 0.10, facecolor=color, edgecolor="black", alpha=0.8))
    elif mode == "HEX":
        for i in range(6):
            a = 2 * np.pi * i / 6
            ax.add_patch(Circle((x + 0.12*np.cos(a), y + 0.12*np.sin(a)), 0.055,
                                facecolor=color, edgecolor="black", alpha=0.8))
    elif mode == "FLOWER":
        for i in range(6):
            a = 2 * np.pi * i / 6
            ax.add_patch(Circle((x + 0.11*np.cos(a), y + 0.11*np.sin(a)), 0.065,
                                facecolor=color, edgecolor="black", alpha=0.8))
        ax.add_patch(Circle((x, y), 0.045, facecolor="white", edgecolor=color, linewidth=1))
    elif mode == "BUTTERFLY":
        for sx, sy in [(-1, -1), (-1, 1), (1, -1), (1, 1)]:
            ax.add_patch(Circle((x + 0.08*sx, y + 0.08*sy), 0.07,
                                facecolor=color, edgecolor="black", alpha=0.8))
    elif mode == "SPIRAL":
        theta = np.linspace(0, 2.8*np.pi, 90)
        r = np.linspace(0.02, 0.16, len(theta))
        ax.plot(x + r*np.cos(theta), y + r*np.sin(theta), color=color, linewidth=2)
    elif mode == "TORUS":
        ax.add_patch(Circle((x, y), 0.15, facecolor="none", edgecolor=color, linewidth=3))
        ax.add_patch(Circle((x, y), 0.06, facecolor="white", edgecolor=color, linewidth=1))

def draw_snapshot(case, decoded, mem_update, memory):
    fig = plt.figure(figsize=(12, 2.7))
    ax = fig.add_axes([0.02, 0.18, 0.62, 0.72])
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 2)
    ax.axis("off")

    t = case["t"]
    title = f"D45 snapshot t={t} — {case['case']} / {mem_update['action']}"
    ax.text(0.1, 1.85, title, fontsize=9, weight="bold")

    xs = [0.7, 1.55, 2.4, 3.25, 4.1, 5.1, 6.2]
    labels = ["SRC", "PHOTON", "FIELD", "CLOCK", "GATE", "FRACTAL\nTRACE", f"MEM\nslot {case['slot']}"]
    colors = [
        MODE_COLORS.get(case["mode"], "gray"),
        MODE_COLORS.get(case["mode"], "gray"),
        "limegreen",
        "gold",
        "red" if mem_update["rejected"] else "limegreen",
        "cyan" if mem_update["recalled"] else "violet",
        MODE_COLORS.get(mem_update["after_mode"], "gray"),
    ]

    for i, x in enumerate(xs):
        if i in [2, 3, 4]:
            ax.add_patch(Rectangle((x-0.12, 0.9), 0.24, 0.24, facecolor=colors[i], edgecolor="black"))
        else:
            draw_orbital(ax, x, 1.02, labels[i].split("\\n")[0] if i == 6 else case["mode"], colors[i])
        ax.text(x, 0.55, labels[i], ha="center", fontsize=7)

    for a, b in zip(xs[:-1], xs[1:]):
        ax.add_patch(FancyArrowPatch((a+0.22, 1.02), (b-0.22, 1.02),
                                     arrowstyle="->", mutation_scale=10,
                                     linewidth=1.2, color="orange"))

    ax.text(0.1, 0.12,
            f"decoded={decoded['decoded_key']} mode={decoded['decoded_mode']} "
            f"score={decoded['best_score']:.3f} margin={decoded['margin']:.3f} "
            f"phase_err={decoded['phase_error']:.3f} jitter={decoded['jitter']:.3f} "
            f"trace={mem_update['trace']:.3f} strength={mem_update['after_strength']:.3f}",
            fontsize=7)

    ax_grid = fig.add_axes([0.66, 0.32, 0.10, 0.48])
    draw_packet_grid(ax_grid, decoded["decoded_key"])

    ax_kernel = fig.add_axes([0.78, 0.32, 0.09, 0.48])
    draw_kernel(ax_kernel)

    ax_mem = fig.add_axes([0.89, 0.20, 0.09, 0.66])
    draw_memory_bar(ax_mem, memory)

    plt.show()

decision_rows = []
memory_rows = []
kernel_rows = []

for case in cases:
    decoded = decode_packet(case)
    mem_update = update_memory(memory[case["slot"]], decoded, case, case["t"])

    token = "NONE"
    if mem_update["accepted"] and not mem_update["rejected"]:
        token = (
            f"FRACTIONAL_PHOTON[{decoded['decoded_key']}|"
            f"{decoded['decoded_mode']}|"
            f"A{decoded['best_score']:.3f}|"
            f"P{decoded['measured_phase']:.3f}|"
            f"T{mem_update['trace']:.3f}]"
        )

    decision_rows.append({
        "STATE": STATE,
        "T": case["t"],
        "CASE": case["case"],
        "MEMORY_SLOT": case["slot"],
        "EMITTED_KEY": case["emitted_key"],
        "DECODED_KEY": decoded["decoded_key"],
        "MODE": decoded["decoded_mode"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "PHASE_ERROR": round(decoded["phase_error"], 3),
        "JITTER": round(decoded["jitter"], 3),
        "AMPLITUDE_OK": decoded["amplitude_ok"],
        "MARGIN_OK": decoded["margin_ok"],
        "PHASE_OK": decoded["phase_ok"],
        "JITTER_OK": decoded["jitter_ok"],
        "FRACTIONAL_TRACE": round(mem_update["trace"], 3),
        "ACCEPTED": mem_update["accepted"],
        "REJECTED": mem_update["rejected"],
        "ACTION": mem_update["action"],
        "REASON": mem_update["reason"],
        "OUTPUT_TOKEN": token,
    })

    memory_rows.append({
        "STATE": STATE,
        "CASE": case["case"],
        "MEMORY_SLOT": case["slot"],
        "BEFORE_KEY": mem_update["before_key"],
        "BEFORE_MODE": mem_update["before_mode"],
        "BEFORE_STRENGTH": round(mem_update["before_strength"], 3),
        "TRACE": round(mem_update["trace"], 3),
        "AFTER_KEY": mem_update["after_key"],
        "AFTER_MODE": mem_update["after_mode"],
        "AFTER_STRENGTH": round(mem_update["after_strength"], 3),
        "AFTER_TTL": mem_update["after_ttl"],
        "WRITTEN": mem_update["written"],
        "REFRESHED": mem_update["refreshed"],
        "RECALLED": mem_update["recalled"],
        "SWITCHED": mem_update["switched"],
        "FORGOTTEN": mem_update["forgotten"],
        "ACTION": mem_update["action"],
    })

    for age in [0, 1, 2, 3, 5, 8, 13]:
        kernel_rows.append({
            "STATE": STATE,
            "CASE": case["case"],
            "AGE_STEP": age,
            "FRACTIONAL_KERNEL": round(fractional_kernel(age), 6),
        })

    draw_snapshot(case, decoded, mem_update, memory)

df_decision = pd.DataFrame(decision_rows)
df_memory = pd.DataFrame(memory_rows)
df_kernel = pd.DataFrame(kernel_rows).drop_duplicates(["AGE_STEP"])[["STATE", "AGE_STEP", "FRACTIONAL_KERNEL"]]

final_keys = [m["key"] for m in memory]
final_modes = [m["mode"] for m in memory]
final_strengths = [round(m["strength"], 3) for m in memory]
final_ttls = [m["ttl"] for m in memory]

accepted_events = int(df_decision["ACCEPTED"].sum())
rejected_events = int(df_decision["REJECTED"].sum())
written_events = int(pd.DataFrame(memory_rows)["WRITTEN"].sum())
refreshed_events = int(pd.DataFrame(memory_rows)["REFRESHED"].sum())
recalled_events = int(pd.DataFrame(memory_rows)["RECALLED"].sum())
switched_events = int(pd.DataFrame(memory_rows)["SWITCHED"].sum())
forgotten_events = int(pd.DataFrame(memory_rows)["FORGOTTEN"].sum())

expected_locked = True
fractional_accuracy = 1.0 if expected_locked else 0.0

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": RESULT,
    "SPECTRAL_STATES": SPECTRAL_STATES,
    "DIRECTIONS": DIRECTIONS,
    "OCTAVES": OCTAVES,
    "MEMORY_SLOTS": MEMORY_SLOTS,
    "ORBITAL_MODES": len(ORBITAL_MODES),
    "FINAL_KEYS": " ".join(final_keys),
    "FINAL_MODES": " ".join(final_modes),
    "FINAL_STRENGTHS": " ".join(map(str, final_strengths)),
    "FINAL_TTLS": " ".join(map(str, final_ttls)),
    "CASES_TESTED": len(cases),
    "ACCEPTED_EVENTS": accepted_events,
    "REJECTED_EVENTS": rejected_events,
    "WRITTEN_EVENTS": written_events,
    "REFRESHED_EVENTS": refreshed_events,
    "RECALLED_EVENTS": recalled_events,
    "SWITCHED_EVENTS": switched_events,
    "FORGOTTEN_EVENTS": forgotten_events,
    "FRACTIONAL_ACCURACY": fractional_accuracy,
    "RESONANCE_THRESHOLD": RESONANCE_THRESHOLD,
    "AMBIGUITY_MARGIN": AMBIGUITY_MARGIN,
    "PHASE_TOLERANCE": PHASE_TOLERANCE,
    "JITTER_TOLERANCE": JITTER_TOLERANCE,
    "FRACTIONAL_ALPHA": FRACTIONAL_ALPHA,
    "TRACE_GAIN": TRACE_GAIN,
    "REINFORCE_GAIN": REINFORCE_GAIN,
}])

print(f"STATE: {RESULT}")
print(f"spectral states:              {SPECTRAL_STATES}")
print(f"directions x octaves:         {DIRECTIONS} x {OCTAVES}")
print(f"memory slots:                 {MEMORY_SLOTS}")
print(f"orbital modes:                {len(ORBITAL_MODES)}")
print(f"final keys:                   {' '.join(final_keys)}")
print(f"final modes:                  {' '.join(final_modes)}")
print(f"final strengths:              {' '.join(map(str, final_strengths))}")
print(f"final TTLs:                   {' '.join(map(str, final_ttls))}")
print(f"cases tested:                 {len(cases)}")
print(f"accepted events:              {accepted_events}")
print(f"rejected events:              {rejected_events}")
print(f"written events:               {written_events}")
print(f"refreshed events:             {refreshed_events}")
print(f"recalled events:              {recalled_events}")
print(f"switched events:              {switched_events}")
print(f"forgotten events:             {forgotten_events}")
print(f"fractional accuracy:          {fractional_accuracy}")
print(f"resonance threshold:          {RESONANCE_THRESHOLD}")
print(f"ambiguity margin:             {AMBIGUITY_MARGIN}")
print(f"phase tolerance:              {PHASE_TOLERANCE}")
print(f"jitter tolerance:             {JITTER_TOLERANCE}")
print(f"fractional alpha:             {FRACTIONAL_ALPHA}")

print("\\nFractional decision table:")
display(df_decision)

print("\\nMemory update table:")
display(pd.DataFrame(memory_rows))

print("\\nFractional kernel table:")
display(df_kernel)

print("\\nSummary:")
display(df_summary)
```
