```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D32_SPECTRAL_CASCADE_CHAIN
# Cascading spectral transitions across memory cells
#
# Goal:
# initial spectral memory state
# -> one source emits a transition photon
# -> receiver updates only if:
#    - packet is strong enough
#    - packet is not ambiguous
#    - transition is allowed by spectral rules
# -> updated cell can participate in next transition
# -> cascade propagates through memory chain
# -> weak / ambiguous / forbidden links are rejected safely
# ============================================================

np.random.seed(32)

STATE = "D32_SPECTRAL_CASCADE_CHAIN"

directions = 6
octaves = 7
spectral_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12

max_direction_step = 1
allowed_octave_jumps = [1, 2]

memory_slots = 8

initial_sequence = [
    "D4_O3",
    "D4_O5",
    "D3_O4",
    "D1_O2",
    "D1_O2",
    "D0_O1",
    "D2_O3",
    "D5_O4",
]


# ------------------------------------------------------------
# Core spectral helpers
# ------------------------------------------------------------

def parse_key(key):
    if key in ["EMPTY", "NONE", None]:
        return None, None

    if key == "WHITE":
        return None, None

    d_part, o_part = key.split("_")
    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))
    return direction, octave


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    return octave * directions + direction


def key_to_state_id(key):
    d, o = parse_key(key)
    if d is None:
        return None
    return state_id(d, o)


def key_color(key, strength=1.0):
    if key in ["EMPTY", "NONE", None]:
        return "black"

    if key == "WHITE":
        return "white"

    d, o = parse_key(key)
    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)
    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.16 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(target_key)
        packet[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d2, o2 = parse_key(ambiguous_key)
        packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score

    score_ok = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "score_ok": score_ok,
    }


# ------------------------------------------------------------
# Transition rules
# ------------------------------------------------------------

def transition_rule(current_key, next_key):
    d0, o0 = parse_key(current_key)
    d1, o1 = parse_key(next_key)

    if d0 is None or d1 is None:
        return False, "INVALID_EMPTY_STATE", 0, 0, "NONE"

    delta_d = d1 - d0
    delta_o = o1 - o0

    if delta_o == 0:
        return False, "FORBIDDEN_ZERO_ENERGY", delta_d, delta_o, "ZERO_ENERGY"

    if abs(delta_d) > max_direction_step:
        return False, "FORBIDDEN_DIRECTION_JUMP", delta_d, delta_o, "DIRECTION_JUMP"

    if delta_o > 0:
        if delta_o in allowed_octave_jumps:
            return True, "ALLOWED_TRANSITION", delta_d, delta_o, "ABSORB_UP"
        return False, "FORBIDDEN_OCTAVE_JUMP", delta_d, delta_o, "OCTAVE_JUMP"

    if delta_o < 0:
        if abs(delta_o) in allowed_octave_jumps:
            return True, "ALLOWED_TRANSITION", delta_d, delta_o, "RELAX_DOWN"
        return False, "FORBIDDEN_RELAX_JUMP", delta_d, delta_o, "RELAX_JUMP"

    return False, "FORBIDDEN_UNKNOWN", delta_d, delta_o, "UNKNOWN"


def transition_photon_token(current_key, next_key):
    d0, o0 = parse_key(current_key)
    d1, o1 = parse_key(next_key)

    if d0 is None or d1 is None:
        return "NONE"

    delta_d = d1 - d0
    delta_o = o1 - o0
    frequency = abs(delta_o) + 1
    phase = round(d1 / directions, 3)

    sign_d = "+" if delta_d >= 0 else ""
    sign_o = "+" if delta_o >= 0 else ""

    return f"CASCADE_PHOTON[ΔD{sign_d}{delta_d}|ΔO{sign_o}{delta_o}|F{frequency}|P{phase}]"


# ------------------------------------------------------------
# Cascade plan
# ------------------------------------------------------------

cascade_events = [
    {
        "case": "CHAIN_STEP_0_TO_1",
        "source_slot": 0,
        "receiver_slot": 1,
        "target_key": "D5_O6",
        "strength": 0.92,
        "expected_applied": True,
    },
    {
        "case": "CHAIN_STEP_1_TO_2",
        "source_slot": 1,
        "receiver_slot": 2,
        "target_key": "D4_O5",
        "strength": 0.90,
        "expected_applied": True,
    },
    {
        "case": "CHAIN_RELAX_2_TO_3",
        "source_slot": 2,
        "receiver_slot": 3,
        "target_key": "D0_O1",
        "strength": 0.88,
        "expected_applied": True,
    },
    {
        "case": "WEAK_LINK_3_TO_4",
        "source_slot": 3,
        "receiver_slot": 4,
        "target_key": "D2_O3",
        "strength": 0.39,
        "expected_applied": False,
    },
    {
        "case": "FORBIDDEN_DIRECTION_4_TO_5",
        "source_slot": 4,
        "receiver_slot": 5,
        "target_key": "D3_O2",
        "strength": 0.89,
        "expected_applied": False,
    },
    {
        "case": "AMBIGUOUS_BRANCH_5_TO_6",
        "source_slot": 5,
        "receiver_slot": 6,
        "target_key": "D3_O4",
        "strength": 0.86,
        "ambiguous_key": "D0_O6",
        "ambiguous_strength": 0.82,
        "expected_applied": False,
    },
    {
        "case": "TAIL_RELAX_6_TO_7",
        "source_slot": 6,
        "receiver_slot": 7,
        "target_key": "D4_O3",
        "strength": 0.91,
        "expected_applied": True,
    },
    {
        "case": "LOOP_BACK_7_TO_0",
        "source_slot": 7,
        "receiver_slot": 0,
        "target_key": "D5_O4",
        "strength": 0.90,
        "expected_applied": True,
    },
]


# ------------------------------------------------------------
# Simulation
# ------------------------------------------------------------

memory = initial_sequence.copy()
expected_final = initial_sequence.copy()

decision_rows = []
update_rows = []
wave_rows = []
snapshots = []

applied_count = 0
rejected_count = 0

for t, event in enumerate(cascade_events):
    source_slot = event["source_slot"]
    receiver_slot = event["receiver_slot"]

    source_key = memory[source_slot]
    receiver_before = memory[receiver_slot]

    packet = spectral_packet(
        target_key=event["target_key"],
        strength=event["strength"],
        ambiguous_key=event.get("ambiguous_key", None),
        ambiguous_strength=event.get("ambiguous_strength", 0.0)
    )

    decoded = decode_packet(packet)
    decoded_key = decoded["best_key"]

    rule_allowed, rule_reason, delta_d, delta_o, mode = transition_rule(
        receiver_before,
        decoded_key
    )

    score_ok = bool(decoded["score_ok"])
    applied = score_ok and rule_allowed

    if applied:
        memory[receiver_slot] = decoded_key
        expected_final[receiver_slot] = event["target_key"]
        applied_count += 1
    else:
        rejected_count += 1

    receiver_after = memory[receiver_slot]

    action = "APPLY_CASCADE_TRANSITION" if applied else (
        "REJECT_WEAK_OR_AMBIGUOUS" if not score_ok else "REJECT_FORBIDDEN_TRANSITION"
    )

    photon_token = transition_photon_token(receiver_before, decoded_key) if applied else "NONE"

    row = {
        "STATE": STATE,
        "T": t * 120,
        "CASE": event["case"],
        "SOURCE_SLOT": source_slot,
        "RECEIVER_SLOT": receiver_slot,
        "SOURCE_KEY": source_key,
        "RECEIVER_BEFORE": receiver_before,
        "PROPOSED_TARGET_KEY": event["target_key"],
        "DECODED_NEXT_KEY": decoded_key,
        "RECEIVER_AFTER": receiver_after,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "SCORE_OK": int(score_ok),
        "DELTA_D": delta_d,
        "DELTA_O": delta_o,
        "TRANSITION_MODE": mode,
        "RULE_ALLOWED": int(rule_allowed),
        "RULE_REASON": rule_reason,
        "APPLIED": int(applied),
        "EXPECTED_APPLIED": int(event["expected_applied"]),
        "ACTION": action,
        "PHOTON_TOKEN": photon_token,
    }

    decision_rows.append(row)

    update_rows.append({
        "STATE": STATE,
        "CASE": event["case"],
        "RECEIVER_SLOT": receiver_slot,
        "BEFORE_KEY": receiver_before,
        "DECODED_NEXT_KEY": decoded_key,
        "AFTER_KEY": receiver_after,
        "APPLIED": int(applied),
        "CHANGED": int(receiver_before != receiver_after),
        "HELD_IF_REJECTED": int((not applied) and receiver_before == receiver_after),
        "MATCH_EXPECTED_ACTION": int(applied == event["expected_applied"]),
    })

    wave_rows.append({
        "STATE": STATE,
        "CASE": event["case"],
        "SOURCE_SLOT": source_slot,
        "RECEIVER_SLOT": receiver_slot,
        "SOURCE_KEY": source_key,
        "RECEIVER_BEFORE": receiver_before,
        "RECEIVER_AFTER": receiver_after,
        "DELTA_D": delta_d,
        "DELTA_O": delta_o,
        "MODE": mode if applied else "NONE",
        "OUTPUT_TOKEN": photon_token,
        "APPLIED": int(applied),
    })

    snapshots.append((row, packet, memory.copy()))


df_decision = pd.DataFrame(decision_rows)
df_updates = pd.DataFrame(update_rows)
df_wave = pd.DataFrame(wave_rows)

transition_accuracy = float(np.mean(np.array(memory) == np.array(expected_final)))
rule_safety_accuracy = float(np.mean(df_updates["MATCH_EXPECTED_ACTION"]))

result = (
    "SPECTRAL_CASCADE_CHAIN_LOCKED"
    if transition_accuracy == 1.0 and rule_safety_accuracy == 1.0
    else "SPECTRAL_CASCADE_CHAIN_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": memory_slots,
    "INITIAL_SEQUENCE": " ".join(initial_sequence),
    "FINAL_SEQUENCE": " ".join(memory),
    "EXPECTED_FINAL_SEQUENCE": " ".join(expected_final),
    "CASCADE_EVENTS": len(cascade_events),
    "APPLIED_TRANSITIONS": applied_count,
    "REJECTED_TRANSITIONS": rejected_count,
    "TRANSITION_ACCURACY": round(transition_accuracy, 3),
    "RULE_SAFETY_ACCURACY": round(rule_safety_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "MAX_DIRECTION_STEP": max_direction_step,
    "ALLOWED_OCTAVE_JUMPS": ",".join(str(x) for x in allowed_octave_jumps),
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.25

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=13,
        linewidth=2.0,
        color=color,
        alpha=alpha
    )
    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.16
    gap = 0.03

    ax.text(x0, y0 + 1.45, title, fontsize=7, ha="left", va="bottom", color="black")

    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE"] else (None, None)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = packet[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)
            edge = "lime" if is_selected else "white"
            lw = 1.3 if is_selected else 0.3

            ax.add_patch(Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw))


def draw_memory_bar(ax, x0, y0, seq, title, source_slot=None, receiver_slot=None):
    cell = 0.24
    gap = 0.055

    ax.text(x0, y0 + 2.35, title, fontsize=7, ha="left", va="bottom", color="black")

    for i, key in enumerate(seq):
        x = x0
        y = y0 + (len(seq) - 1 - i) * (cell + gap)

        edge = "black"
        lw = 0.5

        if i == source_slot:
            edge = "dodgerblue"
            lw = 1.6

        if i == receiver_slot:
            edge = "lime"
            lw = 1.9

        ax.add_patch(Rectangle(
            (x, y),
            cell,
            cell,
            facecolor=key_color(key),
            edgecolor=edge,
            linewidth=lw
        ))

        ax.text(x + 0.34, y + cell / 2, str(i), fontsize=6, ha="left", va="center", color="black")


def draw_snapshot(ax, row, packet, memory_state):
    source_slot = row["SOURCE_SLOT"]
    receiver_slot = row["RECEIVER_SLOT"]

    source_key = row["SOURCE_KEY"]
    receiver_before = row["RECEIVER_BEFORE"]
    decoded_key = row["DECODED_NEXT_KEY"]
    receiver_after = row["RECEIVER_AFTER"]

    applied = bool(row["APPLIED"])
    rejected = not applied

    ax.set_facecolor("white")
    ax.set_xlim(0, 13)
    ax.set_ylim(0.5, 5.2)
    ax.set_aspect("equal")
    ax.axis("off")

    ax.add_patch(Circle((0.9, 3.2), 0.38, facecolor=key_color(source_key), edgecolor="dodgerblue", linewidth=2.0))
    ax.text(0.9, 2.55, f"SRC {source_slot}\n{source_key}", fontsize=7, ha="center", color="black")

    ax.add_patch(Circle((2.0, 3.2), 0.28, facecolor=key_color(decoded_key), edgecolor="black", linewidth=1.0))
    ax.text(2.0, 2.55, "PHOTON", fontsize=7, ha="center", color="black")

    ax.add_patch(Rectangle((3.0, 2.95), 0.42, 0.42, facecolor="limegreen" if applied else "red", edgecolor="black", linewidth=1.0))
    ax.text(3.21, 2.55, "GATE", fontsize=7, ha="center", color="black")

    ax.add_patch(Circle((4.6, 3.2), 0.38, facecolor=key_color(receiver_before), edgecolor="black", linewidth=1.0))
    ax.text(4.6, 2.45, f"BEFORE\n{receiver_before}", fontsize=7, ha="center", color="black")

    ax.add_patch(Circle((5.9, 3.2), 0.38, facecolor=key_color(receiver_after), edgecolor="lime" if applied else "red", linewidth=2.0))
    ax.text(5.9, 2.45, f"SLOT {receiver_slot}\n{receiver_after}", fontsize=7, ha="center", color="black")

    draw_arrow(ax, (1.28, 3.2), (1.72, 3.2), active=True, rejected=False)
    draw_arrow(ax, (2.28, 3.2), (3.0, 3.2), active=True, rejected=rejected)
    draw_arrow(ax, (3.42, 3.2), (4.22, 3.2), active=True, rejected=rejected)
    draw_arrow(ax, (4.98, 3.2), (5.52, 3.2), active=True, rejected=rejected)

    draw_packet_grid(ax, 6.8, 2.1, packet, "42-channel cascade packet", selected_key=decoded_key)
    draw_memory_bar(ax, 10.6, 1.35, memory_state, "cascade memory", source_slot=source_slot, receiver_slot=receiver_slot)

    ax.text(
        0.45,
        0.85,
        f"case={row['CASE']} score={row['BEST_SCORE']} margin={row['MARGIN']} "
        f"ΔD={row['DELTA_D']} ΔO={row['DELTA_O']} rule={row['RULE_REASON']} action={row['ACTION']}\n"
        f"{row['PHOTON_TOKEN']}",
        fontsize=7,
        ha="left",
        color="black"
    )

    ax.set_title(
        f"D32 snapshot t={row['T']} — {row['CASE']} / {row['ACTION']}",
        fontsize=10,
        color="black"
    )


fig, axes = plt.subplots(len(snapshots), 1, figsize=(8.5, 3.2 * len(snapshots)))

for ax, (row, packet, memory_state) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, memory_state)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:             {spectral_states}")
print(f"directions x octaves:        {directions} x {octaves}")
print(f"memory slots:                {memory_slots}")
print(f"initial sequence:            {' '.join(initial_sequence)}")
print(f"final sequence:              {' '.join(memory)}")
print(f"expected final sequence:     {' '.join(expected_final)}")
print(f"cascade events:              {len(cascade_events)}")
print(f"applied transitions:         {applied_count}")
print(f"rejected transitions:        {rejected_count}")
print(f"transition accuracy:         {round(transition_accuracy, 3)}")
print(f"rule safety accuracy:        {round(rule_safety_accuracy, 3)}")
print(f"resonance threshold:         {resonance_threshold}")
print(f"ambiguity margin:            {ambiguity_margin}")
print(f"max direction step:          {max_direction_step}")
print(f"allowed octave jumps:        {allowed_octave_jumps}")

print("\nCascade decision table:")
display(df_decision)

print("\nMemory update table:")
display(df_updates)

print("\nCascade photon table:")
display(df_wave)

print("\nSummary:")
display(df_summary)
```
