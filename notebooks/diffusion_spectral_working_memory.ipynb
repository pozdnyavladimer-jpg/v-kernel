```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D44_SPECTRAL_WORKING_MEMORY
# Active working memory after attention gate:
# attention-selected packets can enter, refresh, switch,
# decay, or be rejected.
#
# Continuation after:
# D36 coherence bus
# D37 orbital intent
# D38 honeycomb lattice
# D39 honeycomb memory wave
# D40 macro resonance loop
# D41 field memory decay / reinforcement
# D42 self-tuning filter
# D43 spectral attention gate
#
# Goal:
# - memory is not only long-term field storage
# - working memory holds active intent across time
# - coherent attended packets refresh active state
# - strong trusted packets can switch active focus
# - weak / ambiguous / phase-unlocked packets are rejected
# - idle cells decay gradually
# ============================================================

np.random.seed(44)

STATE = "D44_SPECTRAL_WORKING_MEMORY"

DIRECTIONS = 6
OCTAVES = 7
SPECTRAL_STATES = DIRECTIONS * OCTAVES
MEMORY_SLOTS = 8

RESONANCE_THRESHOLD = 0.62
ATTENTION_THRESHOLD = 0.58
WORKING_THRESHOLD = 0.55
SWITCH_MARGIN = 0.14
AMBIGUITY_MARGIN = 0.12
PHASE_TOLERANCE = 0.18
DECAY_RATE = 0.12
REFRESH_GAIN = 0.18
TRUST_GAIN = 0.12
MAX_TTL = 4

ORBITAL_MODES = [
    "NONE",
    "SPHERE",
    "RING",
    "TWO_LOBE",
    "BUTTERFLY",
    "FLOWER",
    "HEX",
    "SPIRAL",
]

MODE_COLOR = {
    "NONE": "black",
    "SPHERE": "cyan",
    "RING": "lime",
    "TWO_LOBE": "dodgerblue",
    "BUTTERFLY": "orange",
    "FLOWER": "violet",
    "HEX": "deepskyblue",
    "SPIRAL": "hotpink",
}

def key(d, o):
    if d is None or o is None:
        return "EMPTY"
    return f"D{d}_O{o}"

def parse_key(k):
    if k in ["EMPTY", "NONE", None]:
        return None, None
    d = int(k.split("_")[0][1:])
    o = int(k.split("_")[1][1:])
    return d, o

def state_id(d, o):
    return o * DIRECTIONS + d

def phase_of_key(k):
    d, o = parse_key(k)
    if d is None:
        return None
    return d / DIRECTIONS

def circular_phase_error(a, b):
    if a is None or b is None:
        return 1.0
    diff = abs(a - b) % 1.0
    return min(diff, 1.0 - diff)

def make_token(k, mode, amp, phase, ttl):
    if k in ["EMPTY", "NONE"]:
        return "NONE"
    d, o = parse_key(k)
    return f"WM_PHOTON[D{d}|O{o}|{mode}|A{amp:.2f}|P{phase:.3f}|TTL{ttl}]"

def spectral_packet(k, mode, amp=0.9, noise=0.02):
    packet = np.zeros((OCTAVES, DIRECTIONS))
    d, o = parse_key(k)
    if d is None:
        return packet
    packet[o, d] = amp
    for _ in range(3):
        rd = np.random.randint(0, DIRECTIONS)
        ro = np.random.randint(0, OCTAVES)
        packet[ro, rd] += np.random.uniform(0.0, noise)
    return packet

def decode_packet(packet):
    flat = packet.flatten()
    best_id = int(np.argmax(flat))
    best_score = float(flat[best_id])
    second_id = int(np.argsort(flat)[-2])
    second_score = float(flat[second_id])
    best_o = best_id // DIRECTIONS
    best_d = best_id % DIRECTIONS
    second_o = second_id // DIRECTIONS
    second_d = second_id % DIRECTIONS
    decoded = key(best_d, best_o)
    second = key(second_d, second_o)
    margin = best_score - second_score
    return decoded, best_id, best_score, second, second_score, margin

def mode_match(a, b):
    if a == "NONE" or b == "NONE":
        return False
    return a == b

def orbital_priority(mode):
    order = {
        "NONE": 0,
        "SPHERE": 1,
        "RING": 2,
        "TWO_LOBE": 3,
        "BUTTERFLY": 4,
        "FLOWER": 5,
        "HEX": 6,
        "SPIRAL": 4,
    }
    return order.get(mode, 0)

# ----------------------------
# Working memory initial state
# ----------------------------

working_memory = [
    {
        "slot": 0,
        "key": "D4_O5",
        "mode": "FLOWER",
        "strength": 0.74,
        "ttl": 3,
        "trust": 0.68,
        "phase": phase_of_key("D4_O5"),
    },
    {
        "slot": 1,
        "key": "D1_O2",
        "mode": "RING",
        "strength": 0.58,
        "ttl": 2,
        "trust": 0.50,
        "phase": phase_of_key("D1_O2"),
    },
    {
        "slot": 2,
        "key": "EMPTY",
        "mode": "NONE",
        "strength": 0.00,
        "ttl": 0,
        "trust": 0.00,
        "phase": None,
    },
    {
        "slot": 3,
        "key": "D3_O6",
        "mode": "HEX",
        "strength": 0.62,
        "ttl": 1,
        "trust": 0.54,
        "phase": phase_of_key("D3_O6"),
    },
    {
        "slot": 4,
        "key": "D0_O1",
        "mode": "SPHERE",
        "strength": 0.39,
        "ttl": 1,
        "trust": 0.32,
        "phase": phase_of_key("D0_O1"),
    },
    {
        "slot": 5,
        "key": "EMPTY",
        "mode": "NONE",
        "strength": 0.00,
        "ttl": 0,
        "trust": 0.00,
        "phase": None,
    },
    {
        "slot": 6,
        "key": "D2_O3",
        "mode": "BUTTERFLY",
        "strength": 0.57,
        "ttl": 2,
        "trust": 0.49,
        "phase": phase_of_key("D2_O3"),
    },
    {
        "slot": 7,
        "key": "EMPTY",
        "mode": "NONE",
        "strength": 0.00,
        "ttl": 0,
        "trust": 0.00,
        "phase": None,
    },
]

initial_sequence = " ".join([m["key"] for m in working_memory])

# ----------------------------
# Test events
# ----------------------------

events = [
    {
        "case": "REFRESH_ACTIVE_FLOWER_0",
        "slot": 0,
        "incoming_key": "D4_O5",
        "incoming_mode": "FLOWER",
        "amp": 0.92,
        "attention": 0.91,
        "measured_phase": 0.667,
        "source": "ATTENDED",
        "expected_action": "REFRESH_WORKING_MEMORY",
    },
    {
        "case": "DECAY_IDLE_RING_1",
        "slot": 1,
        "incoming_key": "NONE",
        "incoming_mode": "NONE",
        "amp": 0.00,
        "attention": 0.00,
        "measured_phase": None,
        "source": "IDLE",
        "expected_action": "PASSIVE_DECAY_ONLY",
    },
    {
        "case": "WRITE_EMPTY_HEX_2",
        "slot": 2,
        "incoming_key": "D3_O6",
        "incoming_mode": "HEX",
        "amp": 0.88,
        "attention": 0.83,
        "measured_phase": 0.492,
        "source": "ATTENDED",
        "expected_action": "WRITE_WORKING_MEMORY",
    },
    {
        "case": "SWITCH_WEAK_SPHERE_TO_RING_4",
        "slot": 4,
        "incoming_key": "D1_O2",
        "incoming_mode": "RING",
        "amp": 0.86,
        "attention": 0.78,
        "measured_phase": 0.170,
        "source": "ATTENDED",
        "expected_action": "SWITCH_ACTIVE_INTENT",
    },
    {
        "case": "REJECT_WRONG_INTENT_6",
        "slot": 6,
        "incoming_key": "D2_O3",
        "incoming_mode": "SPHERE",
        "amp": 0.83,
        "attention": 0.75,
        "measured_phase": 0.336,
        "source": "ATTENDED",
        "expected_action": "REJECT_INTENT_MISMATCH",
    },
    {
        "case": "REJECT_PHASE_UNLOCKED_3",
        "slot": 3,
        "incoming_key": "D3_O6",
        "incoming_mode": "HEX",
        "amp": 0.89,
        "attention": 0.81,
        "measured_phase": 0.840,
        "source": "ATTENDED",
        "expected_action": "REJECT_PHASE_UNLOCKED",
    },
    {
        "case": "WEAK_PACKET_FORGET_SLOT_4",
        "slot": 4,
        "incoming_key": "D1_O2",
        "incoming_mode": "RING",
        "amp": 0.34,
        "attention": 0.30,
        "measured_phase": 0.168,
        "source": "WEAK",
        "expected_action": "REJECT_WEAK_AND_DECAY",
    },
    {
        "case": "AMBIGUOUS_FLOWER_HOLD_0",
        "slot": 0,
        "incoming_key": "D4_O5",
        "incoming_mode": "FLOWER",
        "amp": 0.76,
        "attention": 0.60,
        "measured_phase": 0.672,
        "force_ambiguous": True,
        "source": "AMBIGUOUS",
        "expected_action": "HOLD_WORKING_MEMORY",
    },
    {
        "case": "TRUSTED_MEMORY_WINS_0",
        "slot": 0,
        "incoming_key": "D4_O5",
        "incoming_mode": "FLOWER",
        "amp": 0.73,
        "attention": 0.62,
        "measured_phase": 0.665,
        "trusted_memory": True,
        "source": "TRUSTED",
        "expected_action": "REINFORCE_TRUSTED_MEMORY",
    },
    {
        "case": "OVERWRITE_EXPIRED_SLOT_7",
        "slot": 7,
        "incoming_key": "D5_O4",
        "incoming_mode": "SPIRAL",
        "amp": 0.90,
        "attention": 0.84,
        "measured_phase": 0.835,
        "source": "ATTENDED",
        "expected_action": "WRITE_WORKING_MEMORY",
    },
]

decision_rows = []
memory_rows = []
wave_rows = []
snapshots = []

def decay_slot(cell):
    cell["strength"] = max(0.0, cell["strength"] - DECAY_RATE)
    cell["ttl"] = max(0, cell["ttl"] - 1)
    if cell["ttl"] == 0 and cell["strength"] < WORKING_THRESHOLD:
        cell["key"] = "EMPTY"
        cell["mode"] = "NONE"
        cell["trust"] = max(0.0, cell["trust"] - DECAY_RATE)
        cell["phase"] = None
    return cell

def draw_orbital_icon(ax, x, y, mode, scale=0.18, edge="black", alpha=0.9):
    color = MODE_COLOR.get(mode, "gray")

    if mode == "NONE":
        ax.add_patch(Circle((x, y), scale * 0.45, facecolor="black", edgecolor=edge, lw=1.2, alpha=alpha))
        return

    if mode == "SPHERE":
        ax.add_patch(Circle((x, y), scale, facecolor=color, edgecolor=edge, lw=1.2, alpha=alpha))
        return

    if mode == "RING":
        ax.add_patch(Circle((x, y), scale, facecolor="none", edgecolor=color, lw=3.0, alpha=alpha))
        ax.add_patch(Circle((x, y), scale * 0.45, facecolor="white", edgecolor=edge, lw=1.0, alpha=alpha))
        return

    if mode == "TWO_LOBE":
        ax.add_patch(Circle((x - scale * 0.55, y), scale * 0.55, facecolor=color, edgecolor=edge, lw=1.0, alpha=alpha))
        ax.add_patch(Circle((x + scale * 0.55, y), scale * 0.55, facecolor=color, edgecolor=edge, lw=1.0, alpha=alpha))
        return

    if mode == "BUTTERFLY":
        for dx, dy in [(-0.45, 0.45), (0.45, 0.45), (-0.45, -0.45), (0.45, -0.45)]:
            ax.add_patch(Circle((x + dx * scale, y + dy * scale), scale * 0.42, facecolor=color, edgecolor=edge, lw=1.0, alpha=alpha))
        return

    if mode == "FLOWER":
        for a in np.linspace(0, 2*np.pi, 6, endpoint=False):
            ax.add_patch(Circle((x + np.cos(a) * scale * 0.62, y + np.sin(a) * scale * 0.62),
                                scale * 0.35, facecolor=color, edgecolor=edge, lw=0.9, alpha=alpha))
        ax.add_patch(Circle((x, y), scale * 0.28, facecolor="white", edgecolor=edge, lw=1.0, alpha=alpha))
        return

    if mode == "HEX":
        for a in np.linspace(0, 2*np.pi, 6, endpoint=False):
            ax.plot([x, x + np.cos(a) * scale], [y, y + np.sin(a) * scale], color=color, lw=1.8, alpha=alpha)
            ax.add_patch(Circle((x + np.cos(a) * scale, y + np.sin(a) * scale),
                                scale * 0.16, facecolor=color, edgecolor=edge, lw=0.8, alpha=alpha))
        ax.add_patch(Circle((x, y), scale * 0.18, facecolor="white", edgecolor=edge, lw=1.0, alpha=alpha))
        return

    if mode == "SPIRAL":
        t = np.linspace(0, 3.8*np.pi, 120)
        r = scale * t / t.max()
        ax.plot(x + r*np.cos(t), y + r*np.sin(t), color=color, lw=2.4, alpha=alpha)
        ax.add_patch(Circle((x, y), scale * 0.12, facecolor=color, edgecolor=edge, lw=0.8, alpha=alpha))
        return

def draw_packet_grid(ax, packet, x0, y0, w=0.72, h=0.46):
    cell_w = w / DIRECTIONS
    cell_h = h / OCTAVES
    maxv = max(1e-9, float(packet.max()))
    for o in range(OCTAVES):
        for d in range(DIRECTIONS):
            val = packet[o, d] / maxv
            col = plt.cm.viridis(val)
            ax.add_patch(Rectangle((x0 + d*cell_w, y0 + (OCTAVES-1-o)*cell_h),
                                   cell_w*0.9, cell_h*0.9,
                                   facecolor=col, edgecolor="white", lw=0.3))
    ax.text(x0 + w/2, y0 + h + 0.03, "42-channel packet", ha="center", va="bottom", fontsize=7)

def draw_memory_bar(ax, memory, x0, y0, w=0.10, h=0.62):
    slot_h = h / MEMORY_SLOTS
    for i, cell in enumerate(memory):
        mode = cell["mode"]
        color = MODE_COLOR.get(mode, "gray")
        strength = cell["strength"]
        yy = y0 + (MEMORY_SLOTS-1-i)*slot_h
        ax.add_patch(Rectangle((x0, yy), w, slot_h*0.82, facecolor=color, edgecolor="black", lw=0.4, alpha=0.85 if strength > 0 else 0.25))
        ax.text(x0 + w + 0.02, yy + slot_h*0.4, str(i), fontsize=6, va="center")
    ax.text(x0 + w/2, y0 + h + 0.03, "working memory", ha="center", va="bottom", fontsize=7, rotation=0)

def draw_strength_meter(ax, x, y, value, label="strength"):
    ax.text(x, y + 0.08, label, fontsize=7, ha="left")
    ax.add_patch(Rectangle((x, y), 0.45, 0.045, facecolor="white", edgecolor="black", lw=0.7))
    ax.add_patch(Rectangle((x, y), 0.45 * max(0, min(1, value)), 0.045, facecolor="limegreen", edgecolor="none", alpha=0.8))
    ax.text(x + 0.48, y + 0.02, f"{value:.2f}", fontsize=7, va="center")

def draw_pipeline_snapshot(ax, event, before_cell, after_cell, packet, decoded_key, action, best_score, margin, phase_ok, accepted, t):
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 3)
    ax.axis("off")

    title = f"D44 snapshot t={t} — {event['case']} / {action}"
    ax.text(0.02, 2.86, title, fontsize=8, ha="left", va="top")

    src_key = event["incoming_key"]
    src_mode = event["incoming_mode"]

    # incoming source
    draw_orbital_icon(ax, 0.55, 1.78, src_mode if src_mode != "NONE" else "NONE", scale=0.22)
    ax.text(0.55, 1.30, src_key, fontsize=7, ha="center")
    ax.text(0.55, 1.12, "INPUT", fontsize=6, ha="center")

    # attention gate
    ax.add_patch(Circle((1.45, 1.78), 0.17, facecolor="gold", edgecolor="black", lw=1.0))
    ax.text(1.45, 1.32, "ATTN", fontsize=6, ha="center")
    ax.text(1.45, 1.12, f"{event['attention']:.2f}", fontsize=6, ha="center")

    # field
    ax.add_patch(Rectangle((2.25, 1.62), 0.26, 0.26, facecolor="limegreen", edgecolor="black", lw=0.8))
    ax.text(2.38, 1.32, "FIELD", fontsize=6, ha="center")

    # clock
    clock_color = "limegreen" if phase_ok else "red"
    ax.add_patch(Rectangle((3.10, 1.62), 0.26, 0.26, facecolor=clock_color, edgecolor="black", lw=0.8))
    ax.text(3.23, 1.32, "CLOCK", fontsize=6, ha="center")

    # gate
    gate_color = "limegreen" if accepted else "red"
    ax.add_patch(Rectangle((3.95, 1.62), 0.26, 0.26, facecolor=gate_color, edgecolor="black", lw=0.8))
    ax.text(4.08, 1.32, "WM GATE", fontsize=6, ha="center")

    # memory before / after
    draw_orbital_icon(ax, 5.05, 1.78, before_cell["mode"], scale=0.23)
    ax.text(5.05, 1.30, before_cell["key"], fontsize=7, ha="center")
    ax.text(5.05, 1.10, "before", fontsize=6, ha="center")

    draw_orbital_icon(ax, 6.10, 1.78, after_cell["mode"], scale=0.23, edge="green" if accepted else "black")
    ax.text(6.10, 1.30, after_cell["key"], fontsize=7, ha="center")
    ax.text(6.10, 1.10, f"slot {event['slot']}", fontsize=6, ha="center")

    # arrows
    for x1, x2 in [(0.80, 1.24), (1.66, 2.20), (2.54, 3.08), (3.38, 3.92), (4.25, 4.83), (5.28, 5.88)]:
        ax.add_patch(FancyArrowPatch((x1, 1.78), (x2, 1.78), arrowstyle="->", mutation_scale=12, lw=1.2, color="goldenrod"))

    draw_strength_meter(ax, 0.25, 0.42, after_cell["strength"], label="wm strength")
    draw_packet_grid(ax, packet, 7.05, 1.33, w=0.75, h=0.48)
    draw_memory_bar(ax, working_memory, 8.28, 1.22, w=0.12, h=0.68)

    ax.text(
        0.25,
        0.08,
        f"decoded={decoded_key} | score={best_score:.3f} | margin={margin:.3f} | "
        f"phase_ok={int(phase_ok)} | accepted={int(accepted)} | ttl={after_cell['ttl']} | trust={after_cell['trust']:.2f}",
        fontsize=6,
        ha="left",
        va="bottom",
    )

# ----------------------------
# Main simulation
# ----------------------------

for idx, event in enumerate(events):
    t = idx * 120
    slot = event["slot"]

    before_cell = working_memory[slot].copy()

    if event["incoming_key"] == "NONE":
        packet = np.zeros((OCTAVES, DIRECTIONS))
        decoded_key = "NONE"
        best_id = None
        best_score = 0.0
        second_key = "NONE"
        second_score = 0.0
        margin = 0.0
    else:
        packet = spectral_packet(event["incoming_key"], event["incoming_mode"], amp=event["amp"], noise=0.025)

        if event.get("force_ambiguous", False):
            d, o = parse_key(event["incoming_key"])
            packet[o, d] = event["amp"]
            alt_d = (d + 2) % DIRECTIONS
            alt_o = max(0, min(OCTAVES - 1, o - 1))
            packet[alt_o, alt_d] = event["amp"] - 0.04

        decoded_key, best_id, best_score, second_key, second_score, margin = decode_packet(packet)

    expected_phase = phase_of_key(decoded_key)
    measured_phase = event["measured_phase"]
    phase_error = circular_phase_error(measured_phase, expected_phase)
    phase_ok = phase_error <= PHASE_TOLERANCE if decoded_key != "NONE" else False

    score_ok = best_score >= RESONANCE_THRESHOLD
    attention_ok = event["attention"] >= ATTENTION_THRESHOLD
    margin_ok = margin >= AMBIGUITY_MARGIN
    packet_accepted = score_ok and attention_ok and margin_ok and phase_ok

    same_key = decoded_key == before_cell["key"]
    same_mode = mode_match(event["incoming_mode"], before_cell["mode"])
    empty_slot = before_cell["key"] == "EMPTY"
    expired = before_cell["ttl"] <= 0 or before_cell["strength"] < WORKING_THRESHOLD
    incoming_priority = orbital_priority(event["incoming_mode"])
    current_priority = orbital_priority(before_cell["mode"])

    switch_power = event["amp"] + event["attention"] + incoming_priority * 0.03
    current_power = before_cell["strength"] + before_cell["trust"] + current_priority * 0.03
    switch_allowed = packet_accepted and (switch_power > current_power + SWITCH_MARGIN)

    action = "HOLD_WORKING_MEMORY"
    accepted = False
    reason = "NO_CHANGE"

    after_cell = before_cell.copy()

    if event["incoming_key"] == "NONE":
        after_cell = decay_slot(after_cell)
        action = "PASSIVE_DECAY_ONLY"
        reason = "NO_INPUT_DECAY"
        accepted = False

    elif not score_ok or not attention_ok or not margin_ok:
        after_cell = decay_slot(after_cell)
        action = "REJECT_WEAK_OR_AMBIGUOUS_AND_DECAY"
        reason = "WEAK_ATTENTION_OR_AMBIGUOUS"
        accepted = False

    elif not phase_ok:
        after_cell = decay_slot(after_cell)
        action = "REJECT_PHASE_UNLOCKED"
        reason = "PHASE_UNLOCKED"
        accepted = False

    elif empty_slot or expired:
        after_cell["key"] = decoded_key
        after_cell["mode"] = event["incoming_mode"]
        after_cell["strength"] = min(1.0, best_score + 0.03)
        after_cell["ttl"] = MAX_TTL
        after_cell["trust"] = min(1.0, event["attention"] * 0.75)
        after_cell["phase"] = measured_phase
        action = "WRITE_WORKING_MEMORY"
        reason = "EMPTY_OR_EXPIRED_SLOT"
        accepted = True

    elif same_key and same_mode:
        after_cell["strength"] = min(1.0, before_cell["strength"] + REFRESH_GAIN + event["attention"] * 0.08)
        after_cell["ttl"] = MAX_TTL
        after_cell["trust"] = min(1.0, before_cell["trust"] + TRUST_GAIN)
        after_cell["phase"] = measured_phase

        if event.get("trusted_memory", False):
            action = "REINFORCE_TRUSTED_MEMORY"
            reason = "TRUSTED_REPEAT"
        else:
            action = "REFRESH_WORKING_MEMORY"
            reason = "SAME_KEY_SAME_MODE_REFRESH"

        accepted = True

    elif same_key and not same_mode:
        after_cell = decay_slot(after_cell)
        action = "REJECT_INTENT_MISMATCH"
        reason = "SAME_KEY_DIFFERENT_ORBITAL_INTENT"
        accepted = False

    elif switch_allowed:
        after_cell["key"] = decoded_key
        after_cell["mode"] = event["incoming_mode"]
        after_cell["strength"] = min(1.0, best_score + event["attention"] * 0.07)
        after_cell["ttl"] = MAX_TTL
        after_cell["trust"] = min(1.0, event["attention"] * 0.80)
        after_cell["phase"] = measured_phase
        action = "SWITCH_ACTIVE_INTENT"
        reason = "NEW_PACKET_STRONGER_THAN_CURRENT_MEMORY"
        accepted = True

    else:
        after_cell = decay_slot(after_cell)
        action = "HOLD_WORKING_MEMORY"
        reason = "CURRENT_MEMORY_WINS"
        accepted = False

    working_memory[slot] = after_cell.copy()

    expected_action = event["expected_action"]
    match_expected = int(
        expected_action == action
        or (expected_action == "REJECT_WEAK_AND_DECAY" and action == "REJECT_WEAK_OR_AMBIGUOUS_AND_DECAY")
        or (expected_action == "HOLD_WORKING_MEMORY" and action == "HOLD_WORKING_MEMORY")
    )

    output_token = make_token(after_cell["key"], after_cell["mode"], after_cell["strength"], after_cell["phase"] if after_cell["phase"] is not None else 0.0, after_cell["ttl"]) if accepted else "NONE"

    decision_rows.append({
        "STATE": STATE,
        "T": t,
        "CASE": event["case"],
        "MEMORY_SLOT": slot,
        "BEFORE_KEY": before_cell["key"],
        "BEFORE_MODE": before_cell["mode"],
        "INCOMING_KEY": event["incoming_key"],
        "INCOMING_MODE": event["incoming_mode"],
        "DECODED_KEY": decoded_key,
        "BEST_STATE_ID": best_id if best_id is not None else "NONE",
        "BEST_SCORE": round(best_score, 3),
        "SECOND_KEY": second_key,
        "SECOND_SCORE": round(second_score, 3),
        "MARGIN": round(margin, 3),
        "ATTENTION": round(event["attention"], 3),
        "SCORE_OK": int(score_ok),
        "ATTENTION_OK": int(attention_ok),
        "MARGIN_OK": int(margin_ok),
        "EXPECTED_PHASE": None if expected_phase is None else round(expected_phase, 3),
        "MEASURED_PHASE": None if measured_phase is None else round(measured_phase, 3),
        "PHASE_ERROR": round(phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "PACKET_ACCEPTED": int(packet_accepted),
        "SWITCH_ALLOWED": int(switch_allowed),
        "AFTER_KEY": after_cell["key"],
        "AFTER_MODE": after_cell["mode"],
        "AFTER_STRENGTH": round(after_cell["strength"], 3),
        "AFTER_TTL": after_cell["ttl"],
        "AFTER_TRUST": round(after_cell["trust"], 3),
        "ACTION": action,
        "REASON": reason,
        "EXPECTED_ACTION": expected_action,
        "MATCH_EXPECTED_ACTION": match_expected,
        "OUTPUT_TOKEN": output_token,
    })

    memory_rows.append({
        "STATE": STATE,
        "CASE": event["case"],
        "MEMORY_SLOT": slot,
        "BEFORE_KEY": before_cell["key"],
        "BEFORE_MODE": before_cell["mode"],
        "BEFORE_STRENGTH": round(before_cell["strength"], 3),
        "BEFORE_TTL": before_cell["ttl"],
        "DECODED_KEY": decoded_key,
        "INCOMING_MODE": event["incoming_mode"],
        "AFTER_KEY": after_cell["key"],
        "AFTER_MODE": after_cell["mode"],
        "AFTER_STRENGTH": round(after_cell["strength"], 3),
        "AFTER_TTL": after_cell["ttl"],
        "AFTER_TRUST": round(after_cell["trust"], 3),
        "ACTION": action,
        "CHANGED": int(before_cell["key"] != after_cell["key"] or before_cell["mode"] != after_cell["mode"]),
        "REFRESHED": int(action in ["REFRESH_WORKING_MEMORY", "REINFORCE_TRUSTED_MEMORY"]),
        "HELD_OR_DECAYED": int(action in ["HOLD_WORKING_MEMORY", "PASSIVE_DECAY_ONLY", "REJECT_WEAK_OR_AMBIGUOUS_AND_DECAY", "REJECT_PHASE_UNLOCKED", "REJECT_INTENT_MISMATCH"]),
    })

    wave_rows.append({
        "STATE": STATE,
        "CASE": event["case"],
        "INCOMING_KEY": event["incoming_key"],
        "INCOMING_MODE": event["incoming_mode"],
        "DECODED_KEY": decoded_key,
        "ATTENTION": round(event["attention"], 3),
        "BEST_SCORE": round(best_score, 3),
        "PHASE_OK": int(phase_ok),
        "ACTION": action,
        "OUTPUT_TOKEN": output_token,
    })

    snapshots.append((event.copy(), before_cell.copy(), after_cell.copy(), packet.copy(), decoded_key, action, best_score, margin, phase_ok, accepted, t))

df_decisions = pd.DataFrame(decision_rows)
df_memory = pd.DataFrame(memory_rows)
df_wave = pd.DataFrame(wave_rows)

final_sequence = " ".join([m["key"] for m in working_memory])
final_modes = " ".join([m["mode"] for m in working_memory])
final_strengths = " ".join([f"{m['strength']:.2f}" for m in working_memory])
final_ttls = " ".join([str(m["ttl"]) for m in working_memory])

accepted_events = int(df_decisions["PACKET_ACCEPTED"].sum())
written_events = int((df_decisions["ACTION"] == "WRITE_WORKING_MEMORY").sum())
refreshed_events = int(df_decisions["ACTION"].isin(["REFRESH_WORKING_MEMORY", "REINFORCE_TRUSTED_MEMORY"]).sum())
switch_events = int((df_decisions["ACTION"] == "SWITCH_ACTIVE_INTENT").sum())
rejected_events = int(df_decisions["ACTION"].str.contains("REJECT").sum())
decay_events = int(df_decisions["ACTION"].str.contains("DECAY").sum())
held_events = int((df_decisions["ACTION"] == "HOLD_WORKING_MEMORY").sum())

working_accuracy = round(float(df_decisions["MATCH_EXPECTED_ACTION"].mean()), 3)
active_slots = sum(1 for m in working_memory if m["key"] != "EMPTY")
mean_strength = round(float(np.mean([m["strength"] for m in working_memory])), 3)
mean_trust = round(float(np.mean([m["trust"] for m in working_memory])), 3)

result = "SPECTRAL_WORKING_MEMORY_LOCKED" if working_accuracy == 1.0 and active_slots >= 4 else "SPECTRAL_WORKING_MEMORY_PARTIAL"

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": SPECTRAL_STATES,
    "DIRECTIONS": DIRECTIONS,
    "OCTAVES": OCTAVES,
    "MEMORY_SLOTS": MEMORY_SLOTS,
    "ORBITAL_MODES": len(ORBITAL_MODES),
    "INITIAL_SEQUENCE": initial_sequence,
    "FINAL_SEQUENCE": final_sequence,
    "FINAL_MODES": final_modes,
    "FINAL_STRENGTHS": final_strengths,
    "FINAL_TTLS": final_ttls,
    "CASES_TESTED": len(events),
    "ACCEPTED_EVENTS": accepted_events,
    "WRITTEN_EVENTS": written_events,
    "REFRESHED_EVENTS": refreshed_events,
    "SWITCH_EVENTS": switch_events,
    "REJECTED_EVENTS": rejected_events,
    "DECAY_EVENTS": decay_events,
    "HELD_EVENTS": held_events,
    "ACTIVE_SLOTS": active_slots,
    "MEAN_STRENGTH": mean_strength,
    "MEAN_TRUST": mean_trust,
    "WORKING_MEMORY_ACCURACY": working_accuracy,
    "RESONANCE_THRESHOLD": RESONANCE_THRESHOLD,
    "ATTENTION_THRESHOLD": ATTENTION_THRESHOLD,
    "WORKING_THRESHOLD": WORKING_THRESHOLD,
    "SWITCH_MARGIN": SWITCH_MARGIN,
    "AMBIGUITY_MARGIN": AMBIGUITY_MARGIN,
    "PHASE_TOLERANCE": PHASE_TOLERANCE,
    "DECAY_RATE": DECAY_RATE,
    "REFRESH_GAIN": REFRESH_GAIN,
    "TRUST_GAIN": TRUST_GAIN,
    "MAX_TTL": MAX_TTL,
}])

# ----------------------------
# Visual snapshots
# ----------------------------

fig_h = max(3.0 * len(snapshots), 16)
fig, axes = plt.subplots(len(snapshots), 1, figsize=(10, fig_h))

if len(snapshots) == 1:
    axes = [axes]

for ax, snap in zip(axes, snapshots):
    draw_pipeline_snapshot(ax, *snap)

plt.tight_layout()
plt.show()

# ----------------------------
# Text output
# ----------------------------

print(f"STATE: {result}")
print(f"spectral states:              {SPECTRAL_STATES}")
print(f"directions x octaves:         {DIRECTIONS} x {OCTAVES}")
print(f"memory slots:                 {MEMORY_SLOTS}")
print(f"orbital modes:                {len(ORBITAL_MODES)}")
print(f"initial sequence:             {initial_sequence}")
print(f"final sequence:               {final_sequence}")
print(f"final modes:                  {final_modes}")
print(f"final strengths:              {final_strengths}")
print(f"final TTLs:                   {final_ttls}")
print(f"cases tested:                 {len(events)}")
print(f"accepted events:              {accepted_events}")
print(f"written events:               {written_events}")
print(f"refreshed events:             {refreshed_events}")
print(f"switch events:                {switch_events}")
print(f"rejected events:              {rejected_events}")
print(f"decay events:                 {decay_events}")
print(f"held events:                  {held_events}")
print(f"active slots:                 {active_slots}")
print(f"mean strength:                {mean_strength}")
print(f"mean trust:                   {mean_trust}")
print(f"working memory accuracy:      {working_accuracy}")
print(f"resonance threshold:          {RESONANCE_THRESHOLD}")
print(f"attention threshold:          {ATTENTION_THRESHOLD}")
print(f"working threshold:            {WORKING_THRESHOLD}")
print(f"switch margin:                {SWITCH_MARGIN}")
print(f"ambiguity margin:             {AMBIGUITY_MARGIN}")
print(f"phase tolerance:              {PHASE_TOLERANCE}")
print(f"decay rate:                   {DECAY_RATE}")
print(f"refresh gain:                 {REFRESH_GAIN}")
print(f"trust gain:                   {TRUST_GAIN}")
print(f"max TTL:                      {MAX_TTL}")

print("\nWorking memory decision table:")
display(df_decisions)

print("\nWorking memory update table:")
display(df_memory)

print("\nWorking memory photon table:")
display(df_wave)

print("\nSummary:")
display(df_summary)
```
