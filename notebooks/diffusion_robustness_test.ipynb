```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# D16_ROBUSTNESS_TEST
# Stress-test for the D15 full field I/O loop.
#
# Tests:
# - noise
# - signal loss
# - delay drift
# - threshold drift
# - memory decay
# - combined stress
# - break point
# ============================================================

INPUT_SEQUENCE = "10110101"

BIT_LEN = 120
BASE_DECODE_DELAY = 35
BASE_THRESHOLD = 0.38
CARRIER_PERIOD = 18
SEED = 16


def lowpass(x, alpha=0.25):
    y = np.zeros_like(x, dtype=float)
    y[0] = x[0]

    for i in range(1, len(x)):
        y[i] = (1 - alpha) * y[i - 1] + alpha * x[i]

    return y


def shift_signal(x, delay):
    delay = int(delay)
    y = np.zeros_like(x, dtype=float)

    if delay > 0:
        y[delay:] = x[:-delay]
    elif delay < 0:
        y[:delay] = x[-delay:]
    else:
        y[:] = x

    return y


def build_leader_wave(sequence, bit_len=BIT_LEN, high=0.92, low=0.22):
    total = len(sequence) * bit_len + 2 * bit_len
    signal = np.zeros(total, dtype=float)

    for i, bit in enumerate(sequence):
        start = i * bit_len
        end = start + bit_len

        amp = high if bit == "1" else low
        local_t = np.arange(bit_len)

        carrier = 0.5 + 0.5 * np.sin(
            2 * np.pi * local_t / CARRIER_PERIOD
        )

        signal[start:end] = amp * (0.62 + 0.38 * carrier)

    return signal


def decode_bits(follower_signal, sequence_len, bit_len, decode_delay, threshold):
    decoded = []
    means = []

    for i in range(sequence_len):
        start = i * bit_len + decode_delay
        end = start + bit_len

        window = follower_signal[
            max(0, start):min(len(follower_signal), end)
        ]

        if len(window) == 0:
            mean_value = 0.0
        else:
            mean_value = float(np.mean(window))

        bit = 1 if mean_value >= threshold else 0

        means.append(mean_value)
        decoded.append(bit)

    return decoded, means


def write_memory(decoded_bits, noise, memory_decay, rng):
    memory_values = []

    for bit in decoded_bits:
        if bit == 1:
            value = 1.0 - memory_decay + rng.normal(0.0, noise * 0.25)
        else:
            value = 0.0 + memory_decay * 0.28 + rng.normal(0.0, noise * 0.25)

        memory_values.append(float(np.clip(value, 0.0, 1.0)))

    stored_bits = [
        1 if value >= 0.5 else 0
        for value in memory_values
    ]

    return memory_values, stored_bits


def read_memory(memory_values, threshold, noise, rng):
    output_means = []
    readout_bits = []

    for value in memory_values:
        output_mean = 0.20 + 0.58 * value + rng.normal(0.0, noise * 0.2)
        output_mean = float(np.clip(output_mean, 0.0, 1.0))

        output_means.append(output_mean)
        readout_bits.append(1 if output_mean >= threshold else 0)

    return output_means, readout_bits


def run_case(
    name,
    noise=0.0,
    signal_loss=0.0,
    delay_drift=0,
    threshold_drift=0.0,
    memory_decay=0.0,
    bridge_alpha=0.25,
    seed=SEED,
):
    rng = np.random.default_rng(seed)

    leader = build_leader_wave(INPUT_SEQUENCE)

    bridge = lowpass(leader, alpha=bridge_alpha)
    bridge = np.clip(bridge * (1.0 - signal_loss), 0.0, 1.0)

    actual_delay = BASE_DECODE_DELAY + delay_drift

    follower = shift_signal(bridge, actual_delay)
    follower = follower + rng.normal(0.0, noise, size=follower.shape)
    follower = np.clip(follower, 0.0, 1.0)

    decode_threshold = BASE_THRESHOLD + threshold_drift

    decoded_bits, received_means = decode_bits(
        follower,
        sequence_len=len(INPUT_SEQUENCE),
        bit_len=BIT_LEN,
        decode_delay=BASE_DECODE_DELAY,
        threshold=decode_threshold,
    )

    memory_values, stored_bits = write_memory(
        decoded_bits,
        noise=noise,
        memory_decay=memory_decay,
        rng=rng,
    )

    output_means, readout_bits = read_memory(
        memory_values,
        threshold=decode_threshold,
        noise=noise,
        rng=rng,
    )

    input_bits = [int(bit) for bit in INPUT_SEQUENCE]

    decode_accuracy = float(np.mean([
        a == b for a, b in zip(input_bits, decoded_bits)
    ]))

    memory_accuracy = float(np.mean([
        a == b for a, b in zip(input_bits, stored_bits)
    ]))

    readout_accuracy = float(np.mean([
        a == b for a, b in zip(input_bits, readout_bits)
    ]))

    full_loop_accuracy = float(np.mean([
        a == b for a, b in zip(input_bits, readout_bits)
    ]))

    result = (
        "ROBUSTNESS_LOCKED"
        if full_loop_accuracy == 1.0
        else "ROBUSTNESS_PARTIAL"
    )

    bit_rows = []

    for i, input_bit in enumerate(input_bits):
        bit_rows.append({
            "STATE": "D16_ROBUSTNESS_TEST",
            "CASE": name,
            "BIT_INDEX": i,
            "INPUT_BIT": input_bit,
            "RECEIVED_MEAN": round(received_means[i], 3),
            "DECODED_BIT": decoded_bits[i],
            "MEMORY_VALUE": round(memory_values[i], 3),
            "STORED_BIT": stored_bits[i],
            "OUTPUT_MEAN": round(output_means[i], 3),
            "READOUT_BIT": readout_bits[i],
            "MATCH": int(input_bit == readout_bits[i]),
        })

    summary = {
        "STATE": "D16_ROBUSTNESS_TEST",
        "CASE": name,
        "RESULT": result,
        "INPUT_SEQUENCE": INPUT_SEQUENCE,
        "DECODED_SEQUENCE": "".join(map(str, decoded_bits)),
        "STORED_SEQUENCE": "".join(map(str, stored_bits)),
        "READOUT_SEQUENCE": "".join(map(str, readout_bits)),
        "DECODE_ACCURACY": decode_accuracy,
        "MEMORY_ACCURACY": memory_accuracy,
        "READOUT_ACCURACY": readout_accuracy,
        "FULL_LOOP_ACCURACY": full_loop_accuracy,
        "NOISE": noise,
        "SIGNAL_LOSS": signal_loss,
        "DELAY_DRIFT": delay_drift,
        "THRESHOLD": round(decode_threshold, 3),
        "MEMORY_DECAY": memory_decay,
    }

    signals = {
        "leader": leader,
        "bridge": bridge,
        "follower": follower,
        "memory_values": memory_values,
        "decoded_bits": decoded_bits,
        "stored_bits": stored_bits,
        "readout_bits": readout_bits,
    }

    return summary, bit_rows, signals


def make_frame(signals, t, memory_values):
    size_y, size_x = 140, 210
    frame = np.zeros((size_y, size_x), dtype=float)

    yy, xx = np.mgrid[0:size_y, 0:size_x]

    leader_center = (70, 62)
    follower_center = (70, 118)
    radius = 23

    leader_value = signals["leader"][min(t, len(signals["leader"]) - 1)]
    follower_value = signals["follower"][min(t, len(signals["follower"]) - 1)]
    bridge_value = signals["bridge"][min(t, len(signals["bridge"]) - 1)]

    leader_mask = (
        (yy - leader_center[0]) ** 2
        + (xx - leader_center[1]) ** 2
        <= radius ** 2
    )

    follower_mask = (
        (yy - follower_center[0]) ** 2
        + (xx - follower_center[1]) ** 2
        <= radius ** 2
    )

    bridge_mask = (
        (yy >= 65)
        & (yy <= 75)
        & (xx >= 86)
        & (xx <= 94)
    )

    frame[leader_mask] = leader_value
    frame[follower_mask] = follower_value
    frame[bridge_mask] = bridge_value

    slot_x0 = 155
    slot_y0 = 18
    slot_size = 10
    gap = 6

    for i, value in enumerate(memory_values):
        row = i // 2
        col = i % 2

        y0 = slot_y0 + row * (slot_size + gap)
        x0 = slot_x0 + col * (slot_size + gap)

        frame[y0:y0 + slot_size, x0:x0 + slot_size] = value

    return frame


CASES = [
    {
        "name": "BASELINE",
        "noise": 0.00,
        "signal_loss": 0.00,
        "delay_drift": 0,
        "threshold_drift": 0.00,
        "memory_decay": 0.00,
    },
    {
        "name": "NOISE_LOW",
        "noise": 0.035,
        "signal_loss": 0.00,
        "delay_drift": 0,
        "threshold_drift": 0.00,
        "memory_decay": 0.02,
    },
    {
        "name": "NOISE_MEDIUM",
        "noise": 0.070,
        "signal_loss": 0.04,
        "delay_drift": 0,
        "threshold_drift": 0.00,
        "memory_decay": 0.05,
    },
    {
        "name": "DELAY_SHIFT",
        "noise": 0.035,
        "signal_loss": 0.03,
        "delay_drift": 16,
        "threshold_drift": 0.00,
        "memory_decay": 0.03,
    },
    {
        "name": "THRESHOLD_DRIFT",
        "noise": 0.035,
        "signal_loss": 0.04,
        "delay_drift": 0,
        "threshold_drift": 0.08,
        "memory_decay": 0.04,
    },
    {
        "name": "MEMORY_DECAY",
        "noise": 0.035,
        "signal_loss": 0.02,
        "delay_drift": 0,
        "threshold_drift": 0.00,
        "memory_decay": 0.22,
    },
    {
        "name": "COMBINED_STRESS",
        "noise": 0.070,
        "signal_loss": 0.08,
        "delay_drift": 14,
        "threshold_drift": 0.04,
        "memory_decay": 0.14,
    },
    {
        "name": "BREAK_POINT",
        "noise": 0.180,
        "signal_loss": 0.28,
        "delay_drift": 38,
        "threshold_drift": 0.10,
        "memory_decay": 0.38,
    },
]


summaries = []
all_bit_rows = []
saved_signals = {}

for index, case in enumerate(CASES):
    summary, bit_rows, signals = run_case(
        seed=SEED + index,
        **case,
    )

    summaries.append(summary)
    all_bit_rows.extend(bit_rows)
    saved_signals[case["name"]] = signals


df_summary = pd.DataFrame(summaries)
df_bits = pd.DataFrame(all_bit_rows)

normal_cases = df_summary[df_summary["CASE"] != "BREAK_POINT"]

normal_worst_accuracy = float(normal_cases["FULL_LOOP_ACCURACY"].min())
overall_worst_accuracy = float(df_summary["FULL_LOOP_ACCURACY"].min())
mean_accuracy = float(df_summary["FULL_LOOP_ACCURACY"].mean())

if normal_worst_accuracy == 1.0 and overall_worst_accuracy < 1.0:
    final_state = "ROBUSTNESS_RANGE_LOCKED"
elif overall_worst_accuracy == 1.0:
    final_state = "ROBUSTNESS_FULL_LOCKED"
elif normal_worst_accuracy >= 0.875:
    final_state = "ROBUSTNESS_PARTIAL_LOCK"
else:
    final_state = "ROBUSTNESS_FAIL"


print("STATE:", final_state)
print("input sequence:", INPUT_SEQUENCE)
print("cases tested:", len(CASES))
print("normal worst accuracy:", round(normal_worst_accuracy, 3))
print("overall worst accuracy:", round(overall_worst_accuracy, 3))
print("mean full-loop accuracy:", round(mean_accuracy, 3))
print()

print("Summary table:")
display(df_summary)

plt.figure(figsize=(9, 3))
plt.plot(df_summary["CASE"], df_summary["DECODE_ACCURACY"], marker="o", label="decode")
plt.plot(df_summary["CASE"], df_summary["MEMORY_ACCURACY"], marker="o", label="memory")
plt.plot(df_summary["CASE"], df_summary["READOUT_ACCURACY"], marker="o", label="readout")
plt.plot(df_summary["CASE"], df_summary["FULL_LOOP_ACCURACY"], marker="o", label="full loop")

plt.title("D16_ROBUSTNESS_TEST — accuracy under stress")
plt.ylabel("accuracy")
plt.ylim(-0.05, 1.05)
plt.xticks(rotation=35, ha="right")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()


selected_case = "COMBINED_STRESS"
signals = saved_signals[selected_case]

plt.figure(figsize=(9, 3))
plt.plot(signals["leader"], label="input leader wave")
plt.plot(signals["bridge"], label="bridge channel")
plt.plot(signals["follower"], label="stressed follower wave")

plt.title("D16_ROBUSTNESS_TEST — stressed wave transport")
plt.xlabel("time step")
plt.ylabel("activation")
plt.legend()
plt.tight_layout()
plt.show()


snapshot_times = [0, 80, 160, 260, 360, 480, 600, 720, 840, 959]

for t in snapshot_times:
    frame = make_frame(signals, t, signals["memory_values"])

    plt.figure(figsize=(5, 4))
    plt.imshow(frame, cmap="inferno", vmin=0, vmax=1)
    plt.title(f"D16 snapshot t={t} — {selected_case}")
    plt.axis("off")
    plt.show()


print("Stress-case bit table:")
display(
    df_bits[df_bits["CASE"] == selected_case]
    .reset_index(drop=True)
)

print("Break-point bit table:")
display(
    df_bits[df_bits["CASE"] == "BREAK_POINT"]
    .reset_index(drop=True)
)

print("Final summary:")
display(df_summary)
```
