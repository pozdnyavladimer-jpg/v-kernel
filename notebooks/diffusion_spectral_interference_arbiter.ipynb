```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D34_SPECTRAL_INTERFERENCE_ARBITER
# Wave-interference arbitration for spectral / photon-like memory packets
#
# Goal:
# multiple spectral packets arrive at one receiver field
# -> packets superpose as complex amplitudes
# -> constructive interference strengthens a valid key
# -> destructive / weak / ambiguous / wrong-phase packets are rejected
# -> if several keys collide, the strongest valid resonant key wins
# -> receiver writes only when key + phase + margin are safe
# ============================================================

np.random.seed(34)

STATE = "D34_SPECTRAL_INTERFERENCE_ARBITER"

directions = 6
octaves = 7
spectral_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12
phase_tolerance = 0.18
absorption_threshold = 0.62

memory_slots = 8

receiver_target_sequence = [
    "D4_O5",
    "D1_O2",
    "D3_O6",
    "D5_O4",
    "D0_O1",
    "D2_O3",
    "D4_O5",
    "D1_O2",
]

receiver_initial_sequence = ["EMPTY"] * memory_slots


# ------------------------------------------------------------
# Key / spectral utilities
# ------------------------------------------------------------

def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE", "NOISE", "MIXED"]:
        return None, None
    if not isinstance(key, str) or "_" not in key:
        return None, None

    d_part, o_part = key.split("_")
    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))

    return direction, octave


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    return octave * directions + direction


def expected_phase_for_key(key):
    d, o = parse_key(key)
    if d is None:
        return None
    return round(d / directions, 3)


def circular_phase_distance(a, b):
    if a is None or b is None:
        return None

    diff = abs((a - b + 0.5) % 1.0 - 0.5)
    return float(diff)


def wave_token(key, phase=None, strength=None):
    d, o = parse_key(key)

    if d is None:
        return "NONE"

    if phase is None:
        phase = expected_phase_for_key(key)

    frequency = o + 1

    if strength is None:
        return f"INTERFERENCE_PHOTON[D{d}|O{o}|F{frequency}|P{round(phase, 3)}]"

    return f"INTERFERENCE_PHOTON[D{d}|O{o}|F{frequency}|P{round(phase, 3)}|A{round(strength, 3)}]"


def key_color(key, strength=1.0):
    if key in [None, "EMPTY", "NONE", "NOISE"]:
        return "black"

    if key == "WHITE":
        return "white"

    d, o = parse_key(key)

    if d is None:
        return "gray"

    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)

    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


# ------------------------------------------------------------
# Complex spectral packet model
# ------------------------------------------------------------

def empty_complex_packet(noise=0.015):
    amp = np.abs(np.random.normal(0.0, noise, size=(octaves, directions)))
    phase = np.random.uniform(0.0, 1.0, size=(octaves, directions))

    return amp * np.exp(1j * 2 * np.pi * phase)


def add_component(packet, key, strength, phase_cycles=None):
    if key in [None, "EMPTY", "NONE"]:
        return packet

    if key == "WHITE":
        broad_amp = np.clip(
            np.random.normal(0.12, 0.025, size=(octaves, directions)),
            0.0,
            1.0
        )
        broad_phase = np.random.uniform(0.0, 1.0, size=(octaves, directions))
        packet += broad_amp * np.exp(1j * 2 * np.pi * broad_phase)

        return packet

    d, o = parse_key(key)

    if d is None:
        return packet

    if phase_cycles is None:
        phase_cycles = expected_phase_for_key(key)

    packet[o, d] += strength * np.exp(1j * 2 * np.pi * phase_cycles)

    return packet


def spectral_packet(components, noise=0.015):
    packet = empty_complex_packet(noise=noise)

    for comp in components:
        packet = add_component(
            packet,
            comp.get("key"),
            comp.get("strength", 0.0),
            comp.get("phase", None)
        )

    return packet


def combine_packets(packets):
    total = np.zeros((octaves, directions), dtype=np.complex128)

    for packet in packets:
        total += packet

    return total


def decode_interference(packet):
    magnitude = np.abs(packet)
    flat = magnitude.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    best_complex = packet[best_octave, best_direction]
    best_phase = (np.angle(best_complex) / (2 * np.pi)) % 1.0

    margin = best_score - second_score
    amplitude_ok = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "best_phase": float(best_phase),
        "amplitude_ok": bool(amplitude_ok),
    }


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.3

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=14,
        linewidth=2.0,
        color=color,
        alpha=alpha
    )

    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.16
    gap = 0.03

    ax.text(
        x0,
        y0 + 1.55,
        title,
        fontsize=7,
        ha="left",
        va="bottom",
        color="black"
    )

    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE"] else (None, None)
    magnitude = np.clip(np.abs(packet), 0.0, 1.0)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = magnitude[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)
            edge = "lime" if is_selected else "white"
            lw = 1.4 if is_selected else 0.25

            ax.add_patch(Rectangle(
                (x, y),
                cell,
                cell,
                facecolor=color,
                edgecolor=edge,
                linewidth=lw
            ))


def draw_memory_slots(ax, x0, y0, seq, title, active_slot=None):
    cell = 0.20
    gap = 0.055

    ax.text(
        x0,
        y0 + 2.05,
        title,
        fontsize=7,
        ha="left",
        va="bottom",
        color="black"
    )

    for i, key in enumerate(seq):
        x = x0
        y = y0 + (len(seq) - 1 - i) * (cell + gap)

        color = key_color(key)
        edge = "lime" if i == active_slot else "black"
        lw = 1.6 if i == active_slot else 0.5

        ax.add_patch(Rectangle(
            (x, y),
            cell,
            cell,
            facecolor=color,
            edgecolor=edge,
            linewidth=lw
        ))

        ax.text(
            x + 0.32,
            y + cell / 2,
            str(i),
            fontsize=6,
            ha="left",
            va="center",
            color="black"
        )


def draw_phase_meter(ax, x0, y0, best_phase, expected_phase, phase_ok):
    ax.text(x0, y0 + 0.35, "phase gate", fontsize=7, ha="left", color="black")

    ax.add_patch(Rectangle(
        (x0, y0),
        1.1,
        0.12,
        facecolor="lightgray",
        edgecolor="black",
        linewidth=0.5
    ))

    if expected_phase is not None:
        ex = x0 + expected_phase * 1.1
        ax.add_patch(Rectangle(
            (ex - 0.015, y0 - 0.05),
            0.03,
            0.22,
            facecolor="lime",
            edgecolor="black",
            linewidth=0.3
        ))

    bx = x0 + best_phase * 1.1
    color = "lime" if phase_ok else "red"

    ax.add_patch(Circle(
        (bx, y0 + 0.06),
        0.06,
        facecolor=color,
        edgecolor="black",
        linewidth=0.4
    ))


def draw_snapshot(ax, case_row, packet, receiver_memory):
    slot = int(case_row["MEMORY_SLOT"])

    key_a = case_row["SOURCE_A_KEY"]
    key_b = case_row["SOURCE_B_KEY"]

    expected_key = case_row["EXPECTED_RECEIVER_KEY"]
    decoded_key = case_row["DECODED_KEY"]
    final_key = receiver_memory[slot]

    absorbed = bool(case_row["ABSORBED"])
    phase_ok = bool(case_row["PHASE_OK"])

    ax.set_facecolor("white")
    ax.set_xlim(0, 12.5)
    ax.set_ylim(0.55, 5.2)
    ax.set_aspect("equal")
    ax.axis("off")

    ax.add_patch(Circle(
        (0.8, 3.65),
        0.34,
        facecolor=key_color(key_a),
        edgecolor="black",
        linewidth=1.0
    ))

    ax.text(
        0.8,
        3.08,
        f"SRC A\n{key_a}",
        fontsize=6.5,
        ha="center",
        color="black"
    )

    ax.add_patch(Circle(
        (0.8, 2.6),
        0.34,
        facecolor=key_color(key_b),
        edgecolor="black",
        linewidth=1.0
    ))

    ax.text(
        0.8,
        2.03,
        f"SRC B\n{key_b}",
        fontsize=6.5,
        ha="center",
        color="black"
    )

    ax.add_patch(Rectangle(
        (2.0, 2.9),
        0.45,
        0.45,
        facecolor="limegreen",
        edgecolor="black",
        linewidth=1.0
    ))

    ax.text(
        2.23,
        2.55,
        "FIELD\nSUM",
        fontsize=6.5,
        ha="center",
        color="black"
    )

    ax.add_patch(Circle(
        (3.75, 3.12),
        0.42,
        facecolor=key_color(decoded_key),
        edgecolor="lime" if case_row["AMPLITUDE_OK"] else "red",
        linewidth=2.0
    ))

    ax.text(
        3.75,
        2.42,
        f"BEST\n{decoded_key}",
        fontsize=6.5,
        ha="center",
        color="black"
    )

    ax.add_patch(Rectangle(
        (5.0, 2.9),
        0.45,
        0.45,
        facecolor="limegreen" if phase_ok else "red",
        edgecolor="black",
        linewidth=1.0
    ))

    ax.text(
        5.22,
        2.55,
        "PHASE\nGATE",
        fontsize=6.5,
        ha="center",
        color="black"
    )

    ax.add_patch(Circle(
        (6.45, 3.12),
        0.42,
        facecolor=key_color(final_key),
        edgecolor="lime" if absorbed else "red",
        linewidth=2.0
    ))

    ax.text(
        6.45,
        2.42,
        f"MEM\nslot {slot}",
        fontsize=6.5,
        ha="center",
        color="black"
    )

    rejected = not absorbed

    draw_arrow(ax, (1.15, 3.65), (2.0, 3.2), active=True, rejected=False)
    draw_arrow(ax, (1.15, 2.6), (2.0, 3.05), active=True, rejected=False)
    draw_arrow(ax, (2.45, 3.12), (3.32, 3.12), active=True, rejected=rejected)
    draw_arrow(ax, (4.18, 3.12), (5.0, 3.12), active=True, rejected=not phase_ok)
    draw_arrow(ax, (5.45, 3.12), (6.03, 3.12), active=True, rejected=rejected)

    draw_phase_meter(
        ax,
        2.95,
        1.05,
        case_row["BEST_PHASE"],
        case_row["EXPECTED_PHASE"],
        phase_ok
    )

    draw_packet_grid(
        ax,
        7.55,
        2.15,
        packet,
        "42-channel interference field",
        selected_key=decoded_key
    )

    draw_memory_slots(
        ax,
        10.55,
        1.35,
        receiver_memory,
        "receiver memory",
        active_slot=slot
    )

    ax.text(
        0.45,
        0.72,
        f"case={case_row['CASE']} slot={slot} expected={expected_key} decoded={decoded_key} "
        f"score={case_row['BEST_SCORE']} margin={case_row['MARGIN']} "
        f"phase_err={case_row['PHASE_ERROR']} action={case_row['ACTION']}\n"
        f"{case_row['OUTPUT_TOKEN']}",
        fontsize=6.5,
        ha="left",
        color="black"
    )

    ax.set_title(
        f"D34 snapshot t={case_row['T']} — {case_row['CASE']} / {case_row['ACTION']}",
        fontsize=10,
        color="black"
    )


# ------------------------------------------------------------
# Interference test cases
# ------------------------------------------------------------

def p(key, strength, phase=None):
    if phase is None:
        phase = expected_phase_for_key(key)

    return {
        "key": key,
        "strength": strength,
        "phase": phase
    }


cases = [
    {
        "case": "CONSTRUCTIVE_SAME_PHASE",
        "slot": 0,
        "expected": "D4_O5",
        "components_a": [p("D4_O5", 0.54)],
        "components_b": [p("D4_O5", 0.52)],
        "expected_absorb": True,
    },
    {
        "case": "DESTRUCTIVE_OPPOSITE_PHASE",
        "slot": 1,
        "expected": "D1_O2",
        "components_a": [p("D1_O2", 0.58, expected_phase_for_key("D1_O2"))],
        "components_b": [p("D1_O2", 0.56, (expected_phase_for_key("D1_O2") + 0.5) % 1.0)],
        "expected_absorb": False,
    },
    {
        "case": "STRONGER_RESONANCE_WINS",
        "slot": 2,
        "expected": "D3_O6",
        "components_a": [p("D3_O6", 0.82)],
        "components_b": [p("D4_O5", 0.58)],
        "expected_absorb": True,
    },
    {
        "case": "SAME_SLOT_COLLISION_ARBITRATED",
        "slot": 3,
        "expected": "D5_O4",
        "components_a": [p("D0_O1", 0.73)],
        "components_b": [p("D5_O4", 0.88)],
        "expected_absorb": True,
    },
    {
        "case": "WRONG_PHASE_REJECT",
        "slot": 4,
        "expected": "D0_O1",
        "components_a": [p("D0_O1", 0.86, 0.33)],
        "components_b": [],
        "expected_absorb": False,
    },
    {
        "case": "AMBIGUOUS_INTERFERENCE_HOLD",
        "slot": 5,
        "expected": "D2_O3",
        "components_a": [p("D2_O3", 0.84)],
        "components_b": [p("D4_O5", 0.78)],
        "expected_absorb": False,
    },
    {
        "case": "WHITE_BACKGROUND_WITH_TARGET",
        "slot": 6,
        "expected": "D4_O5",
        "components_a": [{"key": "WHITE", "strength": 0.18, "phase": None}],
        "components_b": [p("D4_O5", 0.83)],
        "expected_absorb": True,
    },
    {
        "case": "ORTHOGONAL_NOISE_REJECT",
        "slot": 7,
        "expected": "D1_O2",
        "components_a": [p("D5_O6", 0.55)],
        "components_b": [p("D0_O0", 0.52)],
        "expected_absorb": False,
    },
]

receiver_memory = receiver_initial_sequence.copy()

decision_rows = []
component_rows = []
absorption_rows = []
wave_rows = []
snapshots = []

absorbed_count = 0
rejected_count = 0
accepted_count = 0

for i, item in enumerate(cases):
    packet_a = spectral_packet(item["components_a"])
    packet_b = spectral_packet(item["components_b"])

    field_packet = combine_packets([packet_a, packet_b])
    decoded = decode_interference(field_packet)

    decoded_key = decoded["best_key"]
    expected_key = item["expected"]

    expected_phase = expected_phase_for_key(expected_key)
    phase_error = circular_phase_distance(decoded["best_phase"], expected_phase)

    amplitude_ok = bool(decoded["amplitude_ok"])
    phase_ok = bool(phase_error is not None and phase_error <= phase_tolerance)
    resonance_match = decoded_key == expected_key

    absorbed = bool(amplitude_ok and phase_ok and resonance_match)

    before_key = receiver_memory[item["slot"]]

    if absorbed:
        receiver_memory[item["slot"]] = decoded_key
        absorbed_count += 1

    if amplitude_ok:
        accepted_count += 1
    else:
        rejected_count += 1

    action = "ABSORB_AND_WRITE" if absorbed else (
        "REJECT_WEAK_OR_AMBIGUOUS" if not amplitude_ok else
        "REJECT_WRONG_PHASE" if not phase_ok else
        "REJECT_FILTER_MISMATCH"
    )

    source_a_key = item["components_a"][0]["key"] if len(item["components_a"]) else "NONE"
    source_b_key = item["components_b"][0]["key"] if len(item["components_b"]) else "NONE"

    output_token = wave_token(
        decoded_key,
        decoded["best_phase"],
        decoded["best_score"]
    ) if amplitude_ok else "NONE"

    row = {
        "STATE": STATE,
        "T": i * 120,
        "CASE": item["case"],
        "MEMORY_SLOT": item["slot"],
        "SOURCE_A_KEY": source_a_key,
        "SOURCE_B_KEY": source_b_key,
        "EXPECTED_RECEIVER_KEY": expected_key,
        "DECODED_KEY": decoded_key,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "BEST_PHASE": round(decoded["best_phase"], 3),
        "EXPECTED_PHASE": expected_phase,
        "PHASE_ERROR": round(phase_error, 3) if phase_error is not None else None,
        "AMPLITUDE_OK": int(amplitude_ok),
        "PHASE_OK": int(phase_ok),
        "RESONANCE_MATCH": int(resonance_match),
        "ABSORBED": int(absorbed),
        "EXPECTED_ABSORBED": int(item["expected_absorb"]),
        "ACTION": action,
        "OUTPUT_TOKEN": output_token,
    }

    decision_rows.append(row)

    for src_name, comps in [("A", item["components_a"]), ("B", item["components_b"])]:
        if len(comps) == 0:
            component_rows.append({
                "STATE": STATE,
                "CASE": item["case"],
                "SOURCE": src_name,
                "COMPONENT_KEY": "NONE",
                "STRENGTH": 0.0,
                "PHASE": None,
            })
        else:
            for comp in comps:
                component_rows.append({
                    "STATE": STATE,
                    "CASE": item["case"],
                    "SOURCE": src_name,
                    "COMPONENT_KEY": comp["key"],
                    "STRENGTH": comp["strength"],
                    "PHASE": round(comp["phase"], 3) if comp["phase"] is not None else "broad",
                })

    absorption_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "MEMORY_SLOT": item["slot"],
        "BEFORE_KEY": before_key,
        "EXPECTED_KEY": expected_key,
        "DECODED_KEY": decoded_key,
        "AFTER_KEY": receiver_memory[item["slot"]],
        "AMPLITUDE_OK": int(amplitude_ok),
        "PHASE_OK": int(phase_ok),
        "RESONANCE_MATCH": int(resonance_match),
        "ABSORBED": int(absorbed),
        "MATCH_EXPECTED_ACTION": int(absorbed == item["expected_absorb"]),
    })

    wave_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "DECODED_KEY": decoded_key,
        "BEST_SCORE": round(decoded["best_score"], 3),
        "BEST_PHASE": round(decoded["best_phase"], 3),
        "EXPECTED_PHASE": expected_phase,
        "OUTPUT_TOKEN": output_token,
        "ABSORBED": int(absorbed),
    })

    snapshots.append((row, field_packet, receiver_memory.copy()))


df_decision = pd.DataFrame(decision_rows)
df_components = pd.DataFrame(component_rows)
df_absorption = pd.DataFrame(absorption_rows)
df_wave = pd.DataFrame(wave_rows)

expected_final = receiver_initial_sequence.copy()

for item in cases:
    if item["expected_absorb"]:
        expected_final[item["slot"]] = item["expected"]

interference_accuracy = float(np.mean(np.array(receiver_memory) == np.array(expected_final)))
arbiter_accuracy = float(np.mean(df_absorption["MATCH_EXPECTED_ACTION"].astype(int).values))

result = (
    "SPECTRAL_INTERFERENCE_ARBITER_LOCKED"
    if interference_accuracy == 1.0 and arbiter_accuracy == 1.0
    else "SPECTRAL_INTERFERENCE_ARBITER_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": memory_slots,
    "RECEIVER_TARGET_SEQUENCE": " ".join(receiver_target_sequence),
    "RECEIVER_FINAL_SEQUENCE": " ".join(receiver_memory),
    "EXPECTED_FINAL_SEQUENCE": " ".join(expected_final),
    "CASES_TESTED": len(cases),
    "AMPLITUDE_ACCEPTED_FIELDS": accepted_count,
    "AMPLITUDE_REJECTED_FIELDS": rejected_count,
    "ABSORBED_FIELDS": absorbed_count,
    "INTERFERENCE_ACCURACY": round(interference_accuracy, 3),
    "ARBITER_ACCURACY": round(arbiter_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "ABSORPTION_THRESHOLD": absorption_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "PHASE_TOLERANCE": phase_tolerance,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8.5, 3.2 * len(snapshots)))

if len(snapshots) == 1:
    axes = [axes]

for ax, (row, packet, memory_state) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, memory_state)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"memory slots:                 {memory_slots}")
print(f"receiver target sequence:     {' '.join(receiver_target_sequence)}")
print(f"receiver final sequence:      {' '.join(receiver_memory)}")
print(f"expected final sequence:      {' '.join(expected_final)}")
print(f"cases tested:                 {len(cases)}")
print(f"amplitude accepted fields:    {accepted_count}")
print(f"amplitude rejected fields:    {rejected_count}")
print(f"absorbed fields:              {absorbed_count}")
print(f"interference accuracy:        {round(interference_accuracy, 3)}")
print(f"arbiter accuracy:             {round(arbiter_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"phase tolerance:              {phase_tolerance}")

print("\nInterference decision table:")
display(df_decision)

print("\nIncoming component table:")
display(df_components)

print("\nReceiver absorption table:")
display(df_absorption)

print("\nInterference photon table:")
display(df_wave)

print("\nSummary:")
display(df_summary)
```
