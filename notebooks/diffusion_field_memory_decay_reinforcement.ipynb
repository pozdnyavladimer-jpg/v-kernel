```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D41_FIELD_MEMORY_DECAY_AND_REINFORCEMENT
# Coherent field memory:
# repeated intent strengthens memory,
# unused / weak / incoherent signals decay and eventually forget.
#
# Continuation after:
# D36 coherence bus
# D37 orbital intent
# D38 honeycomb lattice
# D39 honeycomb memory wave
# D40 macro resonance loop
#
# Goal:
# - memory cell is not a static bit
# - memory strength evolves with time
# - coherent repeated spectral+orbital+phase packets reinforce it
# - weak / wrong / ambiguous / unlocked packets are rejected
# - unused memory naturally decays
# - forgotten cells become EMPTY
# ============================================================

np.random.seed(41)

STATE = "D41_FIELD_MEMORY_DECAY_AND_REINFORCEMENT"

directions = 6
octaves = 7
spectral_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12
phase_tolerance = 0.18
jitter_tolerance = 0.16
resync_window = 0.36

passive_decay_rate = 0.025
forget_threshold = 0.46
stable_threshold = 0.60
reinforcement_gain = 0.22
coherence_bonus = 0.08
new_write_base = 0.58
saturation = 1.00

receiver_filter_sequence = [
    "D4_O5",
    "D1_O2",
    "D3_O6",
    "D5_O4",
    "D0_O1",
    "D2_O3",
    "D4_O5",
    "D1_O2",
]

receiver_intent_sequence = [
    "FLOWER",
    "RING",
    "HEX",
    "SPIRAL",
    "SEED",
    "BUTTERFLY",
    "FLOWER",
    "RING",
]

memory_keys = [
    "D4_O5",
    "D1_O2",
    "D3_O6",
    "EMPTY",
    "D0_O1",
    "D2_O3",
    "D4_O5",
    "EMPTY",
]

memory_intents = [
    "FLOWER",
    "RING",
    "HEX",
    "NONE",
    "SEED",
    "BUTTERFLY",
    "FLOWER",
    "NONE",
]

memory_strengths = np.array(
    [0.74, 0.58, 0.64, 0.00, 0.52, 0.50, 0.70, 0.00],
    dtype=float
)


def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE", "IDLE"]:
        return None, None

    parts = str(key).split("_")

    if len(parts) != 2:
        return None, None

    d_part, o_part = parts
    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))

    return direction, octave


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def phase_for_key(key):
    d, _ = parse_key(key)

    if d is None:
        return None

    return round(d / directions, 3)


def circular_phase_error(a, b):
    if a is None or b is None:
        return 1.0

    diff = abs(a - b) % 1.0
    return min(diff, 1.0 - diff)


def key_color(key, strength=1.0):
    if key in [None, "EMPTY", "NONE", "IDLE"]:
        return "black"

    if key == "WHITE":
        return "white"

    d, o = parse_key(key)

    if d is None:
        return "gray"

    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)

    base = plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)

    strength = float(np.clip(strength, 0.0, 1.0))

    return (
        base[0] * strength,
        base[1] * strength,
        base[2] * strength,
        1.0
    )


def intent_marker(intent):
    mapping = {
        "FLOWER": "FLR",
        "RING": "RNG",
        "BUTTERFLY": "BTF",
        "SPIRAL": "SPR",
        "HEX": "HEX",
        "WAVE": "WAV",
        "SEED": "SED",
        "NONE": "---",
    }

    return mapping.get(intent, str(intent)[:3].upper())


def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.16 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE", "IDLE"]:
        d, o = parse_key(target_key)

        if d is not None:
            packet[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE", "IDLE"]:
        d2, o2 = parse_key(ambiguous_key)

        if d2 is not None:
            packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score

    accepted = (
        best_score >= resonance_threshold
        and margin >= ambiguity_margin
    )

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "accepted": accepted,
    }


def apply_passive_decay(keys, intents, strengths):
    decayed_slots = []
    forgotten_slots = []

    for i in range(len(strengths)):
        if keys[i] not in ["EMPTY", "NONE"]:
            strengths[i] = strengths[i] * (1.0 - passive_decay_rate)
            decayed_slots.append(i)

            if strengths[i] < forget_threshold:
                keys[i] = "EMPTY"
                intents[i] = "NONE"
                strengths[i] = 0.0
                forgotten_slots.append(i)

    return decayed_slots, forgotten_slots


def reinforce_strength(old_strength, best_score, is_new):
    if is_new:
        new_strength = new_write_base + reinforcement_gain * best_score
    else:
        new_strength = old_strength + reinforcement_gain * best_score + coherence_bonus

    return float(np.clip(new_strength, 0.0, saturation))


def packet_token(key, intent, amp, phase, jitter, action):
    if key in [None, "EMPTY", "NONE", "IDLE"]:
        return "NONE"

    d, o = parse_key(key)

    if d is None:
        return "NONE"

    return (
        f"MEMORY_PHOTON[D{d}|O{o}|F{o+1}|P{round(phase, 3)}|"
        f"A{round(amp, 3)}|J{round(jitter, 3)}|I={intent}|{action}]"
    )


def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.25

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=12,
        linewidth=1.7,
        color=color,
        alpha=alpha
    )

    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.13
    gap = 0.025

    ax.text(
        x0,
        y0 + 1.25,
        title,
        fontsize=6,
        ha="left",
        va="bottom",
        color="black"
    )

    selected = (
        parse_key(selected_key)
        if selected_key not in [None, "EMPTY", "NONE"]
        else (None, None)
    )

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = packet[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)

            edge = "lime" if is_selected else "white"
            lw = 1.1 if is_selected else 0.25

            ax.add_patch(
                Rectangle(
                    (x, y),
                    cell,
                    cell,
                    facecolor=color,
                    edgecolor=edge,
                    linewidth=lw
                )
            )


def draw_memory_bar(
    ax,
    x0,
    y0,
    keys,
    intents,
    strengths,
    active_slot=None,
    title="field memory"
):
    cell = 0.17
    gap = 0.055

    ax.text(
        x0,
        y0 + 2.0,
        title,
        fontsize=6,
        ha="left",
        va="bottom",
        color="black"
    )

    for i, key in enumerate(keys):
        x = x0
        y = y0 + (len(keys) - 1 - i) * (cell + gap)

        color = key_color(key, strengths[i])

        edge = "lime" if i == active_slot else "black"
        lw = 1.5 if i == active_slot else 0.5

        ax.add_patch(
            Rectangle(
                (x, y),
                cell,
                cell,
                facecolor=color,
                edgecolor=edge,
                linewidth=lw
            )
        )

        ax.text(
            x + 0.24,
            y + cell / 2,
            f"{i}:{intent_marker(intents[i])}",
            fontsize=5.5,
            ha="left",
            va="center",
            color="black"
        )


def draw_strength_gauge(ax, x0, y0, strength, title):
    ax.text(
        x0,
        y0 + 0.48,
        title,
        fontsize=6,
        ha="center",
        color="black"
    )

    ax.add_patch(
        Rectangle(
            (x0 - 0.45, y0),
            0.9,
            0.12,
            facecolor="lightgray",
            edgecolor="black",
            linewidth=0.5
        )
    )

    ax.add_patch(
        Rectangle(
            (x0 - 0.45, y0),
            0.9 * strength,
            0.12,
            facecolor="limegreen",
            edgecolor="none"
        )
    )

    ax.text(
        x0,
        y0 - 0.16,
        f"{strength:.2f}",
        fontsize=6,
        ha="center",
        color="black"
    )


def draw_snapshot(
    ax,
    row,
    packet,
    memory_keys_snapshot,
    memory_intents_snapshot,
    memory_strengths_snapshot
):
    slot = row["MEMORY_SLOT"]

    source_key = row["SOURCE_KEY"]
    source_intent = row["SOURCE_INTENT"]

    decoded_key = row["DECODED_KEY"]

    expected_key = row["FILTER_KEY"]
    expected_intent = row["FILTER_INTENT"]

    before_key = row["BEFORE_KEY"]
    after_key = row["AFTER_KEY"]

    action = row["ACTION"]

    absorbed = bool(row["APPLIED"])
    rejected = not absorbed and row["CASE_TYPE"] != "IDLE"

    ax.set_facecolor("white")
    ax.set_xlim(0, 13.3)
    ax.set_ylim(0.5, 4.9)
    ax.set_aspect("equal")
    ax.axis("off")

    source_strength = (
        row["BEST_SCORE"]
        if row["BEST_SCORE"] != "none"
        else 0.2
    )

    source_color = key_color(source_key, source_strength)

    decoded_strength = (
        row["BEST_SCORE"]
        if row["BEST_SCORE"] != "none"
        else 0.2
    )

    decoded_color = key_color(decoded_key, decoded_strength)
    memory_color = key_color(after_key, row["AFTER_STRENGTH"])

    ax.add_patch(
        Circle(
            (0.8, 3.15),
            0.34,
            facecolor=source_color,
            edgecolor="black",
            linewidth=1.0
        )
    )

    ax.text(
        0.8,
        2.62,
        f"SRC\n{source_key}\n{intent_marker(source_intent)}",
        fontsize=6,
        ha="center",
        color="black"
    )

    ax.add_patch(
        Circle(
            (1.85, 3.15),
            0.25,
            facecolor=source_color,
            edgecolor="black",
            linewidth=0.8
        )
    )

    ax.text(
        1.85,
        2.62,
        "PHOTON",
        fontsize=6,
        ha="center",
        color="black"
    )

    ax.add_patch(
        Rectangle(
            (2.55, 2.95),
            0.35,
            0.35,
            facecolor="limegreen",
            edgecolor="black",
            linewidth=0.8
        )
    )

    ax.text(
        2.73,
        2.62,
        "FIELD",
        fontsize=6,
        ha="center",
        color="black"
    )

    ax.add_patch(
        Rectangle(
            (3.35, 2.95),
            0.35,
            0.35,
            facecolor="gold",
            edgecolor="black",
            linewidth=0.8
        )
    )

    ax.text(
        3.53,
        2.62,
        "CLOCK",
        fontsize=6,
        ha="center",
        color="black"
    )

    edge_decode = "lime" if row["AMPLITUDE_OK"] == 1 else "red"

    ax.add_patch(
        Circle(
            (4.35, 3.15),
            0.34,
            facecolor=decoded_color,
            edgecolor=edge_decode,
            linewidth=1.8
        )
    )

    ax.text(
        4.35,
        2.52,
        f"DECODE\n{decoded_key}\n{intent_marker(source_intent)}",
        fontsize=6,
        ha="center",
        color="black"
    )

    edge_mem = "lime" if absorbed else "red" if rejected else "black"

    ax.add_patch(
        Circle(
            (5.55, 3.15),
            0.34,
            facecolor=memory_color,
            edgecolor=edge_mem,
            linewidth=1.8
        )
    )

    ax.text(
        5.55,
        2.52,
        f"MEM\nslot {slot}\n{after_key}",
        fontsize=6,
        ha="center",
        color="black"
    )

    draw_arrow(ax, (1.14, 3.15), (1.6, 3.15), active=row["CASE_TYPE"] != "IDLE")
    draw_arrow(ax, (2.1, 3.15), (2.55, 3.15), active=row["CASE_TYPE"] != "IDLE")
    draw_arrow(ax, (2.9, 3.15), (3.35, 3.15), active=row["CASE_TYPE"] != "IDLE")
    draw_arrow(ax, (3.7, 3.15), (4.0, 3.15), active=row["CASE_TYPE"] != "IDLE", rejected=rejected)
    draw_arrow(ax, (4.7, 3.15), (5.2, 3.15), active=row["CASE_TYPE"] != "IDLE", rejected=rejected)

    draw_strength_gauge(
        ax,
        6.65,
        3.05,
        row["AFTER_STRENGTH"],
        "strength"
    )

    draw_packet_grid(
        ax,
        7.55,
        2.35,
        packet,
        "42-channel packet",
        selected_key=decoded_key
    )

    draw_memory_bar(
        ax,
        10.0,
        1.4,
        memory_keys_snapshot,
        memory_intents_snapshot,
        memory_strengths_snapshot,
        active_slot=slot,
        title="memory state"
    )

    ax.text(
        0.45,
        0.75,
        f"case={row['CASE']} filter={expected_key}/{intent_marker(expected_intent)} "
        f"before={before_key}:{row['BEFORE_STRENGTH']} after={after_key}:{row['AFTER_STRENGTH']} "
        f"decayed={row['DECAYED_SLOTS']} forgotten={row['FORGOTTEN_SLOTS']}\n"
        f"amp_ok={row['AMPLITUDE_OK']} phase_ok={row['PHASE_OK']} jitter_ok={row['JITTER_OK']} "
        f"intent_ok={row['INTENT_MATCH']} action={action}",
        fontsize=6,
        ha="left",
        color="black"
    )

    ax.set_title(
        f"D41 snapshot t={row['T']} — {row['CASE']} / {action}",
        fontsize=9,
        color="black"
    )


events = [
    {
        "case": "COHERENT_REPEAT_FLOWER_0",
        "case_type": "PACKET",
        "slot": 0,
        "source_key": "D4_O5",
        "source_intent": "FLOWER",
        "strength": 0.91,
        "phase_offset": 0.02,
        "jitter": 0.02,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "IDLE_DECAY_RING_1",
        "case_type": "IDLE",
        "slot": 1,
        "source_key": "IDLE",
        "source_intent": "NONE",
        "strength": 0.0,
        "phase_offset": 0.0,
        "jitter": 0.0,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "COHERENT_REFRESH_RING_1",
        "case_type": "PACKET",
        "slot": 1,
        "source_key": "D1_O2",
        "source_intent": "RING",
        "strength": 0.90,
        "phase_offset": 0.03,
        "jitter": 0.03,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "WRONG_INTENT_REJECT_2",
        "case_type": "PACKET",
        "slot": 2,
        "source_key": "D3_O6",
        "source_intent": "SPIRAL",
        "strength": 0.91,
        "phase_offset": 0.02,
        "jitter": 0.03,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "PHASE_DRIFT_REJECT_2",
        "case_type": "PACKET",
        "slot": 2,
        "source_key": "D3_O6",
        "source_intent": "HEX",
        "strength": 0.90,
        "phase_offset": 0.34,
        "jitter": 0.03,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "RESYNC_HEX_RECOVER_2",
        "case_type": "PACKET",
        "slot": 2,
        "source_key": "D3_O6",
        "source_intent": "HEX",
        "strength": 0.88,
        "phase_offset": 0.31,
        "jitter": 0.04,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": True,
    },
    {
        "case": "NEW_WRITE_SPIRAL_3",
        "case_type": "PACKET",
        "slot": 3,
        "source_key": "D5_O4",
        "source_intent": "SPIRAL",
        "strength": 0.92,
        "phase_offset": 0.02,
        "jitter": 0.03,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "WEAK_PACKET_FORGET_SEED_4",
        "case_type": "PACKET",
        "slot": 4,
        "source_key": "D0_O1",
        "source_intent": "SEED",
        "strength": 0.38,
        "phase_offset": 0.01,
        "jitter": 0.02,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "REWRITE_BUTTERFLY_5",
        "case_type": "PACKET",
        "slot": 5,
        "source_key": "D2_O3",
        "source_intent": "BUTTERFLY",
        "strength": 0.89,
        "phase_offset": 0.02,
        "jitter": 0.03,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "AMBIGUOUS_FLOWER_HOLD_6",
        "case_type": "PACKET",
        "slot": 6,
        "source_key": "D4_O5",
        "source_intent": "FLOWER",
        "strength": 0.86,
        "phase_offset": 0.02,
        "jitter": 0.02,
        "broad": False,
        "ambiguous_key": "D1_O2",
        "ambiguous_strength": 0.81,
        "clock_resync_allowed": False,
    },
    {
        "case": "STRONG_REINFORCE_FLOWER_6",
        "case_type": "PACKET",
        "slot": 6,
        "source_key": "D4_O5",
        "source_intent": "FLOWER",
        "strength": 0.93,
        "phase_offset": 0.02,
        "jitter": 0.02,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "EMPTY_IDLE_HOLD_7",
        "case_type": "IDLE",
        "slot": 7,
        "source_key": "IDLE",
        "source_intent": "NONE",
        "strength": 0.0,
        "phase_offset": 0.0,
        "jitter": 0.0,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
    {
        "case": "LATE_RING_WRITE_7",
        "case_type": "PACKET",
        "slot": 7,
        "source_key": "D1_O2",
        "source_intent": "RING",
        "strength": 0.87,
        "phase_offset": 0.03,
        "jitter": 0.03,
        "broad": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
        "clock_resync_allowed": False,
    },
]

decision_rows = []
memory_rows = []
strength_history_rows = []
snapshots = []

applied_count = 0
rejected_count = 0
decay_only_count = 0
forgotten_total = 0
reinforced_count = 0
new_write_count = 0
resync_count = 0

for t, item in enumerate(events):
    slot = item["slot"]

    decayed_slots, forgotten_slots = apply_passive_decay(
        memory_keys,
        memory_intents,
        memory_strengths
    )

    forgotten_total += len(forgotten_slots)

    before_key = memory_keys[slot]
    before_intent = memory_intents[slot]
    before_strength = float(memory_strengths[slot])

    expected_key = receiver_filter_sequence[slot]
    expected_intent = receiver_intent_sequence[slot]

    if item["case_type"] == "IDLE":
        packet = np.zeros((octaves, directions))

        decoded = {
            "best_key": "NONE",
            "best_state_id": -1,
            "best_score": 0.0,
            "second_key": "NONE",
            "second_state_id": -1,
            "second_score": 0.0,
            "margin": 0.0,
            "accepted": False,
        }

        measured_phase = None
        corrected_phase = None
        raw_phase_error = 1.0
        corrected_phase_error = 1.0

        clock_resync = False
        phase_ok = False
        jitter_ok = True
        amplitude_ok = False
        filter_match = False
        intent_match = False
        applied = False

        action = "PASSIVE_DECAY_ONLY"
        output_token = "NONE"

        decay_only_count += 1

    else:
        packet = spectral_packet(
            target_key=item["source_key"],
            strength=item["strength"],
            noise=0.025,
            broad=item["broad"],
            ambiguous_key=item["ambiguous_key"],
            ambiguous_strength=item["ambiguous_strength"],
        )

        decoded = decode_packet(packet)

        amplitude_ok = bool(decoded["accepted"])

        decoded_key = decoded["best_key"]
        expected_phase = phase_for_key(expected_key)
        base_phase = phase_for_key(decoded_key)

        measured_phase = (
            None
            if base_phase is None
            else (base_phase + item["phase_offset"]) % 1.0
        )

        raw_phase_error = circular_phase_error(
            measured_phase,
            expected_phase
        )

        clock_resync = False
        corrected_phase = measured_phase
        corrected_phase_error = raw_phase_error

        if item["clock_resync_allowed"] and raw_phase_error <= resync_window:
            clock_resync = True
            corrected_phase = expected_phase + 0.003

            if corrected_phase >= 1.0:
                corrected_phase -= 1.0

            corrected_phase_error = circular_phase_error(
                corrected_phase,
                expected_phase
            )

            resync_count += 1

        phase_ok = corrected_phase_error <= phase_tolerance
        jitter_ok = item["jitter"] <= jitter_tolerance
        filter_match = decoded_key == expected_key
        intent_match = item["source_intent"] == expected_intent

        applied = (
            amplitude_ok
            and phase_ok
            and jitter_ok
            and filter_match
            and intent_match
        )

        if applied:
            is_new = before_key in ["EMPTY", "NONE"] or before_strength == 0.0

            after_strength = reinforce_strength(
                before_strength,
                decoded["best_score"],
                is_new=is_new
            )

            memory_keys[slot] = decoded_key
            memory_intents[slot] = item["source_intent"]
            memory_strengths[slot] = after_strength

            if is_new:
                action = "WRITE_NEW_FIELD_MEMORY"
                new_write_count += 1
            else:
                action = "REINFORCE_FIELD_MEMORY"
                reinforced_count += 1

            applied_count += 1

        else:
            if not amplitude_ok:
                action = "REJECT_WEAK_OR_AMBIGUOUS"
            elif not phase_ok:
                action = "REJECT_PHASE_UNLOCKED"
            elif not jitter_ok:
                action = "REJECT_DECOHERENCE"
            elif not filter_match:
                action = "REJECT_FILTER_MISMATCH"
            elif not intent_match:
                action = "REJECT_INTENT_MISMATCH"
            else:
                action = "REJECT_HOLD"

            rejected_count += 1

        output_token = packet_token(
            decoded["best_key"] if applied else "NONE",
            item["source_intent"] if applied else "NONE",
            decoded["best_score"],
            corrected_phase if corrected_phase is not None else 0.0,
            item["jitter"],
            action
        )

    after_key = memory_keys[slot]
    after_intent = memory_intents[slot]
    after_strength = float(memory_strengths[slot])

    row = {
        "STATE": STATE,
        "T": t * 120,
        "CASE": item["case"],
        "CASE_TYPE": item["case_type"],
        "MEMORY_SLOT": slot,
        "SOURCE_KEY": item["source_key"],
        "SOURCE_INTENT": item["source_intent"],
        "FILTER_KEY": expected_key,
        "FILTER_INTENT": expected_intent,
        "DECODED_KEY": decoded["best_key"],
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3) if item["case_type"] != "IDLE" else "none",
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3) if item["case_type"] != "IDLE" else "none",
        "MARGIN": round(decoded["margin"], 3) if item["case_type"] != "IDLE" else "none",
        "AMPLITUDE_OK": int(amplitude_ok),
        "MEASURED_PHASE": round(measured_phase, 3) if measured_phase is not None else "none",
        "RAW_PHASE_ERROR": round(raw_phase_error, 3),
        "CLOCK_RESYNC": int(clock_resync),
        "CORRECTED_PHASE": round(corrected_phase, 3) if corrected_phase is not None else "none",
        "CORRECTED_PHASE_ERROR": round(corrected_phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "JITTER": item["jitter"],
        "JITTER_OK": int(jitter_ok),
        "FILTER_MATCH": int(filter_match),
        "INTENT_MATCH": int(intent_match),
        "BEFORE_KEY": before_key,
        "BEFORE_INTENT": before_intent,
        "BEFORE_STRENGTH": round(before_strength, 3),
        "AFTER_KEY": after_key,
        "AFTER_INTENT": after_intent,
        "AFTER_STRENGTH": round(after_strength, 3),
        "DECAYED_SLOTS": ",".join(map(str, decayed_slots)) if decayed_slots else "none",
        "FORGOTTEN_SLOTS": ",".join(map(str, forgotten_slots)) if forgotten_slots else "none",
        "APPLIED": int(applied),
        "ACTION": action,
        "OUTPUT_TOKEN": output_token,
    }

    decision_rows.append(row)

    memory_rows.append({
        "STATE": STATE,
        "T": t * 120,
        "CASE": item["case"],
        "MEMORY_SLOT": slot,
        "BEFORE_KEY": before_key,
        "BEFORE_INTENT": before_intent,
        "BEFORE_STRENGTH": round(before_strength, 3),
        "AFTER_KEY": after_key,
        "AFTER_INTENT": after_intent,
        "AFTER_STRENGTH": round(after_strength, 3),
        "APPLIED": int(applied),
        "ACTION": action,
    })

    for s in range(len(memory_keys)):
        strength_history_rows.append({
            "STATE": STATE,
            "T": t * 120,
            "CASE": item["case"],
            "MEMORY_SLOT": s,
            "KEY": memory_keys[s],
            "INTENT": memory_intents[s],
            "STRENGTH": round(float(memory_strengths[s]), 3),
            "STABLE": int(memory_strengths[s] >= stable_threshold),
            "EMPTY": int(memory_keys[s] == "EMPTY"),
        })

    snapshots.append((
        row,
        packet.copy(),
        memory_keys.copy(),
        memory_intents.copy(),
        memory_strengths.copy()
    ))

df_decision = pd.DataFrame(decision_rows)
df_memory = pd.DataFrame(memory_rows)
df_history = pd.DataFrame(strength_history_rows)

df_final = pd.DataFrame([
    {
        "STATE": STATE,
        "MEMORY_SLOT": i,
        "FILTER_KEY": receiver_filter_sequence[i],
        "FILTER_INTENT": receiver_intent_sequence[i],
        "FINAL_KEY": memory_keys[i],
        "FINAL_INTENT": memory_intents[i],
        "FINAL_STRENGTH": round(float(memory_strengths[i]), 3),
        "STABLE": int(memory_strengths[i] >= stable_threshold),
        "FORGOTTEN": int(memory_keys[i] == "EMPTY"),
        "MATCH_FILTER_IF_PRESENT": int(memory_keys[i] == receiver_filter_sequence[i]) if memory_keys[i] != "EMPTY" else "empty",
        "MATCH_INTENT_IF_PRESENT": int(memory_intents[i] == receiver_intent_sequence[i]) if memory_keys[i] != "EMPTY" else "empty",
    }
    for i in range(len(memory_keys))
])

applied_rows = df_decision[df_decision["APPLIED"] == 1]
rejected_rows = df_decision[
    (df_decision["APPLIED"] == 0)
    & (df_decision["CASE_TYPE"] == "PACKET")
]

applied_safety = float(np.mean(
    (applied_rows["FILTER_MATCH"] == 1)
    & (applied_rows["INTENT_MATCH"] == 1)
    & (applied_rows["PHASE_OK"] == 1)
    & (applied_rows["JITTER_OK"] == 1)
)) if len(applied_rows) else 1.0

reject_safety = float(np.mean(
    rejected_rows["BEFORE_KEY"].values == rejected_rows["AFTER_KEY"].values
)) if len(rejected_rows) else 1.0

stable_present = df_final[df_final["FINAL_KEY"] != "EMPTY"]

retention_accuracy = float(np.mean(
    stable_present["FINAL_STRENGTH"] >= stable_threshold
)) if len(stable_present) else 1.0

forgotten_rows = df_final[df_final["FORGOTTEN"] == 1]

forget_accuracy = float(np.mean(
    forgotten_rows["FINAL_STRENGTH"] < forget_threshold
)) if len(forgotten_rows) else 1.0

memory_accuracy = float(np.mean([
    applied_safety == 1.0,
    reject_safety == 1.0,
    retention_accuracy == 1.0,
    forget_accuracy == 1.0,
]))

result = (
    "FIELD_MEMORY_DECAY_REINFORCEMENT_LOCKED"
    if memory_accuracy == 1.0
    else "FIELD_MEMORY_DECAY_REINFORCEMENT_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": len(memory_keys),
    "EVENTS_TESTED": len(events),
    "APPLIED_WRITES": applied_count,
    "REINFORCED_MEMORIES": reinforced_count,
    "NEW_WRITES": new_write_count,
    "REJECTED_PACKETS": rejected_count,
    "PASSIVE_DECAY_EVENTS": decay_only_count,
    "FORGOTTEN_TOTAL_EVENTS": forgotten_total,
    "CLOCK_RESYNC_EVENTS": resync_count,
    "APPLIED_SAFETY": round(applied_safety, 3),
    "REJECT_SAFETY": round(reject_safety, 3),
    "RETENTION_ACCURACY": round(retention_accuracy, 3),
    "FORGET_ACCURACY": round(forget_accuracy, 3),
    "MEMORY_ACCURACY": round(memory_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "PHASE_TOLERANCE": phase_tolerance,
    "JITTER_TOLERANCE": jitter_tolerance,
    "PASSIVE_DECAY_RATE": passive_decay_rate,
    "FORGET_THRESHOLD": forget_threshold,
    "STABLE_THRESHOLD": stable_threshold,
    "REINFORCEMENT_GAIN": reinforcement_gain,
}])

# ------------------------------------------------------------
# Visualization snapshots
# ------------------------------------------------------------

fig, axes = plt.subplots(
    len(snapshots),
    1,
    figsize=(9, 2.75 * len(snapshots))
)

if len(snapshots) == 1:
    axes = [axes]

for ax, (row, packet, keys_snapshot, intents_snapshot, strengths_snapshot) in zip(axes, snapshots):
    draw_snapshot(
        ax,
        row,
        packet,
        keys_snapshot,
        intents_snapshot,
        strengths_snapshot
    )

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# Memory strength timeline
# ------------------------------------------------------------

pivot = df_history.pivot(
    index="T",
    columns="MEMORY_SLOT",
    values="STRENGTH"
)

plt.figure(figsize=(9, 4.8))

for slot in pivot.columns:
    plt.plot(
        pivot.index,
        pivot[slot],
        marker="o",
        linewidth=1.6,
        label=f"slot {slot}"
    )

plt.axhline(
    stable_threshold,
    linestyle="--",
    linewidth=1.0,
    label="stable threshold"
)

plt.axhline(
    forget_threshold,
    linestyle=":",
    linewidth=1.0,
    label="forget threshold"
)

plt.title("D41 field memory strength over time")
plt.xlabel("time")
plt.ylabel("memory strength")
plt.ylim(0, 1.05)
plt.legend(loc="upper right", fontsize=7, ncol=2)
plt.grid(True, alpha=0.25)
plt.show()

# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"memory slots:                 {len(memory_keys)}")
print(f"events tested:                {len(events)}")
print(f"applied writes:               {applied_count}")
print(f"reinforced memories:          {reinforced_count}")
print(f"new writes:                   {new_write_count}")
print(f"rejected packets:             {rejected_count}")
print(f"passive decay events:         {decay_only_count}")
print(f"forgotten total events:       {forgotten_total}")
print(f"clock resync events:          {resync_count}")
print(f"applied safety:               {round(applied_safety, 3)}")
print(f"reject safety:                {round(reject_safety, 3)}")
print(f"retention accuracy:           {round(retention_accuracy, 3)}")
print(f"forget accuracy:              {round(forget_accuracy, 3)}")
print(f"memory accuracy:              {round(memory_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"phase tolerance:              {phase_tolerance}")
print(f"jitter tolerance:             {jitter_tolerance}")
print(f"passive decay rate:           {passive_decay_rate}")
print(f"forget threshold:             {forget_threshold}")
print(f"stable threshold:             {stable_threshold}")
print(f"reinforcement gain:           {reinforcement_gain}")

print("\nDecay / reinforcement decision table:")
display(df_decision)

print("\nMemory update table:")
display(df_memory)

print("\nFinal field memory table:")
display(df_final)

print("\nStrength history table:")
display(df_history)

print("\nSummary:")
display(df_summary)
```
