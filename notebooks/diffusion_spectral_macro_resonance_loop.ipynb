```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D40_SPECTRAL_MACRO_RESONANCE_LOOP
# Closed honeycomb thought-loop / standing-wave attractor
#
# D39 = wave can propagate through honeycomb tissue.
# D40 = wave must close a full loop and return to center:
# C0 -> N0 -> N1 -> N2 -> N3 -> N4 -> N5 -> N0 -> C0
# ============================================================

np.random.seed(40)

STATE = "D40_SPECTRAL_MACRO_RESONANCE_LOOP"

directions = 6
octaves = 7
spectral_states = directions * octaves

TARGET_KEY = "D4_O5"
TARGET_INTENT = "FLOWER"

resonance_threshold = 0.62
ambiguity_margin = 0.12
phase_tolerance = 0.18
jitter_tolerance = 0.16
loop_closure_tolerance = 0.18
resync_window = 0.38
hop_loss = 0.035
macro_pulse_hz = 7.83

nodes = ["C0"] + [f"N{i}" for i in range(6)] + [f"O{i}" for i in range(6)]
ring_route = ["C0", "N0", "N1", "N2", "N3", "N4", "N5", "N0", "C0"]

angles = np.linspace(np.pi/2, np.pi/2 - 2*np.pi, 6, endpoint=False)
pos = {"C0": (0.0, 0.0)}

for i, a in enumerate(angles):
    pos[f"N{i}"] = (np.cos(a), np.sin(a))
    pos[f"O{i}"] = (1.75*np.cos(a), 1.75*np.sin(a))

edges = set()
for i in range(6):
    edges |= {
        ("C0", f"N{i}"), (f"N{i}", "C0"),
        (f"N{i}", f"N{(i+1)%6}"), (f"N{(i+1)%6}", f"N{i}"),
        (f"N{i}", f"O{i}"), (f"O{i}", f"N{i}"),
    }

receiver_filter = {n: TARGET_KEY for n in nodes}
receiver_intent = {n: TARGET_INTENT for n in nodes}


def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE"]:
        return None, None
    d, o = key.split("_")
    return int(d.replace("D", "")), int(o.replace("O", ""))


def make_key(d, o):
    return f"D{d}_O{o}"


def phase_of(key):
    d, _ = parse_key(key)
    return None if d is None else d / directions


def phase_error(a, b):
    diff = abs((a - b) % 1.0)
    return min(diff, 1.0 - diff)


def key_color(key):
    if key in [None, "EMPTY", "NONE"]:
        return "black"
    if key == "WHITE":
        return "white"
    d, o = parse_key(key)
    return plt.cm.hsv(((d + 1) / (directions + 1) + 0.08 * ((o + 1) / (octaves + 1))) % 1.0)


def intent_color(intent):
    return {
        "FLOWER": "lime",
        "HEX": "orange",
        "RING": "gold",
        "SPIRAL": "magenta",
        "TWO_LOBE": "cyan",
        "DOT": "gray",
        "EMPTY": "black",
        "NONE": "black",
    }.get(intent, "gray")


def spectral_packet(target_key, strength, noise=0.025, broad=False,
                    ambiguous_key=None, ambiguous_strength=0.0):
    p = np.clip(np.random.normal(0, noise, size=(octaves, directions)), 0, 1)

    if broad:
        p = np.clip(p + 0.16 + np.random.normal(0, noise, size=(octaves, directions)), 0, 1)

    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(target_key)
        p[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(ambiguous_key)
        p[o, d] = ambiguous_strength

    return np.clip(p, 0, 1)


def decode_packet(packet):
    flat = packet.flatten()
    order = np.argsort(flat)[::-1]
    best_id, second_id = int(order[0]), int(order[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_o, best_d = best_id // directions, best_id % directions
    second_o, second_d = second_id // directions, second_id % directions

    margin = best_score - second_score

    return {
        "best_key": make_key(best_d, best_o),
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": make_key(second_d, second_o),
        "second_score": second_score,
        "margin": margin,
        "accepted": bool(best_score >= resonance_threshold and margin >= ambiguity_margin),
    }


def loop_token(key, intent, phase, amp, hops, locked):
    if not locked:
        return "NONE"
    d, o = parse_key(key)
    return f"LOOP_PHOTON[D{d}|O{o}|{intent}|F{o+1}|P{round(phase,3)}|A{round(amp,3)}|H{hops}]"


def run_case(case):
    memory_key = {n: "EMPTY" for n in nodes}
    memory_intent = {n: "EMPTY" for n in nodes}
    active_nodes = set(["C0"])
    rejected_nodes = set()

    src_key = case["source_key"]
    expected_key = case.get("expected_key", TARGET_KEY)
    intent = case["intent"]
    expected_intent = case.get("expected_intent", TARGET_INTENT)

    emit_key = expected_key if src_key == "WHITE" else src_key

    memory_key["C0"] = emit_key
    memory_intent["C0"] = intent

    usable_edges = set(edges)
    for broken in case.get("broken_edges", []):
        usable_edges.discard(broken)

    strength = case.get("strength", 0.92)
    base_phase = case.get("phase", phase_of(emit_key))
    jitter = case.get("jitter", 0.02)
    drift = case.get("phase_drift_per_hop", 0.0)

    success_hops = 0
    reject_hops = 0
    resync_events = 0
    hop_rows = []

    last_decoded = None
    last_packet = None
    last_phase = base_phase
    last_action = "LOOP_NOT_STARTED"

    for hop, (a, b) in enumerate(zip(ring_route[:-1], ring_route[1:]), start=1):
        amp = max(0.0, strength - hop_loss * (hop - 1))
        measured_phase = (base_phase + drift * hop) % 1.0

        packet = spectral_packet(
            emit_key,
            amp,
            noise=case.get("noise", 0.025),
            broad=case.get("broad", False),
            ambiguous_key=case.get("ambiguous_key", None),
            ambiguous_strength=case.get("ambiguous_strength", 0.0),
        )

        decoded = decode_packet(packet)
        decoded_key = decoded["best_key"]

        raw_err = phase_error(measured_phase, phase_of(expected_key))
        corrected_phase = measured_phase
        corrected_err = raw_err
        clock_resync = 0

        phase_ok = raw_err <= phase_tolerance

        if (not phase_ok) and case.get("clock_resync", False) and raw_err <= resync_window:
            corrected_phase = (phase_of(expected_key) + 0.003) % 1.0
            corrected_err = phase_error(corrected_phase, phase_of(expected_key))
            phase_ok = corrected_err <= phase_tolerance
            clock_resync = 1
            resync_events += 1

        edge_ok = (a, b) in usable_edges
        amp_ok = decoded["accepted"]
        jitter_ok = jitter <= jitter_tolerance
        filter_ok = decoded_key == receiver_filter[b]
        intent_ok = intent == receiver_intent[b]

        route_ok = edge_ok and amp_ok and phase_ok and jitter_ok and filter_ok and intent_ok

        if route_ok:
            memory_key[b] = decoded_key
            memory_intent[b] = intent
            active_nodes.add(b)
            success_hops += 1
            action = "WRITE_AND_RELAY"
        else:
            rejected_nodes.add(b)
            reject_hops += 1
            action = (
                "REJECT_BROKEN_EDGE" if not edge_ok else
                "REJECT_WEAK_OR_AMBIGUOUS" if not amp_ok else
                "REJECT_PHASE_UNLOCKED" if not phase_ok else
                "REJECT_DECOHERENCE" if not jitter_ok else
                "REJECT_FILTER_MISMATCH" if not filter_ok else
                "REJECT_INTENT_MISMATCH"
            )

        hop_rows.append({
            "STATE": STATE,
            "CASE": case["case"],
            "HOP": hop,
            "SRC_NODE": a,
            "DST_NODE": b,
            "DECODED_KEY": decoded_key,
            "BEST_SCORE": round(decoded["best_score"], 3),
            "SECOND_KEY": decoded["second_key"],
            "SECOND_SCORE": round(decoded["second_score"], 3),
            "MARGIN": round(decoded["margin"], 3),
            "EDGE_OK": int(edge_ok),
            "AMPLITUDE_OK": int(amp_ok),
            "PHASE": round(measured_phase, 3),
            "RAW_PHASE_ERROR": round(raw_err, 3),
            "CLOCK_RESYNC": clock_resync,
            "CORRECTED_PHASE_ERROR": round(corrected_err, 3),
            "PHASE_OK": int(phase_ok),
            "JITTER": round(jitter, 3),
            "JITTER_OK": int(jitter_ok),
            "FILTER_OK": int(filter_ok),
            "INTENT_OK": int(intent_ok),
            "ROUTE_OK": int(route_ok),
            "ACTION": action,
        })

        last_decoded = decoded
        last_packet = packet
        last_phase = corrected_phase
        last_action = action

        if not route_ok:
            break

    full_hops = len(ring_route) - 1
    closure_phase_err = phase_error(last_phase, phase_of(expected_key))

    loop_locked = (
        success_hops == full_hops
        and memory_key["C0"] == expected_key
        and memory_intent["C0"] == expected_intent
        and closure_phase_err <= loop_closure_tolerance
    )

    expected_locked = bool(case["expected_locked"])
    action = "MACRO_RESONANCE_LOOP_LOCK" if loop_locked else last_action

    token = loop_token(
        last_decoded["best_key"] if last_decoded else "NONE",
        intent,
        last_phase,
        last_decoded["best_score"] if last_decoded else 0.0,
        success_hops,
        loop_locked,
    )

    row = {
        "STATE": STATE,
        "T": case["t"],
        "CASE": case["case"],
        "SOURCE_KEY": src_key,
        "EMITTED_KEY": emit_key,
        "ORBITAL_INTENT": intent,
        "EXPECTED_KEY": expected_key,
        "EXPECTED_INTENT": expected_intent,
        "DECODED_KEY": last_decoded["best_key"] if last_decoded else "NONE",
        "BEST_SCORE": round(last_decoded["best_score"], 3) if last_decoded else 0,
        "SECOND_KEY": last_decoded["second_key"] if last_decoded else "NONE",
        "MARGIN": round(last_decoded["margin"], 3) if last_decoded else 0,
        "SUCCESSFUL_HOPS": success_hops,
        "FULL_LOOP_HOPS": full_hops,
        "REJECTED_HOPS": reject_hops,
        "CLOCK_RESYNC_EVENTS": resync_events,
        "CLOSURE_PHASE_ERROR": round(closure_phase_err, 3),
        "LOOP_LOCKED": int(loop_locked),
        "EXPECTED_LOCKED": int(expected_locked),
        "MATCH_EXPECTED_ACTION": int(loop_locked == expected_locked),
        "ACTION": action,
        "OUTPUT_TOKEN": token,
    }

    return row, hop_rows, memory_key, memory_intent, active_nodes, rejected_nodes, last_packet


def draw_packet_grid(ax, packet, selected_key, x0=4.25, y0=-1.9):
    cell = 0.10
    gap = 0.02

    ax.text(x0, y0 + 0.92, "42-channel packet", fontsize=6)
    selected = parse_key(selected_key)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)
            is_sel = selected == (d, o)

            ax.add_patch(plt.Rectangle(
                (x, y), cell, cell,
                facecolor=plt.cm.plasma(packet[o, d]),
                edgecolor="lime" if is_sel else "white",
                linewidth=1.0 if is_sel else 0.25,
            ))


def draw_snapshot(ax, row, memory_key, memory_intent, active, rejected, packet):
    ax.set_aspect("equal")
    ax.axis("off")
    ax.set_xlim(-2.7, 5.2)
    ax.set_ylim(-2.45, 2.45)

    ax.add_patch(Circle((0, 0), 2.05, fill=False, linestyle="--", edgecolor="goldenrod", alpha=0.5))
    ax.text(0, 2.18, f"D40 t={row['T']} — {row['CASE']} / {row['ACTION']}", ha="center", fontsize=9)
    ax.text(0, 2.00, f"macro pulse ≈ {macro_pulse_hz} Hz", ha="center", fontsize=6)

    for a, b in edges:
        x1, y1 = pos[a]
        x2, y2 = pos[b]
        ax.plot([x1, x2], [y1, y2], color="lightgray", linewidth=0.7, zorder=0)

    for a, b in zip(ring_route[:-1], ring_route[1:]):
        x1, y1 = pos[a]
        x2, y2 = pos[b]
        color = "limegreen" if b in active else "red" if b in rejected else "gray"

        ax.add_patch(FancyArrowPatch(
            (x1, y1), (x2, y2),
            arrowstyle="->",
            mutation_scale=10,
            color=color,
            linewidth=1.3,
            alpha=0.75,
        ))

    for n in nodes:
        x, y = pos[n]
        key = memory_key[n]
        intent = memory_intent[n]

        edge = "lime" if n in active else "red" if n in rejected else "black"
        lw = 2.0 if n in active or n in rejected else 0.7

        ax.add_patch(Circle((x, y), 0.18, facecolor=key_color(key), edgecolor=edge, linewidth=lw))
        ax.add_patch(Circle((x, y - 0.28), 0.07, facecolor=intent_color(intent), edgecolor="black", linewidth=0.4))
        ax.text(x, y + 0.28, n, fontsize=6, ha="center")
        ax.text(x, y - 0.48, key, fontsize=5, ha="center")

    ax.add_patch(Circle((-2.35, 1.55), 0.18, facecolor=key_color(row["EMITTED_KEY"]), edgecolor="black"))
    ax.text(-2.35, 1.25, "SRC", fontsize=6, ha="center")

    ax.add_patch(Circle((-1.85, 1.55), 0.12, facecolor=intent_color(row["ORBITAL_INTENT"]), edgecolor="black"))
    ax.text(-1.85, 1.25, "INTENT", fontsize=6, ha="center")

    ax.add_patch(FancyArrowPatch((-2.15, 1.55), (-2.00, 1.55), arrowstyle="->", mutation_scale=10, color="orange"))
    ax.add_patch(FancyArrowPatch((-1.72, 1.55), (-1.45, 1.35), arrowstyle="->", mutation_scale=10, color="orange"))

    draw_packet_grid(ax, packet, row["DECODED_KEY"])

    ax.text(
        -2.55,
        -2.18,
        f"decoded={row['DECODED_KEY']} score={row['BEST_SCORE']} margin={row['MARGIN']} "
        f"hops={row['SUCCESSFUL_HOPS']}/{row['FULL_LOOP_HOPS']} "
        f"closure_phase_error={row['CLOSURE_PHASE_ERROR']}\n{row['OUTPUT_TOKEN']}",
        fontsize=6,
        ha="left",
    )


cases = [
    {
        "t": 0,
        "case": "CLOSED_FLOWER_LOOP_LOCK",
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.96,
        "phase": 4/6,
        "jitter": 0.02,
        "expected_locked": True,
    },
    {
        "t": 120,
        "case": "PHASE_DRIFT_LOOP_REJECT",
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.94,
        "phase": 4/6,
        "phase_drift_per_hop": 0.065,
        "jitter": 0.02,
        "expected_locked": False,
    },
    {
        "t": 240,
        "case": "RESYNC_FLOWER_LOOP_LOCK",
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.93,
        "phase": (4/6 + 0.31) % 1.0,
        "clock_resync": True,
        "jitter": 0.04,
        "expected_locked": True,
    },
    {
        "t": 360,
        "case": "WRONG_INTENT_LOOP_REJECT",
        "source_key": "D4_O5",
        "intent": "RING",
        "strength": 0.94,
        "phase": 4/6,
        "jitter": 0.02,
        "expected_locked": False,
    },
    {
        "t": 480,
        "case": "WEAK_LOOP_DECAY",
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.55,
        "phase": 4/6,
        "jitter": 0.02,
        "expected_locked": False,
    },
    {
        "t": 600,
        "case": "BROKEN_EDGE_LOOP_REJECT",
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.95,
        "phase": 4/6,
        "jitter": 0.02,
        "broken_edges": [("N3", "N4")],
        "expected_locked": False,
    },
    {
        "t": 720,
        "case": "AMBIGUOUS_LOOP_REJECT",
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.86,
        "ambiguous_key": "D1_O2",
        "ambiguous_strength": 0.81,
        "phase": 4/6,
        "jitter": 0.02,
        "expected_locked": False,
    },
    {
        "t": 840,
        "case": "WHITE_WITH_FLOWER_TARGET_LOCK",
        "source_key": "WHITE",
        "expected_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.92,
        "phase": 4/6,
        "jitter": 0.03,
        "broad": True,
        "expected_locked": True,
    },
]

decision_rows = []
hop_rows_all = []
snapshots = []

for case in cases:
    row, hop_rows, mem_key, mem_intent, active, rejected, packet = run_case(case)
    decision_rows.append(row)
    hop_rows_all.extend(hop_rows)
    snapshots.append((row, mem_key, mem_intent, active, rejected, packet))

df_decision = pd.DataFrame(decision_rows)
df_hops = pd.DataFrame(hop_rows_all)

memory_rows = []
for row, mem_key, mem_intent, *_ in snapshots:
    for n in nodes:
        memory_rows.append({
            "STATE": STATE,
            "CASE": row["CASE"],
            "NODE": n,
            "FINAL_KEY": mem_key[n],
            "FINAL_INTENT": mem_intent[n],
            "ACTIVE": int(mem_key[n] != "EMPTY"),
        })

df_memory = pd.DataFrame(memory_rows)

loop_accuracy = float(df_decision["MATCH_EXPECTED_ACTION"].mean())
locked_loops = int(df_decision["LOOP_LOCKED"].sum())
rejected_loops = int((df_decision["LOOP_LOCKED"] == 0).sum())
max_good_hops = int(df_decision[df_decision["EXPECTED_LOCKED"] == 1]["SUCCESSFUL_HOPS"].max())
clock_resync_events = int(df_decision["CLOCK_RESYNC_EVENTS"].sum())

result = (
    "SPECTRAL_MACRO_RESONANCE_LOOP_LOCKED"
    if loop_accuracy == 1.0 and locked_loops >= 2 and max_good_hops >= len(ring_route) - 1
    else "SPECTRAL_MACRO_RESONANCE_LOOP_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "HONEYCOMB_NODES": len(nodes),
    "RING_ROUTE": " -> ".join(ring_route),
    "FULL_LOOP_HOPS": len(ring_route) - 1,
    "TARGET_KEY": TARGET_KEY,
    "TARGET_ORBITAL_INTENT": TARGET_INTENT,
    "MACRO_PULSE_HZ": macro_pulse_hz,
    "CASES_TESTED": len(cases),
    "LOCKED_LOOPS": locked_loops,
    "REJECTED_LOOPS": rejected_loops,
    "MAX_GOOD_HOPS": max_good_hops,
    "CLOCK_RESYNC_EVENTS": clock_resync_events,
    "LOOP_ACCURACY": round(loop_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "PHASE_TOLERANCE": phase_tolerance,
    "JITTER_TOLERANCE": jitter_tolerance,
    "LOOP_CLOSURE_TOLERANCE": loop_closure_tolerance,
    "HOP_LOSS": hop_loss,
}])

# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8, 4.2 * len(snapshots)))

if len(snapshots) == 1:
    axes = [axes]

for ax, snap in zip(axes, snapshots):
    draw_snapshot(ax, *snap)

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"honeycomb nodes:              {len(nodes)}")
print(f"ring route:                   {' -> '.join(ring_route)}")
print(f"full loop hops:               {len(ring_route) - 1}")
print(f"target key:                   {TARGET_KEY}")
print(f"target orbital intent:        {TARGET_INTENT}")
print(f"macro pulse reference:        {macro_pulse_hz} Hz")
print(f"cases tested:                 {len(cases)}")
print(f"locked loops:                 {locked_loops}")
print(f"rejected loops:               {rejected_loops}")
print(f"max good hops:                {max_good_hops}")
print(f"clock resync events:          {clock_resync_events}")
print(f"loop accuracy:                {round(loop_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"phase tolerance:              {phase_tolerance}")
print(f"jitter tolerance:             {jitter_tolerance}")
print(f"loop closure tolerance:       {loop_closure_tolerance}")
print(f"hop loss:                     {hop_loss}")

print("\nMacro resonance loop decision table:")
display(df_decision)

print("\nLoop hop table:")
display(df_hops)

print("\nLoop memory table:")
display(df_memory)

print("\nSummary:")
display(df_summary)
```
