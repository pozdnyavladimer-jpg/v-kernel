D31 adds rule-gated spectral state transitions. Each memory cell updates only when the incoming photon-like packet is strong, unambiguous, and satisfies allowed transition rules across direction and octave space.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D31_SPECTRAL_TRANSITION_RULES
# Current spectral state + photon packet -> allowed/forbidden transition
#
# Goal:
# receiver cell has CURRENT spectral state
# -> incoming photon-like packet proposes NEXT spectral state
# -> system computes delta: current -> next
# -> selection rules decide if transition is legal
# -> only legal + strong + unambiguous transitions update memory
# ============================================================

np.random.seed(31)

STATE = "D31_SPECTRAL_TRANSITION_RULES"

directions = 6
octaves = 7
spectral_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12
retention_threshold = 0.50

# Transition rule core:
# - octave must change: photon carries transition energy
# - direction may stay or move by one neighboring vector
# - octave jump may be 1 or 2 levels in this simplified V-Kernel model
max_direction_step = 1
allowed_octave_jumps = {1, 2}


# ------------------------------------------------------------
# Core spectral helpers
# ------------------------------------------------------------

def parse_key(key):
    """
    Robust spectral key parser.

    Valid key: D4_O5
    Safe non-state keys: EMPTY, NONE, WHITE
    """
    if key in [None, "EMPTY", "NONE", "WHITE"]:
        return None, None

    if not isinstance(key, str):
        return None, None

    if "_" not in key:
        return None, None

    try:
        d_part, o_part = key.split("_")
        direction = int(d_part.replace("D", ""))
        octave = int(o_part.replace("O", ""))
    except Exception:
        return None, None

    if not (0 <= direction < directions and 0 <= octave < octaves):
        return None, None

    return direction, octave


def make_key(direction, octave):
    if direction is None or octave is None:
        return "NONE"
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    if direction is None or octave is None:
        return None
    return octave * directions + direction


def key_color(key, strength=1.0):
    d, o = parse_key(key)

    if key == "WHITE":
        return "white"

    if d is None or o is None:
        return "black"

    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)

    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


def cyclic_direction_delta(d0, d1):
    """
    Shortest signed direction step on a 6-direction ring.
    Example:
    D5 -> D0 is +1, not -5.
    """
    if d0 is None or d1 is None:
        return None

    raw = (d1 - d0) % directions

    if raw > directions // 2:
        raw -= directions

    return int(raw)


def transition_delta(current_key, next_key):
    d0, o0 = parse_key(current_key)
    d1, o1 = parse_key(next_key)

    if d0 is None or d1 is None:
        return None, None

    delta_d = cyclic_direction_delta(d0, d1)
    delta_o = int(o1 - o0)

    return delta_d, delta_o


def transition_allowed(current_key, next_key):
    """
    Simplified selection rules.

    This is the important D31 part:
    the packet is accepted only if the transition itself is legal.
    """
    d0, o0 = parse_key(current_key)
    d1, o1 = parse_key(next_key)

    if d0 is None or d1 is None:
        return False, "INVALID_STATE"

    delta_d, delta_o = transition_delta(current_key, next_key)

    if delta_o == 0:
        return False, "FORBIDDEN_ZERO_ENERGY"

    if abs(delta_d) > max_direction_step:
        return False, "FORBIDDEN_DIRECTION_JUMP"

    if abs(delta_o) not in allowed_octave_jumps:
        return False, "FORBIDDEN_OCTAVE_JUMP"

    if not (0 <= o1 < octaves and 0 <= d1 < directions):
        return False, "OUT_OF_BOUNDS"

    return True, "ALLOWED_TRANSITION"


def transition_mode(delta_o):
    if delta_o is None:
        return "NONE"

    if delta_o > 0:
        return "ABSORB_UP"

    if delta_o < 0:
        return "RELAX_DOWN"

    return "ZERO_ENERGY"


def photon_token(current_key, next_key, accepted=True):
    d0, o0 = parse_key(current_key)
    d1, o1 = parse_key(next_key)

    if not accepted or d0 is None or d1 is None:
        return "NONE"

    delta_d, delta_o = transition_delta(current_key, next_key)

    energy = abs(delta_o)
    phase = round((d1 % directions) / directions, 3)
    frequency = energy + 1

    sign_d = "+" if delta_d >= 0 else ""
    sign_o = "+" if delta_o >= 0 else ""

    return f"TRANSITION_PHOTON[ΔD{sign_d}{delta_d}|ΔO{sign_o}{delta_o}|F{frequency}|P{phase}]"


# ------------------------------------------------------------
# Photon packet / decoder
# ------------------------------------------------------------

def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.18 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    d, o = parse_key(target_key)
    if d is not None:
        packet[o, d] = strength

    d2, o2 = parse_key(ambiguous_key)
    if d2 is not None:
        packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score

    score_ok = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "score_ok": score_ok,
    }


# ------------------------------------------------------------
# Visualization helpers
# ------------------------------------------------------------

def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.30

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=14,
        linewidth=2.0,
        color=color,
        alpha=alpha
    )
    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None, current_key=None):
    cell = 0.18
    gap = 0.035

    ax.text(x0, y0 + 1.65, title, fontsize=7, ha="left", va="bottom", color="black")

    selected = parse_key(selected_key)
    current = parse_key(current_key)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = packet[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)
            is_current = current == (d, o)

            if is_selected:
                edge = "lime"
                lw = 1.6
            elif is_current:
                edge = "cyan"
                lw = 1.2
            else:
                edge = "white"
                lw = 0.3

            ax.add_patch(
                Rectangle(
                    (x, y),
                    cell,
                    cell,
                    facecolor=color,
                    edgecolor=edge,
                    linewidth=lw
                )
            )


def draw_memory_slots(ax, x0, y0, seq, title, active_slot=None):
    cell = 0.22
    gap = 0.06

    ax.text(x0, y0 + 2.25, title, fontsize=7, ha="left", va="bottom", color="black")

    for i, key in enumerate(seq):
        x = x0
        y = y0 + (len(seq) - 1 - i) * (cell + gap)

        color = key_color(key)

        edge = "lime" if i == active_slot else "black"
        lw = 1.6 if i == active_slot else 0.5

        ax.add_patch(
            Rectangle(
                (x, y),
                cell,
                cell,
                facecolor=color,
                edgecolor=edge,
                linewidth=lw
            )
        )

        ax.text(
            x + 0.34,
            y + cell / 2,
            str(i),
            fontsize=6,
            ha="left",
            va="center",
            color="black"
        )


def draw_snapshot(ax, row, packet, memory_state):
    slot = int(row["MEMORY_SLOT"])

    current_key = row["CURRENT_KEY"]
    decoded_key = row["DECODED_NEXT_KEY"]
    final_key = row["FINAL_KEY"]

    rule_ok = bool(row["RULE_ALLOWED"])
    score_ok = bool(row["SCORE_OK"])
    applied = bool(row["APPLIED"])

    rejected = not applied

    ax.set_facecolor("white")
    ax.set_xlim(0, 12.2)
    ax.set_ylim(0.6, 5.2)
    ax.set_aspect("equal")
    ax.axis("off")

    current_color = key_color(current_key)
    decoded_color = key_color(decoded_key)
    final_color = key_color(final_key)

    ax.add_patch(
        Circle(
            (1.0, 3.2),
            0.45,
            facecolor=current_color,
            edgecolor="black",
            linewidth=1.0
        )
    )
    ax.text(1.0, 2.45, f"CURRENT\n{current_key}", fontsize=7, ha="center", color="black")

    ax.add_patch(
        Circle(
            (2.3, 3.2),
            0.30,
            facecolor=decoded_color,
            edgecolor="black",
            linewidth=1.0
        )
    )
    ax.text(2.3, 2.45, "PHOTON\nnext", fontsize=7, ha="center", color="black")

    gate_color = "limegreen" if rule_ok and score_ok else "red"

    ax.add_patch(
        Rectangle(
            (3.35, 2.95),
            0.50,
            0.50,
            facecolor=gate_color,
            edgecolor="black",
            linewidth=1.0
        )
    )
    ax.text(3.60, 2.45, "RULE\nGATE", fontsize=7, ha="center", color="black")

    ax.add_patch(
        Circle(
            (5.15, 3.2),
            0.45,
            facecolor=decoded_color,
            edgecolor="lime" if rule_ok else "red",
            linewidth=2.0
        )
    )
    ax.text(5.15, 2.35, f"DECODED\n{decoded_key}", fontsize=7, ha="center", color="black")

    ax.add_patch(
        Circle(
            (6.70, 3.2),
            0.45,
            facecolor=final_color,
            edgecolor="lime" if applied else "red",
            linewidth=2.0
        )
    )
    ax.text(6.70, 2.35, f"SLOT {slot}\n{final_key}", fontsize=7, ha="center", color="black")

    draw_arrow(ax, (1.45, 3.2), (2.0, 3.2), active=True, rejected=False)
    draw_arrow(ax, (2.6, 3.2), (3.35, 3.2), active=True, rejected=not score_ok)
    draw_arrow(ax, (3.85, 3.2), (4.70, 3.2), active=True, rejected=rejected)
    draw_arrow(ax, (5.60, 3.2), (6.25, 3.2), active=True, rejected=rejected)

    draw_packet_grid(
        ax,
        7.70,
        2.1,
        packet,
        "42-channel proposed next-state",
        selected_key=decoded_key,
        current_key=current_key
    )

    draw_memory_slots(
        ax,
        10.85,
        1.4,
        memory_state,
        "transition memory",
        active_slot=slot
    )

    ax.text(
        0.55,
        0.9,
        f"case={row['CASE']} slot={slot} ΔD={row['DELTA_D']} ΔO={row['DELTA_O']} "
        f"score={row['BEST_SCORE']} margin={row['MARGIN']} "
        f"rule={row['RULE_REASON']} action={row['ACTION']}\n"
        f"{row['PHOTON_TOKEN']}",
        fontsize=7,
        ha="left",
        color="black"
    )

    ax.set_title(
        f"D31 snapshot t={row['T']} — {row['CASE']} / {row['ACTION']}",
        fontsize=10,
        color="black"
    )


# ------------------------------------------------------------
# Test cases
# ------------------------------------------------------------

initial_memory = [
    "D3_O2",
    "D4_O3",
    "D4_O5",
    "D1_O2",
    "D1_O2",
    "D0_O1",
    "D2_O3",
    "D4_O3",
]

cases = [
    {
        "case": "ALLOWED_ABSORB_STEP",
        "slot": 0,
        "target_next": "D4_O3",
        "packet": spectral_packet("D4_O3", strength=0.92),
    },
    {
        "case": "ALLOWED_ABSORB_JUMP",
        "slot": 1,
        "target_next": "D4_O5",
        "packet": spectral_packet("D4_O5", strength=0.90),
    },
    {
        "case": "ALLOWED_RELAX_DOWN",
        "slot": 2,
        "target_next": "D3_O4",
        "packet": spectral_packet("D3_O4", strength=0.89),
    },
    {
        "case": "FORBIDDEN_ZERO_ENERGY",
        "slot": 3,
        "target_next": "D2_O2",
        "packet": spectral_packet("D2_O2", strength=0.91),
    },
    {
        "case": "FORBIDDEN_DIRECTION_JUMP",
        "slot": 4,
        "target_next": "D4_O3",
        "packet": spectral_packet("D4_O3", strength=0.88),
    },
    {
        "case": "WEAK_TRANSITION",
        "slot": 5,
        "target_next": "D1_O2",
        "packet": spectral_packet("D1_O2", strength=0.39),
    },
    {
        "case": "AMBIGUOUS_TRANSITION",
        "slot": 6,
        "target_next": "D3_O4",
        "packet": spectral_packet(
            "D3_O4",
            strength=0.86,
            ambiguous_key="D2_O4",
            ambiguous_strength=0.81
        ),
    },
    {
        "case": "WHITE_ALLOWED_TARGET",
        "slot": 7,
        "target_next": "D5_O4",
        "packet": spectral_packet("D5_O4", strength=0.90, broad=True),
    },
]

memory_state = initial_memory.copy()

decision_rows = []
update_rows = []
wave_rows = []
snapshots = []

applied_count = 0
rejected_count = 0

for t, item in enumerate(cases):
    slot = item["slot"]
    current_key = memory_state[slot]

    decoded = decode_packet(item["packet"])
    decoded_key = decoded["best_key"]

    delta_d, delta_o = transition_delta(current_key, decoded_key)
    rule_ok, rule_reason = transition_allowed(current_key, decoded_key)

    score_ok = bool(decoded["score_ok"])
    applied = bool(score_ok and rule_ok)

    before_key = memory_state[slot]

    if applied:
        memory_state[slot] = decoded_key
        applied_count += 1
        action = "APPLY_TRANSITION"
    else:
        rejected_count += 1

        if not score_ok:
            action = "REJECT_WEAK_OR_AMBIGUOUS"
        elif not rule_ok:
            action = "REJECT_FORBIDDEN_TRANSITION"
        else:
            action = "REJECT_HOLD"

    after_key = memory_state[slot]
    mode = transition_mode(delta_o)

    row = {
        "STATE": STATE,
        "T": t * 120,
        "CASE": item["case"],
        "MEMORY_SLOT": slot,
        "CURRENT_KEY": before_key,
        "PROPOSED_TARGET_KEY": item["target_next"],
        "DECODED_NEXT_KEY": decoded_key,
        "FINAL_KEY": after_key,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "SCORE_OK": int(score_ok),
        "DELTA_D": delta_d if delta_d is not None else "invalid",
        "DELTA_O": delta_o if delta_o is not None else "invalid",
        "TRANSITION_MODE": mode,
        "RULE_ALLOWED": int(rule_ok),
        "RULE_REASON": rule_reason,
        "APPLIED": int(applied),
        "ACTION": action,
        "PHOTON_TOKEN": photon_token(before_key, decoded_key, accepted=applied),
    }

    decision_rows.append(row)

    update_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "MEMORY_SLOT": slot,
        "BEFORE_KEY": before_key,
        "DECODED_NEXT_KEY": decoded_key,
        "AFTER_KEY": after_key,
        "APPLIED": int(applied),
        "CHANGED": int(before_key != after_key),
        "HELD_IF_REJECTED": int((not applied) and before_key == after_key),
    })

    wave_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "CURRENT_KEY": before_key,
        "NEXT_KEY": decoded_key if applied else "NONE",
        "DELTA_D": delta_d if delta_d is not None else "invalid",
        "DELTA_O": delta_o if delta_o is not None else "invalid",
        "MODE": mode if applied else "NONE",
        "OUTPUT_TOKEN": photon_token(before_key, decoded_key, accepted=applied),
        "APPLIED": int(applied),
    })

    snapshots.append((row, item["packet"], memory_state.copy()))

df_decision = pd.DataFrame(decision_rows)
df_updates = pd.DataFrame(update_rows)
df_wave = pd.DataFrame(wave_rows)

expected_final = initial_memory.copy()

for item in cases:
    slot = item["slot"]
    current_key = expected_final[slot]

    decoded = decode_packet(item["packet"])
    decoded_key = decoded["best_key"]

    rule_ok, _ = transition_allowed(current_key, decoded_key)

    if decoded["score_ok"] and rule_ok:
        expected_final[slot] = decoded_key

transition_accuracy = float(np.mean(np.array(memory_state) == np.array(expected_final)))

rule_safety_accuracy = float(np.mean([
    1 if row["APPLIED"] == 1 else
    1 if row["APPLIED"] == 0 and row["HELD_IF_REJECTED"] == 1 else
    0
    for row in update_rows
]))

result = (
    "SPECTRAL_TRANSITION_RULES_LOCKED"
    if transition_accuracy == 1.0 and rule_safety_accuracy == 1.0
    else "SPECTRAL_TRANSITION_RULES_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": len(memory_state),
    "INITIAL_SEQUENCE": " ".join(initial_memory),
    "FINAL_SEQUENCE": " ".join(memory_state),
    "EXPECTED_FINAL_SEQUENCE": " ".join(expected_final),
    "CASES_TESTED": len(cases),
    "APPLIED_TRANSITIONS": applied_count,
    "REJECTED_TRANSITIONS": rejected_count,
    "TRANSITION_ACCURACY": round(transition_accuracy, 3),
    "RULE_SAFETY_ACCURACY": round(rule_safety_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "MAX_DIRECTION_STEP": max_direction_step,
    "ALLOWED_OCTAVE_JUMPS": ",".join(str(x) for x in sorted(allowed_octave_jumps)),
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8, 3.2 * len(snapshots)))

for ax, (row, packet, mem) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, mem)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:             {spectral_states}")
print(f"directions x octaves:        {directions} x {octaves}")
print(f"memory slots:                {len(memory_state)}")
print(f"initial sequence:            {' '.join(initial_memory)}")
print(f"final sequence:              {' '.join(memory_state)}")
print(f"expected final sequence:     {' '.join(expected_final)}")
print(f"cases tested:                {len(cases)}")
print(f"applied transitions:         {applied_count}")
print(f"rejected transitions:        {rejected_count}")
print(f"transition accuracy:         {round(transition_accuracy, 3)}")
print(f"rule safety accuracy:        {round(rule_safety_accuracy, 3)}")
print(f"resonance threshold:         {resonance_threshold}")
print(f"ambiguity margin:            {ambiguity_margin}")
print(f"max direction step:          {max_direction_step}")
print(f"allowed octave jumps:        {sorted(allowed_octave_jumps)}")

print("\nTransition decision table:")
display(df_decision)

print("\nMemory update table:")
display(df_updates)

print("\nTransition photon table:")
display(df_wave)

print("\nSummary:")
display(df_summary)
```
