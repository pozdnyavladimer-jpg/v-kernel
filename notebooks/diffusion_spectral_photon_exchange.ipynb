```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D30_SPECTRAL_PHOTON_EXCHANGE
# Photon-like spectral exchange between memory cells
#
# Goal:
# source memory cell holds spectral state
# -> readout emits photon-like spectral wave packet
# -> receiver filter accepts only matching direction+octave
# -> accepted packet updates receiver memory
# -> wrong direction / wrong octave / weak / ambiguous packets rejected
# ============================================================

np.random.seed(30)

STATE = "D30_SPECTRAL_PHOTON_EXCHANGE"

directions = 6
octaves = 7
spectral_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12
absorption_threshold = 0.62

source_sequence = [
    "D4_O5",
    "D1_O2",
    "D4_O5",
    "D3_O6",
    "D0_O1",
    "D5_O4",
    "D2_O3",
    "D4_O5",
]

receiver_target_sequence = [
    "D4_O5",
    "D1_O2",
    "D4_O5",
    "D3_O6",
    "D0_O1",
    "D5_O4",
    "D2_O3",
    "D4_O5",
]

receiver_initial_sequence = [
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
    "EMPTY",
]


# ------------------------------------------------------------
# Spectral key helpers
# ------------------------------------------------------------

def parse_key(key):
    """
    Valid spectral key format: D4_O5
    Special non-spectral labels: EMPTY, NONE, WHITE
    """
    if key in ["EMPTY", "NONE", "WHITE", None]:
        return None, None

    if "_" not in key:
        return None, None

    d_part, o_part = key.split("_")
    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))

    return direction, octave


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    return octave * directions + direction


def wave_token(key):
    if key == "WHITE":
        return "PHOTON[WHITE|BROAD_SPECTRUM]"

    direction, octave = parse_key(key)

    if direction is None:
        return "NONE"

    phase = round(direction / directions, 3)
    frequency = octave + 1

    return f"PHOTON[D{direction}|O{octave}|F{frequency}|P{phase}]"


def key_color(key, strength=1.0):
    """
    Safe color mapper.
    Handles normal spectral keys and special labels.
    """
    if key in ["EMPTY", "NONE", None]:
        return "black"

    if key == "WHITE":
        return "white"

    d, o = parse_key(key)

    if d is None or o is None:
        return "gray"

    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)

    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


# ------------------------------------------------------------
# Spectral packet generation / decoding
# ------------------------------------------------------------

def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    # WHITE / broad spectrum background
    if broad:
        packet += 0.18 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    # Main spectral peak
    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(target_key)
        packet[o, d] = strength

    # Competing spectral peak
    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d2, o2 = parse_key(ambiguous_key)
        packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score

    accepted = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "accepted": accepted,
    }


def receiver_accepts(decoded_key, expected_key):
    return decoded_key == expected_key


# ------------------------------------------------------------
# Visualization helpers
# ------------------------------------------------------------

def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.18
    gap = 0.035

    ax.text(x0, y0 + 1.65, title, fontsize=7, ha="left", va="bottom", color="black")

    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE", "WHITE"] else (None, None)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = packet[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)
            edge = "lime" if is_selected else "white"
            lw = 1.4 if is_selected else 0.3

            ax.add_patch(
                Rectangle(
                    (x, y),
                    cell,
                    cell,
                    facecolor=color,
                    edgecolor=edge,
                    linewidth=lw
                )
            )


def draw_memory_slots(ax, x0, y0, seq, title, active_slot=None):
    cell = 0.22
    gap = 0.06

    ax.text(x0, y0 + 2.25, title, fontsize=7, ha="left", va="bottom", color="black")

    for i, key in enumerate(seq):
        x = x0
        y = y0 + (len(seq) - 1 - i) * (cell + gap)

        color = key_color(key)
        edge = "lime" if i == active_slot else "black"
        lw = 1.6 if i == active_slot else 0.5

        ax.add_patch(
            Rectangle(
                (x, y),
                cell,
                cell,
                facecolor=color,
                edgecolor=edge,
                linewidth=lw
            )
        )

        ax.text(
            x + 0.34,
            y + cell / 2,
            str(i),
            fontsize=6,
            ha="left",
            va="center",
            color="black"
        )


def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.3

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=14,
        linewidth=2.0,
        color=color,
        alpha=alpha
    )

    ax.add_patch(arrow)


def draw_snapshot(ax, case_row, packet, receiver_memory):
    slot = case_row["MEMORY_SLOT"]
    source_key = case_row["SOURCE_KEY"]
    emitted_key = case_row["EMITTED_KEY"]
    decoded_key = case_row["DECODED_KEY"]
    accepted = bool(case_row["ACCEPTED"])
    absorbed = bool(case_row["ABSORBED"])

    ax.set_facecolor("white")
    ax.set_xlim(0, 12)
    ax.set_ylim(0.6, 5.2)
    ax.set_aspect("equal")
    ax.axis("off")

    source_color = key_color(source_key)
    photon_color = key_color(emitted_key)
    decoded_color = key_color(decoded_key)

    receiver_key = receiver_memory[slot] if absorbed else "EMPTY"
    receiver_color = key_color(receiver_key)

    # Source cell
    ax.add_patch(
        Circle(
            (1.0, 3.2),
            0.45,
            facecolor=source_color,
            edgecolor="black",
            linewidth=1.0
        )
    )
    ax.text(1.0, 2.55, f"SRC\n{source_key}", fontsize=7, ha="center", color="black")

    # Emitted photon
    ax.add_patch(
        Circle(
            (2.4, 3.2),
            0.30,
            facecolor=photon_color,
            edgecolor="black",
            linewidth=1.0
        )
    )
    ax.text(2.4, 2.55, "EMIT", fontsize=7, ha="center", color="black")

    # Field / transport
    ax.add_patch(
        Rectangle(
            (3.45, 2.95),
            0.45,
            0.45,
            facecolor="limegreen",
            edgecolor="black",
            linewidth=1.0
        )
    )
    ax.text(3.68, 2.55, "FIELD", fontsize=7, ha="center", color="black")

    # Receiver decoder
    ax.add_patch(
        Circle(
            (5.2, 3.2),
            0.45,
            facecolor=decoded_color,
            edgecolor="lime" if accepted else "red",
            linewidth=2.0
        )
    )
    ax.text(5.2, 2.45, f"RECV\n{decoded_key}", fontsize=7, ha="center", color="black")

    # Receiver memory cell
    ax.add_patch(
        Circle(
            (6.7, 3.2),
            0.45,
            facecolor=receiver_color,
            edgecolor="lime" if absorbed else "red",
            linewidth=2.0
        )
    )
    ax.text(6.7, 2.45, f"MEM\nslot {slot}", fontsize=7, ha="center", color="black")

    rejected = not absorbed

    draw_arrow(ax, (1.45, 3.2), (2.1, 3.2), active=True, rejected=False)
    draw_arrow(ax, (2.7, 3.2), (3.45, 3.2), active=True, rejected=False)
    draw_arrow(ax, (3.9, 3.2), (4.75, 3.2), active=True, rejected=rejected)
    draw_arrow(ax, (5.65, 3.2), (6.25, 3.2), active=True, rejected=rejected)

    draw_packet_grid(
        ax,
        7.6,
        2.1,
        packet,
        "42-channel photon packet",
        selected_key=decoded_key
    )

    draw_memory_slots(
        ax,
        10.6,
        1.4,
        receiver_memory,
        "receiver memory",
        active_slot=slot
    )

    wave = wave_token(emitted_key)

    ax.text(
        0.6,
        0.9,
        f"case={case_row['CASE']} slot={slot} emitted={emitted_key} decoded={decoded_key} "
        f"score={case_row['BEST_SCORE']} action={case_row['ACTION']}\n{wave}",
        fontsize=7,
        ha="left",
        color="black"
    )

    ax.set_title(
        f"D30 snapshot t={case_row['T']} — {case_row['CASE']} / {case_row['ACTION']}",
        fontsize=10,
        color="black"
    )


# ------------------------------------------------------------
# Test cases
# ------------------------------------------------------------

cases = [
    {
        "case": "MATCHING_PHOTON",
        "slot": 0,
        "source_key": "D4_O5",
        "emitted_key": "D4_O5",
        "expected_key": "D4_O5",
        "packet": spectral_packet("D4_O5", strength=0.91),
    },
    {
        "case": "WRONG_DIRECTION",
        "slot": 1,
        "source_key": "D0_O2",
        "emitted_key": "D0_O2",
        "expected_key": "D1_O2",
        "packet": spectral_packet("D0_O2", strength=0.90),
    },
    {
        "case": "WRONG_OCTAVE",
        "slot": 2,
        "source_key": "D4_O1",
        "emitted_key": "D4_O1",
        "expected_key": "D4_O5",
        "packet": spectral_packet("D4_O1", strength=0.88),
    },
    {
        "case": "WHITE_WITH_TARGET",
        "slot": 3,
        "source_key": "WHITE",
        "emitted_key": "D3_O6",
        "expected_key": "D3_O6",
        "packet": spectral_packet("D3_O6", strength=0.89, broad=True),
    },
    {
        "case": "WEAK_PHOTON",
        "slot": 4,
        "source_key": "D0_O1",
        "emitted_key": "D0_O1",
        "expected_key": "D0_O1",
        "packet": spectral_packet("D0_O1", strength=0.39),
    },
    {
        "case": "AMBIGUOUS_PHOTON",
        "slot": 5,
        "source_key": "D5_O4",
        "emitted_key": "D5_O4",
        "expected_key": "D5_O4",
        "packet": spectral_packet(
            "D5_O4",
            strength=0.86,
            ambiguous_key="D2_O3",
            ambiguous_strength=0.81
        ),
    },
    {
        "case": "MATCHING_PHOTON_2",
        "slot": 6,
        "source_key": "D2_O3",
        "emitted_key": "D2_O3",
        "expected_key": "D2_O3",
        "packet": spectral_packet("D2_O3", strength=0.93),
    },
    {
        "case": "MATCHING_PHOTON_3",
        "slot": 7,
        "source_key": "D4_O5",
        "emitted_key": "D4_O5",
        "expected_key": "D4_O5",
        "packet": spectral_packet("D4_O5", strength=0.87),
    },
]


# ------------------------------------------------------------
# Simulation
# ------------------------------------------------------------

receiver_memory = receiver_initial_sequence.copy()

decision_rows = []
absorption_rows = []
wave_rows = []
snapshots = []

accepted_count = 0
rejected_count = 0
absorbed_count = 0

for t, item in enumerate(cases):
    decoded = decode_packet(item["packet"])

    decoded_key = decoded["best_key"]
    expected_key = item["expected_key"]

    accepted = bool(decoded["accepted"])
    matches_receiver = receiver_accepts(decoded_key, expected_key)
    absorbed = accepted and matches_receiver

    before_key = receiver_memory[item["slot"]]

    if absorbed:
        receiver_memory[item["slot"]] = decoded_key
        absorbed_count += 1

    if accepted:
        accepted_count += 1
    else:
        rejected_count += 1

    action = "ABSORB_AND_WRITE" if absorbed else "REJECT_HOLD"

    row = {
        "STATE": STATE,
        "T": t * 120,
        "CASE": item["case"],
        "MEMORY_SLOT": item["slot"],
        "SOURCE_KEY": item["source_key"],
        "EMITTED_KEY": item["emitted_key"],
        "EXPECTED_RECEIVER_KEY": expected_key,
        "DECODED_KEY": decoded_key,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "ACCEPTED": int(accepted),
        "RESONANCE_MATCH": int(matches_receiver),
        "ABSORBED": int(absorbed),
        "ACTION": action,
    }

    decision_rows.append(row)

    absorption_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "MEMORY_SLOT": item["slot"],
        "BEFORE_KEY": before_key,
        "EXPECTED_KEY": expected_key,
        "AFTER_KEY": receiver_memory[item["slot"]],
        "ACCEPTED": int(accepted),
        "ABSORBED": int(absorbed),
        "MATCH_IF_ABSORBED": int(receiver_memory[item["slot"]] == expected_key) if absorbed else "not_written",
    })

    wave_rows.append({
        "STATE": STATE,
        "CASE": item["case"],
        "EMITTED_KEY": item["emitted_key"],
        "DECODED_KEY": decoded_key,
        "OUTPUT_TOKEN": wave_token(decoded_key if accepted else "NONE"),
        "ACCEPTED": int(accepted),
        "ABSORBED": int(absorbed),
    })

    snapshots.append((row, item["packet"], receiver_memory.copy()))


df_decision = pd.DataFrame(decision_rows)
df_absorption = pd.DataFrame(absorption_rows)
df_wave = pd.DataFrame(wave_rows)


# ------------------------------------------------------------
# Expected final receiver memory
# ------------------------------------------------------------

expected_final = receiver_initial_sequence.copy()

for item in cases:
    decoded = decode_packet(item["packet"])

    if decoded["accepted"] and decoded["best_key"] == item["expected_key"]:
        expected_final[item["slot"]] = item["expected_key"]


exchange_accuracy = float(
    np.mean(np.array(receiver_memory) == np.array(expected_final))
)

absorption_accuracy = float(
    np.mean([
        1 if row["ABSORBED"] == 1 and row["AFTER_KEY"] == row["EXPECTED_KEY"] else
        1 if row["ABSORBED"] == 0 and row["AFTER_KEY"] == row["BEFORE_KEY"] else
        0
        for row in absorption_rows
    ])
)

result = (
    "SPECTRAL_PHOTON_EXCHANGE_LOCKED"
    if exchange_accuracy == 1.0 and absorption_accuracy == 1.0
    else "SPECTRAL_PHOTON_EXCHANGE_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": len(receiver_memory),
    "SOURCE_SEQUENCE": " ".join(source_sequence),
    "RECEIVER_TARGET_SEQUENCE": " ".join(receiver_target_sequence),
    "RECEIVER_FINAL_SEQUENCE": " ".join(receiver_memory),
    "EXPECTED_FINAL_SEQUENCE": " ".join(expected_final),
    "CASES_TESTED": len(cases),
    "ACCEPTED_PACKETS": accepted_count,
    "REJECTED_PACKETS": rejected_count,
    "ABSORBED_PACKETS": absorbed_count,
    "EXCHANGE_ACCURACY": round(exchange_accuracy, 3),
    "ABSORPTION_ACCURACY": round(absorption_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "ABSORPTION_THRESHOLD": absorption_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8, 3.2 * len(snapshots)))

for ax, (row, packet, memory_state) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, memory_state)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"memory slots:                 {len(receiver_memory)}")
print(f"source sequence:              {' '.join(source_sequence)}")
print(f"receiver target sequence:     {' '.join(receiver_target_sequence)}")
print(f"receiver final sequence:      {' '.join(receiver_memory)}")
print(f"expected final sequence:      {' '.join(expected_final)}")
print(f"cases tested:                 {len(cases)}")
print(f"accepted packets:             {accepted_count}")
print(f"rejected packets:             {rejected_count}")
print(f"absorbed packets:             {absorbed_count}")
print(f"exchange accuracy:            {round(exchange_accuracy, 3)}")
print(f"absorption accuracy:          {round(absorption_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")

print("\nPhoton exchange decision table:")
display(df_decision)

print("\nReceiver absorption table:")
display(df_absorption)

print("\nPhoton wave table:")
display(df_wave)

print("\nSummary:")
display(df_summary)
```
