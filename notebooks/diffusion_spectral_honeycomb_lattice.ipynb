```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D38_SPECTRAL_HONEYCOMB_LATTICE
# Honeycomb fabric of orbital-intent spectral memory cells
#
# D36 = phase-coherent bus
# D37 = orbital intent memory cell
# D38 = many orbital cells connected as honeycomb lattice
#
# Rule:
# packet writes only if:
#   spectral key matches
#   orbital intent matches
#   phase clock is locked / resynced
#   jitter is safe
#   source and receiver are adjacent in lattice
# ============================================================

np.random.seed(38)

STATE = "D38_SPECTRAL_HONEYCOMB_LATTICE"

directions = 6
octaves = 7
spectral_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12
phase_tolerance = 0.18
jitter_tolerance = 0.16
intent_threshold = 0.70

DRAW_SNAPSHOTS = True

cell_ids = ["C0", "N0", "N1", "N2", "N3", "N4", "N5"]

receiver_filters = {
    "C0": "D4_O5",
    "N0": "D5_O6",
    "N1": "D1_O2",
    "N2": "D3_O6",
    "N3": "D0_O1",
    "N4": "D2_O3",
    "N5": "D4_O5",
}

receiver_intents = {
    "C0": "FLOWER",
    "N0": "HEX",
    "N1": "TWO_LOBE",
    "N2": "FLOWER",
    "N3": "RING",
    "N4": "BUTTERFLY",
    "N5": "FLOWER",
}

memory = {c: "EMPTY" for c in cell_ids}
orbital_memory = {c: "NONE" for c in cell_ids}

positions = {"C0": (0.0, 0.0)}
for i in range(6):
    a = np.pi / 3 * i
    positions[f"N{i}"] = (np.cos(a), np.sin(a))

edges = []
for i in range(6):
    edges.append(("C0", f"N{i}"))
    edges.append((f"N{i}", f"N{(i + 1) % 6}"))

edge_set = set(tuple(sorted(e)) for e in edges)


def adjacent(a, b):
    if a == "EXT":
        return True
    return tuple(sorted([a, b])) in edge_set


def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE"]:
        return None, None
    parts = str(key).split("_")
    if len(parts) != 2:
        return None, None
    d = int(parts[0].replace("D", ""))
    o = int(parts[1].replace("O", ""))
    return d, o


def make_key(d, o):
    return f"D{d}_O{o}"


def phase_of(key):
    d, _ = parse_key(key)
    if d is None:
        return 0.0
    return round(d / directions, 3)


def circular_phase_error(a, b):
    diff = abs((a - b) % 1.0)
    return min(diff, 1.0 - diff)


def key_color(key):
    if key in [None, "EMPTY", "NONE"]:
        return "black"
    if key == "WHITE":
        return "white"
    d, o = parse_key(key)
    hue = (d + 1) / (directions + 1)
    octave_shift = 0.08 * ((o + 1) / (octaves + 1))
    return plt.cm.hsv((hue + octave_shift) % 1.0)


def intent_color(intent):
    return {
        "FLOWER": "gold",
        "HEX": "limegreen",
        "RING": "cyan",
        "TWO_LOBE": "orange",
        "BUTTERFLY": "magenta",
        "SPIRAL": "violet",
        "NONE": "black",
    }.get(intent, "gray")


def spectral_packet(target_key=None, strength=0.9, noise=0.025,
                    broad=False, ambiguous_key=None, ambiguous_strength=0.0):
    p = np.random.normal(0.0, noise, size=(octaves, directions))
    p = np.clip(p, 0.0, 1.0)

    if broad:
        p += 0.18 + np.random.normal(0.0, noise, size=(octaves, directions))
        p = np.clip(p, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(target_key)
        p[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(ambiguous_key)
        p[o, d] = ambiguous_strength

    return np.clip(p, 0.0, 1.0)


def decode_packet(p):
    flat = p.flatten()
    order = np.argsort(flat)[::-1]

    best_id = int(order[0])
    second_id = int(order[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_o = best_id // directions
    best_d = best_id % directions

    second_o = second_id // directions
    second_d = second_id % directions

    best_key = make_key(best_d, best_o)
    second_key = make_key(second_d, second_o)

    margin = best_score - second_score
    accepted = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_score": second_score,
        "margin": margin,
        "accepted": accepted,
    }


def photon_token(key, phase, amp, jitter, intent):
    d, o = parse_key(key)
    if d is None:
        return "NONE"
    return f"HONEYCOMB_PHOTON[D{d}|O{o}|F{o+1}|P{round(phase,3)}|A{round(amp,3)}|J{round(jitter,3)}|I:{intent}]"


def resync_phase(expected_key, measured_phase, allow_resync):
    expected = phase_of(expected_key)
    raw_err = circular_phase_error(measured_phase, expected)

    if allow_resync and raw_err > phase_tolerance:
        corrected = (expected + np.random.normal(0.0, 0.004)) % 1.0
        corrected_err = circular_phase_error(corrected, expected)
        return corrected, raw_err, corrected_err, True

    return measured_phase, raw_err, raw_err, False


events = [
    {
        "case": "BOOTSTRAP_CENTER_FLOWER",
        "source": "EXT", "receiver": "C0",
        "source_key": "WHITE", "emit": "D4_O5", "intent": "FLOWER",
        "strength": 0.92, "drift": 0.01, "jitter": 0.02,
        "resync": False, "expected_absorbed": True,
    },
    {
        "case": "CENTER_TO_N0_HEX",
        "source": "C0", "receiver": "N0",
        "source_key": "D4_O5", "emit": "D5_O6", "intent": "HEX",
        "strength": 0.91, "drift": -0.02, "jitter": 0.03,
        "resync": False, "expected_absorbed": True,
    },
    {
        "case": "CENTER_TO_N1_TWO_LOBE",
        "source": "C0", "receiver": "N1",
        "source_key": "D4_O5", "emit": "D1_O2", "intent": "TWO_LOBE",
        "strength": 0.90, "drift": 0.03, "jitter": 0.03,
        "resync": False, "expected_absorbed": True,
    },
    {
        "case": "WRONG_DIRECTION_TO_N1_REJECT",
        "source": "C0", "receiver": "N1",
        "source_key": "D4_O5", "emit": "D0_O2", "intent": "TWO_LOBE",
        "strength": 0.90, "drift": 0.01, "jitter": 0.03,
        "resync": False, "expected_absorbed": False,
    },
    {
        "case": "WRONG_ORBITAL_INTENT_N4_REJECT",
        "source": "C0", "receiver": "N4",
        "source_key": "D4_O5", "emit": "D2_O3", "intent": "RING",
        "strength": 0.89, "drift": 0.02, "jitter": 0.03,
        "resync": False, "expected_absorbed": False,
    },
    {
        "case": "NON_NEIGHBOR_ROUTE_REJECT",
        "source": "N0", "receiver": "N3",
        "source_key": "D5_O6", "emit": "D0_O1", "intent": "RING",
        "strength": 0.88, "drift": 0.02, "jitter": 0.03,
        "resync": False, "expected_absorbed": False,
    },
    {
        "case": "LARGE_PHASE_DRIFT_REJECT",
        "source": "C0", "receiver": "N2",
        "source_key": "D4_O5", "emit": "D3_O6", "intent": "FLOWER",
        "strength": 0.91, "drift": 0.34, "jitter": 0.03,
        "resync": False, "expected_absorbed": False,
    },
    {
        "case": "RESYNC_N2_FLOWER_ACCEPT",
        "source": "C0", "receiver": "N2",
        "source_key": "D4_O5", "emit": "D3_O6", "intent": "FLOWER",
        "strength": 0.90, "drift": 0.31, "jitter": 0.04,
        "resync": True, "expected_absorbed": True,
    },
    {
        "case": "DECOHERENCE_JITTER_REJECT",
        "source": "C0", "receiver": "N3",
        "source_key": "D4_O5", "emit": "D0_O1", "intent": "RING",
        "strength": 0.90, "drift": 0.02, "jitter": 0.29,
        "resync": False, "expected_absorbed": False,
    },
    {
        "case": "WEAK_ORBITAL_PACKET_REJECT",
        "source": "C0", "receiver": "N4",
        "source_key": "D4_O5", "emit": "D2_O3", "intent": "BUTTERFLY",
        "strength": 0.39, "drift": 0.02, "jitter": 0.02,
        "resync": False, "expected_absorbed": False,
    },
    {
        "case": "AMBIGUOUS_ORBITAL_HOLD",
        "source": "C0", "receiver": "N5",
        "source_key": "D4_O5", "emit": "D4_O5", "intent": "FLOWER",
        "strength": 0.86, "ambiguous_key": "D1_O2", "ambiguous_strength": 0.81,
        "drift": 0.02, "jitter": 0.02,
        "resync": False, "expected_absorbed": False,
    },
    {
        "case": "CLEAN_N3_RING_ACCEPT",
        "source": "C0", "receiver": "N3",
        "source_key": "D4_O5", "emit": "D0_O1", "intent": "RING",
        "strength": 0.89, "drift": -0.015, "jitter": 0.03,
        "resync": False, "expected_absorbed": True,
    },
    {
        "case": "CLEAN_N4_BUTTERFLY_ACCEPT",
        "source": "C0", "receiver": "N4",
        "source_key": "D4_O5", "emit": "D2_O3", "intent": "BUTTERFLY",
        "strength": 0.93, "drift": 0.02, "jitter": 0.03,
        "resync": False, "expected_absorbed": True,
    },
    {
        "case": "CLEAN_N5_FLOWER_ACCEPT",
        "source": "C0", "receiver": "N5",
        "source_key": "D4_O5", "emit": "D4_O5", "intent": "FLOWER",
        "strength": 0.90, "drift": -0.02, "jitter": 0.03,
        "resync": False, "expected_absorbed": True,
    },
]


rows = []
routing_rows = []
clock_rows = []
memory_rows = []
snapshots = []

absorbed_count = 0
rejected_count = 0
resync_count = 0
amp_accept_count = 0
phase_lock_count = 0
intent_lock_count = 0
adjacency_reject_count = 0

for i, ev in enumerate(events):
    p = spectral_packet(
        target_key=ev["emit"],
        strength=ev["strength"],
        ambiguous_key=ev.get("ambiguous_key"),
        ambiguous_strength=ev.get("ambiguous_strength", 0.0),
    )

    decoded = decode_packet(p)

    src = ev["source"]
    rec = ev["receiver"]
    filter_key = receiver_filters[rec]
    filter_intent = receiver_intents[rec]

    decoded_key = decoded["best_key"]
    decoded_intent = ev["intent"]

    measured_phase = (phase_of(decoded_key) + ev["drift"]) % 1.0
    corrected_phase, raw_phase_error, phase_error, did_resync = resync_phase(
        filter_key,
        measured_phase,
        ev["resync"]
    )

    amplitude_ok = bool(decoded["accepted"])
    phase_ok = phase_error <= phase_tolerance
    jitter_ok = ev["jitter"] <= jitter_tolerance
    intent_ok = decoded_intent == filter_intent
    filter_match = decoded_key == filter_key
    adjacency_ok = adjacent(src, rec)

    absorbed = (
        amplitude_ok
        and phase_ok
        and jitter_ok
        and intent_ok
        and filter_match
        and adjacency_ok
    )

    before_key = memory[rec]
    before_intent = orbital_memory[rec]

    if absorbed:
        memory[rec] = decoded_key
        orbital_memory[rec] = decoded_intent
        absorbed_count += 1
        action = "HONEYCOMB_ROUTE_AND_WRITE"
    else:
        rejected_count += 1
        if not amplitude_ok:
            action = "REJECT_WEAK_OR_AMBIGUOUS"
        elif not adjacency_ok:
            action = "REJECT_NON_NEIGHBOR_ROUTE"
            adjacency_reject_count += 1
        elif not filter_match:
            action = "REJECT_FILTER_MISMATCH"
        elif not intent_ok:
            action = "REJECT_INTENT_MISMATCH"
        elif not phase_ok:
            action = "REJECT_PHASE_UNLOCKED"
        elif not jitter_ok:
            action = "REJECT_DECOHERENCE"
        else:
            action = "REJECT_HOLD"

    if amplitude_ok:
        amp_accept_count += 1
    if phase_ok:
        phase_lock_count += 1
    if intent_ok:
        intent_lock_count += 1
    if did_resync:
        resync_count += 1

    token = photon_token(decoded_key, corrected_phase, decoded["best_score"], ev["jitter"], decoded_intent) if amplitude_ok else "NONE"

    row = {
        "STATE": STATE,
        "T": i * 120,
        "CASE": ev["case"],
        "SOURCE_CELL": src,
        "RECEIVER_CELL": rec,
        "SOURCE_KEY": ev["source_key"],
        "EMITTED_KEY": ev["emit"],
        "RECEIVER_FILTER_KEY": filter_key,
        "DECODED_KEY": decoded_key,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "PACKET_ACCEPTED": int(amplitude_ok),
        "EMITTED_INTENT": ev["intent"],
        "RECEIVER_INTENT": filter_intent,
        "INTENT_OK": int(intent_ok),
        "EXPECTED_PHASE": round(phase_of(filter_key), 3),
        "MEASURED_PHASE": round(measured_phase, 3),
        "RAW_PHASE_ERROR": round(raw_phase_error, 3),
        "CLOCK_RESYNC": int(did_resync),
        "CORRECTED_PHASE": round(corrected_phase, 3),
        "PHASE_ERROR": round(phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "JITTER": round(ev["jitter"], 3),
        "JITTER_OK": int(jitter_ok),
        "ADJACENCY_OK": int(adjacency_ok),
        "FILTER_MATCH": int(filter_match),
        "ABSORBED": int(absorbed),
        "EXPECTED_ABSORBED": int(ev["expected_absorbed"]),
        "BEFORE_KEY": before_key,
        "AFTER_KEY": memory[rec],
        "ACTION": action,
        "OUTPUT_TOKEN": token,
    }

    rows.append(row)

    routing_rows.append({
        "STATE": STATE,
        "CASE": ev["case"],
        "SOURCE_CELL": src,
        "RECEIVER_CELL": rec,
        "ADJACENT": int(adjacency_ok),
        "FILTER_KEY": filter_key,
        "DECODED_KEY": decoded_key,
        "FILTER_MATCH": int(filter_match),
        "INTENT_MATCH": int(intent_ok),
        "ABSORBED": int(absorbed),
        "ACTION": action,
    })

    clock_rows.append({
        "STATE": STATE,
        "CASE": ev["case"],
        "RECEIVER_CELL": rec,
        "DECODED_KEY": decoded_key,
        "EXPECTED_PHASE": round(phase_of(filter_key), 3),
        "MEASURED_PHASE": round(measured_phase, 3),
        "RAW_PHASE_ERROR": round(raw_phase_error, 3),
        "CLOCK_RESYNC": int(did_resync),
        "CORRECTED_PHASE": round(corrected_phase, 3),
        "PHASE_ERROR": round(phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "JITTER": round(ev["jitter"], 3),
        "JITTER_OK": int(jitter_ok),
    })

    memory_rows.append({
        "STATE": STATE,
        "CASE": ev["case"],
        "CELL": rec,
        "BEFORE_KEY": before_key,
        "BEFORE_INTENT": before_intent,
        "AFTER_KEY": memory[rec],
        "AFTER_INTENT": orbital_memory[rec],
        "ABSORBED": int(absorbed),
        "HELD_IF_REJECTED": int((not absorbed) and before_key == memory[rec]),
    })

    snapshots.append((row.copy(), p.copy(), memory.copy(), orbital_memory.copy()))


df_events = pd.DataFrame(rows)
df_routing = pd.DataFrame(routing_rows)
df_clock = pd.DataFrame(clock_rows)
df_memory = pd.DataFrame(memory_rows)

expected_final = receiver_filters.copy()
expected_orbital = receiver_intents.copy()

lattice_accuracy = float(np.mean([memory[c] == expected_final[c] for c in cell_ids]))
orbital_accuracy = float(np.mean([orbital_memory[c] == expected_orbital[c] for c in cell_ids]))
safety_accuracy = float(np.mean([r["ABSORBED"] == r["EXPECTED_ABSORBED"] for r in rows]))

result = (
    "SPECTRAL_HONEYCOMB_LATTICE_LOCKED"
    if lattice_accuracy == 1.0 and orbital_accuracy == 1.0 and safety_accuracy == 1.0
    else "SPECTRAL_HONEYCOMB_LATTICE_PARTIAL"
)

df_final = pd.DataFrame([
    {
        "STATE": STATE,
        "CELL": c,
        "FILTER_KEY": receiver_filters[c],
        "FINAL_KEY": memory[c],
        "EXPECTED_KEY": expected_final[c],
        "ORBITAL_MODE": receiver_intents[c],
        "FINAL_ORBITAL_MODE": orbital_memory[c],
        "MATCH_EXPECTED": int(memory[c] == expected_final[c] and orbital_memory[c] == expected_orbital[c]),
    }
    for c in cell_ids
])

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "HONEYCOMB_CELLS": len(cell_ids),
    "FILTER_SEQUENCE": " ".join([receiver_filters[c] for c in cell_ids]),
    "FINAL_SEQUENCE": " ".join([memory[c] for c in cell_ids]),
    "EXPECTED_SEQUENCE": " ".join([expected_final[c] for c in cell_ids]),
    "FINAL_ORBITAL_SEQUENCE": " ".join([orbital_memory[c] for c in cell_ids]),
    "EXPECTED_ORBITAL_SEQUENCE": " ".join([expected_orbital[c] for c in cell_ids]),
    "EVENTS_TESTED": len(events),
    "AMPLITUDE_ACCEPTED_PACKETS": amp_accept_count,
    "PHASE_LOCKED_PACKETS": phase_lock_count,
    "INTENT_LOCKED_PACKETS": intent_lock_count,
    "CLOCK_RESYNC_EVENTS": resync_count,
    "ABSORBED_PACKETS": absorbed_count,
    "REJECTED_PACKETS": rejected_count,
    "ADJACENCY_REJECTS": adjacency_reject_count,
    "LATTICE_ACCURACY": round(lattice_accuracy, 3),
    "ORBITAL_ACCURACY": round(orbital_accuracy, 3),
    "SAFETY_ACCURACY": round(safety_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "PHASE_TOLERANCE": phase_tolerance,
    "JITTER_TOLERANCE": jitter_tolerance,
    "INTENT_THRESHOLD": intent_threshold,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

def draw_arrow(ax, a, b, rejected=False):
    color = "red" if rejected else "orange"
    ax.add_patch(FancyArrowPatch(a, b, arrowstyle="->", mutation_scale=13, lw=2, color=color))


def draw_grid(ax, x0, y0, packet, selected_key):
    cell = 0.12
    gap = 0.025
    sel = parse_key(selected_key)
    ax.text(x0, y0 + 1.16, "42-channel packet", fontsize=6, ha="left")
    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)
            edge = "lime" if sel == (d, o) else "white"
            lw = 1.3 if sel == (d, o) else 0.2
            ax.add_patch(Rectangle((x, y), cell, cell, facecolor=plt.cm.plasma(packet[o, d]), edgecolor=edge, lw=lw))


def draw_lattice(ax, x0, y0, mem, orb, active=None, src=None):
    scale = 0.55

    for a, b in edge_set:
        xa, ya = positions[a]
        xb, yb = positions[b]
        ax.plot([x0 + xa * scale, x0 + xb * scale],
                [y0 + ya * scale, y0 + yb * scale],
                color="lightgray", lw=1)

    for c in cell_ids:
        x, y = positions[c]
        px = x0 + x * scale
        py = y0 + y * scale

        edge = "lime" if c == active else "deepskyblue" if c == src else "black"
        lw = 2 if c in [active, src] else 0.6

        ax.add_patch(Circle((px, py), 0.14, facecolor=key_color(mem[c]), edgecolor=edge, lw=lw))
        ax.text(px, py - 0.25, c, fontsize=6, ha="center")

        if orb[c] != "NONE":
            ax.add_patch(Circle((px + 0.10, py + 0.10), 0.045, facecolor=intent_color(orb[c]), edgecolor="black", lw=0.3))


def draw_snapshot(ax, row, packet, mem_state, orb_state):
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 4)
    ax.set_aspect("equal")
    ax.axis("off")

    rejected = not bool(row["ABSORBED"])

    src_key = row["SOURCE_KEY"]
    emit_key = row["EMITTED_KEY"]
    dec_key = row["DECODED_KEY"]
    rec = row["RECEIVER_CELL"]
    src = row["SOURCE_CELL"]

    ax.add_patch(Circle((0.7, 2.4), 0.28, facecolor=key_color(src_key), edgecolor="black"))
    ax.text(0.7, 1.95, f"{src}\n{src_key}", fontsize=6, ha="center")

    ax.add_patch(Circle((1.6, 2.4), 0.20, facecolor=key_color(emit_key), edgecolor="black"))
    ax.text(1.6, 1.95, "PHOTON", fontsize=6, ha="center")

    ax.add_patch(Rectangle((2.25, 2.22), 0.34, 0.34, facecolor="limegreen", edgecolor="black"))
    ax.text(2.42, 1.95, "BUS", fontsize=6, ha="center")

    clock_color = "limegreen" if row["PHASE_OK"] and row["JITTER_OK"] else "red"
    ax.add_patch(Rectangle((3.1, 2.22), 0.34, 0.34, facecolor=clock_color, edgecolor="black"))
    ax.text(3.27, 1.95, "CLOCK", fontsize=6, ha="center")

    ax.add_patch(Circle((4.05, 2.4), 0.28, facecolor=key_color(dec_key), edgecolor="lime" if row["PACKET_ACCEPTED"] else "red", lw=2))
    ax.text(4.05, 1.88, f"DECODE\n{dec_key}", fontsize=6, ha="center")

    ax.add_patch(Circle((5.05, 2.4), 0.28, facecolor=key_color(mem_state[rec]), edgecolor="lime" if row["ABSORBED"] else "red", lw=2))
    ax.text(5.05, 1.88, f"{rec}\n{mem_state[rec]}", fontsize=6, ha="center")

    ax.add_patch(Circle((5.75, 2.4), 0.16, facecolor=intent_color(row["EMITTED_INTENT"]), edgecolor="lime" if row["INTENT_OK"] else "red", lw=1.5))
    ax.text(5.75, 1.95, row["EMITTED_INTENT"], fontsize=5.5, ha="center")

    draw_arrow(ax, (0.98, 2.4), (1.4, 2.4))
    draw_arrow(ax, (1.82, 2.4), (2.25, 2.4))
    draw_arrow(ax, (2.6, 2.4), (3.1, 2.4))
    draw_arrow(ax, (3.45, 2.4), (3.78, 2.4), rejected=rejected)
    draw_arrow(ax, (4.32, 2.4), (4.78, 2.4), rejected=rejected)

    draw_grid(ax, 6.35, 1.75, packet, dec_key)
    draw_lattice(ax, 8.7, 2.3, mem_state, orb_state, active=rec, src=None if src == "EXT" else src)

    ax.text(
        0.35,
        0.45,
        f"case={row['CASE']} source={src} receiver={rec} decoded={dec_key} "
        f"score={row['BEST_SCORE']} phase_ok={row['PHASE_OK']} "
        f"jitter_ok={row['JITTER_OK']} intent_ok={row['INTENT_OK']} "
        f"adjacent={row['ADJACENCY_OK']} action={row['ACTION']}\n{row['OUTPUT_TOKEN']}",
        fontsize=6,
        ha="left"
    )

    ax.set_title(f"D38 snapshot t={row['T']} — {row['CASE']} / {row['ACTION']}", fontsize=9)


if DRAW_SNAPSHOTS:
    fig, axes = plt.subplots(len(snapshots), 1, figsize=(8, 2.85 * len(snapshots)))
    if len(snapshots) == 1:
        axes = [axes]

    for ax, (row, packet, mem_state, orb_state) in zip(axes, snapshots):
        draw_snapshot(ax, row, packet, mem_state, orb_state)

    plt.tight_layout()
    plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"honeycomb cells:              {len(cell_ids)}")
print(f"filter sequence:              {' '.join([receiver_filters[c] for c in cell_ids])}")
print(f"final sequence:               {' '.join([memory[c] for c in cell_ids])}")
print(f"expected sequence:            {' '.join([expected_final[c] for c in cell_ids])}")
print(f"final orbital sequence:       {' '.join([orbital_memory[c] for c in cell_ids])}")
print(f"expected orbital sequence:    {' '.join([expected_orbital[c] for c in cell_ids])}")
print(f"events tested:                {len(events)}")
print(f"amplitude accepted packets:   {amp_accept_count}")
print(f"phase locked packets:         {phase_lock_count}")
print(f"intent locked packets:        {intent_lock_count}")
print(f"clock resync events:          {resync_count}")
print(f"absorbed packets:             {absorbed_count}")
print(f"rejected packets:             {rejected_count}")
print(f"adjacency rejects:            {adjacency_reject_count}")
print(f"lattice accuracy:             {round(lattice_accuracy, 3)}")
print(f"orbital accuracy:             {round(orbital_accuracy, 3)}")
print(f"safety accuracy:              {round(safety_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"phase tolerance:              {phase_tolerance}")
print(f"jitter tolerance:             {jitter_tolerance}")
print(f"intent threshold:             {intent_threshold}")

print("\nHoneycomb event table:")
display(df_events)

print("\nRouting / adjacency table:")
display(df_routing)

print("\nClock / phase table:")
display(df_clock)

print("\nOrbital memory update table:")
display(df_memory)

print("\nFinal honeycomb memory table:")
display(df_final)

print("\nSummary:")
display(df_summary)
```
