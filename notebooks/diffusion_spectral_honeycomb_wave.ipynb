```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D39_SPECTRAL_HONEYCOMB_MEMORY_WAVE
# Honeycomb orbital-memory wave propagation / macro resonance
#
# C0 holds a spectral orbital intent, ex: FLOWER.
# It emits a phase-coherent wave into a honeycomb lattice.
# Neighbor cells accept only matching:
#   spectral key + phase + jitter + orbital intent + adjacency.
#
# This is the next layer after D38:
# not single packet routing, but wave propagation through tissue.
# ============================================================

np.random.seed(39)

STATE = "D39_SPECTRAL_HONEYCOMB_MEMORY_WAVE"

directions = 6
octaves = 7
spectral_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12
phase_tolerance = 0.18
jitter_tolerance = 0.16
intent_threshold = 0.82

schumann_hz = 7.83
hop_loss = 0.105
phase_drift_per_hop = 0.035
intent_noise_per_hop = 0.035

target_key = "D4_O5"
target_intent = "FLOWER"

# ------------------------------------------------------------
# Key helpers
# ------------------------------------------------------------

def parse_key(key):
    if key is None:
        return None, None
    if not isinstance(key, str):
        return None, None
    if key in ["EMPTY", "NONE", "WHITE"]:
        return None, None
    if "_" not in key:
        return None, None

    try:
        d_part, o_part = key.split("_")
        d = int(d_part.replace("D", ""))
        o = int(o_part.replace("O", ""))

        if d < 0 or d >= directions:
            return None, None
        if o < 0 or o >= octaves:
            return None, None

        return d, o
    except Exception:
        return None, None


def make_key(direction, octave):
    return f"D{int(direction)}_O{int(octave)}"


def expected_phase(key):
    d, o = parse_key(key)
    if d is None:
        return 0.0
    return round(d / directions, 3)


def phase_error(a, b):
    diff = abs(float(a) - float(b)) % 1.0
    return min(diff, 1.0 - diff)


def key_color(key):
    if key in ["EMPTY", "NONE", None]:
        return "black"
    if key == "WHITE":
        return "white"

    d, o = parse_key(key)
    if d is None:
        return "black"

    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)
    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


def wave_token(key, mode, phase, amplitude, hop, jitter):
    d, o = parse_key(key)
    if d is None:
        return "NONE"

    return (
        f"HONEYCOMB_WAVE[D{d}|O{o}|F{o+1}|"
        f"P{round(float(phase), 3)}|"
        f"A{round(float(amplitude), 3)}|"
        f"J{round(float(jitter), 3)}|"
        f"H{hop}|M{mode}]"
    )

# ------------------------------------------------------------
# Spectral packet
# ------------------------------------------------------------

def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.16 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(target_key)
        if d is not None:
            packet[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d2, o2 = parse_key(ambiguous_key)
        if d2 is not None:
            packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score

    accepted = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "accepted": bool(accepted),
    }

# ------------------------------------------------------------
# Orbital intent
# ------------------------------------------------------------

def orbital_signature(mode):
    mode = str(mode).upper()

    if mode == "FLOWER":
        sig = np.array([1.00, 0.88, 1.00, 0.88, 1.00, 0.88])
    elif mode == "HEX":
        sig = np.array([1.00, 0.25, 1.00, 0.25, 1.00, 0.25])
    elif mode == "RING":
        sig = np.array([0.70, 0.70, 0.70, 0.70, 0.70, 0.70])
    elif mode == "BUTTERFLY":
        sig = np.array([1.00, 0.15, 0.20, 1.00, 0.15, 0.20])
    elif mode == "SPIRAL":
        sig = np.array([0.20, 0.38, 0.62, 0.88, 1.00, 0.55])
    elif mode == "TWO_LOBE":
        sig = np.array([1.00, 0.10, 0.10, 1.00, 0.10, 0.10])
    else:
        sig = np.zeros(6)

    return sig / (np.linalg.norm(sig) + 1e-9)


def degrade_signature(mode, hop, extra_noise=0.0):
    sig = orbital_signature(mode).copy()
    noise = np.random.normal(
        0.0,
        intent_noise_per_hop * hop + extra_noise,
        size=sig.shape
    )
    sig = np.clip(sig + noise, 0.0, None)
    return sig / (np.linalg.norm(sig) + 1e-9)


def intent_score(received_mode, expected_mode, hop, extra_noise=0.0):
    a = degrade_signature(received_mode, hop, extra_noise=extra_noise)
    b = orbital_signature(expected_mode)
    return float(np.dot(a, b))


def draw_orbital_icon(ax, x, y, mode, key, scale=0.22, accepted=True, alpha=1.0):
    sig = orbital_signature(mode)
    base_color = key_color(key)
    edge = "lime" if accepted else "red"

    ax.add_patch(
        Circle(
            (x, y),
            scale * 0.35,
            facecolor=base_color,
            edgecolor=edge,
            linewidth=1.0,
            alpha=alpha
        )
    )

    for i, amp in enumerate(sig):
        theta = 2 * np.pi * i / directions
        r = scale * (0.85 + 0.55 * amp)

        lx = x + r * np.cos(theta)
        ly = y + r * np.sin(theta)

        ax.add_patch(
            Circle(
                (lx, ly),
                scale * (0.22 + 0.18 * amp),
                facecolor=base_color,
                edgecolor=edge,
                linewidth=0.7,
                alpha=0.65 * alpha
            )
        )

# ------------------------------------------------------------
# Honeycomb lattice
# ------------------------------------------------------------

def build_honeycomb():
    nodes = {}
    nodes["C0"] = (0.0, 0.0)

    for i in range(6):
        theta = np.pi / 6 + i * np.pi / 3
        nodes[f"N{i}"] = (1.7 * np.cos(theta), 1.7 * np.sin(theta))
        nodes[f"O{i}"] = (3.0 * np.cos(theta), 3.0 * np.sin(theta))

    edges = []

    for i in range(6):
        edges.append(("C0", f"N{i}"))
        edges.append((f"N{i}", f"N{(i + 1) % 6}"))
        edges.append((f"N{i}", f"O{i}"))

    adjacency = {n: set() for n in nodes}

    for a, b in edges:
        adjacency[a].add(b)
        adjacency[b].add(a)

    return nodes, edges, adjacency


node_pos, lattice_edges, adjacency = build_honeycomb()
node_names = list(node_pos.keys())

node_filter_key = {n: target_key for n in node_names}
node_filter_mode = {n: target_intent for n in node_names}

# deliberate mismatches to prove rejection
node_filter_key["O3"] = "D1_O2"
node_filter_mode["O4"] = "HEX"
node_filter_key["O5"] = "D5_O4"

memory_key = {n: "EMPTY" for n in node_names}
memory_mode = {n: "NONE" for n in node_names}

memory_key["C0"] = target_key
memory_mode["C0"] = target_intent

# ------------------------------------------------------------
# Drawing helpers
# ------------------------------------------------------------

def draw_arrow(ax, start, end, active=True, rejected=False, alpha=1.0):
    color = "red" if rejected else "orange" if active else "gray"

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=11,
        linewidth=1.8,
        color=color,
        alpha=alpha if active else 0.2
    )

    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.10
    gap = 0.018

    ax.text(
        x0,
        y0 + 0.98,
        title,
        fontsize=6,
        ha="left",
        va="bottom",
        color="black"
    )

    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE"] else (None, None)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)

            amp = packet[o, d]
            color = plt.cm.plasma(amp)

            is_selected = selected == (d, o)
            edge = "lime" if is_selected else "white"
            lw = 1.1 if is_selected else 0.25

            ax.add_patch(
                Rectangle(
                    (x, y),
                    cell,
                    cell,
                    facecolor=color,
                    edgecolor=edge,
                    linewidth=lw
                )
            )


def draw_lattice(ax, memory_key_state, memory_mode_state, active_nodes=None, rejected_nodes=None, title=""):
    active_nodes = set(active_nodes or [])
    rejected_nodes = set(rejected_nodes or [])

    ax.set_facecolor("white")
    ax.set_xlim(-4.8, 6.0)
    ax.set_ylim(-3.8, 3.8)
    ax.set_aspect("equal")
    ax.axis("off")

    ax.add_patch(
        Circle(
            (0, 0),
            3.45,
            fill=False,
            edgecolor="goldenrod",
            linewidth=1.0,
            linestyle="--",
            alpha=0.75
        )
    )

    ax.text(
        0,
        3.55,
        f"macro pulse ≈ {schumann_hz} Hz",
        fontsize=8,
        ha="center",
        color="black"
    )

    for a, b in lattice_edges:
        xa, ya = node_pos[a]
        xb, yb = node_pos[b]
        ax.plot([xa, xb], [ya, yb], color="lightgray", linewidth=1.0, zorder=0)

    for n, (x, y) in node_pos.items():
        stored_key = memory_key_state.get(n, "EMPTY")
        stored_mode = memory_mode_state.get(n, "NONE")
        filter_key = node_filter_key.get(n, "EMPTY")

        if n in active_nodes:
            edge = "lime"
            lw = 2.6
        elif n in rejected_nodes:
            edge = "red"
            lw = 2.0
        else:
            edge = "black"
            lw = 0.6

        face = key_color(stored_key)

        ax.add_patch(
            Circle(
                (x, y),
                0.22,
                facecolor=face,
                edgecolor=edge,
                linewidth=lw,
                zorder=2
            )
        )

        ax.text(x, y - 0.43, n, fontsize=6, ha="center", color="black")
        ax.text(x, y + 0.34, filter_key, fontsize=5, ha="center", color="black")

        if stored_mode not in ["NONE", "EMPTY"]:
            draw_orbital_icon(
                ax,
                x,
                y,
                stored_mode,
                stored_key,
                scale=0.12,
                accepted=(n in active_nodes),
                alpha=0.85
            )

    ax.set_title(title, fontsize=10, color="black")


def draw_snapshot(ax, snap):
    draw_lattice(
        ax,
        snap["memory_key"],
        snap["memory_mode"],
        active_nodes=snap["active_nodes"],
        rejected_nodes=snap["rejected_nodes"],
        title=f"D39 t={snap['T']} — {snap['CASE']} / hop {snap['HOP']} / {snap['ACTION']}"
    )

    y = -3.25
    x0 = -4.45

    source_color = key_color(snap["EMITTED_KEY"])
    decoded_color = key_color(snap["DECODED_KEY"])

    ax.add_patch(
        Circle(
            (x0, y),
            0.18,
            facecolor=source_color,
            edgecolor="black",
            linewidth=0.8
        )
    )

    ax.text(x0, y - 0.33, snap["EMITTED_KEY"], fontsize=6, ha="center", color="black")

    ax.add_patch(
        Rectangle(
            (x0 + 0.65, y - 0.12),
            0.24,
            0.24,
            facecolor="limegreen",
            edgecolor="black",
            linewidth=0.8
        )
    )

    ax.text(x0 + 0.77, y - 0.33, "BUS", fontsize=6, ha="center", color="black")

    ax.add_patch(
        Rectangle(
            (x0 + 1.25, y - 0.12),
            0.24,
            0.24,
            facecolor="orange",
            edgecolor="black",
            linewidth=0.8
        )
    )

    ax.text(x0 + 1.37, y - 0.33, "CLOCK", fontsize=6, ha="center", color="black")

    ax.add_patch(
        Circle(
            (x0 + 2.05, y),
            0.18,
            facecolor=decoded_color,
            edgecolor="lime" if snap["PACKET_OK"] else "red",
            linewidth=1.6
        )
    )

    ax.text(x0 + 2.05, y - 0.33, snap["DECODED_KEY"], fontsize=6, ha="center", color="black")

    draw_arrow(ax, (x0 + 0.18, y), (x0 + 0.64, y), active=True, rejected=False)
    draw_arrow(ax, (x0 + 0.90, y), (x0 + 1.25, y), active=True, rejected=not snap["PHASE_OK"])
    draw_arrow(ax, (x0 + 1.50, y), (x0 + 1.87, y), active=True, rejected=not snap["PACKET_OK"])

    draw_packet_grid(
        ax,
        3.35,
        -3.55,
        snap["PACKET"],
        "42-channel wave packet",
        selected_key=snap["DECODED_KEY"]
    )

    ax.text(
        -4.55,
        -3.72,
        f"active={','.join(snap['active_nodes']) if snap['active_nodes'] else 'none'} "
        f"rejected={','.join(snap['rejected_nodes']) if snap['rejected_nodes'] else 'none'} "
        f"score={snap['BEST_SCORE']} "
        f"intent={snap['INTENT_SCORE']} "
        f"phase_err={snap['PHASE_ERROR']} "
        f"action={snap['ACTION']}\n{snap['OUTPUT_TOKEN']}",
        fontsize=6,
        ha="left",
        color="black"
    )

# ------------------------------------------------------------
# Wave events
# ------------------------------------------------------------

events = [
    {
        "case": "FLOWER_CENTER_WAVE",
        "source": "C0",
        "emitted_key": "D4_O5",
        "mode": "FLOWER",
        "strength": 0.94,
        "phase_offset": 0.010,
        "jitter": 0.020,
        "max_hops": 2,
        "broad": False,
        "resync": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "WRONG_DIRECTION_REJECT",
        "source": "C0",
        "emitted_key": "D0_O5",
        "mode": "FLOWER",
        "strength": 0.91,
        "phase_offset": 0.010,
        "jitter": 0.030,
        "max_hops": 1,
        "broad": False,
        "resync": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "WRONG_INTENT_REJECT",
        "source": "C0",
        "emitted_key": "D4_O5",
        "mode": "RING",
        "strength": 0.92,
        "phase_offset": 0.015,
        "jitter": 0.030,
        "max_hops": 1,
        "broad": False,
        "resync": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "LARGE_PHASE_DRIFT_REJECT",
        "source": "C0",
        "emitted_key": "D4_O5",
        "mode": "FLOWER",
        "strength": 0.93,
        "phase_offset": 0.340,
        "jitter": 0.030,
        "max_hops": 1,
        "broad": False,
        "resync": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "DECOHERENCE_JITTER_REJECT",
        "source": "C0",
        "emitted_key": "D4_O5",
        "mode": "FLOWER",
        "strength": 0.91,
        "phase_offset": 0.020,
        "jitter": 0.290,
        "max_hops": 1,
        "broad": False,
        "resync": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "WEAK_WAVE_REJECT",
        "source": "C0",
        "emitted_key": "D4_O5",
        "mode": "FLOWER",
        "strength": 0.39,
        "phase_offset": 0.010,
        "jitter": 0.020,
        "max_hops": 1,
        "broad": False,
        "resync": False,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
    {
        "case": "AMBIGUOUS_WAVE_REJECT",
        "source": "C0",
        "emitted_key": "D4_O5",
        "mode": "FLOWER",
        "strength": 0.86,
        "phase_offset": 0.010,
        "jitter": 0.020,
        "max_hops": 1,
        "broad": False,
        "resync": False,
        "ambiguous_key": "D1_O2",
        "ambiguous_strength": 0.81,
    },
    {
        "case": "RESYNC_FLOWER_LATE_ACCEPT",
        "source": "N0",
        "emitted_key": "D4_O5",
        "mode": "FLOWER",
        "strength": 0.90,
        "phase_offset": 0.310,
        "jitter": 0.040,
        "max_hops": 1,
        "broad": False,
        "resync": True,
        "ambiguous_key": None,
        "ambiguous_strength": 0.0,
    },
]

# ------------------------------------------------------------
# Propagation engine
# ------------------------------------------------------------

route_rows = []
event_rows = []
snapshots = []

absorbed_total = 0
rejected_total = 0
routed_total = 0
resync_total = 0

for event_index, ev in enumerate(events):
    source = ev["source"]
    emitted_key = ev["emitted_key"]
    mode = ev["mode"]

    frontier = [source]
    visited = set([source])

    case_absorbed = []
    case_rejected = []
    case_max_absorbed_hop = 0

    for hop in range(1, ev["max_hops"] + 1):
        next_frontier = []
        active_nodes = []
        rejected_nodes = []

        amplitude = max(0.0, ev["strength"] - hop_loss * (hop - 1))
        phase = (
            expected_phase(emitted_key)
            + ev["phase_offset"]
            + phase_drift_per_hop * (hop - 1)
        ) % 1.0

        jitter = ev["jitter"] + 0.015 * (hop - 1)

        packet = spectral_packet(
            emitted_key,
            strength=amplitude,
            noise=0.025 + 0.005 * hop,
            broad=ev["broad"],
            ambiguous_key=ev["ambiguous_key"],
            ambiguous_strength=ev["ambiguous_strength"],
        )

        decoded = decode_packet(packet)
        decoded_key = decoded["best_key"]

        raw_phase_error = phase_error(phase, expected_phase(decoded_key))

        clock_resync = 0
        corrected_phase = phase
        corrected_phase_error = raw_phase_error

        if ev["resync"] and raw_phase_error <= 0.36:
            corrected_phase = (expected_phase(decoded_key) + 0.003) % 1.0
            corrected_phase_error = phase_error(corrected_phase, expected_phase(decoded_key))
            clock_resync = 1
            resync_total += 1

        phase_ok = corrected_phase_error <= phase_tolerance
        jitter_ok = jitter <= jitter_tolerance
        score_ok = bool(decoded["accepted"])

        candidate_nodes = []

        for src in frontier:
            for nb in adjacency[src]:
                if nb not in visited:
                    candidate_nodes.append((src, nb))

        hop_absorbed_count = 0

        for src, nb in candidate_nodes:
            filter_key = node_filter_key[nb]
            filter_mode = node_filter_mode[nb]

            filter_match = decoded_key == filter_key

            score_intent = intent_score(mode, filter_mode, hop)

            # strict orbital intent check:
            # RING cannot pass as FLOWER even if cosine similarity is high.
            intent_ok = score_intent >= intent_threshold and mode == filter_mode

            absorbed = (
                score_ok
                and phase_ok
                and jitter_ok
                and filter_match
                and intent_ok
            )

            if absorbed:
                memory_key[nb] = decoded_key
                memory_mode[nb] = mode

                next_frontier.append(nb)
                active_nodes.append(nb)
                case_absorbed.append(nb)

                absorbed_total += 1
                hop_absorbed_count += 1
                case_max_absorbed_hop = max(case_max_absorbed_hop, hop)

                action = "HONEYCOMB_ROUTE_AND_WRITE"
            else:
                rejected_nodes.append(nb)
                case_rejected.append(nb)

                rejected_total += 1

                if not score_ok:
                    action = "REJECT_WEAK_OR_AMBIGUOUS"
                elif not phase_ok:
                    action = "REJECT_PHASE_UNLOCKED"
                elif not jitter_ok:
                    action = "REJECT_DECOHERENCE"
                elif not filter_match:
                    action = "REJECT_FILTER_MISMATCH"
                elif not intent_ok:
                    action = "REJECT_INTENT_MISMATCH"
                else:
                    action = "REJECT_HOLD"

            route_rows.append({
                "STATE": STATE,
                "CASE": ev["case"],
                "HOP": hop,
                "SOURCE_NODE": src,
                "RECEIVER_NODE": nb,
                "FILTER_KEY": filter_key,
                "FILTER_MODE": filter_mode,
                "EMITTED_KEY": emitted_key,
                "EMITTED_MODE": mode,
                "DECODED_KEY": decoded_key,
                "BEST_SCORE": round(decoded["best_score"], 3),
                "SECOND_KEY": decoded["second_key"],
                "SECOND_SCORE": round(decoded["second_score"], 3),
                "MARGIN": round(decoded["margin"], 3),
                "PHASE_ERROR": round(corrected_phase_error, 3),
                "JITTER": round(jitter, 3),
                "SCORE_OK": int(score_ok),
                "PHASE_OK": int(phase_ok),
                "JITTER_OK": int(jitter_ok),
                "FILTER_MATCH": int(filter_match),
                "INTENT_SCORE": round(score_intent, 3),
                "INTENT_OK": int(intent_ok),
                "CLOCK_RESYNC": int(clock_resync),
                "ABSORBED": int(absorbed),
                "ACTION": action,
            })

        if hop_absorbed_count > 0:
            routed_total += 1

        snapshot_action = (
            "PROPAGATE_HONEYCOMB_WAVE"
            if hop_absorbed_count > 0
            else "WAVE_DECAY_OR_REJECT"
        )

        if candidate_nodes:
            avg_intent = np.mean([
                intent_score(mode, node_filter_mode[nb], hop)
                for _, nb in candidate_nodes
            ])
        else:
            avg_intent = 0.0

        snapshots.append({
            "T": event_index * 240 + hop * 120,
            "CASE": ev["case"],
            "HOP": hop,
            "ACTION": snapshot_action,
            "EMITTED_KEY": emitted_key,
            "DECODED_KEY": decoded_key,
            "PACKET": packet.copy(),
            "PACKET_OK": score_ok,
            "PHASE_OK": phase_ok,
            "BEST_SCORE": round(decoded["best_score"], 3),
            "INTENT_SCORE": round(float(avg_intent), 3),
            "PHASE_ERROR": round(corrected_phase_error, 3),
            "OUTPUT_TOKEN": wave_token(
                decoded_key if score_ok else "NONE",
                mode,
                corrected_phase,
                decoded["best_score"],
                hop,
                jitter
            ),
            "active_nodes": active_nodes,
            "rejected_nodes": rejected_nodes,
            "memory_key": memory_key.copy(),
            "memory_mode": memory_mode.copy(),
        })

        for n in next_frontier:
            visited.add(n)

        frontier = next_frontier

        if len(frontier) == 0:
            break

    event_rows.append({
        "STATE": STATE,
        "CASE": ev["case"],
        "SOURCE_NODE": source,
        "EMITTED_KEY": emitted_key,
        "EMITTED_MODE": mode,
        "MAX_HOPS_REQUESTED": ev["max_hops"],
        "MAX_HOP_REACHED": case_max_absorbed_hop,
        "ABSORBED_NODES": ",".join(case_absorbed) if case_absorbed else "none",
        "REJECTED_NODES": ",".join(case_rejected) if case_rejected else "none",
        "ROUTED": int(case_max_absorbed_hop > 0),
        "ORBITAL_INTENT_SURVIVED_2_HOPS": int(
            ev["case"] == "FLOWER_CENTER_WAVE"
            and case_max_absorbed_hop >= 2
        ),
    })

df_routes = pd.DataFrame(route_rows)
df_events = pd.DataFrame(event_rows)

df_memory = pd.DataFrame([
    {
        "STATE": STATE,
        "NODE": n,
        "FILTER_KEY": node_filter_key[n],
        "FILTER_MODE": node_filter_mode[n],
        "FINAL_KEY": memory_key[n],
        "FINAL_MODE": memory_mode[n],
        "MATCH_FILTER_IF_WRITTEN": int(memory_key[n] == node_filter_key[n]) if memory_key[n] != "EMPTY" else "empty",
        "INTENT_MATCH_IF_WRITTEN": int(memory_mode[n] == node_filter_mode[n]) if memory_mode[n] != "NONE" else "empty",
    }
    for n in node_names
])

good_wave_hops = int(
    df_events.loc[
        df_events["CASE"] == "FLOWER_CENTER_WAVE",
        "MAX_HOP_REACHED"
    ].iloc[0]
)

bad_absorptions = int(
    df_routes[
        (
            df_routes["CASE"].isin([
                "WRONG_DIRECTION_REJECT",
                "WRONG_INTENT_REJECT",
                "LARGE_PHASE_DRIFT_REJECT",
                "DECOHERENCE_JITTER_REJECT",
                "WEAK_WAVE_REJECT",
                "AMBIGUOUS_WAVE_REJECT",
            ])
        )
        & (df_routes["ABSORBED"] == 1)
    ].shape[0]
)

written_rows = df_memory[df_memory["FINAL_KEY"] != "EMPTY"]

filter_accuracy = (
    float(np.mean(written_rows["MATCH_FILTER_IF_WRITTEN"].astype(int)))
    if len(written_rows)
    else 1.0
)

intent_accuracy = (
    float(np.mean(written_rows["INTENT_MATCH_IF_WRITTEN"].astype(int)))
    if len(written_rows)
    else 1.0
)

wave_locked = (
    good_wave_hops >= 2
    and bad_absorptions == 0
    and filter_accuracy == 1.0
    and intent_accuracy == 1.0
)

result = (
    "SPECTRAL_HONEYCOMB_WAVE_LOCKED"
    if wave_locked
    else "SPECTRAL_HONEYCOMB_WAVE_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "HONEYCOMB_NODES": len(node_names),
    "HONEYCOMB_EDGES": len(lattice_edges),
    "SCHUMANN_REFERENCE_HZ": schumann_hz,
    "TARGET_KEY": target_key,
    "TARGET_ORBITAL_INTENT": target_intent,
    "EVENTS_TESTED": len(events),
    "GOOD_WAVE_MAX_HOP": good_wave_hops,
    "ROUTED_WAVE_EVENTS": routed_total,
    "ABSORBED_TOTAL": absorbed_total,
    "REJECTED_TOTAL": rejected_total,
    "BAD_ABSORPTIONS": bad_absorptions,
    "CLOCK_RESYNC_EVENTS": resync_total,
    "FILTER_ACCURACY": round(filter_accuracy, 3),
    "INTENT_ACCURACY": round(intent_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "PHASE_TOLERANCE": phase_tolerance,
    "JITTER_TOLERANCE": jitter_tolerance,
    "INTENT_THRESHOLD": intent_threshold,
    "HOP_LOSS": hop_loss,
}])

# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

snapshots_to_draw = snapshots[:10]

fig, axes = plt.subplots(
    len(snapshots_to_draw),
    1,
    figsize=(8, 4.8 * len(snapshots_to_draw))
)

axes = np.atleast_1d(axes)

for ax, snap in zip(axes, snapshots_to_draw):
    draw_snapshot(ax, snap)

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"honeycomb nodes:              {len(node_names)}")
print(f"honeycomb edges:              {len(lattice_edges)}")
print(f"Schumann reference:           {schumann_hz} Hz")
print(f"target key:                   {target_key}")
print(f"target orbital intent:        {target_intent}")
print(f"events tested:                {len(events)}")
print(f"good wave max hop:            {good_wave_hops}")
print(f"routed wave events:           {routed_total}")
print(f"absorbed total:               {absorbed_total}")
print(f"rejected total:               {rejected_total}")
print(f"bad absorptions:              {bad_absorptions}")
print(f"clock resync events:          {resync_total}")
print(f"filter accuracy:              {round(filter_accuracy, 3)}")
print(f"intent accuracy:              {round(intent_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"phase tolerance:              {phase_tolerance}")
print(f"jitter tolerance:             {jitter_tolerance}")
print(f"intent threshold:             {intent_threshold}")
print(f"hop loss:                     {hop_loss}")

print("\nHoneycomb event table:")
display(df_events)

print("\nHoneycomb route table:")
display(df_routes)

print("\nFinal honeycomb memory table:")
display(df_memory)

print("\nSummary:")
display(df_summary)
```
