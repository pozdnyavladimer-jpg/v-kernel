```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D42_SELF_TUNING_RESONANCE_FILTER
# Adaptive field-memory filters after D41 decay/reinforcement.
#
# Idea:
# - memory is not static
# - every slot has its own resonance filter
# - coherent repeated intent lowers the local threshold and raises trust
# - bad / weak / ambiguous / wrong-phase packets make the filter stricter
# - unused memory decays and can be forgotten
#
# Success:
# SPECTRAL_SELF_TUNING_FILTER_LOCKED
# ============================================================

np.random.seed(42)

STATE = "D42_SELF_TUNING_RESONANCE_FILTER"

directions = 6
octaves = 7
spectral_states = directions * octaves
memory_slots = 8

base_resonance_threshold = 0.62
base_ambiguity_margin = 0.12
base_phase_tolerance = 0.18
base_jitter_tolerance = 0.16

min_threshold = 0.50
max_threshold = 0.82
min_margin = 0.08
max_margin = 0.22
min_phase_tolerance = 0.10
max_phase_tolerance = 0.26
forget_threshold = 0.24

reinforce_gain = 0.075
reject_penalty = 0.04
decay_amount = 0.035

orbital_modes = ["FLOWER", "RING", "HEX", "SPIRAL", "SEED", "BUTTERFLY", "WAVE", "NONE"]

slot_filters = [
    {"key": "D4_O5", "intent": "FLOWER"},
    {"key": "D1_O2", "intent": "RING"},
    {"key": "D3_O6", "intent": "HEX"},
    {"key": "D5_O4", "intent": "SPIRAL"},
    {"key": "D0_O1", "intent": "SEED"},
    {"key": "D2_O3", "intent": "BUTTERFLY"},
    {"key": "D4_O5", "intent": "FLOWER"},
    {"key": "D1_O2", "intent": "RING"},
]

memory = [
    {"key": "D4_O5", "intent": "FLOWER", "strength": 0.72, "age": 0, "writes": 1},
    {"key": "D1_O2", "intent": "RING", "strength": 0.50, "age": 2, "writes": 1},
    {"key": "D3_O6", "intent": "HEX", "strength": 0.58, "age": 1, "writes": 1},
    {"key": "D5_O4", "intent": "SPIRAL", "strength": 0.45, "age": 4, "writes": 1},
    {"key": "D0_O1", "intent": "SEED", "strength": 0.27, "age": 5, "writes": 1},
    {"key": "D2_O3", "intent": "BUTTERFLY", "strength": 0.63, "age": 1, "writes": 1},
    {"key": "EMPTY", "intent": "NONE", "strength": 0.00, "age": 0, "writes": 0},
    {"key": "D1_O2", "intent": "RING", "strength": 0.41, "age": 3, "writes": 1},
]

adaptive_filters = []
for i in range(memory_slots):
    adaptive_filters.append({
        "slot": i,
        "filter_key": slot_filters[i]["key"],
        "filter_intent": slot_filters[i]["intent"],
        "threshold": base_resonance_threshold,
        "margin": base_ambiguity_margin,
        "phase_tolerance": base_phase_tolerance,
        "jitter_tolerance": base_jitter_tolerance,
        "trust": 0.50,
        "accepts": 0,
        "rejects": 0,
    })


def clamp(value, lo, hi):
    return float(np.clip(value, lo, hi))


def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE"]:
        return None, None
    d_part, o_part = key.split("_")
    return int(d_part.replace("D", "")), int(o_part.replace("O", ""))


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    return octave * directions + direction


def expected_phase(key):
    d, _ = parse_key(key)
    if d is None:
        return None
    return round(d / directions, 3)


def circular_phase_error(a, b):
    diff = abs(a - b) % 1.0
    return min(diff, 1.0 - diff)


def intent_marker(intent):
    markers = {
        "FLOWER": "✿",
        "RING": "○",
        "HEX": "⬡",
        "SPIRAL": "◌",
        "SEED": "•",
        "BUTTERFLY": "∞",
        "WAVE": "~",
        "NONE": "",
    }
    return markers.get(intent, "?")


def key_color(key, strength=1.0):
    if key in [None, "EMPTY", "NONE"]:
        return "black"
    if key == "WHITE":
        return "white"
    d, o = parse_key(key)
    hue_value = (d + 1) / (directions + 1)
    octave_value = (o + 1) / (octaves + 1)
    rgba = list(plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0))
    rgba[3] = clamp(0.35 + 0.65 * strength, 0.35, 1.0)
    return tuple(rgba)


def spectral_packet(
    target_key=None,
    strength=0.90,
    noise=0.025,
    ambiguous_key=None,
    ambiguous_strength=0.0,
    broad=False,
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad or target_key == "WHITE":
        packet += 0.14 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(target_key)
        packet[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE"]:
        d2, o2 = parse_key(ambiguous_key)
        packet[o2, d2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet, threshold, margin_threshold):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions
    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)
    margin = best_score - second_score

    accepted = best_score >= threshold and margin >= margin_threshold

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "accepted": accepted,
    }


def apply_decay(mem, amount=decay_amount):
    if mem["key"] in ["EMPTY", "NONE"]:
        return False

    mem["strength"] = clamp(mem["strength"] - amount, 0.0, 1.0)
    mem["age"] += 1

    if mem["strength"] < forget_threshold:
        mem["key"] = "EMPTY"
        mem["intent"] = "NONE"
        mem["strength"] = 0.0
        mem["writes"] = 0
        return True

    return False


def adapt_filter(filter_state, absorbed, action, memory_strength):
    old = filter_state.copy()

    if absorbed:
        filter_state["accepts"] += 1
        filter_state["trust"] = clamp(filter_state["trust"] + reinforce_gain, 0.0, 1.0)
        filter_state["threshold"] = clamp(filter_state["threshold"] - 0.012, min_threshold, max_threshold)
        filter_state["margin"] = clamp(filter_state["margin"] + 0.006, min_margin, max_margin)
        filter_state["phase_tolerance"] = clamp(
            filter_state["phase_tolerance"] + 0.004,
            min_phase_tolerance,
            max_phase_tolerance,
        )
    else:
        filter_state["rejects"] += 1

        if action in [
            "REJECT_WEAK_OR_AMBIGUOUS",
            "REJECT_FILTER_MISMATCH",
            "REJECT_INTENT_MISMATCH",
            "REJECT_PHASE_UNLOCKED",
            "REJECT_DECOHERENCE",
        ]:
            filter_state["trust"] = clamp(filter_state["trust"] - reject_penalty, 0.0, 1.0)
            filter_state["threshold"] = clamp(filter_state["threshold"] + 0.018, min_threshold, max_threshold)
            filter_state["margin"] = clamp(filter_state["margin"] + 0.008, min_margin, max_margin)
            filter_state["phase_tolerance"] = clamp(
                filter_state["phase_tolerance"] - 0.006,
                min_phase_tolerance,
                max_phase_tolerance,
            )

        if action in ["PASSIVE_DECAY_ONLY", "PASSIVE_FORGET"]:
            pressure = 0.025 * (1.0 - memory_strength)
            filter_state["threshold"] = clamp(filter_state["threshold"] + pressure, min_threshold, max_threshold)
            filter_state["trust"] = clamp(filter_state["trust"] - 0.012, 0.0, 1.0)

    return old, filter_state.copy()


def tune_token(decoded_key, intent, phase, amplitude, action):
    if action.startswith("REJECT") or action.startswith("PASSIVE"):
        return "NONE"
    if decoded_key in [None, "EMPTY", "NONE"]:
        return "NONE"
    d, o = parse_key(decoded_key)
    return f"TUNE_PHOTON[D{d}|O{o}|F{o+1}|P{round(phase,3)}|A{round(amplitude,3)}|I:{intent}]"


def draw_arrow(ax, start, end, ok=True, active=True):
    color = "limegreen" if ok else "red"
    alpha = 1.0 if active else 0.25
    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=10,
        linewidth=1.5,
        color=color,
        alpha=alpha,
    )
    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.11
    gap = 0.022
    ax.text(x0, y0 + 1.02, title, fontsize=6, ha="left", va="bottom", color="black")

    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE"] else (None, None)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)
            amp = packet[o, d]
            edge = "lime" if selected == (d, o) else "white"
            lw = 1.1 if selected == (d, o) else 0.25
            ax.add_patch(
                Rectangle(
                    (x, y),
                    cell,
                    cell,
                    facecolor=plt.cm.plasma(amp),
                    edgecolor=edge,
                    linewidth=lw,
                )
            )


def draw_strength_bar(ax, x0, y0, value, title):
    ax.text(x0, y0 + 0.28, title, fontsize=6, ha="left", va="bottom", color="black")
    ax.add_patch(Rectangle((x0, y0), 0.82, 0.12, facecolor="white", edgecolor="black", linewidth=0.5))
    ax.add_patch(
        Rectangle(
            (x0, y0),
            0.82 * clamp(value, 0, 1),
            0.12,
            facecolor="limegreen",
            edgecolor="none",
        )
    )
    ax.text(x0 + 0.88, y0 + 0.06, f"{value:.2f}", fontsize=6, ha="left", va="center", color="black")


def draw_filter_panel(ax, x0, y0, filt):
    ax.text(x0, y0 + 1.05, "adaptive filter", fontsize=6, ha="left", va="bottom", color="black")
    rows = [
        ("thr", filt["threshold"], min_threshold, max_threshold),
        ("mrg", filt["margin"], min_margin, max_margin),
        ("ph", filt["phase_tolerance"], min_phase_tolerance, max_phase_tolerance),
        ("trust", filt["trust"], 0.0, 1.0),
    ]
    for i, (name, value, lo, hi) in enumerate(rows):
        yy = y0 + 0.78 - i * 0.20
        frac = (value - lo) / (hi - lo) if hi > lo else 0
        ax.text(x0, yy, name, fontsize=6, ha="left", va="center", color="black")
        ax.add_patch(
            Rectangle(
                (x0 + 0.42, yy - 0.05),
                0.58,
                0.10,
                facecolor="white",
                edgecolor="black",
                linewidth=0.4,
            )
        )
        ax.add_patch(
            Rectangle(
                (x0 + 0.42, yy - 0.05),
                0.58 * clamp(frac, 0, 1),
                0.10,
                facecolor="limegreen",
                edgecolor="none",
            )
        )
        ax.text(x0 + 1.05, yy, f"{value:.2f}", fontsize=6, ha="left", va="center", color="black")


def draw_memory_column(ax, x0, y0, memory_state, active_slot):
    cell = 0.16
    gap = 0.04
    ax.text(x0, y0 + 1.65, "field memory", fontsize=6, ha="left", va="bottom", color="black")

    for i, mem in enumerate(memory_state):
        x = x0
        y = y0 + (len(memory_state) - 1 - i) * (cell + gap)
        edge = "lime" if i == active_slot else "black"
        lw = 1.5 if i == active_slot else 0.5
        ax.add_patch(
            Rectangle(
                (x, y),
                cell,
                cell,
                facecolor=key_color(mem["key"], mem["strength"]),
                edgecolor=edge,
                linewidth=lw,
            )
        )
        ax.text(x + 0.23, y + cell / 2, str(i), fontsize=5, ha="left", va="center", color="black")


def draw_snapshot(ax, row, packet, memory_state, filter_state):
    slot = int(row["MEMORY_SLOT"])
    source_key = row["SOURCE_KEY"]
    decoded_key = row["DECODED_KEY"]
    filter_key = row["FILTER_KEY"]
    intent = row["PACKET_INTENT"]
    action = row["ACTION"]
    absorbed = bool(row["ABSORBED"])
    packet_ok = bool(row["PACKET_ACCEPTED"])
    phase_ok = bool(row["PHASE_OK"])

    ax.set_facecolor("white")
    ax.set_xlim(0, 12)
    ax.set_ylim(0.45, 4.2)
    ax.set_aspect("equal")
    ax.axis("off")

    ax.add_patch(Circle((0.70, 2.75), 0.25, facecolor=key_color(source_key), edgecolor="black", linewidth=0.8))
    ax.text(0.70, 2.28, f"SRC\n{source_key}", fontsize=6, ha="center", color="black")

    ax.add_patch(Circle((1.35, 2.75), 0.18, facecolor=key_color(source_key), edgecolor="black", linewidth=0.8))
    ax.text(1.35, 2.28, "PHOTON", fontsize=5, ha="center", color="black")

    ax.add_patch(Rectangle((1.95, 2.60), 0.28, 0.28, facecolor="limegreen", edgecolor="black", linewidth=0.8))
    ax.text(2.09, 2.28, "FIELD", fontsize=5, ha="center", color="black")

    clock_color = "limegreen" if phase_ok else "red"
    ax.add_patch(Rectangle((2.65, 2.60), 0.28, 0.28, facecolor=clock_color, edgecolor="black", linewidth=0.8))
    ax.text(2.79, 2.28, "CLOCK", fontsize=5, ha="center", color="black")

    recv_edge = "lime" if packet_ok else "red"
    ax.add_patch(Circle((3.45, 2.75), 0.25, facecolor=key_color(decoded_key), edgecolor=recv_edge, linewidth=1.6))
    ax.text(3.45, 2.22, f"DECODE\n{decoded_key}", fontsize=6, ha="center", color="black")

    mem_edge = "lime" if absorbed else "red" if action.startswith("REJECT") else "black"
    mem_key = memory_state[slot]["key"]
    mem_strength = memory_state[slot]["strength"]
    ax.add_patch(Circle((4.25, 2.75), 0.25, facecolor=key_color(mem_key, mem_strength), edgecolor=mem_edge, linewidth=1.6))
    ax.text(4.25, 2.22, f"MEM\nslot {slot}", fontsize=6, ha="center", color="black")

    draw_arrow(ax, (0.95, 2.75), (1.17, 2.75), ok=True)
    draw_arrow(ax, (1.54, 2.75), (1.95, 2.75), ok=True)
    draw_arrow(ax, (2.23, 2.75), (2.65, 2.75), ok=True)
    draw_arrow(ax, (2.93, 2.75), (3.20, 2.75), ok=packet_ok and phase_ok)
    draw_arrow(ax, (3.70, 2.75), (4.00, 2.75), ok=absorbed)

    ax.add_patch(Circle((5.10, 2.75), 0.25, facecolor="white", edgecolor="black", linewidth=0.8))
    ax.text(
        5.10,
        2.75,
        intent_marker(intent),
        fontsize=12,
        ha="center",
        va="center",
        color="limegreen" if absorbed else "red",
    )
    ax.text(5.10, 2.22, intent, fontsize=6, ha="center", color="black")

    draw_strength_bar(ax, 5.85, 2.55, row["AFTER_STRENGTH"], "strength")
    draw_filter_panel(ax, 7.00, 2.10, filter_state)
    draw_packet_grid(ax, 8.80, 2.00, packet, "42-channel packet", selected_key=decoded_key)
    draw_memory_column(ax, 10.60, 1.55, memory_state, slot)

    ax.text(
        0.55,
        0.72,
        f"case={row['CASE']} slot={slot} filter={filter_key}/{row['FILTER_INTENT']} decoded={decoded_key} "
        f"score={row['BEST_SCORE']} margin={row['MARGIN']} action={action}\n"
        f"phase_error={row['PHASE_ERROR']} jitter={row['JITTER']} trust={row['TRUST_AFTER']} output={row['OUTPUT_TOKEN']}",
        fontsize=6,
        ha="left",
        color="black",
    )

    ax.set_title(
        f"D42 snapshot t={row['T']} — {row['CASE']} / {action}",
        fontsize=9,
        color="black",
    )


# ------------------------------------------------------------
# Events
# ------------------------------------------------------------

events = [
    {
        "case": "COHERENT_REPEAT_FLOWER_0",
        "slot": 0,
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.93,
        "phase": expected_phase("D4_O5") + 0.012,
        "jitter": 0.02,
    },
    {
        "case": "IDLE_DECAY_RING_1",
        "slot": 1,
        "source_key": "NONE",
        "intent": "NONE",
        "strength": 0.00,
        "phase": 0.0,
        "jitter": 0.0,
        "idle": True,
    },
    {
        "case": "COHERENT_REFRESH_RING_1",
        "slot": 1,
        "source_key": "D1_O2",
        "intent": "RING",
        "strength": 0.90,
        "phase": expected_phase("D1_O2") + 0.028,
        "jitter": 0.03,
    },
    {
        "case": "WRONG_INTENT_REJECT_2",
        "slot": 2,
        "source_key": "D3_O6",
        "intent": "FLOWER",
        "strength": 0.91,
        "phase": expected_phase("D3_O6") + 0.010,
        "jitter": 0.03,
    },
    {
        "case": "PHASE_DRIFT_REJECT_2",
        "slot": 2,
        "source_key": "D3_O6",
        "intent": "HEX",
        "strength": 0.90,
        "phase": expected_phase("D3_O6") + 0.345,
        "jitter": 0.03,
    },
    {
        "case": "RESYNC_HEX_RECOVER_2",
        "slot": 2,
        "source_key": "D3_O6",
        "intent": "HEX",
        "strength": 0.89,
        "phase": expected_phase("D3_O6") + 0.310,
        "jitter": 0.04,
        "resync": True,
    },
    {
        "case": "NEW_WRITE_FLOWER_6",
        "slot": 6,
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.92,
        "phase": expected_phase("D4_O5") + 0.008,
        "jitter": 0.02,
    },
    {
        "case": "WEAK_PACKET_FORGET_SEED_4",
        "slot": 4,
        "source_key": "D0_O1",
        "intent": "SEED",
        "strength": 0.38,
        "phase": expected_phase("D0_O1") + 0.018,
        "jitter": 0.03,
    },
    {
        "case": "REWRITE_BUTTERFLY_5",
        "slot": 5,
        "source_key": "D2_O3",
        "intent": "BUTTERFLY",
        "strength": 0.91,
        "phase": expected_phase("D2_O3") + 0.020,
        "jitter": 0.03,
    },
    {
        "case": "AMBIGUOUS_FLOWER_HOLD_0",
        "slot": 0,
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.86,
        "ambiguous_key": "D1_O2",
        "ambiguous_strength": 0.82,
        "phase": expected_phase("D4_O5") + 0.020,
        "jitter": 0.02,
    },
    {
        "case": "STRONG_REINFORCE_FLOWER_0",
        "slot": 0,
        "source_key": "D4_O5",
        "intent": "FLOWER",
        "strength": 0.97,
        "phase": expected_phase("D4_O5") + 0.004,
        "jitter": 0.02,
    },
    {
        "case": "EMPTY_IDLE_HOLD_7",
        "slot": 7,
        "source_key": "NONE",
        "intent": "NONE",
        "strength": 0.00,
        "phase": 0.0,
        "jitter": 0.0,
        "idle": True,
    },
    {
        "case": "LATE_RING_WRITE_7",
        "slot": 7,
        "source_key": "D1_O2",
        "intent": "RING",
        "strength": 0.88,
        "phase": expected_phase("D1_O2") - 0.015,
        "jitter": 0.03,
    },
]


# ------------------------------------------------------------
# Simulation
# ------------------------------------------------------------

decision_rows = []
filter_rows = []
memory_rows = []
snapshots = []

absorbed_count = 0
rejected_count = 0
adapted_count = 0
forgotten_count = 0
resync_count = 0

for t, event in enumerate(events):
    slot = event["slot"]
    filt_before = adaptive_filters[slot].copy()
    mem_before = memory[slot].copy()
    packet = np.zeros((octaves, directions))

    if event.get("idle", False):
        decoded = {
            "best_key": "NONE",
            "best_state_id": -1,
            "best_score": 0.0,
            "second_key": "NONE",
            "second_score": 0.0,
            "margin": 0.0,
            "accepted": False,
        }
        packet_accepted = False
        filter_match = False
        intent_match = False
        phase_error = 0.0
        phase_ok = True
        jitter_ok = True
        absorbed = False
        forgotten = apply_decay(memory[slot], amount=0.045)
        action = "PASSIVE_FORGET" if forgotten else "PASSIVE_DECAY_ONLY"
        output = "NONE"

    else:
        packet = spectral_packet(
            target_key=event["source_key"],
            strength=event["strength"],
            ambiguous_key=event.get("ambiguous_key"),
            ambiguous_strength=event.get("ambiguous_strength", 0.0),
            broad=event.get("broad", False),
        )

        decoded = decode_packet(packet, filt_before["threshold"], filt_before["margin"])
        decoded_key = decoded["best_key"]

        packet_accepted = bool(decoded["accepted"])
        filter_match = decoded_key == filt_before["filter_key"]
        intent_match = event["intent"] == filt_before["filter_intent"]

        exp_phase = expected_phase(filt_before["filter_key"])
        measured_phase = event["phase"] % 1.0
        raw_phase_error = circular_phase_error(measured_phase, exp_phase)

        if event.get("resync", False):
            corrected_phase = (exp_phase + 0.003) % 1.0
            resync_count += 1
        else:
            corrected_phase = measured_phase

        phase_error = circular_phase_error(corrected_phase, exp_phase)
        phase_ok = phase_error <= filt_before["phase_tolerance"]
        jitter_ok = event["jitter"] <= filt_before["jitter_tolerance"]

        absorbed = bool(packet_accepted and filter_match and intent_match and phase_ok and jitter_ok)
        forgotten = False

        if absorbed:
            if memory[slot]["key"] in ["EMPTY", "NONE"]:
                memory[slot] = {
                    "key": decoded_key,
                    "intent": event["intent"],
                    "strength": clamp(event["strength"], 0.0, 1.0),
                    "age": 0,
                    "writes": 1,
                }
                action = "WRITE_NEW_FIELD_MEMORY"
            else:
                memory[slot]["key"] = decoded_key
                memory[slot]["intent"] = event["intent"]
                memory[slot]["strength"] = clamp(
                    memory[slot]["strength"] + reinforce_gain + 0.04 * event["strength"],
                    0.0,
                    1.0,
                )
                memory[slot]["age"] = 0
                memory[slot]["writes"] += 1
                action = "REINFORCE_AND_TUNE_FILTER"

            absorbed_count += 1

        else:
            forgotten = apply_decay(memory[slot], amount=0.04)

            if not packet_accepted:
                action = "REJECT_WEAK_OR_AMBIGUOUS"
            elif not filter_match:
                action = "REJECT_FILTER_MISMATCH"
            elif not intent_match:
                action = "REJECT_INTENT_MISMATCH"
            elif not phase_ok:
                action = "REJECT_PHASE_UNLOCKED"
            elif not jitter_ok:
                action = "REJECT_DECOHERENCE"
            else:
                action = "REJECT_UNKNOWN"

            if forgotten:
                action = action + "_AND_FORGET"

            rejected_count += 1

        output = tune_token(decoded_key, event["intent"], corrected_phase, decoded["best_score"], action)

    old_filter, new_filter = adapt_filter(
        adaptive_filters[slot],
        absorbed=absorbed,
        action=action.replace("_AND_FORGET", ""),
        memory_strength=memory[slot]["strength"],
    )
    adaptive_filters[slot] = new_filter

    threshold_delta = round(new_filter["threshold"] - filt_before["threshold"], 4)
    margin_delta = round(new_filter["margin"] - filt_before["margin"], 4)
    trust_delta = round(new_filter["trust"] - filt_before["trust"], 4)

    adapted = int(abs(threshold_delta) > 0 or abs(margin_delta) > 0 or abs(trust_delta) > 0)
    adapted_count += adapted
    forgotten_count += int(forgotten)

    intent_scan = "".join([
        intent_marker(event["intent"]) if k == slot else "·"
        for k in range(memory_slots)
    ])

    row = {
        "STATE": STATE,
        "T": t * 120,
        "CASE": event["case"],
        "MEMORY_SLOT": slot,
        "SOURCE_KEY": event["source_key"],
        "FILTER_KEY": filt_before["filter_key"],
        "FILTER_INTENT": filt_before["filter_intent"],
        "PACKET_INTENT": event["intent"],
        "BEFORE_KEY": mem_before["key"],
        "BEFORE_INTENT": mem_before["intent"],
        "BEFORE_STRENGTH": round(mem_before["strength"], 3),
        "DECODED_KEY": decoded["best_key"],
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "THRESHOLD_BEFORE": round(filt_before["threshold"], 3),
        "MARGIN_BEFORE": round(filt_before["margin"], 3),
        "PACKET_ACCEPTED": int(packet_accepted),
        "FILTER_MATCH": int(filter_match),
        "INTENT_MATCH": int(intent_match),
        "PHASE_ERROR": round(phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "JITTER": round(event["jitter"], 3),
        "JITTER_OK": int(jitter_ok),
        "CLOCK_RESYNC": int(bool(event.get("resync", False))),
        "ABSORBED": int(absorbed),
        "FORGOTTEN": int(forgotten),
        "AFTER_KEY": memory[slot]["key"],
        "AFTER_INTENT": memory[slot]["intent"],
        "AFTER_STRENGTH": round(memory[slot]["strength"], 3),
        "THRESHOLD_AFTER": round(new_filter["threshold"], 3),
        "MARGIN_AFTER": round(new_filter["margin"], 3),
        "TRUST_AFTER": round(new_filter["trust"], 3),
        "FILTER_ADAPTED": adapted,
        "ACTION": action,
        "INTENT_SCAN": intent_scan,
        "OUTPUT_TOKEN": output,
    }
    decision_rows.append(row)

    filter_rows.append({
        "STATE": STATE,
        "CASE": event["case"],
        "MEMORY_SLOT": slot,
        "FILTER_KEY": filt_before["filter_key"],
        "FILTER_INTENT": filt_before["filter_intent"],
        "THRESHOLD_BEFORE": round(filt_before["threshold"], 3),
        "THRESHOLD_AFTER": round(new_filter["threshold"], 3),
        "THRESHOLD_DELTA": threshold_delta,
        "MARGIN_BEFORE": round(filt_before["margin"], 3),
        "MARGIN_AFTER": round(new_filter["margin"], 3),
        "MARGIN_DELTA": margin_delta,
        "TRUST_BEFORE": round(filt_before["trust"], 3),
        "TRUST_AFTER": round(new_filter["trust"], 3),
        "TRUST_DELTA": trust_delta,
        "ACCEPTS": new_filter["accepts"],
        "REJECTS": new_filter["rejects"],
        "FILTER_ADAPTED": adapted,
    })

    memory_rows.append({
        "STATE": STATE,
        "CASE": event["case"],
        "MEMORY_SLOT": slot,
        "BEFORE_KEY": mem_before["key"],
        "BEFORE_INTENT": mem_before["intent"],
        "BEFORE_STRENGTH": round(mem_before["strength"], 3),
        "AFTER_KEY": memory[slot]["key"],
        "AFTER_INTENT": memory[slot]["intent"],
        "AFTER_STRENGTH": round(memory[slot]["strength"], 3),
        "ABSORBED": int(absorbed),
        "FORGOTTEN": int(forgotten),
        "ACTION": action,
    })

    snapshots.append((row, packet.copy(), [m.copy() for m in memory], new_filter.copy()))


df_decision = pd.DataFrame(decision_rows)
df_filters = pd.DataFrame(filter_rows)
df_memory_events = pd.DataFrame(memory_rows)

df_final_memory = pd.DataFrame([
    {
        "STATE": STATE,
        "MEMORY_SLOT": i,
        "FILTER_KEY": slot_filters[i]["key"],
        "FILTER_INTENT": slot_filters[i]["intent"],
        "FINAL_KEY": memory[i]["key"],
        "FINAL_INTENT": memory[i]["intent"],
        "FINAL_STRENGTH": round(memory[i]["strength"], 3),
        "WRITES": memory[i]["writes"],
        "AGE": memory[i]["age"],
        "MATCH_FILTER_OR_EMPTY": int(
            (memory[i]["key"] == slot_filters[i]["key"] and memory[i]["intent"] == slot_filters[i]["intent"])
            or memory[i]["key"] == "EMPTY"
        ),
    }
    for i in range(memory_slots)
])

df_final_filters = pd.DataFrame([
    {
        "STATE": STATE,
        "MEMORY_SLOT": f["slot"],
        "FILTER_KEY": f["filter_key"],
        "FILTER_INTENT": f["filter_intent"],
        "THRESHOLD": round(f["threshold"], 3),
        "MARGIN": round(f["margin"], 3),
        "PHASE_TOLERANCE": round(f["phase_tolerance"], 3),
        "JITTER_TOLERANCE": round(f["jitter_tolerance"], 3),
        "TRUST": round(f["trust"], 3),
        "ACCEPTS": f["accepts"],
        "REJECTS": f["rejects"],
    }
    for f in adaptive_filters
])

memory_accuracy = float(df_final_memory["MATCH_FILTER_OR_EMPTY"].mean())
filter_adaptation_rate = float(df_decision["FILTER_ADAPTED"].mean())

safety_accuracy = float(np.mean([
    1 if (
        (row["ACTION"] in ["REINFORCE_AND_TUNE_FILTER", "WRITE_NEW_FIELD_MEMORY"] and row["ABSORBED"] == 1)
        or (str(row["ACTION"]).startswith("REJECT") and row["ABSORBED"] == 0)
        or str(row["ACTION"]).startswith("PASSIVE")
    ) else 0
    for _, row in df_decision.iterrows()
]))

result = (
    "SPECTRAL_SELF_TUNING_FILTER_LOCKED"
    if memory_accuracy == 1.0 and safety_accuracy == 1.0 and adapted_count >= 8
    else "SPECTRAL_SELF_TUNING_FILTER_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": memory_slots,
    "EVENTS_TESTED": len(events),
    "ABSORBED_PACKETS": absorbed_count,
    "REJECTED_PACKETS": rejected_count,
    "FILTER_ADAPTATIONS": adapted_count,
    "FORGOTTEN_EVENTS": forgotten_count,
    "CLOCK_RESYNC_EVENTS": resync_count,
    "MEMORY_ACCURACY": round(memory_accuracy, 3),
    "FILTER_ADAPTATION_RATE": round(filter_adaptation_rate, 3),
    "SAFETY_ACCURACY": round(safety_accuracy, 3),
    "BASE_RESONANCE_THRESHOLD": base_resonance_threshold,
    "BASE_AMBIGUITY_MARGIN": base_ambiguity_margin,
    "BASE_PHASE_TOLERANCE": base_phase_tolerance,
    "FORGET_THRESHOLD": forget_threshold,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8.5, 2.65 * len(snapshots)))
if len(snapshots) == 1:
    axes = [axes]

for ax, (row, packet, mem_state, filt_state) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, mem_state, filt_state)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"memory slots:                 {memory_slots}")
print(f"events tested:                {len(events)}")
print(f"absorbed packets:             {absorbed_count}")
print(f"rejected packets:             {rejected_count}")
print(f"filter adaptations:           {adapted_count}")
print(f"forgotten events:             {forgotten_count}")
print(f"clock resync events:          {resync_count}")
print(f"memory accuracy:              {round(memory_accuracy, 3)}")
print(f"filter adaptation rate:       {round(filter_adaptation_rate, 3)}")
print(f"safety accuracy:              {round(safety_accuracy, 3)}")
print(f"base resonance threshold:     {base_resonance_threshold}")
print(f"base ambiguity margin:        {base_ambiguity_margin}")
print(f"base phase tolerance:         {base_phase_tolerance}")
print(f"forget threshold:             {forget_threshold}")

print("\nSelf-tuning decision table:")
display(df_decision)

print("\nFilter adaptation table:")
display(df_filters)

print("\nMemory event table:")
display(df_memory_events)

print("\nFinal memory table:")
display(df_final_memory)

print("\nFinal adaptive filter table:")
display(df_final_filters)

print("\nSummary:")
display(df_summary)
```
