```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D36_SPECTRAL_COHERENCE_BUS
# Phase-coherent spectral bus for photon-like field memory
#
# Goal:
# source emits spectral packet
# -> shared BUS carries packet
# -> coherence clock checks phase + jitter
# -> decoder extracts spectral key
# -> receiver filters absorb only matching direction+octave
# -> fanout can write to multiple receivers
# -> weak / ambiguous / wrong phase / decoherent / wrong filter are rejected
# ============================================================

np.random.seed(36)

STATE = "D36_SPECTRAL_COHERENCE_BUS"

directions = 6
octaves = 7
spectral_states = directions * octaves
memory_slots = 8

resonance_threshold = 0.62
ambiguity_margin = 0.12
phase_tolerance = 0.18
jitter_tolerance = 0.16
resync_capture_window = 0.36
default_fanout = 2

receiver_filter_sequence = [
    "D4_O5",
    "D1_O2",
    "D3_O6",
    "D5_O4",
    "D0_O1",
    "D2_O3",
    "D4_O5",
    "D1_O2",
]

receiver_memory = ["EMPTY"] * memory_slots


# ------------------------------------------------------------
# Spectral helpers
# ------------------------------------------------------------

def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE"]:
        return None, None

    if not isinstance(key, str) or "_" not in key:
        return None, None

    d_part, o_part = key.split("_", 1)
    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))

    return direction, octave


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def state_id(direction, octave):
    if direction is None or octave is None:
        return None
    return octave * directions + direction


def circular_error(a, b):
    if a is None or b is None:
        return None

    diff = abs(a - b) % 1.0
    return min(diff, 1.0 - diff)


def expected_phase(key):
    direction, octave = parse_key(key)

    if direction is None:
        return None

    return round(direction / directions, 3)


def safe_phase_add(phase, drift):
    return round((phase + drift) % 1.0, 3)


def key_color(key, strength=1.0):
    if key in [None, "EMPTY", "NONE"]:
        return "black"

    if key == "WHITE":
        return "white"

    direction, octave = parse_key(key)

    if direction is None:
        return "black"

    hue_value = (direction + 1) / (directions + 1)
    octave_value = (octave + 1) / (octaves + 1)

    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0,
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.15 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        direction, octave = parse_key(target_key)
        packet[octave, direction] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE"]:
        direction_2, octave_2 = parse_key(ambiguous_key)
        packet[octave_2, direction_2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score

    amplitude_ok = (
        best_score >= resonance_threshold
        and margin >= ambiguity_margin
    )

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "amplitude_ok": amplitude_ok,
    }


def correct_phase(raw_phase, expected, resync=False):
    raw_error = circular_error(raw_phase, expected)

    if resync and raw_error is not None and raw_error <= resync_capture_window:
        corrected_phase = safe_phase_add(expected, 0.003)
        corrected_error = circular_error(corrected_phase, expected)
        return corrected_phase, corrected_error, True

    return raw_phase, raw_error, False


def bus_token(key, phase, amplitude, jitter, active=True):
    if not active or key in [None, "EMPTY", "NONE"]:
        return "NONE"

    direction, octave = parse_key(key)

    if direction is None:
        return "NONE"

    return (
        f"BUS_PHOTON[D{direction}|O{octave}|F{octave + 1}|"
        f"P{round(phase, 3)}|A{round(amplitude, 3)}|J{round(jitter, 3)}]"
    )


def format_slots(slots):
    if not slots:
        return "none"

    return ",".join(str(x) for x in slots)


# ------------------------------------------------------------
# Visualization helpers
# ------------------------------------------------------------

def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.25

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=13,
        linewidth=1.8,
        color=color,
        alpha=alpha,
    )

    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.16
    gap = 0.032

    ax.text(
        x0,
        y0 + 1.52,
        title,
        fontsize=6.5,
        ha="left",
        va="bottom",
        color="black",
    )

    if selected_key not in [None, "EMPTY", "NONE", "WHITE"]:
        selected = parse_key(selected_key)
    else:
        selected = (None, None)

    for octave in range(octaves):
        for direction in range(directions):
            x = x0 + direction * (cell + gap)
            y = y0 + (octaves - 1 - octave) * (cell + gap)

            amp = packet[octave, direction]
            color = plt.cm.plasma(amp)

            is_selected = selected == (direction, octave)
            edge = "lime" if is_selected else "white"
            lw = 1.3 if is_selected else 0.25

            ax.add_patch(
                Rectangle(
                    (x, y),
                    cell,
                    cell,
                    facecolor=color,
                    edgecolor=edge,
                    linewidth=lw,
                )
            )


def draw_memory_slots(ax, x0, y0, seq, title, active_slots=None):
    if active_slots is None:
        active_slots = []

    cell = 0.20
    gap = 0.055

    ax.text(
        x0,
        y0 + 2.15,
        title,
        fontsize=6.5,
        ha="left",
        va="bottom",
        color="black",
    )

    for i, key in enumerate(seq):
        x = x0
        y = y0 + (len(seq) - 1 - i) * (cell + gap)

        color = key_color(key)
        edge = "lime" if i in active_slots else "black"
        lw = 1.5 if i in active_slots else 0.45

        ax.add_patch(
            Rectangle(
                (x, y),
                cell,
                cell,
                facecolor=color,
                edgecolor=edge,
                linewidth=lw,
            )
        )

        ax.text(
            x + 0.30,
            y + cell / 2,
            str(i),
            fontsize=6,
            ha="left",
            va="center",
            color="black",
        )


def draw_bus_receivers(ax, x0, y0, filters, active_slots=None, blocked_slots=None):
    if active_slots is None:
        active_slots = []

    if blocked_slots is None:
        blocked_slots = []

    ax.text(
        x0,
        y0 + 1.35,
        "receiver filters",
        fontsize=6.5,
        ha="left",
        color="black",
    )

    for i, key in enumerate(filters):
        row = i // 4
        col = i % 4

        x = x0 + col * 0.62
        y = y0 + (1 - row) * 0.55

        edge = "lime" if i in active_slots else "red" if i in blocked_slots else "black"
        lw = 1.8 if i in active_slots or i in blocked_slots else 0.5

        ax.add_patch(
            Circle(
                (x, y),
                0.18,
                facecolor=key_color(key),
                edgecolor=edge,
                linewidth=lw,
            )
        )

        ax.text(
            x,
            y - 0.32,
            f"{i}",
            fontsize=6,
            ha="center",
            color="black",
        )


def draw_snapshot(ax, row, packet, memory_state):
    source_key = row["SOURCE_KEY"]
    emitted_key = row["EMITTED_KEY"]
    decoded_key = row["DECODED_KEY"]

    absorbed_slots = (
        []
        if row["ABSORBED_SLOTS"] == "none"
        else [int(x) for x in row["ABSORBED_SLOTS"].split(",")]
    )

    blocked_slots = (
        []
        if row["BLOCKED_SLOTS"] == "none"
        else [int(x) for x in row["BLOCKED_SLOTS"].split(",")]
    )

    rejected = (
        row["ACTION"].startswith("REJECT")
        or row["ACTION"].startswith("BUS_HOLD")
    )

    ax.set_facecolor("white")
    ax.set_xlim(0, 12.2)
    ax.set_ylim(0.6, 5.0)
    ax.set_aspect("equal")
    ax.axis("off")

    ax.add_patch(
        Circle(
            (0.8, 3.1),
            0.38,
            facecolor=key_color(source_key),
            edgecolor="black",
            linewidth=1.0,
        )
    )
    ax.text(0.8, 2.5, f"SRC\n{source_key}", fontsize=6.5, ha="center", color="black")

    ax.add_patch(
        Circle(
            (1.85, 3.1),
            0.24,
            facecolor=key_color(emitted_key),
            edgecolor="black",
            linewidth=1.0,
        )
    )
    ax.text(1.85, 2.5, "PHOTON", fontsize=6.5, ha="center", color="black")

    ax.add_patch(
        Rectangle(
            (2.55, 2.88),
            0.42,
            0.42,
            facecolor="limegreen",
            edgecolor="black",
            linewidth=1.0,
        )
    )
    ax.text(2.76, 2.5, "BUS", fontsize=6.5, ha="center", color="black")

    clock_edge = "lime" if int(row["PHASE_OK"]) and int(row["JITTER_OK"]) else "red"

    ax.add_patch(
        Rectangle(
            (3.55, 2.88),
            0.42,
            0.42,
            facecolor="orange",
            edgecolor=clock_edge,
            linewidth=1.5,
        )
    )
    ax.text(3.76, 2.5, "CLOCK", fontsize=6.5, ha="center", color="black")

    ax.add_patch(
        Circle(
            (4.75, 3.1),
            0.38,
            facecolor=key_color(decoded_key),
            edgecolor="lime" if int(row["AMPLITUDE_OK"]) else "red",
            linewidth=2.0,
        )
    )
    ax.text(4.75, 2.42, f"DECODE\n{decoded_key}", fontsize=6.5, ha="center", color="black")

    mem_color = "limegreen" if absorbed_slots else "black"

    ax.add_patch(
        Circle(
            (6.05, 3.1),
            0.38,
            facecolor=mem_color,
            edgecolor="lime" if absorbed_slots else "red",
            linewidth=2.0,
        )
    )
    ax.text(
        6.05,
        2.42,
        f"WRITE\n{format_slots(absorbed_slots)}",
        fontsize=6.5,
        ha="center",
        color="black",
    )

    draw_arrow(ax, (1.18, 3.1), (1.60, 3.1), active=True, rejected=False)
    draw_arrow(ax, (2.10, 3.1), (2.55, 3.1), active=True, rejected=False)
    draw_arrow(ax, (2.97, 3.1), (3.55, 3.1), active=True, rejected=rejected)
    draw_arrow(ax, (3.97, 3.1), (4.37, 3.1), active=True, rejected=rejected)
    draw_arrow(ax, (5.13, 3.1), (5.67, 3.1), active=True, rejected=rejected)

    draw_bus_receivers(
        ax,
        6.9,
        2.85,
        receiver_filter_sequence,
        active_slots=absorbed_slots,
        blocked_slots=blocked_slots,
    )

    draw_packet_grid(
        ax,
        8.8,
        2.1,
        packet,
        "42-channel bus packet",
        selected_key=decoded_key,
    )

    draw_memory_slots(
        ax,
        11.0,
        1.3,
        memory_state,
        "bus memory",
        active_slots=absorbed_slots,
    )

    ax.text(
        0.45,
        0.85,
        f"case={row['CASE']} decoded={decoded_key} score={row['BEST_SCORE']} "
        f"phase_err={row['PHASE_ERROR']} jitter={row['JITTER']} "
        f"slots={row['ABSORBED_SLOTS']} action={row['ACTION']}\n{row['OUTPUT_TOKEN']}",
        fontsize=6.5,
        ha="left",
        color="black",
    )

    ax.set_title(
        f"D36 snapshot t={row['T']} — {row['CASE']} / {row['ACTION']}",
        fontsize=9,
        color="black",
    )


# ------------------------------------------------------------
# Bus test events
# ------------------------------------------------------------

bus_events = [
    {
        "case": "BUS_FANOUT_D4_TO_0_6",
        "source_key": "D4_O5",
        "emitted_key": "D4_O5",
        "strength": 0.92,
        "fanout": 2,
        "phase_drift": 0.010,
        "jitter": 0.020,
        "resync": False,
        "expected_absorbed": [0, 6],
    },
    {
        "case": "BUS_FANOUT_D1_TO_1_7",
        "source_key": "D1_O2",
        "emitted_key": "D1_O2",
        "strength": 0.90,
        "fanout": 2,
        "phase_drift": 0.030,
        "jitter": 0.030,
        "resync": False,
        "expected_absorbed": [1, 7],
    },
    {
        "case": "WRONG_DIRECTION_BUS_REJECT",
        "source_key": "D0_O2",
        "emitted_key": "D0_O2",
        "strength": 0.90,
        "fanout": 1,
        "phase_drift": 0.010,
        "jitter": 0.030,
        "resync": False,
        "expected_absorbed": [],
    },
    {
        "case": "LARGE_PHASE_DRIFT_REJECT",
        "source_key": "D3_O6",
        "emitted_key": "D3_O6",
        "strength": 0.91,
        "fanout": 1,
        "phase_drift": 0.340,
        "jitter": 0.030,
        "resync": False,
        "expected_absorbed": [],
    },
    {
        "case": "RESYNC_D5_ACCEPT",
        "source_key": "D5_O4",
        "emitted_key": "D5_O4",
        "strength": 0.89,
        "fanout": 1,
        "phase_drift": 0.310,
        "jitter": 0.040,
        "resync": True,
        "expected_absorbed": [3],
    },
    {
        "case": "DECOHERENCE_JITTER_REJECT",
        "source_key": "D0_O1",
        "emitted_key": "D0_O1",
        "strength": 0.90,
        "fanout": 1,
        "phase_drift": 0.020,
        "jitter": 0.290,
        "resync": False,
        "expected_absorbed": [],
    },
    {
        "case": "WEAK_BUS_PACKET_REJECT",
        "source_key": "D2_O3",
        "emitted_key": "D2_O3",
        "strength": 0.39,
        "fanout": 1,
        "phase_drift": 0.010,
        "jitter": 0.020,
        "resync": False,
        "expected_absorbed": [],
    },
    {
        "case": "AMBIGUOUS_BUS_PACKET_REJECT",
        "source_key": "D2_O3",
        "emitted_key": "D2_O3",
        "strength": 0.86,
        "fanout": 1,
        "phase_drift": 0.010,
        "jitter": 0.020,
        "resync": False,
        "ambiguous_key": "D4_O5",
        "ambiguous_strength": 0.81,
        "expected_absorbed": [],
    },
    {
        "case": "CLEAN_D0_LATE_ACCEPT",
        "source_key": "D0_O1",
        "emitted_key": "D0_O1",
        "strength": 0.88,
        "fanout": 1,
        "phase_drift": -0.015,
        "jitter": 0.030,
        "resync": False,
        "expected_absorbed": [4],
    },
    {
        "case": "CLEAN_D2_LATE_ACCEPT",
        "source_key": "D2_O3",
        "emitted_key": "D2_O3",
        "strength": 0.93,
        "fanout": 1,
        "phase_drift": 0.020,
        "jitter": 0.030,
        "resync": False,
        "expected_absorbed": [5],
    },
    {
        "case": "CLEAN_D3_LATE_ACCEPT",
        "source_key": "D3_O6",
        "emitted_key": "D3_O6",
        "strength": 0.90,
        "fanout": 1,
        "phase_drift": -0.020,
        "jitter": 0.030,
        "resync": False,
        "expected_absorbed": [2],
    },
]


# ------------------------------------------------------------
# Simulation
# ------------------------------------------------------------

decision_rows = []
receiver_rows = []
clock_rows = []
snapshots = []

amplitude_accepted = 0
phase_locked = 0
clock_resync_events = 0
bus_routed_packets = 0
rejected_packets = 0
absorbed_receivers = 0
blocked_by_full = 0

expected_final = receiver_memory.copy()

for event_index, event in enumerate(bus_events):
    packet = spectral_packet(
        target_key=event["emitted_key"],
        strength=event["strength"],
        noise=event.get("noise", 0.025),
        broad=event.get("broad", False),
        ambiguous_key=event.get("ambiguous_key", None),
        ambiguous_strength=event.get("ambiguous_strength", 0.0),
    )

    decoded = decode_packet(packet)
    decoded_key = decoded["best_key"]

    exp_phase = expected_phase(decoded_key)
    measured_phase = safe_phase_add(exp_phase, event["phase_drift"])

    corrected_phase, phase_error, resynced = correct_phase(
        measured_phase,
        exp_phase,
        event.get("resync", False),
    )

    phase_ok = phase_error is not None and phase_error <= phase_tolerance
    jitter_ok = event["jitter"] <= jitter_tolerance
    amplitude_ok = bool(decoded["amplitude_ok"])

    candidate_slots = [
        slot
        for slot, filter_key in enumerate(receiver_filter_sequence)
        if filter_key == decoded_key
    ]

    open_candidates = [
        slot
        for slot in candidate_slots
        if receiver_memory[slot] in ["EMPTY", decoded_key]
    ]

    blocked_slots = []
    absorbed_slots = []

    if amplitude_ok and phase_ok and jitter_ok and candidate_slots:
        fanout = int(event.get("fanout", default_fanout))

        for slot in open_candidates:
            if len(absorbed_slots) < fanout and receiver_memory[slot] == "EMPTY":
                receiver_memory[slot] = decoded_key
                absorbed_slots.append(slot)
            elif len(absorbed_slots) >= fanout:
                blocked_slots.append(slot)

        action = "BUS_ROUTE_AND_WRITE" if absorbed_slots else "BUS_HOLD_ALREADY_LOCKED"

    elif not amplitude_ok:
        action = "REJECT_WEAK_OR_AMBIGUOUS"

    elif not phase_ok:
        action = "REJECT_PHASE_UNLOCKED"

    elif not jitter_ok:
        action = "REJECT_DECOHERENCE"

    elif not candidate_slots:
        action = "REJECT_FILTER_MISMATCH"

    else:
        action = "REJECT_BUS_HOLD"

    for slot in event["expected_absorbed"]:
        expected_final[slot] = receiver_filter_sequence[slot]

    if amplitude_ok:
        amplitude_accepted += 1

    if phase_ok:
        phase_locked += 1

    if resynced:
        clock_resync_events += 1

    if absorbed_slots:
        bus_routed_packets += 1
    else:
        rejected_packets += 1

    absorbed_receivers += len(absorbed_slots)
    blocked_by_full += len(blocked_slots)

    t = event_index * 120

    row = {
        "STATE": STATE,
        "T": t,
        "CASE": event["case"],
        "SOURCE_KEY": event["source_key"],
        "EMITTED_KEY": event["emitted_key"],
        "DECODED_KEY": decoded_key,
        "BEST_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_KEY": decoded["second_key"],
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "AMPLITUDE_OK": int(amplitude_ok),
        "EXPECTED_PHASE": round(exp_phase, 3),
        "MEASURED_PHASE": round(measured_phase, 3),
        "CORRECTED_PHASE": round(corrected_phase, 3),
        "PHASE_ERROR": round(phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "JITTER": round(event["jitter"], 3),
        "JITTER_OK": int(jitter_ok),
        "CLOCK_RESYNC": int(resynced),
        "CANDIDATE_SLOTS": format_slots(candidate_slots),
        "ABSORBED_SLOTS": format_slots(absorbed_slots),
        "BLOCKED_SLOTS": format_slots(blocked_slots),
        "FANOUT_LIMIT": int(event.get("fanout", default_fanout)),
        "ACTION": action,
        "OUTPUT_TOKEN": bus_token(
            decoded_key,
            corrected_phase,
            decoded["best_score"],
            event["jitter"],
            active=(amplitude_ok and phase_ok and jitter_ok),
        ),
    }

    decision_rows.append(row)

    if candidate_slots:
        for slot in candidate_slots:
            receiver_rows.append({
                "STATE": STATE,
                "CASE": event["case"],
                "RECEIVER_SLOT": slot,
                "FILTER_KEY": receiver_filter_sequence[slot],
                "DECODED_KEY": decoded_key,
                "FINAL_SLOT_KEY": receiver_memory[slot],
                "AMPLITUDE_OK": int(amplitude_ok),
                "PHASE_OK": int(phase_ok),
                "JITTER_OK": int(jitter_ok),
                "FILTER_MATCH": int(receiver_filter_sequence[slot] == decoded_key),
                "ABSORBED": int(slot in absorbed_slots),
                "BLOCKED_BY_FANOUT": int(slot in blocked_slots),
                "ACTION": "ABSORB_AND_WRITE" if slot in absorbed_slots else "HOLD",
            })
    else:
        receiver_rows.append({
            "STATE": STATE,
            "CASE": event["case"],
            "RECEIVER_SLOT": "none",
            "FILTER_KEY": "none",
            "DECODED_KEY": decoded_key,
            "FINAL_SLOT_KEY": "none",
            "AMPLITUDE_OK": int(amplitude_ok),
            "PHASE_OK": int(phase_ok),
            "JITTER_OK": int(jitter_ok),
            "FILTER_MATCH": 0,
            "ABSORBED": 0,
            "BLOCKED_BY_FANOUT": 0,
            "ACTION": "NO_MATCHING_RECEIVER",
        })

    clock_rows.append({
        "STATE": STATE,
        "CASE": event["case"],
        "DECODED_KEY": decoded_key,
        "EXPECTED_PHASE": round(exp_phase, 3),
        "MEASURED_PHASE": round(measured_phase, 3),
        "RAW_PHASE_ERROR": round(circular_error(measured_phase, exp_phase), 3),
        "CLOCK_RESYNC": int(resynced),
        "CORRECTED_PHASE": round(corrected_phase, 3),
        "CORRECTED_PHASE_ERROR": round(phase_error, 3),
        "PHASE_OK": int(phase_ok),
        "JITTER": round(event["jitter"], 3),
        "JITTER_OK": int(jitter_ok),
    })

    snapshots.append((row, packet, receiver_memory.copy()))


# ------------------------------------------------------------
# Tables and metrics
# ------------------------------------------------------------

df_decision = pd.DataFrame(decision_rows)
df_receivers = pd.DataFrame(receiver_rows)
df_clock = pd.DataFrame(clock_rows)

df_memory = pd.DataFrame([
    {
        "STATE": STATE,
        "MEMORY_SLOT": slot,
        "FILTER_KEY": receiver_filter_sequence[slot],
        "FINAL_KEY": receiver_memory[slot],
        "EXPECTED_KEY": expected_final[slot],
        "MATCH_EXPECTED": int(receiver_memory[slot] == expected_final[slot]),
    }
    for slot in range(memory_slots)
])

bus_accuracy = float(np.mean(np.array(receiver_memory) == np.array(expected_final)))
receiver_safety_accuracy = float(np.mean(df_memory["MATCH_EXPECTED"]))

expected_absorbed_total = sum(len(event["expected_absorbed"]) for event in bus_events)

absorbed_accuracy = (
    1.0
    if absorbed_receivers == expected_absorbed_total and bus_accuracy == 1.0
    else 0.0
)

result = (
    "SPECTRAL_COHERENCE_BUS_LOCKED"
    if bus_accuracy == 1.0
    and receiver_safety_accuracy == 1.0
    and absorbed_accuracy == 1.0
    else "SPECTRAL_COHERENCE_BUS_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": memory_slots,
    "RECEIVER_FILTER_SEQUENCE": " ".join(receiver_filter_sequence),
    "RECEIVER_FINAL_SEQUENCE": " ".join(receiver_memory),
    "EXPECTED_FINAL_SEQUENCE": " ".join(expected_final),
    "BUS_EVENTS": len(bus_events),
    "AMPLITUDE_ACCEPTED_PACKETS": amplitude_accepted,
    "PHASE_LOCKED_PACKETS": phase_locked,
    "CLOCK_RESYNC_EVENTS": clock_resync_events,
    "BUS_ROUTED_PACKETS": bus_routed_packets,
    "REJECTED_PACKETS": rejected_packets,
    "ABSORBED_RECEIVERS": absorbed_receivers,
    "BLOCKED_BY_FANOUT": blocked_by_full,
    "BUS_ACCURACY": round(bus_accuracy, 3),
    "RECEIVER_SAFETY_ACCURACY": round(receiver_safety_accuracy, 3),
    "ABSORPTION_ACCURACY": round(absorbed_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "PHASE_TOLERANCE": phase_tolerance,
    "JITTER_TOLERANCE": jitter_tolerance,
    "DEFAULT_FANOUT": default_fanout,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8.5, 2.9 * len(snapshots)))

if len(snapshots) == 1:
    axes = [axes]

for ax, (row, packet, memory_state) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, memory_state)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"memory slots:                 {memory_slots}")
print(f"receiver filter sequence:     {' '.join(receiver_filter_sequence)}")
print(f"receiver final sequence:      {' '.join(receiver_memory)}")
print(f"expected final sequence:      {' '.join(expected_final)}")
print(f"bus events:                   {len(bus_events)}")
print(f"amplitude accepted packets:   {amplitude_accepted}")
print(f"phase locked packets:         {phase_locked}")
print(f"clock resync events:          {clock_resync_events}")
print(f"bus routed packets:           {bus_routed_packets}")
print(f"rejected packets:             {rejected_packets}")
print(f"absorbed receivers:           {absorbed_receivers}")
print(f"blocked by fanout:            {blocked_by_full}")
print(f"bus accuracy:                 {round(bus_accuracy, 3)}")
print(f"receiver safety accuracy:     {round(receiver_safety_accuracy, 3)}")
print(f"absorption accuracy:          {round(absorbed_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"phase tolerance:              {phase_tolerance}")
print(f"jitter tolerance:             {jitter_tolerance}")

print("\nCoherence bus decision table:")
display(df_decision)

print("\nReceiver routing table:")
display(df_receivers)

print("\nClock / phase table:")
display(df_clock)

print("\nFinal bus memory table:")
display(df_memory)

print("\nSummary:")
display(df_summary)
```
