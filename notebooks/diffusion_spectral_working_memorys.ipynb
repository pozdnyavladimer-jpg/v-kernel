```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D44_SPECTRAL_WORKING_MEMORY — FULL FIXED COLAB CELL
# Active working memory after D43 Attention Gate.
#
# Locked behavior:
# - attention can WRITE a new active intent
# - repeated same key+mode REFRESHES memory
# - weak old memory can SWITCH to a stronger attended intent
# - ambiguous packet over strong same memory is HELD, not erased
# - weak / wrong / phase-unlocked / idle packets decay
# ============================================================

np.random.seed(44)

STATE = "D44_SPECTRAL_WORKING_MEMORY"

directions = 6
octaves = 7
spectral_states = directions * octaves

orbital_modes = ["NONE", "SPHERE", "RING", "HEX", "FLOWER", "SPIRAL", "BUTTERFLY", "SEED"]

RESONANCE_THRESHOLD = 0.62
AMBIGUITY_MARGIN = 0.12
ATTENTION_THRESHOLD = 0.58
PHASE_TOLERANCE = 0.18
JITTER_TOLERANCE = 0.16
WORKING_THRESHOLD = 0.55
FORGET_THRESHOLD = 0.24
REFRESH_GAIN = 0.10
TRUST_GAIN = 0.10
DECAY_RATE = 0.12
MAX_TTL = 4


# ------------------------------------------------------------
# Helpers
# ------------------------------------------------------------

def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE"]:
        return None, None
    if not isinstance(key, str) or "_" not in key:
        return None, None
    d_part, o_part = key.split("_")
    return int(d_part.replace("D", "")), int(o_part.replace("O", ""))


def make_key(d, o):
    return f"D{d}_O{o}"


def expected_phase(key):
    d, _ = parse_key(key)
    if d is None:
        return 0.0
    return round(d / directions, 3)


def phase_error(a, b):
    if a is None or b is None:
        return 1.0
    diff = abs((a - b) % 1.0)
    return min(diff, 1.0 - diff)


def state_id(d, o):
    return o * directions + d


def key_color(key):
    if key in [None, "EMPTY", "NONE"]:
        return "black"
    if key == "WHITE":
        return "white"
    d, o = parse_key(key)
    if d is None:
        return "black"
    hue = (d + 1) / (directions + 1)
    octave_shift = (o + 1) / (octaves + 1)
    return plt.cm.hsv((hue + 0.08 * octave_shift) % 1.0)


def mode_color(mode):
    if mode not in orbital_modes or mode == "NONE":
        return "white"
    return plt.cm.Set2((orbital_modes.index(mode) - 1) / max(1, len(orbital_modes) - 2))


def make_cell(key="EMPTY", mode="NONE", strength=0.0, ttl=0, trust=0.0, phase=None):
    if phase is None and key not in ["EMPTY", "NONE"]:
        phase = expected_phase(key)
    return {
        "key": key,
        "mode": mode,
        "strength": float(strength),
        "ttl": int(ttl),
        "trust": float(trust),
        "phase": phase,
    }


def clone_cell(c):
    return {
        "key": c["key"],
        "mode": c["mode"],
        "strength": float(c["strength"]),
        "ttl": int(c["ttl"]),
        "trust": float(c["trust"]),
        "phase": c["phase"],
    }


def decay_cell(c):
    out = clone_cell(c)

    if out["key"] in ["EMPTY", "NONE"]:
        return make_cell()

    out["strength"] = max(0.0, out["strength"] - DECAY_RATE)
    out["trust"] = max(0.0, out["trust"] - DECAY_RATE * 0.45)
    out["ttl"] = max(0, out["ttl"] - 1)

    if out["strength"] < FORGET_THRESHOLD or out["ttl"] <= 0:
        return make_cell()

    return out


def wave_token(key, mode, strength, phase, ttl):
    d, o = parse_key(key)
    if d is None or mode == "NONE":
        return "NONE"
    return f"WM_PHOTON[D{d}|O{o}|F{o+1}|P{round(phase,3)}|A{round(strength,3)}|TTL{ttl}|{mode}]"


# ------------------------------------------------------------
# Spectral packet and decoder
# ------------------------------------------------------------

def spectral_packet(target_key=None, strength=0.9, noise=0.025, ambiguous_key=None, ambiguous_strength=0.0):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(target_key)
        packet[o, d] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE"]:
        d, o = parse_key(ambiguous_key)
        packet[o, d] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    order = np.argsort(flat)[::-1]

    best_id = int(order[0])
    second_id = int(order[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_o = best_id // directions
    best_d = best_id % directions

    second_o = second_id // directions
    second_d = second_id % directions

    return {
        "best_key": make_key(best_d, best_o),
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": make_key(second_d, second_o),
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": best_score - second_score,
    }


# ------------------------------------------------------------
# Small visualization
# ------------------------------------------------------------

def draw_arrow(ax, a, b, reject=False):
    color = "red" if reject else "orange"
    ax.add_patch(FancyArrowPatch(a, b, arrowstyle="->", mutation_scale=12, linewidth=1.7, color=color))


def draw_packet_grid(ax, x0, y0, packet, selected_key=None):
    cell = 0.12
    gap = 0.023
    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE"] else (None, None)

    for o in range(octaves):
        for d in range(directions):
            x = x0 + d * (cell + gap)
            y = y0 + (octaves - 1 - o) * (cell + gap)
            edge = "lime" if selected == (d, o) else "white"
            lw = 1.1 if selected == (d, o) else 0.25
            ax.add_patch(Rectangle((x, y), cell, cell, facecolor=plt.cm.plasma(packet[o, d]),
                                   edgecolor=edge, linewidth=lw))


def draw_memory_bar(ax, x0, y0, memory, active_slot):
    cell = 0.14
    gap = 0.03
    for i, m in enumerate(memory):
        x = x0
        y = y0 + (len(memory) - 1 - i) * (cell + gap)
        edge = "lime" if i == active_slot else "black"
        ax.add_patch(Rectangle((x, y), cell, cell, facecolor=key_color(m["key"]), edgecolor=edge, linewidth=0.7))
        ax.add_patch(Circle((x + cell + 0.07, y + cell / 2), 0.04,
                            facecolor=mode_color(m["mode"]), edgecolor="black", linewidth=0.2))
        ax.text(x + cell + 0.13, y + cell / 2, str(i), fontsize=5, va="center")


def draw_strength(ax, x, y, value):
    ax.text(x, y + 0.16, "strength", fontsize=6, ha="left")
    ax.add_patch(Rectangle((x, y), 0.7, 0.09, facecolor="white", edgecolor="black", linewidth=0.4))
    ax.add_patch(Rectangle((x, y), 0.7 * np.clip(value, 0, 1), 0.09, facecolor="limegreen", edgecolor="none"))
    ax.text(x + 0.76, y + 0.045, f"{value:.2f}", fontsize=6, va="center")


def draw_snapshot(ax, row, packet, before, after, memory):
    slot = int(row["MEMORY_SLOT"])
    action = row["ACTION"]
    reject = action.startswith("REJECT") or action.startswith("PASSIVE")

    ax.set_xlim(0, 12)
    ax.set_ylim(0.45, 3.2)
    ax.set_aspect("equal")
    ax.axis("off")
    ax.set_facecolor("white")

    xs = [0.7, 1.35, 2.0, 2.65, 3.3, 4.1, 4.95]
    labels = ["SRC", "ATTN", "FIELD", "CLOCK", "WM GATE", "BEFORE", "AFTER"]
    colors = [
        key_color(row["INCOMING_KEY"]),
        key_color(row["INCOMING_KEY"]),
        "limegreen",
        "limegreen" if row["PHASE_OK"] else "red",
        "limegreen" if row["ACCEPTED"] else "red",
        key_color(before["key"]),
        key_color(after["key"]),
    ]

    for x, lab, color in zip(xs, labels, colors):
        if lab in ["FIELD", "CLOCK", "WM GATE"]:
            ax.add_patch(Rectangle((x - 0.11, 2.17), 0.22, 0.22, facecolor=color, edgecolor="black", linewidth=0.7))
        else:
            ax.add_patch(Circle((x, 2.28), 0.20, facecolor=color, edgecolor="black", linewidth=0.7))
        ax.text(x, 1.90, lab, fontsize=5.5, ha="center")

    for i in range(len(xs) - 1):
        draw_arrow(ax, (xs[i] + 0.20, 2.28), (xs[i + 1] - 0.20, 2.28), reject=(reject and i >= 3))

    ax.text(0.7, 1.58, f"{row['INCOMING_KEY']}\n{row['INCOMING_MODE']}", fontsize=5.5, ha="center")
    ax.text(4.1, 1.58, f"{before['key']}\n{before['mode']}", fontsize=5.5, ha="center")
    ax.text(4.95, 1.58, f"{after['key']}\n{after['mode']}", fontsize=5.5, ha="center")

    draw_strength(ax, 5.55, 2.22, after["strength"])

    ax.text(6.8, 3.03, "42-channel packet", fontsize=6, ha="left")
    draw_packet_grid(ax, 6.8, 1.83, packet, row["DECODED_KEY"])

    ax.text(9.75, 3.03, "working memory", fontsize=6, ha="left")
    draw_memory_bar(ax, 9.75, 1.43, memory, slot)

    ax.text(
        0.45,
        0.65,
        f"case={row['CASE']} slot={slot} decoded={row['DECODED_KEY']} "
        f"score={row['BEST_SCORE']} margin={row['MARGIN']} action={action}\n"
        f"reason={row['REASON']} token={row['OUTPUT_TOKEN']}",
        fontsize=5.4,
        ha="left"
    )

    ax.set_title(f"D44 t={row['T']} — {row['CASE']} / {action}", fontsize=8)


# ------------------------------------------------------------
# Initial memory
# ------------------------------------------------------------

working_memory = [
    make_cell("EMPTY", "NONE", 0.0, 0, 0.0),
    make_cell("D1_O2", "RING", 0.62, 2, 0.32),
    make_cell("D3_O6", "HEX", 0.80, 3, 0.55),
    make_cell("EMPTY", "NONE", 0.0, 0, 0.0),
    make_cell("D0_O1", "SPHERE", 0.42, 2, 0.22),
    make_cell("EMPTY", "NONE", 0.0, 0, 0.0),
    make_cell("D4_O5", "FLOWER", 0.92, 3, 0.75),
    make_cell("D2_O3", "BUTTERFLY", 0.60, 2, 0.35),
]

initial_sequence = " ".join(m["key"] for m in working_memory)
initial_modes = " ".join(m["mode"] for m in working_memory)


# ------------------------------------------------------------
# Test events
# ------------------------------------------------------------

events = [
    dict(case="WRITE_FLOWER_0", slot=0, incoming_key="D4_O5", incoming_mode="FLOWER",
         strength=0.92, attention=0.94, phase_shift=0.01, jitter=0.02,
         expected_action="WRITE_WORKING_MEMORY", expected_key="D4_O5", expected_mode="FLOWER"),

    dict(case="IDLE_DECAY_RING_1", slot=1, incoming_key="NONE", incoming_mode="NONE",
         strength=0.0, attention=0.0, phase_shift=0.0, jitter=0.02,
         expected_action="PASSIVE_DECAY_ONLY", expected_key="D1_O2", expected_mode="RING"),

    dict(case="COHERENT_REFRESH_FLOWER_0", slot=0, incoming_key="D4_O5", incoming_mode="FLOWER",
         strength=0.90, attention=0.93, phase_shift=0.02, jitter=0.02,
         expected_action="REFRESH_WORKING_MEMORY", expected_key="D4_O5", expected_mode="FLOWER"),

    dict(case="WRONG_INTENT_REJECT_2", slot=2, incoming_key="D3_O6", incoming_mode="SPIRAL",
         strength=0.88, attention=0.90, phase_shift=0.01, jitter=0.02,
         expected_action="REJECT_INTENT_MISMATCH", expected_key="D3_O6", expected_mode="HEX"),

    dict(case="PHASE_DRIFT_REJECT_2", slot=2, incoming_key="D3_O6", incoming_mode="HEX",
         strength=0.90, attention=0.89, phase_shift=0.34, jitter=0.03,
         expected_action="REJECT_PHASE_UNLOCKED", expected_key="D3_O6", expected_mode="HEX"),

    dict(case="RESYNC_HEX_RECOVER_2", slot=2, incoming_key="D3_O6", incoming_mode="HEX",
         strength=0.91, attention=0.92, phase_shift=0.31, jitter=0.04, clock_resync=True,
         expected_action="REFRESH_WORKING_MEMORY", expected_key="D3_O6", expected_mode="HEX"),

    dict(case="NEW_WRITE_RING_5", slot=5, incoming_key="D1_O2", incoming_mode="RING",
         strength=0.91, attention=0.88, phase_shift=0.02, jitter=0.03,
         expected_action="WRITE_WORKING_MEMORY", expected_key="D1_O2", expected_mode="RING"),

    dict(case="WEAK_PACKET_DECAY_SLOT_4", slot=4, incoming_key="D0_O1", incoming_mode="SEED",
         strength=0.39, attention=0.42, phase_shift=0.02, jitter=0.03,
         expected_action="REJECT_WEAK_OR_AMBIGUOUS_AND_DECAY", expected_key="D0_O1", expected_mode="SPHERE"),

    dict(case="SWITCH_WEAK_SPHERE_TO_RING_4", slot=4, incoming_key="D1_O2", incoming_mode="RING",
         strength=0.93, attention=0.93, phase_shift=0.01, jitter=0.02,
         expected_action="SWITCH_ACTIVE_INTENT", expected_key="D1_O2", expected_mode="RING"),

    dict(case="AMBIGUOUS_FLOWER_HOLD_0", slot=0, incoming_key="D4_O5", incoming_mode="FLOWER",
         strength=0.86, ambiguous_key="D1_O2", ambiguous_strength=0.81,
         attention=0.91, phase_shift=0.01, jitter=0.02,
         expected_action="HOLD_WORKING_MEMORY", expected_key="D4_O5", expected_mode="FLOWER"),

    dict(case="TRUSTED_MEMORY_REINFORCE_6", slot=6, incoming_key="D4_O5", incoming_mode="FLOWER",
         strength=0.89, attention=0.95, phase_shift=0.0, jitter=0.02, trusted_memory=True,
         expected_action="REINFORCE_TRUSTED_MEMORY", expected_key="D4_O5", expected_mode="FLOWER"),
]


# ------------------------------------------------------------
# Main simulation
# ------------------------------------------------------------

decision_rows = []
memory_rows = []
snapshots = []

accepted_events = 0
written_events = 0
refreshed_events = 0
switch_events = 0
held_events = 0
decay_events = 0
rejected_events = 0
phase_rejects = 0
weak_rejects = 0
intent_rejects = 0

for t, e in enumerate(events):
    slot = e["slot"]
    before = clone_cell(working_memory[slot])
    incoming_key = e["incoming_key"]
    incoming_mode = e["incoming_mode"]

    if incoming_key == "NONE":
        packet = spectral_packet(None, 0.0)
        dec = dict(best_key="NONE", best_state_id=-1, best_score=0.0,
                   second_key="NONE", second_state_id=-1, second_score=0.0, margin=0.0)
        measured_phase = None
        corrected_phase = None
        raw_phase_error = 0.0
        corrected_phase_error = 0.0
    else:
        packet = spectral_packet(
            incoming_key,
            e["strength"],
            ambiguous_key=e.get("ambiguous_key", None),
            ambiguous_strength=e.get("ambiguous_strength", 0.0),
        )
        dec = decode_packet(packet)
        measured_phase = (expected_phase(dec["best_key"]) + e.get("phase_shift", 0.0)) % 1.0
        raw_phase_error = phase_error(measured_phase, expected_phase(incoming_key))

        if e.get("clock_resync", False):
            corrected_phase = (expected_phase(incoming_key) + 0.003) % 1.0
        else:
            corrected_phase = measured_phase

        corrected_phase_error = phase_error(corrected_phase, expected_phase(incoming_key))

    decoded_key = dec["best_key"]
    best_score = dec["best_score"]
    margin = dec["margin"]

    score_ok = best_score >= RESONANCE_THRESHOLD
    margin_ok = margin >= AMBIGUITY_MARGIN
    attention_ok = e["attention"] >= ATTENTION_THRESHOLD
    phase_ok = corrected_phase_error <= PHASE_TOLERANCE
    jitter_ok = e.get("jitter", 0.0) <= JITTER_TOLERANCE
    key_match = incoming_key == "NONE" or decoded_key == incoming_key

    packet_accepted = bool(score_ok and margin_ok and attention_ok and phase_ok and jitter_ok and key_match)

    empty_slot = before["key"] in ["EMPTY", "NONE"]
    expired = before["ttl"] <= 0 or before["strength"] < FORGET_THRESHOLD

    same_key = before["key"] == decoded_key and decoded_key not in ["EMPTY", "NONE"]
    same_mode = before["mode"] == incoming_mode and incoming_mode != "NONE"

    switch_allowed = (
        packet_accepted
        and not empty_slot
        and not expired
        and not same_key
        and best_score >= before["strength"] + 0.12
        and e["attention"] >= before["trust"] + 0.15
    )

    # D44.1 critical fix
    protected_same_memory = (
        before["key"] != "EMPTY"
        and same_key
        and same_mode
        and before["strength"] >= WORKING_THRESHOLD
    )

    ambiguous_protected_hold = (
        incoming_key != "NONE"
        and score_ok
        and attention_ok
        and not margin_ok
        and phase_ok
        and jitter_ok
        and protected_same_memory
    )

    weak_memory_switch = (
        before["key"] != "EMPTY"
        and packet_accepted
        and before["strength"] < WORKING_THRESHOLD
        and best_score >= before["strength"] + 0.25
    )

    after = clone_cell(before)
    accepted = False

    if incoming_key == "NONE":
        after = decay_cell(after)
        action = "PASSIVE_DECAY_ONLY"
        reason = "NO_INPUT_DECAY"

    elif not phase_ok:
        after = decay_cell(after)
        action = "REJECT_PHASE_UNLOCKED"
        reason = "PHASE_UNLOCKED"
        phase_rejects += 1

    elif ambiguous_protected_hold:
        after["ttl"] = max(after["ttl"], 1)
        after["strength"] = before["strength"]
        after["trust"] = before["trust"]
        after["phase"] = before["phase"]
        action = "HOLD_WORKING_MEMORY"
        reason = "AMBIGUOUS_PACKET_OVER_PROTECTED_ACTIVE_MEMORY"
        held_events += 1

    elif weak_memory_switch:
        after["key"] = decoded_key
        after["mode"] = incoming_mode
        after["strength"] = min(1.0, best_score + e["attention"] * 0.07)
        after["ttl"] = MAX_TTL
        after["trust"] = min(1.0, e["attention"] * 0.80)
        after["phase"] = corrected_phase
        action = "SWITCH_ACTIVE_INTENT"
        reason = "WEAK_CURRENT_MEMORY_SWITCHED_TO_STRONGER_ATTENDED_INTENT"
        accepted = True
        switch_events += 1

    elif not key_match:
        after = decay_cell(after)
        action = "REJECT_FILTER_OR_KEY_MISMATCH"
        reason = "DECODED_KEY_DOES_NOT_MATCH_INCOMING_KEY"
        intent_rejects += 1

    elif not score_ok or not attention_ok or not margin_ok or not jitter_ok:
        after = decay_cell(after)
        action = "REJECT_WEAK_OR_AMBIGUOUS_AND_DECAY"
        reason = "WEAK_ATTENTION_OR_AMBIGUOUS_OR_JITTER"
        weak_rejects += 1

    elif same_key and same_mode:
        after["strength"] = min(1.0, before["strength"] + REFRESH_GAIN + e["attention"] * 0.08)
        after["ttl"] = MAX_TTL
        after["trust"] = min(1.0, before["trust"] + TRUST_GAIN)
        after["phase"] = corrected_phase

        if e.get("trusted_memory", False):
            action = "REINFORCE_TRUSTED_MEMORY"
            reason = "TRUSTED_REPEAT"
        else:
            action = "REFRESH_WORKING_MEMORY"
            reason = "SAME_KEY_SAME_MODE_REFRESH"

        accepted = True
        refreshed_events += 1

    elif same_key and not same_mode:
        after = decay_cell(after)
        action = "REJECT_INTENT_MISMATCH"
        reason = "SAME_KEY_DIFFERENT_ORBITAL_INTENT"
        intent_rejects += 1

    elif switch_allowed:
        after["key"] = decoded_key
        after["mode"] = incoming_mode
        after["strength"] = min(1.0, best_score + e["attention"] * 0.07)
        after["ttl"] = MAX_TTL
        after["trust"] = min(1.0, e["attention"] * 0.80)
        after["phase"] = corrected_phase
        action = "SWITCH_ACTIVE_INTENT"
        reason = "NEW_PACKET_STRONGER_THAN_CURRENT_MEMORY"
        accepted = True
        switch_events += 1

    elif empty_slot or expired:
        after["key"] = decoded_key
        after["mode"] = incoming_mode
        after["strength"] = min(1.0, best_score + 0.03)
        after["ttl"] = MAX_TTL
        after["trust"] = min(1.0, e["attention"] * 0.75)
        after["phase"] = corrected_phase
        action = "WRITE_WORKING_MEMORY"
        reason = "EMPTY_OR_EXPIRED_SLOT"
        accepted = True
        written_events += 1

    else:
        after["ttl"] = max(0, after["ttl"] - 1)
        action = "HOLD_WORKING_MEMORY"
        reason = "CURRENT_MEMORY_WINS"
        held_events += 1

    if accepted:
        accepted_events += 1
    else:
        rejected_events += 1

    if action.startswith("REJECT") or action.startswith("PASSIVE"):
        decay_events += 1

    working_memory[slot] = after

    expected_match = int(
        action == e["expected_action"]
        and after["key"] == e["expected_key"]
        and after["mode"] == e["expected_mode"]
    )

    token = wave_token(after["key"], after["mode"], after["strength"], after["phase"], after["ttl"]) if accepted else "NONE"

    row = {
        "STATE": STATE,
        "T": t * 120,
        "CASE": e["case"],
        "MEMORY_SLOT": slot,
        "BEFORE_KEY": before["key"],
        "BEFORE_MODE": before["mode"],
        "BEFORE_STRENGTH": round(before["strength"], 3),
        "BEFORE_TTL": before["ttl"],
        "INCOMING_KEY": incoming_key,
        "INCOMING_MODE": incoming_mode,
        "ATTENTION": round(e["attention"], 3),
        "DECODED_KEY": decoded_key,
        "BEST_STATE_ID": dec["best_state_id"],
        "BEST_SCORE": round(best_score, 3),
        "SECOND_KEY": dec["second_key"],
        "SECOND_SCORE": round(dec["second_score"], 3),
        "MARGIN": round(margin, 3),
        "SCORE_OK": int(score_ok),
        "MARGIN_OK": int(margin_ok),
        "ATTENTION_OK": int(attention_ok),
        "PHASE_OK": int(phase_ok),
        "JITTER_OK": int(jitter_ok),
        "RAW_PHASE_ERROR": round(raw_phase_error, 3),
        "CORRECTED_PHASE_ERROR": round(corrected_phase_error, 3),
        "CLOCK_RESYNC": int(e.get("clock_resync", False)),
        "SAME_KEY": int(same_key),
        "SAME_MODE": int(same_mode),
        "ACCEPTED": int(accepted),
        "ACTION": action,
        "REASON": reason,
        "AFTER_KEY": after["key"],
        "AFTER_MODE": after["mode"],
        "AFTER_STRENGTH": round(after["strength"], 3),
        "AFTER_TTL": after["ttl"],
        "EXPECTED_ACTION": e["expected_action"],
        "EXPECTED_KEY": e["expected_key"],
        "EXPECTED_MODE": e["expected_mode"],
        "EXPECTED_MATCH": expected_match,
        "OUTPUT_TOKEN": token,
    }

    decision_rows.append(row)

    memory_rows.append({
        "STATE": STATE,
        "CASE": e["case"],
        "MEMORY_SLOT": slot,
        "BEFORE": f"{before['key']}:{before['mode']}:S{round(before['strength'],2)}:T{before['ttl']}",
        "AFTER": f"{after['key']}:{after['mode']}:S{round(after['strength'],2)}:T{after['ttl']}",
        "ACTION": action,
        "EXPECTED_MATCH": expected_match,
        "HELD_IF_REJECTED": int(not accepted and after["key"] == before["key"] and after["key"] != "EMPTY"),
    })

    snapshots.append((row, packet, before, after, [clone_cell(x) for x in working_memory]))


df_decision = pd.DataFrame(decision_rows)
df_memory = pd.DataFrame(memory_rows)

final_sequence = " ".join(m["key"] for m in working_memory)
final_modes = " ".join(m["mode"] for m in working_memory)
final_strengths = " ".join(str(round(m["strength"], 2)) for m in working_memory)
final_ttls = " ".join(str(m["ttl"]) for m in working_memory)

working_memory_accuracy = float(df_decision["EXPECTED_MATCH"].mean())
active_slots = int(sum(1 for m in working_memory if m["key"] != "EMPTY" and m["strength"] >= WORKING_THRESHOLD))

result = (
    "SPECTRAL_WORKING_MEMORY_LOCKED"
    if working_memory_accuracy == 1.0 and switch_events >= 1 and held_events >= 1 and active_slots >= 4
    else "SPECTRAL_WORKING_MEMORY_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": len(working_memory),
    "ORBITAL_MODES": ",".join(orbital_modes),
    "INITIAL_SEQUENCE": initial_sequence,
    "INITIAL_MODES": initial_modes,
    "FINAL_SEQUENCE": final_sequence,
    "FINAL_MODES": final_modes,
    "FINAL_STRENGTHS": final_strengths,
    "FINAL_TTLS": final_ttls,
    "CASES_TESTED": len(events),
    "ACCEPTED_EVENTS": accepted_events,
    "WRITTEN_EVENTS": written_events,
    "REFRESHED_EVENTS": refreshed_events,
    "SWITCH_EVENTS": switch_events,
    "HELD_EVENTS": held_events,
    "DECAY_EVENTS": decay_events,
    "REJECTED_EVENTS": rejected_events,
    "PHASE_REJECTS": phase_rejects,
    "WEAK_REJECTS": weak_rejects,
    "INTENT_REJECTS": intent_rejects,
    "ACTIVE_SLOTS": active_slots,
    "WORKING_MEMORY_ACCURACY": round(working_memory_accuracy, 3),
    "RESONANCE_THRESHOLD": RESONANCE_THRESHOLD,
    "AMBIGUITY_MARGIN": AMBIGUITY_MARGIN,
    "ATTENTION_THRESHOLD": ATTENTION_THRESHOLD,
    "PHASE_TOLERANCE": PHASE_TOLERANCE,
    "JITTER_TOLERANCE": JITTER_TOLERANCE,
    "WORKING_THRESHOLD": WORKING_THRESHOLD,
    "MAX_TTL": MAX_TTL,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8.4, 2.3 * len(snapshots)))

if len(snapshots) == 1:
    axes = [axes]

for ax, (row, packet, before, after, memory) in zip(axes, snapshots):
    draw_snapshot(ax, row, packet, before, after, memory)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:             {spectral_states}")
print(f"directions x octaves:        {directions} x {octaves}")
print(f"memory slots:                {len(working_memory)}")
print(f"orbital modes:               {', '.join(orbital_modes)}")
print(f"initial sequence:            {initial_sequence}")
print(f"initial modes:               {initial_modes}")
print(f"final sequence:              {final_sequence}")
print(f"final modes:                 {final_modes}")
print(f"final strengths:             {final_strengths}")
print(f"final TTLs:                  {final_ttls}")
print(f"cases tested:                {len(events)}")
print(f"accepted events:             {accepted_events}")
print(f"written events:              {written_events}")
print(f"refreshed events:            {refreshed_events}")
print(f"switch events:               {switch_events}")
print(f"held events:                 {held_events}")
print(f"decay events:                {decay_events}")
print(f"rejected events:             {rejected_events}")
print(f"active slots:                {active_slots}")
print(f"working memory accuracy:     {round(working_memory_accuracy, 3)}")
print(f"resonance threshold:         {RESONANCE_THRESHOLD}")
print(f"ambiguity margin:            {AMBIGUITY_MARGIN}")
print(f"attention threshold:         {ATTENTION_THRESHOLD}")
print(f"phase tolerance:             {PHASE_TOLERANCE}")
print(f"jitter tolerance:            {JITTER_TOLERANCE}")
print(f"working threshold:           {WORKING_THRESHOLD}")

print("\nWorking memory decision table:")
display(df_decision)

print("\nMemory update table:")
display(df_memory)

print("\nSummary:")
display(df_summary)
```
