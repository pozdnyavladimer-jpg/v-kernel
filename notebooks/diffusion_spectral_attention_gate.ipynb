```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch, RegularPolygon
from IPython.display import display

# ============================================================
# D43_SPECTRAL_ATTENTION_GATE
# Attention gate for field memory:
#
# Goal:
# many spectral/orbital packets arrive at the same time
# -> decode spectral key from 42-channel packet
# -> check phase / jitter / filter / intent / memory trust
# -> select only the most important coherent packet
# -> suppress noise, wrong intent, wrong phase, weak and ambiguous packets
#
# Continuation after:
# D36 coherence bus
# D37 orbital intent
# D38 honeycomb lattice
# D39 honeycomb memory wave
# D40 macro resonance loop
# D41 field memory decay and reinforcement
# D42 spectral self-tuning filter
# ============================================================

np.random.seed(43)

STATE = "D43_SPECTRAL_ATTENTION_GATE"

directions = 6
octaves = 7
spectral_states = directions * octaves

orbital_modes = [
    "SPHERE",
    "RING",
    "HEX",
    "FLOWER",
    "SPIRAL",
    "BUTTERFLY",
    "NONE",
    "NOISE",
]

resonance_threshold = 0.62
ambiguity_margin = 0.12
phase_tolerance = 0.18
jitter_tolerance = 0.16
attention_threshold = 0.62
selection_margin = 0.02

receiver_filter_keys = [
    "D4_O5",
    "D1_O2",
    "D3_O6",
    "D5_O4",
    "D0_O1",
    "D2_O3",
    "D4_O5",
    "D1_O2",
]

receiver_filter_intents = [
    "FLOWER",
    "RING",
    "HEX",
    "BUTTERFLY",
    "RING",
    "SPIRAL",
    "FLOWER",
    "RING",
]

initial_filter_trust = np.array(
    [0.78, 0.55, 0.20, 0.45, 0.30, 0.25, 0.65, 0.50],
    dtype=float,
)

memory_keys = ["EMPTY"] * 8
memory_intents = ["NONE"] * 8
filter_trust = initial_filter_trust.copy()

expected_memory_keys = ["EMPTY"] * 8
expected_memory_intents = ["NONE"] * 8


def parse_key(key):
    if key in [None, "EMPTY", "NONE", "WHITE", "NOISE"]:
        return None, None
    d_part, o_part = key.split("_")
    direction = int(d_part.replace("D", ""))
    octave = int(o_part.replace("O", ""))
    return direction, octave


def make_key(direction, octave):
    return f"D{direction}_O{octave}"


def base_phase(key):
    direction, octave = parse_key(key)
    if direction is None:
        return 0.0
    return round(direction / directions, 3)


def circular_phase_error(a, b):
    diff = abs(a - b) % 1.0
    return min(diff, 1.0 - diff)


def key_color(key, strength=1.0):
    if key == "WHITE":
        return "white"
    if key in ["EMPTY", "NONE"]:
        return "black"
    if key == "NOISE":
        return "gray"

    direction, octave = parse_key(key)
    if direction is None:
        return "black"

    hue_value = (direction + 1) / (directions + 1)
    octave_value = (octave + 1) / (octaves + 1)
    return plt.cm.hsv((hue_value + 0.08 * octave_value) % 1.0)


def intent_color(intent):
    palette = {
        "FLOWER": "limegreen",
        "RING": "gold",
        "HEX": "royalblue",
        "SPIRAL": "magenta",
        "BUTTERFLY": "turquoise",
        "SPHERE": "orange",
        "NOISE": "gray",
        "NONE": "black",
    }
    return palette.get(intent, "black")


def spectral_packet(
    target_key=None,
    strength=0.9,
    noise=0.025,
    broad=False,
    ambiguous_key=None,
    ambiguous_strength=0.0,
):
    packet = np.random.normal(0.0, noise, size=(octaves, directions))
    packet = np.clip(packet, 0.0, 1.0)

    if broad:
        packet += 0.18 + np.random.normal(0.0, noise, size=(octaves, directions))
        packet = np.clip(packet, 0.0, 1.0)

    if target_key not in [None, "EMPTY", "NONE", "WHITE", "NOISE"]:
        direction, octave = parse_key(target_key)
        packet[octave, direction] = strength

    if ambiguous_key not in [None, "EMPTY", "NONE", "WHITE", "NOISE"]:
        direction2, octave2 = parse_key(ambiguous_key)
        packet[octave2, direction2] = ambiguous_strength

    return np.clip(packet, 0.0, 1.0)


def decode_packet(packet):
    flat = packet.flatten()
    sorted_idx = np.argsort(flat)[::-1]

    best_id = int(sorted_idx[0])
    second_id = int(sorted_idx[1])

    best_score = float(flat[best_id])
    second_score = float(flat[second_id])

    best_octave = best_id // directions
    best_direction = best_id % directions

    second_octave = second_id // directions
    second_direction = second_id % directions

    best_key = make_key(best_direction, best_octave)
    second_key = make_key(second_direction, second_octave)

    margin = best_score - second_score
    accepted = best_score >= resonance_threshold and margin >= ambiguity_margin

    return {
        "best_key": best_key,
        "best_state_id": best_id,
        "best_score": best_score,
        "second_key": second_key,
        "second_state_id": second_id,
        "second_score": second_score,
        "margin": margin,
        "accepted": accepted,
    }


def attention_score(best_score, phase_error, intent_match, key_match, trust_score):
    amp_score = min(1.0, best_score)
    phase_score = max(0.0, 1.0 - phase_error / phase_tolerance)
    novelty = 1.0 - trust_score if key_match and intent_match else 0.0

    score = (
        0.25 * amp_score
        + 0.20 * phase_score
        + 0.20 * int(intent_match)
        + 0.25 * trust_score
        + 0.10 * novelty
    )

    return score, phase_score, novelty


def attention_action_for_reject(evaluated):
    if len(evaluated) == 0:
        return "HOLD_IDLE"

    if any(not e["amp_ok"] for e in evaluated):
        return "REJECT_WEAK_OR_AMBIGUOUS"

    if any(not e["phase_ok"] for e in evaluated):
        return "REJECT_PHASE_UNLOCKED"

    if any(not e["jitter_ok"] for e in evaluated):
        return "REJECT_DECOHERENCE"

    return "REJECT_INTENT_OR_FILTER"


def photon_token(key, intent, attention, phase, jitter):
    if key in ["NONE", "EMPTY"]:
        return "NONE"
    return (
        f"ATTENTION_PHOTON[{key}|{intent}|"
        f"A{round(attention, 3)}|P{round(phase, 3)}|J{round(jitter, 3)}]"
    )


def draw_packet_grid(ax, x0, y0, packet, title, selected_key=None):
    cell = 0.13
    gap = 0.025

    ax.text(x0, y0 + 1.35, title, fontsize=6.5, ha="left", va="bottom", color="black")

    selected = parse_key(selected_key) if selected_key not in [None, "EMPTY", "NONE"] else (None, None)

    for octave in range(octaves):
        for direction in range(directions):
            x = x0 + direction * (cell + gap)
            y = y0 + (octaves - 1 - octave) * (cell + gap)

            amp = packet[octave, direction]
            color = plt.cm.plasma(amp)

            is_selected = selected == (direction, octave)
            edge = "lime" if is_selected else "white"
            lw = 1.3 if is_selected else 0.25

            ax.add_patch(
                Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw)
            )


def draw_arrow(ax, start, end, active=True, rejected=False):
    color = "red" if rejected else "orange" if active else "gray"
    alpha = 1.0 if active else 0.3

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=13,
        linewidth=2.0,
        color=color,
        alpha=alpha,
    )
    ax.add_patch(arrow)


def draw_orbital_icon(ax, cx, cy, mode, scale=0.22, active=True):
    color = intent_color(mode)
    edge = "lime" if active else "red"

    if mode == "FLOWER":
        for k in range(6):
            a = 2 * np.pi * k / 6
            ax.add_patch(
                Circle(
                    (cx + np.cos(a) * scale * 0.75, cy + np.sin(a) * scale * 0.75),
                    scale * 0.32,
                    facecolor="none",
                    edgecolor=color,
                    linewidth=1.0,
                    alpha=0.9,
                )
            )
        ax.add_patch(Circle((cx, cy), scale * 0.2, facecolor=color, edgecolor=edge, linewidth=1.0))

    elif mode == "RING":
        ax.add_patch(Circle((cx, cy), scale, facecolor="none", edgecolor=color, linewidth=1.8))
        ax.add_patch(Circle((cx, cy), scale * 0.38, facecolor="white", edgecolor=color, linewidth=0.8))

    elif mode == "HEX":
        ax.add_patch(
            RegularPolygon(
                (cx, cy),
                numVertices=6,
                radius=scale,
                orientation=np.pi / 6,
                facecolor="none",
                edgecolor=color,
                linewidth=1.8,
            )
        )

    elif mode == "SPIRAL":
        theta = np.linspace(0, 3.2 * np.pi, 90)
        r = np.linspace(0.02, scale, 90)
        ax.plot(cx + r * np.cos(theta), cy + r * np.sin(theta), color=color, linewidth=1.6)

    elif mode == "BUTTERFLY":
        ax.add_patch(Circle((cx - scale * 0.45, cy + scale * 0.22), scale * 0.45, facecolor="none", edgecolor=color, linewidth=1.5))
        ax.add_patch(Circle((cx + scale * 0.45, cy + scale * 0.22), scale * 0.45, facecolor="none", edgecolor=color, linewidth=1.5))
        ax.add_patch(Circle((cx - scale * 0.45, cy - scale * 0.22), scale * 0.45, facecolor="none", edgecolor=color, linewidth=1.5))
        ax.add_patch(Circle((cx + scale * 0.45, cy - scale * 0.22), scale * 0.45, facecolor="none", edgecolor=color, linewidth=1.5))

    elif mode == "SPHERE":
        ax.add_patch(Circle((cx, cy), scale, facecolor=color, edgecolor=edge, linewidth=1.0, alpha=0.5))

    else:
        ax.add_patch(Circle((cx, cy), scale, facecolor="white", edgecolor="black", linewidth=0.7))
        ax.text(cx, cy, "∅", fontsize=8, ha="center", va="center")


def draw_memory_slots(ax, x0, y0, keys, intents, title, active_slot=None):
    cell = 0.18
    gap = 0.045

    ax.text(x0, y0 + 2.05, title, fontsize=6.5, ha="left", va="bottom", color="black")

    for slot, key in enumerate(keys):
        x = x0
        y = y0 + (len(keys) - 1 - slot) * (cell + gap)

        color = key_color(key)
        edge = "lime" if slot == active_slot else "black"
        lw = 1.4 if slot == active_slot else 0.4

        ax.add_patch(Rectangle((x, y), cell, cell, facecolor=color, edgecolor=edge, linewidth=lw))
        ax.text(x + 0.24, y + cell / 2, f"{slot}:{intents[slot][:3]}", fontsize=5.5, ha="left", va="center", color="black")


def draw_snapshot(ax, event_row, evaluated, selected, memory_keys_snapshot, memory_intents_snapshot):
    ax.set_facecolor("white")
    ax.set_xlim(0, 13)
    ax.set_ylim(0.35, 5.1)
    ax.set_aspect("equal")
    ax.axis("off")

    case = event_row["CASE"]
    t = event_row["T"]
    action = event_row["ACTION"]

    ax.text(0.45, 4.75, "incoming packets", fontsize=7, ha="left", color="black")

    if len(evaluated) == 0:
        ax.text(0.7, 3.8, "NO INPUT", fontsize=8, ha="center", color="gray")

    for idx, cand in enumerate(evaluated[:4]):
        y = 4.15 - idx * 0.55

        src_color = key_color(cand["source_key"])
        ax.add_patch(Circle((0.7, y), 0.18, facecolor=src_color, edgecolor="black", linewidth=0.7))
        ax.text(0.7, y - 0.33, cand["id"], fontsize=6, ha="center", color="black")

        draw_orbital_icon(ax, 1.25, y, cand["intent"], scale=0.16, active=cand["valid"])

        ax.text(
            1.65,
            y,
            f"s{cand['slot']} {cand['decoded_key']} {cand['intent']}\nattn={cand['attention_score']}",
            fontsize=5.5,
            ha="left",
            va="center",
            color="black",
        )

    ax.add_patch(Rectangle((3.05, 3.05), 0.34, 0.34, facecolor="limegreen", edgecolor="black", linewidth=0.8))
    ax.text(3.22, 2.72, "FIELD", fontsize=6, ha="center", color="black")

    ax.add_patch(Rectangle((4.0, 3.05), 0.34, 0.34, facecolor="gold", edgecolor="black", linewidth=0.8))
    ax.text(4.17, 2.72, "CLOCK", fontsize=6, ha="center", color="black")

    gate_color = "limegreen" if selected is not None else "red"
    ax.add_patch(Rectangle((4.95, 3.05), 0.42, 0.42, facecolor=gate_color, edgecolor="black", linewidth=0.8))
    ax.text(5.16, 2.72, "ATTN\nGATE", fontsize=6, ha="center", color="black")

    if selected is not None:
        selected_key = selected["decoded_key"]
        selected_intent = selected["intent"]
        selected_slot = selected["slot"]
        selected_color = key_color(selected_key)
        rejected = False
    else:
        selected_key = "NONE"
        selected_intent = "NONE"
        selected_slot = None
        selected_color = "black"
        rejected = True

    ax.add_patch(Circle((6.55, 3.25), 0.32, facecolor=selected_color, edgecolor="lime" if not rejected else "red", linewidth=2.0))
    ax.text(6.55, 2.58, f"SELECT\n{selected_key}", fontsize=6, ha="center", color="black")
    draw_orbital_icon(ax, 7.25, 3.25, selected_intent, scale=0.24, active=not rejected)
    ax.text(7.25, 2.58, selected_intent, fontsize=6, ha="center", color="black")

    draw_arrow(ax, (2.55, 3.25), (3.05, 3.25), active=True, rejected=False)
    draw_arrow(ax, (3.39, 3.25), (4.0, 3.25), active=True, rejected=False)
    draw_arrow(ax, (4.34, 3.25), (4.95, 3.25), active=True, rejected=False)
    draw_arrow(ax, (5.37, 3.25), (6.22, 3.25), active=True, rejected=rejected)

    if selected is not None:
        packet_to_show = selected["packet"]
        grid_key = selected["decoded_key"]
    elif len(evaluated) > 0:
        strongest = sorted(evaluated, key=lambda x: x["attention_score"], reverse=True)[0]
        packet_to_show = strongest["packet"]
        grid_key = strongest["decoded_key"]
    else:
        packet_to_show = np.zeros((octaves, directions))
        grid_key = None

    draw_packet_grid(ax, 8.05, 2.35, packet_to_show, "42-channel attention packet", selected_key=grid_key)
    draw_memory_slots(ax, 10.55, 1.65, memory_keys_snapshot, memory_intents_snapshot, "field memory", active_slot=selected_slot)

    ax.text(
        0.45,
        0.55,
        f"case={case} selected={event_row['SELECTED_CANDIDATE']} expected={event_row['EXPECTED_SELECTED']} "
        f"score={event_row['SELECTED_ATTENTION_SCORE']} margin={event_row['SELECTION_MARGIN']} action={action}\n"
        f"{event_row['OUTPUT_TOKEN']}",
        fontsize=6.4,
        ha="left",
        color="black",
    )

    ax.set_title(
        f"D43 snapshot t={t} — {case} / {action}",
        fontsize=9.5,
        color="black",
    )


events = [
    {
        "case": "FOCUS_FLOWER_OVER_NOISE",
        "expected": "A",
        "candidates": [
            {"id": "A", "slot": 0, "source_key": "D4_O5", "emitted_key": "D4_O5", "intent": "FLOWER", "strength": 0.92, "phase_offset": 0.01, "jitter": 0.02},
            {"id": "B", "slot": 4, "source_key": "D0_O1", "emitted_key": "D0_O1", "intent": "NOISE", "strength": 0.70, "phase_offset": 0.02, "jitter": 0.03},
        ],
    },
    {
        "case": "SAME_KEY_INTENT_SELECT",
        "expected": "A",
        "candidates": [
            {"id": "A", "slot": 1, "source_key": "D1_O2", "emitted_key": "D1_O2", "intent": "RING", "strength": 0.90, "phase_offset": 0.02, "jitter": 0.02},
            {"id": "B", "slot": 1, "source_key": "D1_O2", "emitted_key": "D1_O2", "intent": "SPIRAL", "strength": 0.94, "phase_offset": 0.02, "jitter": 0.02},
        ],
    },
    {
        "case": "TRUSTED_MEMORY_WINS",
        "expected": "A",
        "candidates": [
            {"id": "A", "slot": 6, "source_key": "D4_O5", "emitted_key": "D4_O5", "intent": "FLOWER", "strength": 0.72, "phase_offset": 0.01, "jitter": 0.02},
            {"id": "B", "slot": 2, "source_key": "D3_O6", "emitted_key": "D3_O6", "intent": "HEX", "strength": 0.88, "phase_offset": 0.01, "jitter": 0.02},
        ],
    },
    {
        "case": "NOVEL_HEX_ACCEPT",
        "expected": "A",
        "candidates": [
            {"id": "A", "slot": 2, "source_key": "D3_O6", "emitted_key": "D3_O6", "intent": "HEX", "strength": 0.84, "phase_offset": 0.03, "jitter": 0.03},
            {"id": "B", "slot": 5, "source_key": "D2_O3", "emitted_key": "D2_O3", "intent": "SPIRAL", "strength": 0.66, "phase_offset": 0.02, "jitter": 0.03},
        ],
    },
    {
        "case": "WRONG_INTENT_REJECT",
        "expected": "NONE",
        "candidates": [
            {"id": "A", "slot": 4, "source_key": "D0_O1", "emitted_key": "D0_O1", "intent": "BUTTERFLY", "strength": 0.90, "phase_offset": 0.01, "jitter": 0.02},
        ],
    },
    {
        "case": "PHASE_DRIFT_REJECT",
        "expected": "NONE",
        "candidates": [
            {"id": "A", "slot": 3, "source_key": "D5_O4", "emitted_key": "D5_O4", "intent": "BUTTERFLY", "strength": 0.91, "phase_offset": 0.36, "jitter": 0.03},
        ],
    },
    {
        "case": "WEAK_PACKET_REJECT",
        "expected": "NONE",
        "candidates": [
            {"id": "A", "slot": 5, "source_key": "D2_O3", "emitted_key": "D2_O3", "intent": "SPIRAL", "strength": 0.38, "phase_offset": 0.01, "jitter": 0.02},
        ],
    },
    {
        "case": "AMBIGUOUS_ATTENTION_HOLD",
        "expected": "NONE",
        "candidates": [
            {"id": "A", "slot": 5, "source_key": "D2_O3", "emitted_key": "D2_O3", "intent": "SPIRAL", "strength": 0.86, "phase_offset": 0.01, "jitter": 0.02, "ambiguous_key": "D4_O5", "ambiguous_strength": 0.81},
        ],
    },
    {
        "case": "WHITE_WITH_TARGET_ATTEND",
        "expected": "A",
        "candidates": [
            {"id": "A", "slot": 2, "source_key": "WHITE", "emitted_key": "D3_O6", "intent": "HEX", "strength": 0.88, "phase_offset": 0.01, "jitter": 0.02, "broad": True},
        ],
    },
    {
        "case": "GLOBAL_MULTI_FOCUS",
        "expected": "A",
        "candidates": [
            {"id": "A", "slot": 0, "source_key": "D4_O5", "emitted_key": "D4_O5", "intent": "FLOWER", "strength": 0.82, "phase_offset": 0.01, "jitter": 0.02},
            {"id": "B", "slot": 3, "source_key": "D5_O4", "emitted_key": "D5_O4", "intent": "BUTTERFLY", "strength": 0.90, "phase_offset": 0.01, "jitter": 0.02},
            {"id": "C", "slot": 1, "source_key": "D1_O2", "emitted_key": "D1_O2", "intent": "RING", "strength": 0.85, "phase_offset": 0.01, "jitter": 0.02},
        ],
    },
]

attention_rows = []
candidate_rows = []
memory_rows = []
snapshots = []

attended_count = 0
rejected_count = 0
correct_attention_count = 0

for t, event in enumerate(events):
    evaluated = []

    for cand in event["candidates"]:
        packet = spectral_packet(
            target_key=cand.get("emitted_key"),
            strength=cand.get("strength", 0.9),
            broad=cand.get("broad", False),
            ambiguous_key=cand.get("ambiguous_key"),
            ambiguous_strength=cand.get("ambiguous_strength", 0.0),
        )

        decoded = decode_packet(packet)

        slot = cand["slot"]
        filter_key = receiver_filter_keys[slot]
        filter_intent = receiver_filter_intents[slot]

        expected_phase = base_phase(filter_key)
        measured_phase = (base_phase(decoded["best_key"]) + cand.get("phase_offset", 0.0)) % 1.0
        phase_error = circular_phase_error(measured_phase, expected_phase)

        key_match = decoded["best_key"] == filter_key
        intent_match = cand["intent"] == filter_intent
        amp_ok = bool(decoded["accepted"])
        phase_ok = phase_error <= phase_tolerance
        jitter = cand.get("jitter", 0.0)
        jitter_ok = jitter <= jitter_tolerance

        trust_score = float(filter_trust[slot])
        score, phase_score, novelty = attention_score(
            best_score=decoded["best_score"],
            phase_error=phase_error,
            intent_match=intent_match,
            key_match=key_match,
            trust_score=trust_score,
        )

        valid = (
            amp_ok
            and phase_ok
            and jitter_ok
            and key_match
            and intent_match
            and score >= attention_threshold
        )

        e = {
            "packet": packet,
            "id": cand["id"],
            "slot": slot,
            "source_key": cand["source_key"],
            "emitted_key": cand["emitted_key"],
            "intent": cand["intent"],
            "decoded_key": decoded["best_key"],
            "best_state_id": decoded["best_state_id"],
            "best_score": decoded["best_score"],
            "second_key": decoded["second_key"],
            "second_score": decoded["second_score"],
            "margin": decoded["margin"],
            "filter_key": filter_key,
            "filter_intent": filter_intent,
            "expected_phase": expected_phase,
            "measured_phase": measured_phase,
            "phase_error": phase_error,
            "phase_score": phase_score,
            "jitter": jitter,
            "trust_score": trust_score,
            "novelty": novelty,
            "attention_score": score,
            "amp_ok": amp_ok,
            "phase_ok": phase_ok,
            "jitter_ok": jitter_ok,
            "key_match": key_match,
            "intent_match": intent_match,
            "valid": valid,
        }

        evaluated.append(e)

    valid_sorted = sorted([e for e in evaluated if e["valid"]], key=lambda x: x["attention_score"], reverse=True)

    if len(valid_sorted) == 0:
        selected = None
        selected_id = "NONE"
        action = attention_action_for_reject(evaluated)
        selected_attention_score = 0.0
        score_margin = 0.0
        output_token = "NONE"
        rejected_count += 1

    elif len(valid_sorted) > 1 and (valid_sorted[0]["attention_score"] - valid_sorted[1]["attention_score"]) < selection_margin:
        selected = None
        selected_id = "NONE"
        action = "REJECT_ATTENTION_AMBIGUOUS"
        selected_attention_score = float(valid_sorted[0]["attention_score"])
        score_margin = float(valid_sorted[0]["attention_score"] - valid_sorted[1]["attention_score"])
        output_token = "NONE"
        rejected_count += 1

    else:
        selected = valid_sorted[0]
        selected_id = selected["id"]
        action = "ATTEND_AND_WRITE"
        selected_attention_score = float(selected["attention_score"])

        if len(valid_sorted) > 1:
            score_margin = float(valid_sorted[0]["attention_score"] - valid_sorted[1]["attention_score"])
        else:
            score_margin = 1.0

        slot = selected["slot"]
        memory_keys[slot] = selected["decoded_key"]
        memory_intents[slot] = selected["intent"]

        filter_trust[slot] = min(1.0, filter_trust[slot] + 0.08 + 0.04 * selected["novelty"])

        output_token = photon_token(
            selected["decoded_key"],
            selected["intent"],
            selected_attention_score,
            selected["measured_phase"],
            selected["jitter"],
        )

        attended_count += 1

    if event["expected"] != "NONE":
        expected_eval = [e for e in evaluated if e["id"] == event["expected"]][0]
        expected_memory_keys[expected_eval["slot"]] = expected_eval["decoded_key"]
        expected_memory_intents[expected_eval["slot"]] = expected_eval["intent"]

    correct_selection = selected_id == event["expected"]
    correct_attention_count += int(correct_selection)

    attended_slot = selected["slot"] if selected is not None else None
    for s in range(len(filter_trust)):
        if s != attended_slot:
            filter_trust[s] = max(0.05, filter_trust[s] * 0.995)

    for e in evaluated:
        candidate_rows.append({
            "STATE": STATE,
            "T": t * 120,
            "CASE": event["case"],
            "CANDIDATE": e["id"],
            "MEMORY_SLOT": e["slot"],
            "SOURCE_KEY": e["source_key"],
            "EMITTED_KEY": e["emitted_key"],
            "INTENT": e["intent"],
            "DECODED_KEY": e["decoded_key"],
            "FILTER_KEY": e["filter_key"],
            "FILTER_INTENT": e["filter_intent"],
            "BEST_SCORE": round(e["best_score"], 3),
            "SECOND_KEY": e["second_key"],
            "SECOND_SCORE": round(e["second_score"], 3),
            "MARGIN": round(e["margin"], 3),
            "PHASE_ERROR": round(e["phase_error"], 3),
            "JITTER": round(e["jitter"], 3),
            "TRUST": round(e["trust_score"], 3),
            "NOVELTY": round(e["novelty"], 3),
            "ATTENTION_SCORE": round(e["attention_score"], 3),
            "AMP_OK": int(e["amp_ok"]),
            "PHASE_OK": int(e["phase_ok"]),
            "JITTER_OK": int(e["jitter_ok"]),
            "KEY_MATCH": int(e["key_match"]),
            "INTENT_MATCH": int(e["intent_match"]),
            "VALID_FOR_ATTENTION": int(e["valid"]),
            "SELECTED": int(e["id"] == selected_id),
        })

    attention_rows.append({
        "STATE": STATE,
        "T": t * 120,
        "CASE": event["case"],
        "EXPECTED_SELECTED": event["expected"],
        "SELECTED_CANDIDATE": selected_id,
        "SELECTED_SLOT": selected["slot"] if selected is not None else "NONE",
        "SELECTED_KEY": selected["decoded_key"] if selected is not None else "NONE",
        "SELECTED_INTENT": selected["intent"] if selected is not None else "NONE",
        "SELECTED_ATTENTION_SCORE": round(selected_attention_score, 3),
        "SELECTION_MARGIN": round(score_margin, 3),
        "ACTION": action,
        "MATCH_EXPECTED_SELECTION": int(correct_selection),
        "OUTPUT_TOKEN": output_token,
    })

    memory_rows.append({
        "STATE": STATE,
        "T": t * 120,
        "CASE": event["case"],
        "ACTION": action,
        "MEMORY_KEYS": " ".join(memory_keys),
        "MEMORY_INTENTS": " ".join(memory_intents),
        "FILTER_TRUST": " ".join([str(round(x, 3)) for x in filter_trust]),
        "SELECTED": selected_id,
    })

    snapshots.append((
        attention_rows[-1],
        evaluated,
        selected,
        memory_keys.copy(),
        memory_intents.copy(),
    ))

df_attention = pd.DataFrame(attention_rows)
df_candidates = pd.DataFrame(candidate_rows)
df_memory = pd.DataFrame(memory_rows)

attention_accuracy = float(np.mean(df_attention["MATCH_EXPECTED_SELECTION"].values))

noise_suppression_accuracy = float(np.mean([
    1 if row["EXPECTED_SELECTED"] != "NONE" else int(row["SELECTED_CANDIDATE"] == "NONE")
    for _, row in df_attention.iterrows()
]))

memory_key_accuracy = float(np.mean(np.array(memory_keys) == np.array(expected_memory_keys)))
memory_intent_accuracy = float(np.mean(np.array(memory_intents) == np.array(expected_memory_intents)))
field_memory_accuracy = float((memory_key_accuracy + memory_intent_accuracy) / 2.0)

result = (
    "SPECTRAL_ATTENTION_GATE_LOCKED"
    if attention_accuracy == 1.0 and noise_suppression_accuracy == 1.0 and field_memory_accuracy == 1.0
    else "SPECTRAL_ATTENTION_GATE_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": spectral_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": len(memory_keys),
    "ORBITAL_MODES": len(orbital_modes),
    "CASES_TESTED": len(events),
    "ATTENDED_PACKETS": attended_count,
    "REJECTED_EVENTS": rejected_count,
    "FINAL_MEMORY_KEYS": " ".join(memory_keys),
    "EXPECTED_MEMORY_KEYS": " ".join(expected_memory_keys),
    "FINAL_MEMORY_INTENTS": " ".join(memory_intents),
    "EXPECTED_MEMORY_INTENTS": " ".join(expected_memory_intents),
    "ATTENTION_ACCURACY": round(attention_accuracy, 3),
    "NOISE_SUPPRESSION_ACCURACY": round(noise_suppression_accuracy, 3),
    "FIELD_MEMORY_ACCURACY": round(field_memory_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "ATTENTION_THRESHOLD": attention_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "SELECTION_MARGIN": selection_margin,
    "PHASE_TOLERANCE": phase_tolerance,
    "JITTER_TOLERANCE": jitter_tolerance,
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

fig, axes = plt.subplots(len(snapshots), 1, figsize=(9, 3.1 * len(snapshots)))

if len(snapshots) == 1:
    axes = [axes]

for ax, (event_row, evaluated, selected, mem_keys, mem_intents) in zip(axes, snapshots):
    draw_snapshot(ax, event_row, evaluated, selected, mem_keys, mem_intents)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:              {spectral_states}")
print(f"directions x octaves:         {directions} x {octaves}")
print(f"memory slots:                 {len(memory_keys)}")
print(f"orbital modes:                {len(orbital_modes)}")
print(f"cases tested:                 {len(events)}")
print(f"attended packets:             {attended_count}")
print(f"rejected events:              {rejected_count}")
print(f"final memory keys:            {' '.join(memory_keys)}")
print(f"expected memory keys:         {' '.join(expected_memory_keys)}")
print(f"final memory intents:         {' '.join(memory_intents)}")
print(f"expected memory intents:      {' '.join(expected_memory_intents)}")
print(f"attention accuracy:           {round(attention_accuracy, 3)}")
print(f"noise suppression accuracy:   {round(noise_suppression_accuracy, 3)}")
print(f"field memory accuracy:        {round(field_memory_accuracy, 3)}")
print(f"resonance threshold:          {resonance_threshold}")
print(f"attention threshold:          {attention_threshold}")
print(f"ambiguity margin:             {ambiguity_margin}")
print(f"selection margin:             {selection_margin}")
print(f"phase tolerance:              {phase_tolerance}")
print(f"jitter tolerance:             {jitter_tolerance}")

print("\nAttention decision table:")
display(df_attention)

print("\nCandidate scoring table:")
display(df_candidates)

print("\nAttention memory timeline:")
display(df_memory)

print("\nSummary:")
display(df_summary)
```
