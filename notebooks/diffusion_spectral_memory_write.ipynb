```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch
from IPython.display import display

# ============================================================
# D28_SPECTRAL_MEMORY_WRITE
# Field memory writes and holds spectral states
#
# Goal:
# spectral input sequence
# -> 42-channel resonance bank
# -> decode strongest unambiguous spectral state
# -> write spectral state into memory slot
# -> hold after source is off
# -> reject weak / empty / ambiguous packets
#
# This moves memory from binary bits into 42-state spectral symbols:
# 6 directions x 7 octaves = 42 spectral memory states
# ============================================================

np.random.seed(28)

STATE = "D28_SPECTRAL_MEMORY_WRITE"

directions = 6
octaves = 7
total_states = directions * octaves

resonance_threshold = 0.62
ambiguity_margin = 0.12
retention_threshold = 0.50

input_spectral_sequence = [
    "D4_O5",
    "D1_O2",
    "D4_O5",
    "D3_O6",
    "D0_O1",
    "D5_O4",
    "D2_O3",
    "D4_O5",
]

memory_slots = len(input_spectral_sequence)

# Empty spectral memory:
# -1 means no spectral state written yet
memory_before = [-1 for _ in range(memory_slots)]


def parse_key(key):
    left, right = key.split("_O")
    direction = int(left.replace("D", ""))
    octave = int(right)
    return direction, octave


def state_id(direction, octave):
    return octave * directions + direction


def state_key(direction, octave):
    return f"D{direction}_O{octave}"


def key_from_state_id(sid):
    if sid < 0:
        return "EMPTY"
    octave = sid // directions
    direction = sid % directions
    return state_key(direction, octave)


def spectral_color(direction, octave, strength=1.0):
    if direction < 0 or octave < 0:
        return "black"

    # 42-state color map
    value = (
        0.08
        + 0.68 * (octave / max(1, octaves - 1))
        + 0.18 * (direction / max(1, directions - 1))
    )
    value = float(np.clip(value * strength, 0.04, 0.95))
    return plt.cm.inferno(value)


def make_packet(direction=None, octave=None, mode="VALID"):
    packet = np.random.uniform(0.02, 0.12, size=(octaves, directions))

    if mode == "VALID":
        packet[octave, direction] = 0.90 + np.random.normal(0, 0.025)

        # Add weak neighboring spectral bleed
        neighbor_d = (direction + 1) % directions
        packet[octave, neighbor_d] = 0.28 + np.random.normal(0, 0.02)

        if octave > 0:
            packet[octave - 1, direction] = 0.24 + np.random.normal(0, 0.02)

    elif mode == "LOW_AMPLITUDE":
        packet[octave, direction] = 0.38

    elif mode == "EMPTY_FIELD":
        packet = np.random.uniform(0.00, 0.06, size=(octaves, directions))

    elif mode == "AMBIGUOUS_TWO_KEYS":
        packet[octave, direction] = 0.88

        other_direction = (direction + 2) % directions
        other_octave = octave
        packet[other_octave, other_direction] = 0.83

    elif mode == "WHITE_NO_TARGET":
        packet = np.ones((octaves, directions)) * 0.24
        packet += np.random.uniform(0.00, 0.03, size=(octaves, directions))

    packet = np.clip(packet, 0.0, 1.0)
    return packet


def decode_packet(packet):
    flat = packet.flatten()
    sorted_indexes = np.argsort(flat)[::-1]

    best_flat = int(sorted_indexes[0])
    second_flat = int(sorted_indexes[1])

    best_score = float(flat[best_flat])
    second_score = float(flat[second_flat])

    best_octave = best_flat // directions
    best_direction = best_flat % directions

    best_sid = state_id(best_direction, best_octave)
    best_key = state_key(best_direction, best_octave)

    margin = best_score - second_score

    accepted = (
        best_score >= resonance_threshold
        and margin >= ambiguity_margin
    )

    return {
        "accepted": accepted,
        "best_direction": best_direction,
        "best_octave": best_octave,
        "best_state_id": best_sid,
        "best_key": best_key,
        "best_score": best_score,
        "second_score": second_score,
        "margin": margin,
    }


# ------------------------------------------------------------
# Write spectral sequence into memory
# ------------------------------------------------------------

memory_after_write = memory_before.copy()
memory_strength = [0.0 for _ in range(memory_slots)]

write_rows = []
channel_rows = []

for slot, key in enumerate(input_spectral_sequence):
    direction, octave = parse_key(key)
    expected_sid = state_id(direction, octave)

    packet = make_packet(direction, octave, mode="VALID")
    decoded = decode_packet(packet)

    if decoded["accepted"]:
        memory_after_write[slot] = decoded["best_state_id"]
        memory_strength[slot] = decoded["best_score"]

    written_key = key_from_state_id(memory_after_write[slot])

    write_rows.append({
        "STATE": STATE,
        "MEMORY_SLOT": slot,
        "INPUT_KEY": key,
        "INPUT_STATE_ID": expected_sid,
        "DECODED_KEY": decoded["best_key"],
        "DECODED_STATE_ID": decoded["best_state_id"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "ACCEPTED": int(decoded["accepted"]),
        "STORED_KEY": written_key,
        "STORED_STATE_ID": memory_after_write[slot],
        "MATCH": int(written_key == key)
    })

    for octave_i in range(octaves):
        for direction_i in range(directions):
            channel_rows.append({
                "STATE": STATE,
                "MEMORY_SLOT": slot,
                "INPUT_KEY": key,
                "CHANNEL_KEY": state_key(direction_i, octave_i),
                "CHANNEL_STATE_ID": state_id(direction_i, octave_i),
                "AMPLITUDE": round(float(packet[octave_i, direction_i]), 3),
                "IS_INPUT_CHANNEL": int(direction_i == direction and octave_i == octave)
            })


df_write = pd.DataFrame(write_rows)
df_channels = pd.DataFrame(channel_rows)


# ------------------------------------------------------------
# Retention / readout after source is off
# ------------------------------------------------------------

retained_strength = []
readout_rows = []

for slot, sid in enumerate(memory_after_write):
    strength = memory_strength[slot]

    # Memory keeps its spectral address but loses some amplitude over time.
    retained = float(np.clip(strength - 0.12 + np.random.normal(0, 0.025), 0.0, 1.0))
    retained_strength.append(retained)

    readout_key = key_from_state_id(sid) if retained >= retention_threshold else "FADED"
    expected_key = input_spectral_sequence[slot]

    readout_rows.append({
        "STATE": STATE,
        "MEMORY_SLOT": slot,
        "EXPECTED_KEY": expected_key,
        "STORED_KEY": key_from_state_id(sid),
        "RETAINED_STRENGTH": round(retained, 3),
        "READOUT_KEY": readout_key,
        "HELD": int(retained >= retention_threshold),
        "MATCH": int(readout_key == expected_key)
    })

df_readout = pd.DataFrame(readout_rows)


# ------------------------------------------------------------
# Rejection tests
# ------------------------------------------------------------

reject_cases = [
    ("EMPTY_FIELD", None, None),
    ("LOW_AMPLITUDE", "D4_O5", None),
    ("AMBIGUOUS_TWO_KEYS", "D2_O3", None),
    ("WHITE_NO_TARGET", None, None),
]

reject_rows = []

for case_name, key, _ in reject_cases:
    if key is None:
        packet = make_packet(mode=case_name)
        expected_key = "NONE"
    else:
        direction, octave = parse_key(key)
        packet = make_packet(direction, octave, mode=case_name)
        expected_key = key

    decoded = decode_packet(packet)

    reject_rows.append({
        "STATE": STATE,
        "CASE": case_name,
        "EXPECTED_KEY": expected_key,
        "DECODED_KEY": decoded["best_key"],
        "BEST_SCORE": round(decoded["best_score"], 3),
        "SECOND_SCORE": round(decoded["second_score"], 3),
        "MARGIN": round(decoded["margin"], 3),
        "ACCEPTED": int(decoded["accepted"]),
        "REJECTED_OK": int(not decoded["accepted"])
    })

df_reject = pd.DataFrame(reject_rows)


# ------------------------------------------------------------
# Summary metrics
# ------------------------------------------------------------

write_accuracy = float(df_write["MATCH"].mean())
readout_accuracy = float(df_readout["MATCH"].mean())
reject_accuracy = float(df_reject["REJECTED_OK"].mean())

stored_sequence = " ".join([key_from_state_id(x) for x in memory_after_write])
readout_sequence = " ".join(df_readout["READOUT_KEY"].tolist())

result = (
    "SPECTRAL_MEMORY_WRITE_LOCKED"
    if write_accuracy == 1.0 and readout_accuracy == 1.0 and reject_accuracy == 1.0
    else "SPECTRAL_MEMORY_WRITE_PARTIAL"
)

df_summary = pd.DataFrame([{
    "STATE": STATE,
    "RESULT": result,
    "SPECTRAL_STATES": total_states,
    "DIRECTIONS": directions,
    "OCTAVES": octaves,
    "MEMORY_SLOTS": memory_slots,
    "INPUT_SPECTRAL_SEQUENCE": " ".join(input_spectral_sequence),
    "STORED_SPECTRAL_SEQUENCE": stored_sequence,
    "READOUT_SPECTRAL_SEQUENCE": readout_sequence,
    "WRITE_ACCURACY": round(write_accuracy, 3),
    "READOUT_ACCURACY": round(readout_accuracy, 3),
    "REJECT_ACCURACY": round(reject_accuracy, 3),
    "RESONANCE_THRESHOLD": resonance_threshold,
    "AMBIGUITY_MARGIN": ambiguity_margin,
    "RETENTION_THRESHOLD": retention_threshold
}])


# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------

def draw_arrow(ax, start, end, active=True, reject=False):
    color = "red" if reject else "orange" if active else "gray"
    alpha = 1.0 if active else 0.25

    arrow = FancyArrowPatch(
        start,
        end,
        arrowstyle="->",
        mutation_scale=14,
        linewidth=2.0,
        color=color,
        alpha=alpha
    )
    ax.add_patch(arrow)


def draw_packet_grid(ax, x0, y0, packet, title):
    cell = 0.22
    gap = 0.045

    ax.text(x0, y0 + 2.35, title, fontsize=8, ha="left", va="bottom", color="black")

    decoded = decode_packet(packet)
    best_d = decoded["best_direction"]
    best_o = decoded["best_octave"]

    for octave in range(octaves):
        for direction in range(directions):
            x = x0 + direction * (cell + gap)
            y = y0 + (octaves - 1 - octave) * (cell + gap)

            amp = float(packet[octave, direction])
            face = spectral_color(direction, octave, 0.55 + amp)

            edge = "lime" if direction == best_d and octave == best_o and decoded["accepted"] else "black"
            lw = 2.0 if edge == "lime" else 0.4

            ax.add_patch(Rectangle(
                (x, y),
                cell,
                cell,
                facecolor=face,
                edgecolor=edge,
                linewidth=lw
            ))


def draw_memory_bank(ax, x0, y0, memory_state_ids, title):
    cell_w = 0.32
    cell_h = 0.25
    gap = 0.06

    ax.text(x0, y0 + memory_slots * (cell_h + gap) + 0.18, title, fontsize=8, ha="left", color="black")

    for slot, sid in enumerate(memory_state_ids):
        y = y0 + (memory_slots - 1 - slot) * (cell_h + gap)

        if sid < 0:
            face = "black"
            label = "--"
        else:
            d = sid % directions
            o = sid // directions
            face = spectral_color(d, o, 1.0)
            label = key_from_state_id(sid)

        ax.add_patch(Rectangle(
            (x0, y),
            cell_w,
            cell_h,
            facecolor=face,
            edgecolor="black",
            linewidth=0.6
        ))

        ax.text(x0 + 0.42, y + cell_h / 2, f"{slot}: {label}", fontsize=7, ha="left", va="center", color="black")


def draw_snapshot(ax, slot, key, t, phase, mode="WRITE"):
    direction, octave = parse_key(key)
    packet = make_packet(direction, octave, mode="VALID")
    decoded = decode_packet(packet)

    visible_memory = memory_before.copy()
    for i in range(slot + 1):
        visible_memory[i] = memory_after_write[i]

    ax.set_facecolor("white")
    ax.set_xlim(0, 12)
    ax.set_ylim(1.0, 6.5)
    ax.set_aspect("equal")
    ax.axis("off")

    wave = 0.5 + 0.5 * np.sin(2 * np.pi * phase)

    src_color = spectral_color(direction, octave, 1.0)
    ax.add_patch(Circle((1.2, 4.2), 0.52, facecolor=src_color, edgecolor="black", linewidth=1.0))
    ax.text(1.2, 3.45, "SRC", fontsize=8, ha="center", color="black")

    ax.add_patch(Rectangle((2.25, 3.95), 0.58, 0.5, facecolor=src_color, edgecolor="lime", linewidth=2.0))
    ax.text(2.54, 3.65, "42 BANK", fontsize=7, ha="center", color="black")

    mem_color = spectral_color(direction, octave, 1.0)
    ax.add_patch(Circle((4.2, 4.2), 0.58, facecolor=mem_color, edgecolor="lime", linewidth=2.0))
    ax.text(4.2, 3.45, f"SLOT {slot}", fontsize=8, ha="center", color="black")

    draw_arrow(ax, (1.75, 4.2), (2.25, 4.2), active=True)
    draw_arrow(ax, (2.85, 4.2), (3.6, 4.2), active=True)

    draw_packet_grid(ax, 5.25, 3.0, packet, "42-channel packet")
    draw_memory_bank(ax, 8.0, 2.1, visible_memory, "spectral memory")

    ax.text(
        0.7,
        1.55,
        f"slot={slot} input={key} decoded={decoded['best_key']} score={decoded['best_score']:.3f} action=WRITE_SPECTRAL_STATE",
        fontsize=8,
        ha="left",
        color="black"
    )

    ax.set_title(f"D28 snapshot t={t} — WRITE {key} INTO SLOT {slot}", fontsize=11, color="black")


snapshots = [
    (0, input_spectral_sequence[0], 0, 0.00),
    (1, input_spectral_sequence[1], 120, 0.14),
    (2, input_spectral_sequence[2], 240, 0.28),
    (3, input_spectral_sequence[3], 360, 0.42),
    (4, input_spectral_sequence[4], 480, 0.56),
    (5, input_spectral_sequence[5], 600, 0.70),
    (6, input_spectral_sequence[6], 720, 0.84),
    (7, input_spectral_sequence[7], 840, 0.96),
]

fig, axes = plt.subplots(len(snapshots), 1, figsize=(8, 3.8 * len(snapshots)))

for ax, (slot, key, t, phase) in zip(axes, snapshots):
    draw_snapshot(ax, slot, key, t, phase)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# Output
# ------------------------------------------------------------

print(f"STATE: {result}")
print(f"spectral states:          {total_states}")
print(f"directions x octaves:     {directions} x {octaves}")
print(f"memory slots:             {memory_slots}")
print(f"input spectral sequence:  {' '.join(input_spectral_sequence)}")
print(f"stored sequence:          {stored_sequence}")
print(f"readout sequence:         {readout_sequence}")
print(f"write accuracy:           {round(write_accuracy, 3)}")
print(f"readout accuracy:         {round(readout_accuracy, 3)}")
print(f"reject accuracy:          {round(reject_accuracy, 3)}")
print(f"resonance threshold:      {resonance_threshold}")
print(f"ambiguity margin:         {ambiguity_margin}")
print(f"retention threshold:      {retention_threshold}")

print("\nSpectral write table:")
display(df_write)

print("\nSpectral retention / readout table:")
display(df_readout)

print("\nRejection safety table:")
display(df_reject)

print("\nTarget channel amplitudes:")
display(
    df_channels[
        df_channels["IS_INPUT_CHANNEL"] == 1
    ].reset_index(drop=True)
)

print("\nSummary:")
display(df_summary)
```
